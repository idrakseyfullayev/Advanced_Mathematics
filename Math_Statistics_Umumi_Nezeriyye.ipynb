## Binomial Paylanma Nədir?

### **Binomial Paylanma**

**Binomial paylanma**, iki nəticəli (uğur və ya uğursuzluq) təcrübələrdən ibarət bir sıra nəticələrin ehtimalını təsvir edən statistik paylanmadır. Binomial paylanma adətən təkrarlanan təcrübələrdə istifadə olunur və hər təcrübədə müstəqil olaraq uğur və ya uğursuzluq baş verə bilər.

---

### **Əsas Xüsusiyyətlər:**
1. **Müəyyən edilmiş təcrübələrin sayı ($ n $):**
   Təcrübələr sabit sayda təkrarlanır.
   
2. **Uğur ehtimalı ($ p $):**
   Hər bir təcrübədə uğurun baş vermə ehtimalı sabitdir.
   
3. **Müstəqillik:**
   Təcrübələr bir-birindən müstəqildir, yəni bir təcrübənin nəticəsi digərinə təsir etmir.

4. **İki nəticə:**
   Hər bir təcrübənin yalnız iki mümkün nəticəsi var: uğur ($ k $) və uğursuzluq ($ n-k $).

---

### **Binomial Paylanma Formulu:**

Bir binomial paylanma təcrübəsində $X $-in $ k $ uğur almasının ehtimalı belə verilir:

$$
P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}
$$

Burada:
- $ \binom{n}{k} = \frac{n!}{k! (n-k)!} $: Kombinasiyalar,
- $ p $: Uğur ehtimalı,
- $ 1-p $: Uğursuzluq ehtimalı,
- $ k $: Uğurların sayı,
- $ n $: Təcrübələrin sayı.

---

### **Binomial Paylanmanın Xüsusiyyətləri:**

1. **Orta (Mean):**
   $$
   \mu = n \cdot p
   $$

2. **Dispersiya (Variance):**
   $$
   \sigma^2 = n \cdot p \cdot (1-p)
   $$

3. **Standart Sapma (Standard Deviation):**
   $$
   \sigma = \sqrt{n \cdot p \cdot (1-p)}
   $$

---

### **Binomial Paylanmaya Nümunə:**

**Misal:** Bir tələbə 10 suallıq bir test edir. Hər suala doğru cavab vermə ehtimalı $ p = 0.7 $-dir. Testdə doğru cavabların sayını təsvir etmək üçün binomial paylanma istifadə edilə bilər.

1. **Verilənlər:**
   - $ n = 10 $ (təcrübələrin sayı),
   - $ p = 0.7 $ (uğur ehtimalı),
   - $ k $ (doğru cavabların sayı).

2. **Ehtimalın Hesablanması (Məsələn, $ k = 7 $):**
   $$
   P(X = 7) = \binom{10}{7} (0.7)^7 (1-0.7)^3
   $$
   Burada:
   $$
   \binom{10}{7} = \frac{10!}{7! \cdot (10-7)!} = 120
   $$
   
   
Hesablamanı davam etdirərək nəticə belədir:

1. Kombinasiya ($ \binom{10}{7} $):
   $$
   \binom{10}{7} = 120
   $$

2. Binomial ehtimal:
   $$
   P(X = 7) = 120 \cdot (0.7)^7 \cdot (0.3)^3 \approx 0.2668
   $$

**Cavab:**

**Bu o deməkdir ki, tələbənin  10 sualdan 7-nin doğru cavab verilməsi ehtimalı təqribən **0.2668** (və ya 26.68%) təşkil edir.**

---

### **Python ilə Binomial Paylanma Hesablaması:**

```python
from scipy.stats import binom

# Parametrlər
n = 10  # Təcrübə sayı
p = 0.7  # Uğur ehtimalı
k = 7  # Uğur sayı

# Binomial ehtimal
probability = binom.pmf(k, n, p)
print(f"P(X = {k}) = {probability:.4f}")
```

---

### **Binomial Paylanmanın İstifadə Sahələri:**

1. **Test və sınaq analizləri:** Məsələn, məhsul keyfiyyət testlərində müəyyən bir məhsulun uğurlu olub olmadığını yoxlamaq.
2. **Tibbi təcrübələr:** Müəyyən bir dərmanın uğurlu olub olmadığını müəyyənləşdirmək.
3. **Sosial araşdırmalar:** Seçki nəticələrinin proqnozlaşdırılması üçün.

Binomial paylanma real həyatda geniş istifadə olunan bir ehtimal paylanmasıdır. Əgər əlavə nümunə və ya xüsusi bir mövzu haqqında məlumat istəsəniz, əlavə izah edə bilərəm!

## Beta Paylanma

Beta paylanma, statistikada bir ehtimal paylanmasıdır və [0,1] intervalında bir təsadüfi dəyişənin ehtimallarını modelləşdirmək üçün istifadə olunur. Bu paylanma bir çox sahədə, o cümlədən bayes statistikası, qərar nəzəriyyəsi və simulyasiya modellərində istifadə olunur.

### Beta Paylanmanın Tərifi
Beta paylanma iki müsbət parametrə, $ \alpha $ və $ \beta $-ya bağlıdır və onların qiymətləri paylanmanın formasını təyin edir. Paylanmanın sıxlıq funksiyası (PDF) belə ifadə edilir:

$$
f(x; \alpha, \beta) = \frac{x^{\alpha-1}(1-x)^{\beta-1}}{B(\alpha, \beta)}, \quad 0 < x < 1,
$$

burada 
- $ x $: Təsadüfi dəyişən,
- $ \alpha > 0 $: Forma parametri (əvvəlki müvəffəqiyyətlər),
- $ \beta > 0 $: Forma parametri (əvvəlki uğursuzluqlar),
- $ B(\alpha, \beta) $: Beta funksiyası.

### Beta Funksiyası
Beta funksiyası $ B(\alpha, \beta) $ belə tərif edilir:

$$
B(\alpha, \beta) = \int_0^1 t^{\alpha-1}(1-t)^{\beta-1} \, dt,
$$

və bu, $ \Gamma $-funksiyası ilə əlaqəlidir:
$$
B(\alpha, \beta) = \frac{\Gamma(\alpha)\Gamma(\beta)}{\Gamma(\alpha + \beta)},
$$

burada $ \Gamma(x) $ qamma funksiyasıdır.

### Beta Paylanmanın Xüsusiyyətləri
1. **Orta**:
   $$
   \mu = \frac{\alpha}{\alpha + \beta}
   $$
2. **Dispersiya**:
   $$
   \sigma^2 = \frac{\alpha \beta}{(\alpha + \beta)^2 (\alpha + \beta + 1)}
   $$
3. **Dəyişkənlik Aralığı**: Beta paylanması həmişə [0,1] intervalında dəyərlər alır.

### Beta Paylanmanın Forması
Beta paylanmasının forması $ \alpha $ və $ \beta $-nın qiymətlərindən asılıdır:
- $ \alpha = \beta = 1 $: Uniform paylanma.
- $ \alpha > 1, \beta > 1 $: Belə vəziyyətdə paylanma simmetrik və ortada zirvəli olur.
- $ \alpha < 1, \beta < 1 $: Paylanma hər iki ucda zirvəlidir.
- $ \alpha > \beta $: Paylanma sola meylli.
- $ \beta > \alpha $: Paylanma sağa meylli.

### Beta Paylanmanın Tətbiqləri
1. **Bayes statistikası**: Beta paylanma tez-tez bir Bernoulli təsadüfi dəyişən üçün ilkin paylanma kimi istifadə olunur.
2. **Proseslərin simulyasiyası**: Ehtimal dəyərlərini [0,1] intervalında modelləşdirmək üçün istifadə olunur.
3. **Beta-binomial paylanma**: Binomial paylanma ilə birləşdirilərək istifadə olunur.

### Misal
Tutaq ki, bir sınaqda əvvəlki uğurların sayı $ \alpha = 2 $, uğursuzluqların sayı isə $ \beta = 3 $-dür. Bu halda paylanmanın sıxlıq funksiyası belə olur:
$$
f(x; 2, 3) = \frac{x^{2-1}(1-x)^{3-1}}{B(2, 3)} = \frac{6x(1-x)^2}{1},
$$
burada $ B(2, 3) = \frac{\Gamma(2)\Gamma(3)}{\Gamma(5)} = \frac{1 \cdot 2}{120} = \frac{1}{6} $.

Bu funksiyanı inteqrasiya etməklə müxtəlif ehtimal bölgələrini hesablamaq olar.

## Ədədi Orta (Arithmetic Mean) vs Çəkili Orta (Weighted Mean)

### 1. **Ədədi Orta (Arithmetic Mean):**  
Ədədi orta bir sıra ədədi dəyərlərin sadəcə toplamının həmin dəyərlərin sayına bölünməsi ilə hesablanır. Bu, verilənlərdəki mərkəz meylini göstərən əsas statistik ölçülərdən biridir.

**Düstur:**  
$$
\text{Arithmetic Mean} = \frac{\sum_{i=1}^n x_i}{n}
$$ 

**Burada:**  
- $x_i$ - verilənlərin hər bir dəyəri,  
- $n$ - dəyərlərin sayı,  
- $\sum$ - toplama operatoru.

**Məsələn:**  
Verilənlər: $5, 10, 15$  
$$
\text{Arithmetic Mean} = \frac{5 + 10 + 15}{3} = \frac{30}{3} = 10
$$  

**Xüsusiyyətlər:**  
- Ədədi orta simmetrik paylanmada verilənlərin mərkəzində yerləşir.  
- Hər bir dəyər eyni çəkiyə malikdir.

---

### 2. **Çəkili Orta (Weighted Mean):**  
Çəkili orta, hər bir dəyərin fərqli çəkisi olduğu vəziyyətlər üçün istifadə olunur. Yəni hər dəyərin təsiri (çəki) fərqli ola bilər. Bu, daha əhəmiyyətli dəyərlərə daha çox təsir etməyə imkan verir.

**Düstur:**  
$$
\text{Weighted Mean} = \frac{\sum_{i=1}^n w_i x_i}{\sum_{i=1}^n w_i}
$$  

**Burada:**  
- $x_i$ - verilənlərin hər bir dəyəri,  
- $w_i$ - həmin dəyərlə əlaqəli çəki,  
- $\sum$ - toplama operatoru.

**Məsələn:**  
Verilənlər: $5, 10, 15$, çəkilər: $1, 2, 3$  
$$
\text{Weighted Mean} = \frac{(1 \cdot 5) + (2 \cdot 10) + (3 \cdot 15)}{1 + 2 + 3} = \frac{5 + 20 + 45}{6} = \frac{70}{6} \approx 11.67
$$  

**Xüsusiyyətlər:**  
- Çəkilər fərqli əhəmiyyət səviyyələrini ifadə edir.  
- Çəkilər düzgün normallaşdırılmalıdır (cəmi sıfır olmamalıdır).

---

### **Fərqlər:**  
| Xüsusiyyət             | Ədədi Orta                      | Çəkili Orta                         |
|-------------------------|----------------------------------|--------------------------------------|
| **Çəki paylanması**     | Hər bir dəyər eyni çəkiyə malikdir. | Dəyərlərə fərqli çəkilər verilir.   |
| **Düstur**              | $\frac{\sum x_i}{n}$         | $\frac{\sum w_i x_i}{\sum w_i}$   |
| **İstifadə sahəsi**     | Sadə paylanmalar.              | Çəkilər əhəmiyyətli olduqda.        |

---

### **Hər İki Orta Həyatımızda:**  
- Ədədi orta: Orta ballar, gəlirlər, xərclər kimi sadə hesablamalarda istifadə edilir.  
- Çəkili orta: İmtahan nəticələrinin hesablanmasında, maliyyə analitikalarında, portfel idarəçiliyində istifadə edilir.

## Median (Orta qiymət) vs Moda (Ən çox rast gəlinən dəyər)

### **Median (Orta qiymət)**

**Tərifi:**  
Median verilənlərin sıralanmış halda ortasında yerləşən qiymətdir. O, verilənlər toplusunun mərkəzini müəyyən edir və **dəyərlərin bərabər yarısını ayırır**. Median xüsusilə ekstremal dəyərlərə (outlierlərə) həssas olmadığı üçün statistikanın yayılma ilə bağlı təhlillərində geniş istifadə edilir.

---

**Düsturları:**

1. **Tək sayda verilənlər üçün:**
   $$
   \text{Median} = x_{(n+1)/2}
   $$

2. **Cüt sayda verilənlər üçün:**
   $$
   \text{Median} = \frac{x_{n/2} + x_{(n/2)+1}}{2}
   $$

Burada:
- $ n $: Verilənlərin sayı,
- $ x $: Sıralanmış verilənlər.

---

**Xüsusiyyətləri:**  
1. Dəyərlər sıralanmış olmalıdır.  
2. Outlierlərdən təsirlənmir (orta hesabdan fərqli olaraq).  
3. Verilənlərin mərkəzini müəyyən edir.  

---

**Nümunə:**  
Verilənlər: $ 3, 7, 2, 9, 5 $  
1. **Sıralanmış formada:** $ 2, 3, 5, 7, 9 $  
2. **Tək sayda dəyərlər: Median = 5** (ortadakı dəyər).  

Verilənlər: $ 8, 4, 6, 2 $  
1. **Sıralanmış formada:** $ 2, 4, 6, 8 $  
2. **Cüt sayda dəyərlər: Median = $\frac{4+6}{2} = 5$**.

---

### **Moda (Ən çox rast gəlinən dəyər)**

**Tərifi:**  
Moda verilənlər toplusunda **ən çox rast gəlinən qiyməti** göstərir. Moda həm kəmiyyət, həm də keyfiyyət dəyişənləri üçün istifadə edilə bilər. Verilənlərdə bir neçə moda ola bilər və bu halda verilənlər **bimodal** və ya **multimodal** adlanır.

---

**Xüsusiyyətləri:**  
1. Moda həm say, həm də kateqorik verilənlər üçün uyğundur.  
2. Verilənlərdə modanın olmaması mümkündür (hər dəyər yalnız bir dəfə rast gəlinirsə).  
3. Çox modanın olması mümkündür.  

---

**Düstur:**  
- Verilənlərdə **ən çox təkrarlanan dəyər modadır**.
- Dağılım üçün sıxlıq təhlilində, modanı qrafikdə **pik nöqtə** kimi görmək mümkündür.

---

**Nümunə:**  
Verilənlər: $ 4, 6, 6, 8, 10, 6, 8, 8 $  
1. **Moda = 6** və **Moda = 8** (Bimodal verilənlər).  

Verilənlər: $ 1, 2, 3, 4, 5 $  
1. Moda yoxdur (hər bir dəyər yalnız bir dəfə rast gəlinir).

---

### **Median və Modanın Müqayisəsi**

| Xüsusiyyət      | Median                | Moda                   |
|------------------|-----------------------|------------------------|
| **Nəyi göstərir?** | Ortadakı qiymət        | Ən çox rast gəlinən qiymət |
| **Tətbiq sahəsi**  | Kəmiyyət dəyişənləri   | Kəmiyyət və keyfiyyət dəyişənləri |
| **Outlierlərə təsir** | Təsir etmir           | Təsir etmir            |
| **Bir neçə dəyər**  | Yalnız bir median olur | Çox moda ola bilər      |

---

### **Tətbiqlər:**
1. **Median:**
   - Gəlir, qiymət, zaman analitikası üçün uyğundur.
   - Mərkəzi meyli göstərir.
   
2. **Moda:**
   - Ən çox tələb olunan məhsulu müəyyənləşdirmək üçün istifadə edilir.
   - Ən çox rast gəlinən kateqoriyanı (rəng, marka və s.) təhlil etmək üçün tətbiq olunur.

## Q1 və Q3 hesablanma qaydası 

Q1 (Birinci kvartil) və Q3 (Üçüncü kvartil) verilənlərdəki paylanmanın 25% və 75% nöqtələrini ifadə edir. Onları tapmaq üçün aşağıdakı addımları izləyin:

### Addımlar:

1. **Verilənləri sırala:**
   Verilənləri kiçikdən böyüyə doğru sıralayın.

2. **Medianı tap:**
   Median verilənlərin tam ortasındakı nöqtədir. Verilənlərin sayı cüt olduqda, ortadakı iki ədədin orta qiymətini tapın.

3. **Q1 (Birinci kvartil):**
   - Medianın **solunda** qalan verilənləri götürün.
   - Bu yarının medianını hesablayın. Bu Q1-dir.

4. **Q3 (Üçüncü kvartil):**
   - Medianın **sağında** qalan verilənləri götürün.
   - Bu yarının medianını hesablayın. Bu Q3-dir.

---

### Məsələn:

Verilmiş data:  
`1, 3, 5, 7, 9, 11, 13, 15, 17`

1. Median (ortadakı dəyər): **9**  
2. Sol hissə (Medianın solu): `1, 3, 5, 7`  
   - Bu verilənlərin medianı (Q1): `(3+5)/2 = 4`  
3. Sağ hissə (Medianın sağı): `11, 13, 15, 17`  
   - Bu verilənlərin medianı (Q3): `(13+15)/2 = 14`  

Nəticə:  
Q1 = 4, Q3 = 14

## Q1 və Q3  hesablanması üçün düstur 

Q1 (birinci kvartil) və Q3 (üçüncü kvartil) hesablanması üçün düstur konkret verilənlər siyahısının mövqelərinə əsaslanır. Düsturlar verilənlər ardıcıllığı sıralandıqdan sonra tətbiq olunur.

---

### Düsturlar:

**1. Q1 (Birinci kvartil):**  
$ Q1 = \text{Data at } \left(\frac{n+1}{4}\right)\text{-th position} $

**2. Q3 (Üçüncü kvartil):**  
$ Q3 = \text{Data at } \left(\frac{3(n+1)}{4}\right)\text{-th position} $

Burada:  
- $ n $: verilənlərin ümumi sayı  
- Mövqe tam ədədlə düşmədikdə, interpolasiya (aralıq qiymətlərin tapılması) istifadə olunur.

---

### Nümunə:

**Verilən data:**  
$ 2, 4, 6, 8, 10, 12, 14, 16 $ (sıralanmışdır, $ n = 8 $)

**1. Q1 hesablanması:**  
$ \frac{n+1}{4} = \frac{8+1}{4} = 2.25 \  $ 

2-ci və 3-cü elementlər arasında interpolasiya:  
$ Q1 = 4 + 0.25 \times (6 - 4) = 4 + 0.5 = 4.5 $

**2. Q3 hesablanması:**  
$ \frac{3(n+1)}{4} = \frac{3 \times (8+1)}{4} = 6.75 $  

6-cı və 7-ci elementlər arasında interpolasiya:  
$ Q3 = 12 + 0.75 \times (14 - 12) = 12 + 1.5 = 13.5 $

**Nəticə:**  
Q1 = 4.5, Q3 = 13.5
