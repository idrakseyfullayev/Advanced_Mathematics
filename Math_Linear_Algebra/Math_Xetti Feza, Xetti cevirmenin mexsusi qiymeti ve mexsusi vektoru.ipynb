## Xətti fəza (1)

>**Tərif:** Xətti fəza (vektor fəzası) skalyarlarla vurma və vektorların toplanması əməliyyatlarına görə bağlı olan və bu əməliyyatların əsas xassələrini təmin edən çoxluqdur.  

**Düstur:**  
Vektor fəzasında $ u, v, w $ vektorlar və $ a, b $ skalyarlar olmaqla, əsas şərtlərdən biri:  

$$
a (u + v) = au + av
$$

Bu, skalyarın vektor cəmi üzərində paylanmasını göstərir.


---

>**Tərif:** Xətti fəza, həmçinin vektor fəzası kimi tanınır, xətti cəbr və funksional analizdə əsas anlayışlardan biridir. O, vektor adlanan obyektlərin bir cəmini və skalyarlarla vurulmasını müəyyən edən cəbri quruluşdur.  

### **Xətti fəzaların əsas xüsusiyyətləri**  
Bir çox hallarda $\mathbb{R}^n$ və ya $\mathbb{C}^n$ kimi tanış olduğumuz vektor fəzaları xətti fəzanın xüsusi hallarıdır. Rəsmi olaraq, bir vektor fəzası aşağıdakı şərtləri ödəyən obyektlər çoxluğudur:  

- **Cəmin kommutativliyi**: $ u + v = v + u $  
- **Cəmin assosiativliyi**: $ (u + v) + w = u + (v + w) $  
- **Nəticəçi (zero vektor) mövcuddur**: $ 0 $ elədir ki, $ v + 0 = v $ hər $ v $-yə görə  
- **Əks element mövcuddur**: Hər $ v $-yə görə $ -v $ mövcuddur ki, $ v + (-v) = 0 $  
- **Skalyarla vurmada assosiativlik**: $ a(bv) = (ab)v $  
- **Skalyarların vektorlar üzərində paylanması**:  
  - $ a(u + v) = au + av $  
  - $ (a + b)v = av + bv $  
- **Birim skalyarın təsiri**: $ 1 \cdot v = v $  

Burada $ u, v, w $ vektorlar, $ a, b $ isə skalyarlardır.  

### **Xətti fəzanın əsas nümunələri**  
- **Euklid fəzası** ($\mathbb{R}^n$) – Adi $n$-ölçülü vektorlar fəzası  
- **Polinomlar fəzası** – Verilmiş dərəcədən çox olmayan polinomların çoxluğu  
- **Matrislər fəzası** – Məsələn, bütün $ m \times n $ ölçülü matrislər çoxluğu

## Xətti Çevirmə (1)

### **Xətti Çevirmə Xətti Fəzada**  

> **Tərif:** Xətti çevirmə (Linear Transformation), bir vektor məkanındakı vektorları başqa bir vektor məkanına ötürən xüsusi bir funksiyadır. Bu çevirmə vektorların əsas xüsusiyyətlərini qoruyur və xətti fəzalar daxilində işlənir.

---

## **1. Xətti Fəza (Vector Space) nədir?**  

**Xətti fəza**, vektorların cəmi və skalyara vurma əməliyyatlarına bağlı olan və müəyyən riyazi qaydalara tabe olan vektorlar toplusudur. Bir xətti fəza aşağıdakı xüsusiyyətlərə malikdir:  

✅ **Toplama və skalyara vurma bağlıdır**  
✅ **Sıfır vektor mövcuddur**  
✅ **Hər vektorun tərs elementi var**  
✅ **Distribütiv, assosiativ və kommutativ qanunları ödənir**  

Məsələn, **$ \mathbb{R}^n $** vektor fəzaları ən geniş istifadə olunan xətti fəzalar sırasına daxildir.

---

## **2. Xətti Çevirmə Nədir?**  

Bir **$ T: V \to W $** xətti çevirməsi, **$ V $** xətti fəzadan **$ W $** xətti fəzaya keçid edən bir funksiyadır və aşağıdakı iki əsas şərti ödəmiş olur:  

1️⃣ **Cəmin qorunması**  
$$
T(\mathbf{u} + \mathbf{v}) = T(\mathbf{u}) + T(\mathbf{v})
$$
(yəni vektorların cəmi çevriləndən sonra da saxlanmalıdır).  

2️⃣ **Skalyara vurmanın qorunması**  
$$
T(c \mathbf{v}) = c T(\mathbf{v})
$$
(yəni bir vektor çevrilməzdən əvvəl skalyara vurulsa, nəticə dəyişməz).  

Bu iki şərti ödəyən çevirmə **xətti çevirmədir**.

---

## **3. Xətti Çevirmənin Matris Təmsili**  

Xətti çevirmələr matrislər vasitəsilə ifadə edilir. Məsələn, **$ A $** matrisi ilə təsvir olunan xətti çevirmə:

$$
T(\mathbf{x}) = A \mathbf{x}
$$

Burada:
- **$ A $** — xətti çevirmə matrisidir,
- **$ \mathbf{x} $** — çevrilən vektordur.

Məsələn, **$ \mathbb{R}^2 $** vektorlarını çevirmək üçün **$ 2 \times 2 $** matrisi istifadə edə bilərik:

$$
A = \begin{bmatrix} 2 & 1 \\ 0 & 3 \end{bmatrix}
$$

və vektor **$ \mathbf{x} = \begin{bmatrix} x \\ y \end{bmatrix} $** çevrildikdə:

$$
T(\mathbf{x}) = A \mathbf{x} = \begin{bmatrix} 2x + y \\ 3y \end{bmatrix}
$$

Bu, **hər bir vektorun yeni fəzada necə dəyişdiyini** göstərir.

---

## **4. Xətti Çevirmənin Xüsusi Növləri**  

Xətti çevirmələr müxtəlif formada ola bilər. Ən çox istifadə edilənləri:  

✅ **Ölçü dəyişməsi (Scaling)**  
   $$
   A = \begin{bmatrix} k & 0 \\ 0 & k \end{bmatrix}
   $$
   - Bütün vektorları $ k $ qədər böyüdür və ya kiçildir.  

✅ **Döndərmə (Rotation)**  
   $$
   A = \begin{bmatrix} \cos \theta & -\sin \theta \\ \sin \theta & \cos \theta \end{bmatrix}
   $$
   - Vektorları **$ \theta $** bucağı qədər fırladır.  

✅ **Kəsmə (Shearing)**  
   $$
   A = \begin{bmatrix} 1 & k \\ 0 & 1 \end{bmatrix}
   $$
   - Obyektləri bir istiqamətdə uzadır.  

✅ **Refleksiya (Reflection)**  
   - **X oxuna görə:**  
     $$
     A = \begin{bmatrix} 1 & 0 \\ 0 & -1 \end{bmatrix}
     $$
   - **Y oxuna görə:**  
     $$
     A = \begin{bmatrix} -1 & 0 \\ 0 & 1 \end{bmatrix}
     $$

---

## **5. Xətti Çevirmənin Xüsusi Nöqtələri**  

Xətti çevirmənin bəzi əsas anlayışları var:  

🔹 **Öz vektorlar (Eigenvectors) və öz dəyərlər (Eigenvalues)**  
Əgər **$ T(\mathbf{v}) = \lambda \mathbf{v} $** olarsa,  
- **$ \mathbf{v} $** **öz vektor** adlanır,  
- **$ \lambda $** **öz dəyər** adlanır.  
Bu anlayış **PCA və maşın öyrənmədə** istifadə olunur.  

🔹 **Rang və determinant**  
- **Rang** — çevirmə zamanı vektorların neçəölçülü fəzada qaldığını göstərir.  
- **Determinant** — çevirmənin həcmə və sahəyə təsirini ölçür.  

---

## **6. Xətti Çevirmənin Tətbiq Sahələri**  

📌 **Maşın öyrənməsi və Neyron Şəbəkələr**  
- PCA (Baş komponent analizi) üçün istifadə olunur.  

📌 **Kompüter Qrafikası**  
- Obyektlərin fırlanması və perspektiv transformasiyaları.  

📌 **Fizika və Dinamika**  
- Cisimlərin hərəkəti və koordinat çevrilmələri.  

📌 **Siqnal Emalı**  
- Fourier çevrilmələri və filtrasiya prosesləri.  

---

### **Nəticə**  

✅ **Xətti çevirmə xətti fəzalar daxilində vektorları başqa vektor məkanına keçirən bir funksiyadır.**  
✅ **Xətti çevirmə matrislər vasitəsilə ifadə edilir və müəyyən xüsusiyyətləri qoruyur.**  
✅ **Bu anlayış maşın öyrənməsi, fizika, qrafika və digər sahələrdə geniş istifadə olunur.**

## Xətti Çevirmənin Məxsusi Qiyməti və Məxsusi Vektoru (1)

### **Xətti Çevirmənin Məxsusi Qiyməti və Məxsusi Vektoru**  

Xətti çevirmənin (Linear Transformation) **məxsusi vektoru** (eigenvector) və **məxsusi qiyməti** (eigenvalue) anlayışları xətti cəbrin əsas mövzularından biridir. Bu anlayışlar vektorların çevrilmə nəticəsində necə davrandığını anlamağa kömək edir və **PCA (Baş Komponent Analizi), Markov Zəncirləri, Kvant Mexanikası, Maşın Öyrənməsi və digər sahələrdə** geniş tətbiq olunur.  

---

## **1. Məxsusi Vektor və Məxsusi Qiymət nədir?**  

Bir xətti çevirmə **$ T: V \to V $** üçün **$ \mathbf{v} \neq 0 $** olan bir vektor elə ola bilər ki, çevrilmədən sonra **yalnız miqdarca dəyişsin**, yəni **istiqaməti dəyişməsin**.  

Matris şəklində ifadə etsək:

$$
A \mathbf{v} = \lambda \mathbf{v}
$$

Burada:  
- **$ A $** — xətti çevirməni təmsil edən **$ n \times n $** ölçülü kvadrat matrisdir,  
- **$ \mathbf{v} \neq 0 $** — **məxsusi vektor** (eigenvector) adlanır,  
- **$ \lambda $** — **məxsusi qiymət** (eigenvalue) adlanır.  

Bu tənlik onu göstərir ki, **$ A \mathbf{v} $** sadəcə **$ \lambda $** skalyarına vurulmuş **$ \mathbf{v} $** vektorudur. Yəni **$ \mathbf{v} $** yalnız uzunluqca dəyişir, istiqamət dəyişmir.

---

## **2. Məxsusi Qiymətlərin Tapılması**  

Məxsusi qiymətləri tapmaq üçün aşağıdakı determinant tənliyini həll edirik:

$$
\det(A - \lambda I) = 0
$$

Burada:  
- **$ I $** — vahid matrisdir (**identity matrix**),  
- **$ \lambda I $** çıxmaqla matrisdən diaqonalını dəyişirik,  
- **$ \det(A - \lambda I) $** determinantı sıfıra bərabər edirik.  

#### **Misal:**
Gəlin, aşağıdakı matris üçün məxsusi qiymətləri tapaq:

$$
A = \begin{bmatrix} 4 & 2 \\ 1 & 3 \end{bmatrix}
$$

Məxsusi qiymətləri tapmaq üçün **$ \det(A - \lambda I) = 0 $** determinantını həll edirik:

$$
\det \begin{bmatrix} 4 - \lambda & 2 \\ 1 & 3 - \lambda \end{bmatrix} = 0
$$

Determinantı hesablayaq:

$$
(4 - \lambda)(3 - \lambda) - (2 \cdot 1) = 0
$$

$$
12 - 4\lambda - 3\lambda + \lambda^2 - 2 = 0
$$

$$
\lambda^2 - 7\lambda + 10 = 0
$$

Bu kvadrat tənliyi həll edirik:

$$
(\lambda - 5)(\lambda - 2) = 0
$$

Buradan **məxsusi qiymətlər**:

$$
\lambda_1 = 5, \quad \lambda_2 = 2
$$

---

## **3. Məxsusi Vektorların Tapılması**  

Hər bir **$ \lambda $** üçün **$ A \mathbf{v} = \lambda \mathbf{v} $** tənliyini həll edərək **$ \mathbf{v} $** tapırıq.

### **$ \lambda_1 = 5 $** üçün:

$$
(A - 5I) \mathbf{v} = 0
$$

$$
\begin{bmatrix} -1 & 2 \\ 1 & -2 \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix}
$$

İlk tənlikdən:

$$
-1x + 2y = 0 \quad \Rightarrow \quad x = 2y
$$

$ y $-yə istənilən skalyar dəyər verə bilərik. Məsələn, $ y = 1 $ götürsək:

$$
\mathbf{v}_1 = \begin{bmatrix} 2 \\ 1 \end{bmatrix}
$$

---

### **$ \lambda_2 = 2 $** üçün:

$$
(A - 2I) \mathbf{v} = 0
$$

$$
\begin{bmatrix} 2 & 2 \\ 1 & 1 \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix}
$$

İlk tənlikdən:

$$
2x + 2y = 0 \quad \Rightarrow \quad x = -y
$$

Burada da $ y $-yə istənilən dəyəri verə bilərik. Məsələn, $ y = 1 $ götürsək:

$$
\mathbf{v}_2 = \begin{bmatrix} -1 \\ 1 \end{bmatrix}
$$

---

## **4. Məxsusi Qiymət və Məxsusi Vektorların Əhəmiyyəti**  

✅ **PCA (Baş Komponent Analizi)** — Statistika və maşın öyrənmədə **dimensiya azaldılması** üçün istifadə edilir.  
✅ **Differensial tənliklər və fizikada** — Sistemlərin uzunmüddətli davranışını anlamaq üçün tətbiq edilir.  
✅ **Markov zəncirləri** — Stokastik sistemlərdə tarazlıq vəziyyətini tapmaq üçün istifadə olunur.  
✅ **Kvant mexanikası** — Şrödinger tənliklərində kvant halları məxsusi vektorlarla ifadə olunur.  

---

## **5. Nəticə**  

📌 **Məxsusi qiymət** — xətti çevirmədə bir vektorun **miqdarca** necə dəyişdiyini göstərir.  
📌 **Məxsusi vektor** — xətti çevirmədə **istiqaməti dəyişməyən** vektordur.  
📌 **$ \det(A - \lambda I) = 0 $** tənliyi həll edilərək məxsusi qiymətlər tapılır.  
📌 **$ (A - \lambda I) \mathbf{v} = 0 $** tənliyini həll edərək məxsusi vektorlar tapılır.
