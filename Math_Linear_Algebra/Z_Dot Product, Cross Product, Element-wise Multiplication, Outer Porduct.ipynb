## İki vektorun Skalyar hasili (Dot Product)

### **Tərif:**  
>**Tərif:** İki vektorun **skalyar hasili** (dot product), bir **sətir matrisinin** ($1 \times n$) bir **sütun matrisinə** ($n \times 1$) vurulması ilə hesablanır. Nəticə **$1 \times 1$** ölçülü skalyar bir ədəd olur:  

$$
\mathbf{A} \cdot \mathbf{B} = \sum_{i=1}^{n} A_i B_i
$$

---

### **Bir sətir matrisinin bir sütun matrisinə hasili:**  
$$
\begin{bmatrix} a_1 & a_2 & \dots & a_n \end{bmatrix}  \cdot 
\begin{bmatrix} b_1 \\ b_2 \\ \vdots \\ b_n \end{bmatrix}  
= (a_1 \times b_1) + (a_2 \times b_2) + \dots + (a_n \times b_n)
$$

---

### **Nümunə:** 

Verilmiş vektorlar:  
$$
\begin{bmatrix} 2 & 3 & 4 \end{bmatrix} \quad (1 \times 3)
$$

$$
\begin{bmatrix} 5 \\ 6 \\ 7 \end{bmatrix} \quad (3 \times 1)
$$

$$
\begin{bmatrix} 2 & 3 & 4 \end{bmatrix}  \cdot  
\begin{bmatrix} 5 \\ 6 \\ 7 \end{bmatrix}  
= (2 \times 5) + (3 \times 6) + (4 \times 7) = 10 + 18 + 28 = 56
$$

## İki vektorun Vektorial hasili (Cross Product)



### **Tərif:**  
>**Tərif:** İki vektorun **vektorial hasili** (cross product), **$3 \times 1$** ölçülü **sütun matrisinin** başqa bir **$3 \times 1$** ölçülü **sütun matrisinə** hasilidir. Nəticə yenə **$3 \times 1$** ölçülü **sütun matrisi** olur. Lakin Matrislərin hasil qaydalarına görə, **$3 \times 1$** ölçülü **sütun matrisini** başqa bir **$3 \times 1$** ölçülü **sütun matrisinə** birbaşa vurmaq mümkün deyil. **Çünki matris hasilinin tərifinə görə, birinci matrisin sütun sayı, ikinci matrisin sətr sayına bərabər olmalıdır**. Vektorial hasil xüsusi bir əməliyyatdır və o, determinant ifadəsi ilə müəyyən edilir. Onu matris hasilinə uyğun yazmaq üçün **sətir və sütun matrislərinin xarici hasili (outer product) kimi də düşünmək olar**, amma standart matris hasilindən fərqlənir. İki **$3 \times 1$** ölçülü **sütun matrisinin** vektorial hasili standart matris vurulması qaydalarına uyğun deyil. Onu **determinant metodu** ilə **$3 \times 1$** ölçülü yeni **sütun matris** kimi ifadə etmək olar:  
  
$$
\mathbf{A} \times \mathbf{B} =
\begin{vmatrix} 
\mathbf{i} & \mathbf{j} & \mathbf{k} \\ 
A_1 & A_2 & A_3 \\ 
B_1 & B_2 & B_3 
\end{vmatrix}
$$

Bu determinant açıldıqda nəticə yenə **$3 \times 1$** ölçülü **sütun matrisidir**, amma birbaşa matris hasilinə uyğun gəlmir.

---

### **Nəticə:**  

>**Tərif:** İki vektorun **vektorial hasili** (cross product), yalnız **3 ölçülü vektorlar** üçün müəyyən olunur. Vektorial hasil nəticəsində yeni **3 ölçülü vektor** əmələ gəlir və bu **vektor hər iki ilkin vektora perpendikulyar olur.**  

$$
\mathbf{A} \times \mathbf{B} =
\begin{vmatrix} 
\mathbf{i} & \mathbf{j} & \mathbf{k} \\ 
A_1 & A_2 & A_3 \\ 
B_1 & B_2 & B_3 
\end{vmatrix}
$$

---

### **İki vektorun vektorial hasili:**  
$$
\begin{bmatrix} A_1 \\ A_2 \\ A_3 \end{bmatrix}  
\times  
\begin{bmatrix} B_1 \\ B_2 \\ B_3 \end{bmatrix}  
=
\begin{bmatrix} 
(A_2 B_3 - A_3 B_2) \\ 
(A_3 B_1 - A_1 B_3) \\ 
(A_1 B_2 - A_2 B_1) 
\end{bmatrix}
$$

---

### **Nümunə:** 

Verilmiş vektorlar:  
$$
\begin{bmatrix} 2 \\ 3 \\ 4 \end{bmatrix} \quad (3 \times 1)
$$

$$
\begin{bmatrix} 5 \\ 6 \\ 7 \end{bmatrix} \quad (3 \times 1)
$$




$$
\mathbf{A} \times \mathbf{B} =
\begin{vmatrix} 
\mathbf{i} & \mathbf{j} & \mathbf{k} \\ 
2 & 3 & 4 \\ 
5 & 6 & 7 
\end{vmatrix}
$$




$$
\begin{bmatrix} 2 \\ 3 \\ 4 \end{bmatrix}  
\times  
\begin{bmatrix} 5 \\ 6 \\ 7 \end{bmatrix}  
=
\begin{bmatrix} 
(3 \times 7 - 4 \times 6) \\ 
(4 \times 5 - 2 \times 7) \\ 
(2 \times 6 - 3 \times 5) 
\end{bmatrix}
=
\begin{bmatrix} 
(21 - 24) \\ 
(20 - 14) \\ 
(12 - 15) 
\end{bmatrix}
=
\begin{bmatrix} -3 \\ 6 \\ -3 \end{bmatrix}
$$

---

### **Düzgün izah (matrislər əsasında):**  
Əgər iki **$3 \times 1$** ölçülü sütun matrisini matris hasilinə uyğun ifadə etmək istəsək, onların transpozisiyasından istifadə etməliyik:

- **Skalyar hasildə**:  
  $$
  (\mathbf{A}^T \times \mathbf{B}) \Rightarrow (1 \times 3) \times (3 \times 1) = 1 \times 1
  $$
  Bu, **skalyar ədəd** verir.

- **Vektorial hasildə**:  
  Vektorial hasil üçün xüsusi determinant üsulu istifadə edilir və onu adi matris hasilinə tam uyğun şəkildə ifadə etmək olmur.

---

## Element-wise Multiplication (Hadamard çarpması)

**Element-wise multiplication (Hadamard çarpması)**, **iki eyni ölçülü matrisin** hər bir elementinin bir-biri ilə vurulmasıdır. Bu əməliyyat **iki matrisin** elementlərini **əvvəlcə uyğun mövqelərdəki** dəyərləri vuraraq həyata keçirilir.

Bu, **iki sütun matrisinin** və ya **iki sətir matrisinin** bir-birinə vurulmasından fərqlidir. Hadamard çarpması, matrisin ümumi vurulmasından fərqli olaraq, **matrisin ölçüləri eyni olmalıdır** (yəni, hər iki matrisin sətir və sütun sayıları eyni olmalıdır).

---

### **Tərif:**  
İki **$m \times n$** ölçülü matrisin Hadamard çarpması, **hər bir elementin bir-biri ilə vurulması** ilə əldə edilir:

$$
\mathbf{A} \circ \mathbf{B} = 
\begin{bmatrix} 
A_{11} \cdot B_{11} & A_{12} \cdot B_{12} & \dots & A_{1n} \cdot B_{1n} \\
A_{21} \cdot B_{21} & A_{22} \cdot B_{22} & \dots & A_{2n} \cdot B_{2n} \\
\vdots & \vdots & \ddots & \vdots \\
A_{m1} \cdot B_{m1} & A_{m2} \cdot B_{m2} & \dots & A_{mn} \cdot B_{mn}
\end{bmatrix}
$$

---

### **Nümunə:**
$$
\mathbf{A} = \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}, \quad 
\mathbf{B} = \begin{bmatrix} 5 & 6 \\ 7 & 8 \end{bmatrix}
$$

Hadamard çarpması:

$$
\mathbf{A} \circ \mathbf{B} = 
\begin{bmatrix} 
1 \cdot 5 & 2 \cdot 6 \\
3 \cdot 7 & 4 \cdot 8 
\end{bmatrix} 
= \begin{bmatrix} 
5 & 12 \\
21 & 32
\end{bmatrix}
$$

---

Əlbəttə! **Hadamard çarpması** 3x1 ölçülü sütun vektorları üçün aşağıdakı kimi olacaq:

---

### **Nümunə:**

$$
\mathbf{A} = \begin{bmatrix} 2 \\ 3 \\ 4 \end{bmatrix}, \quad 
\mathbf{B} = \begin{bmatrix} 5 \\ 6 \\ 7 \end{bmatrix}
$$

Hadamard çarpması:

$$
\mathbf{A} \circ \mathbf{B} = 
\begin{bmatrix} 
2 \cdot 5 \\
3 \cdot 6 \\
4 \cdot 7 
\end{bmatrix} 
= \begin{bmatrix} 
10 \\
18 \\
28 
\end{bmatrix}
$$

---

Təbii, **Hadamard çarpması** 1x3 ölçülü iki sətir vektoru üçün aşağıdakı kimi olacaq:

---

### **Nümunə:**

$$
\mathbf{A} = \begin{bmatrix} 2 & 3 & 4 \end{bmatrix}, \quad 
\mathbf{B} = \begin{bmatrix} 5 & 6 & 7 \end{bmatrix}
$$

Hadamard çarpması:

$$
\mathbf{A} \circ \mathbf{B} = 
\begin{bmatrix} 
2 \cdot 5 & 3 \cdot 6 & 4 \cdot 7
\end{bmatrix} 
= \begin{bmatrix} 
10 & 18 & 28
\end{bmatrix}
$$

---



### **Cavab:**  
Hadamard çarpması, **iki eyni ölçülü matrisin** elementləri arasında tətbiq edilən vurma əməliyyatıdır. Bu, **setir matrisinin sütun matrisinə və ya iki sütun matrisinin bir-birinə vurulması** ilə deyil, **iki eyni ölçülü matrisin** element-wise vurulması ilə edilir.

## **Xarici Hasil (Outer Product)**



### **Tərif:**

> **Tərif:** İki vektorun **xarici hasili (outer product)**, bir **sütun vektorunun** (\$m \times 1\$) bir **sətir vektoru** (\$1 \times n\$) ilə vurulması nəticəsində **\$m \times n\$** ölçülü matris əmələ gətirir. Bu, standart matris hasil qaydalarına uyğundur, çünki birinci matrisin sütun sayı (1) ikinci matrisin sətir sayına (1) bərabərdir.

$$
\mathbf{A} \otimes \mathbf{B} = \mathbf{A} \cdot \mathbf{B}^T
$$

---

### **Bir sütun matrisinin bir sətir matrisinə hasili:**

$$
\begin{bmatrix} a_1 \\ a_2 \\ \vdots \\ a_m \end{bmatrix}  
\cdot  
\begin{bmatrix} b_1 & b_2 & \dots & b_n \end{bmatrix}  
= 
\begin{bmatrix} 
a_1 b_1 & a_1 b_2 & \dots & a_1 b_n \\ 
a_2 b_1 & a_2 b_2 & \dots & a_2 b_n \\ 
\vdots & \vdots & \ddots & \vdots \\ 
a_m b_1 & a_m b_2 & \dots & a_m b_n 
\end{bmatrix}
$$

---

### **Nümunə:**

Verilmiş vektorlar:

$$
\mathbf{A} = \begin{bmatrix} 2 \\ 3 \\ 4 \end{bmatrix} \quad (3 \times 1)
$$

$$
\mathbf{B} = \begin{bmatrix} 5 & 6 & 7 \end{bmatrix} \quad (1 \times 3)
$$

Xarici hasil:

$$
\mathbf{A} \otimes \mathbf{B} = 
\begin{bmatrix} 2 \\ 3 \\ 4 \end{bmatrix}  
\cdot  
\begin{bmatrix} 5 & 6 & 7 \end{bmatrix}  
= 
\begin{bmatrix} 
2 \cdot 5 & 2 \cdot 6 & 2 \cdot 7 \\ 
3 \cdot 5 & 3 \cdot 6 & 3 \cdot 7 \\ 
4 \cdot 5 & 4 \cdot 6 & 4 \cdot 7 
\end{bmatrix}  
= 
\begin{bmatrix} 
10 & 12 & 14 \\ 
15 & 18 & 21 \\ 
20 & 24 & 28 
\end{bmatrix}
$$

---

### **Əhəmiyyətli fərqlər:**

1. **Skalyar hasil (Dot Product):**  
   - Nəticə **skalyar** (tək ədəd) olur.  
   - Bir **sətir vektoru** (\$1 \times n\$) ilə bir **sütun vektoru** (\$n \times 1\$) vurulur.  
   - Əgər **iki sütun vektoru** verilibsə, birincinin transpozu (\$\mathbf{A}^T\$) alınmalıdır.

2. **Vektorial hasil (Cross Product):**  
   - Yalnız **3D vektorlar** üçün müəyyən edilib.  
   - Nəticə **yeni vektor** (hər ikisinə perpendikulyar) olur.  
   - Matris vurma qaydalarına tam uyğun deyil, determinant üsulu ilə hesablanır.

3. **Element-wise (Hadamard) Hasil:**  
   - **Eyni ölçülü** matrislərin uyğun elementləri vurulur.  
   - Nəticə eyni ölçülü matris olur.  
   - **Fərqli ölçülər üçün tətbiq edilməz.**

4. **Xarici hasil (Outer Product):**  
   - Bir **sütun vektoru** (\$m \times 1\$) ilə bir **sətir vektoru** (\$1 \times n\$) vurulur.  
   - Nəticə **\$m \times n\$** ölçülü matris olur.  
   - Əgər **iki sütun vektoru** verilibsə, ikincinin transpozu (\$\mathbf{B}^T\$) alınmalıdır.

---

### **Xülasə:**

| Hasil növü         | Giriş ölçüləri       | Nəticə ölçüsü       | Şərtlər                     |
|--------------------|----------------------|---------------------|-----------------------------|
| **Skalyar (Dot)**  | \$1 \times n\$ və \$n \times 1\$ | \$1 \times 1\$       | Sətir × Sütun               |
| **Vektorial (Cross)** | \$3 \times 1\$ və \$3 \times 1\$ | \$3 \times 1\$       | Yalnız 3D vektorlar üçün    |
| **Hadamard**       | \$m \times n\$ və \$m \times n\$ | \$m \times n\$       | Eyni ölçülü olmalıdır       |
| **Xarici (Outer)** | \$m \times 1\$ və \$1 \times n\$ | \$m \times n\$       | Sütun × Sətir               |

---

### **Əlavə qeyd:**
- **Xarici hasil**, **tensor hasilinin** xüsusi halıdır və xətti cəbrdə çoxlu tətbiqatlara malikdir (məsələn, proyeksiya matrisləri, xarici törəmələr).  
- **Skalyar hasil** və **xarici hasil** arasındakı əsas fərq ondadır ki, skalyar hasil **sıxılma (contraction)** əməliyyatıdır (nəticə tək ədəd), xarici hasil isə **genişlənmə (expansion)** əməliyyatıdır (nəticə matris).  

Bu izahatlar vasitəsilə müxtəlif vektor/matris hasil növləri arasındakı fərqləri aydın şəkildə anlamaq mümkündür.

## Xarici hasil (Outer Product) üçün sütun vektorunu sətir vektoruna vurmaq düzgün yoldur

**Xarici hasil (Outer Product)** üçün **sütun vektorunu sətir vektoruna vurmaq** düzgün yoldur, çünki matris vurma qaydalarına görə:

- **Birinci matrisin sütun sayı** (\$n\$) **ikinci matrisin sətir sayına** (\$m\$) bərabər olmalıdır.  
- **Sütun vektoru** (\$m \times 1\$) və **sətir vektoru** (\$1 \times n\$) üçün bu şərt ödənir (\$1 = 1\$).  
- Nəticə **\$m \times n\$** ölçülü matris olacaq.

---

### **Düzgün qayda:**
$$
\underbrace{
\begin{bmatrix} 
a_1 \\ 
a_2 \\ 
\vdots \\ 
a_m 
\end{bmatrix}
}_{\text{sütun vektoru} \ (m \times 1)} 
\cdot 
\underbrace{
\begin{bmatrix} 
b_1 & b_2 & \dots & b_n 
\end{bmatrix}
}_{\text{sətir vektoru} \ (1 \times n)} 
= 
\underbrace{
\begin{bmatrix} 
a_1 b_1 & a_1 b_2 & \dots & a_1 b_n \\ 
a_2 b_1 & a_2 b_2 & \dots & a_2 b_n \\ 
\vdots & \vdots & \ddots & \vdots \\ 
a_m b_1 & a_m b_2 & \dots & a_m b_n 
\end{bmatrix}
}_{\text{xarici hasil} \ (m \times n)}
$$


---

### **Nümunə:**
1. **Sütun × Sətir** (Doğru):  
   $$
   \begin{bmatrix} 2 \\ 3 \end{bmatrix} \cdot \begin{bmatrix} 5 & 6 \end{bmatrix} = \begin{bmatrix} 10 & 12 \\ 15 & 18 \end{bmatrix}
   $$

2. **Sətir × Sütun** (Skalyar hasil verir, xarici hasil deyil!):  
   $$
   \begin{bmatrix} 2 & 3 \end{bmatrix} \cdot \begin{bmatrix} 5 \\ 6 \end{bmatrix} = 2 \cdot 5 + 3 \cdot 6 = 28 \quad (\text{Skalyar})
   $$

---

### **Sualınızın cavabı:**
- **Xarici hasil üçün:**  
  **Sütun vektorunu sətir vektoruna vurmaq lazımdır** (\$\mathbf{A} \cdot \mathbf{B}^T\$). Əksinə etsəniz (sətir × sütun), skalyar hasil alınar.  
- **Səbəb:** Matris vurma qaydalarında sütun-sətir uyğunluğu tələb olunur.  

---

### **Fərqin vizual izahı:**
| Hasil növü       | Giriş forması       | Nəticə             |
|------------------|---------------------|--------------------|
| **Xarici hasil** | Sütun (\$m \times 1\$) × Sətir (\$1 \times n\$) | Matris (\$m \times n\$) |
| **Skalyar hasil** | Sətir (\$1 \times n\$) × Sütun (\$n \times 1\$) | Ədəd (\$1 \times 1\$) |

---

### **Əgər iki sütun vektoru verilibsə?**
1. Birinci sütun vektorunu saxlayıb, ikincini **transpoz** etməlisiniz:  
   $$
   \mathbf{A} \otimes \mathbf{B} = \mathbf{A} \cdot \mathbf{B}^T
   $$  
   **Misal:**  
   $$
   \begin{bmatrix} 2 \\ 3 \end{bmatrix} \otimes \begin{bmatrix} 5 \\ 6 \end{bmatrix} = \begin{bmatrix} 2 \\ 3 \end{bmatrix} \cdot \begin{bmatrix} 5 & 6 \end{bmatrix} = \begin{bmatrix} 10 & 12 \\ 15 & 18 \end{bmatrix}
   $$

---

### **Qısa cavab:**
**Xarici hasil üçün həmişə sütun vektorunu sətir vektoruna vurun!**  
- Əgər ikisi də sütundursa, birini transpoz edin (\$\mathbf{B}^T\$).  
- Əgər sətir × sütun vursanız, skalyar hasil alınar.  

Bu qayda xətti cəbrdə matris vurma standartlarına uyğundur.
