Y…ôni, …ôg…ôr bizd…ô d√ºz x…ôttin √ºz…ôrind…ô bir **ba≈ülanƒüƒ±c n√∂qt…ô** v…ô onun **istiqam…ôt vektoru** m…ôlumdursa, h…ômin d√ºz x…ôttin b√ºt√ºn n√∂qt…ôl…ôrini parametrik ≈ü…ôkild…ô yaza bil…ôrik.

* **N√∂qt…ô** $A(x_0, y_0)$ ‚Äì d√ºz x…ôttin ke√ßdiyi ist…ônil…ôn bir n√∂qt…ô.

## üìå 3. D√ºstur

\end{cases}
\quad (t \in \mathbb{R})
$$

Burada $(a, b, c)$ ‚Äì istiqam…ôt vektoru, $(x_0, y_0, z_0)$ ‚Äì d√ºz x…ôttin ke√ßdiyi n√∂qt…ôdir.

## D√ºz X…ôttin Bucaq ∆èmsallƒ± T…ônliyi

G…ôlin **D√ºz x…ôttin bucaq …ômsallƒ± t…ônliyi**ni addƒ±m-addƒ±m izah ed…ôk.

---

## üìå 1. T…ôrif

>**T…ôrif:** **Bucaq …ômsallƒ± t…ônlik** ‚Äî d√ºz x…ôttin meylini (eƒüimini) n…ôz…ôr…ô alaraq yazƒ±lan v…ô x…ôtt √ºz…ôrind…ôki b√ºt√ºn n√∂qt…ôl…ôrin koordinatlarƒ±nƒ± birba≈üa ifad…ô ed…ôn t…ônlikdir.
Ba≈üqa s√∂zl…ô, x…ôttin $x$ oxu il…ô d√ºz bucaq altƒ±nda etdiyi bucaqdan istifad…ô ed…ôr…ôk t…ôrtib olunur.


## üìå  T…ôrif

>**T…ôrif:** **D√ºz x…ôttin bucaq …ômsallƒ± t…ônliyi**, m√ºst…ôvid…ô x…ôttin meylini g√∂st…ôr…ôn sabit bucaq …ômsalƒ± il…ô $y = kx + b$ formasƒ±nda ifad…ô olunan, x…ôttin yamacƒ±nƒ± v…ô ordinat k…ôsi≈üm…ôsini m√º…ôyy…ôn ed…ôn t…ônlikdir.



---

## üìå 2. D√ºstur (2D m√ºst…ôvi √º√ß√ºn)

∆èg…ôr $\alpha$ ‚Äî d√ºz x…ôttin $x$ oxu il…ô …ôm…ôl…ô g…ôtirdiyi bucaqdƒ±r, v…ô $(x_0, y_0)$ x…ôttin √ºz…ôrind…ô bir n√∂qt…ôdirs…ô, bucaq …ômsallƒ± t…ônlik bel…ôdir:

$$
y - y_0 = \tan \alpha \cdot (x - x_0)
$$

Burada:

* $\tan \alpha = m$ ‚Äî x…ôttin **meyl …ômsalƒ±**,
* $(x_0, y_0)$ ‚Äî x…ôttin √ºz…ôrind…ôki n√∂qt…ô.

Bel…ôlikl…ô, **sad…ôl…ô≈üdirilmi≈ü formada**:

$$
y = m x + q
$$

* $m = \tan \alpha$ ‚Äî meyl …ômsalƒ±,
* $q = y_0 - m x_0$ ‚Äî x…ôttin $y$ oxunu k…ôsdiyi n√∂qt…ô.

$$

## D√ºz X…ôttin Polyar koordinat sistemind…ô T…ônliyi



## üìå 1. T…ôrif

Poylar koordinat sistemind…ô d√ºz x…ôtt, **q√ºtb n√∂qt…ôsind…ôn (O)** x…ôtt…ô q…ôd…ôr olan **…ôn qƒ±sa m…ôsaf…ô** ($\rho$) v…ô h…ômin m…ôsaf…ônin **Ox oxu il…ô …ôm…ôl…ô g…ôtirdiyi bucaq** ($\varphi$) vasit…ôsil…ô ifad…ô olunur.

Y…ôni h…ôr d√ºz x…ôtt **$(\rho, \varphi)$** parametrl…ôri il…ô tam m√º…ôyy…ôn edilir.

---

## üìå 2. D√ºstur

D√ºz x…ôttin polyar koordinatlarda t…ônliyi bel…ôdir:

$$
x \cos \varphi + y \sin \varphi = \rho
$$

Burada:

* $\rho$ ‚Äî q√ºtbd…ôn (O) d√ºz x…ôtt…ô olan m…ôsaf…ô,
* $\varphi$ ‚Äî h…ômin m…ôsaf…ônin $Ox$ oxu il…ô bucaƒüƒ±,
* $(x, y)$ ‚Äî d√ºz x…ôtt √ºz…ôrind…ôki n√∂qt…ônin koordinatlarƒ±.

---

## üìå 3. Poylar koordinatlara ke√ßid

∆èg…ôr n√∂qt…ô polyar sistemd…ô $(r, \theta)$ il…ô verilibs…ô, onda:

$$
x = r \cos \theta, \quad y = r \sin \theta
$$

Bunlarƒ± x…ôtt t…ônliyin…ô qoysaq:

$$
r \cos(\theta - \varphi) = \rho
$$

Bu da **poylar formada d√ºz x…ôtt t…ônliyi**dir.

---

## üìå 4. N√ºmun…ô

M…ôs…ôl…ôn, q√ºtbd…ôn m…ôsaf…ôsi $\rho = 5$, istiqam…ôt bucaƒüƒ± $\varphi = 30^\circ$ olan d√ºz x…ôttin t…ônliyi:

$$
x \cos 30^\circ + y \sin 30^\circ = 5
$$

$$
\frac{\sqrt{3}}{2}x + \frac{1}{2}y = 5
$$

## D√ºz X…ôttin Normal T…ônliyi



## üìå 1. T…ôrif

**Normal t…ônlik** ‚Äì d√ºz x…ôttin q√ºtbd…ôn (ba≈ülanƒüƒ±c n√∂qt…ôd…ôn $O$) x…ôtt…ô q…ôd…ôr √ß…ôkilmi≈ü **perpendikulyar (normal) m…ôsaf…ô** $\rho$ v…ô h…ômin normalƒ±n **Ox oxu il…ô …ôm…ôl…ô g…ôtirdiyi bucaq** $\varphi$ vasit…ôsil…ô yazƒ±lan t…ônlikdir.

Ba≈üqa s√∂zl…ô, x…ôtt √ºz…ôrind…ôki b√ºt√ºn n√∂qt…ôl…ôr $(x,y)$, bu **normal m…ôsaf…ô** v…ô **bucaq** …ôsasƒ±nda ifad…ô olunur.

---

## üìå 2. D√ºstur

$$
x \cos \varphi + y \sin \varphi = \rho
$$

Burada:

* $\rho$ ‚Äî q√ºtbd…ôn d√ºz x…ôtt…ô q…ôd…ôr m…ôsaf…ô,
* $\varphi$ ‚Äî normalƒ±n $Ox$ oxu il…ô bucaƒüƒ±,
* $(x, y)$ ‚Äî d√ºz x…ôtt √ºz…ôrind…ôki n√∂qt…ô.

---

## üìå 3. √úmumi t…ônlikl…ô …ôlaq…ô

∆èg…ôr x…ôttin √ºmumi t…ônliyi bel…ôdirs…ô:

$$
Ax + By + C = 0
$$

onda onu normal t…ônliy…ô g…ôtirm…ôk √º√ß√ºn …ômsallarƒ± **normalla≈üdƒ±rmaq** lazƒ±mdƒ±r:

$$
\frac{A}{\sqrt{A^2+B^2}} \, x + \frac{B}{\sqrt{A^2+B^2}} \, y + \frac{C}{\sqrt{A^2+B^2}} = 0
$$

v…ô burada:

$$
\rho = -\frac{C}{\sqrt{A^2+B^2}}, \quad 
\cos \varphi = \frac{A}{\sqrt{A^2+B^2}}, \quad 
\sin \varphi = \frac{B}{\sqrt{A^2+B^2}}
$$

---

## üìå 4. N√ºmun…ô

D√ºz x…ôttin √ºmumi t…ônliyi:

$$
3x + 4y - 12 = 0
$$

* Normalƒ±n uzunluƒüu:

$$
\rho = \frac{| -12 |}{\sqrt{3^2+4^2}} = \frac{12}{5} = 2.4
$$

* Normal vektor: $(3,4)$, dem…ôli:

$$
\cos \varphi = \tfrac{3}{5}, \quad \sin \varphi = \tfrac{4}{5}
$$

Normal t…ônlik:

$$
x \cdot \frac{3}{5} + y \cdot \frac{4}{5} = 2.4
$$

## D√ºz X…ôttin Vektorial T…ônliyi



## üìå 1. T…ôrif

D√ºz x…ôttin **vektorial t…ônliyi** ‚Äî d√ºz x…ôtt √ºz…ôrind…ôki b√ºt√ºn n√∂qt…ôl…ôrin radius-vektorunu, ba≈ülanƒüƒ±c n√∂qt…ônin radius-vektoru il…ô istiqam…ôt vektorunun x…ôtti kombinasiyasƒ± ≈ü…ôklind…ô ifad…ô ed…ôn t…ônlikdir.

---

## üìå 2. D√ºstur (2D v…ô 3D √º√ß√ºn √ºmumi)

∆èg…ôr d√ºz x…ôtt

* $A(x_0, y_0, z_0)$ n√∂qt…ôsind…ôn ke√ßir (2D-d…ô $z_0$ yoxdur),
* istiqam…ôt vektoru $\vec{v} = (a, b, c)$-dirs…ô,

onda vektorial t…ônlik:

$$
\vec{r} = \vec{r_0} + t \vec{v}, \quad t \in \mathbb{R}
$$

Burada:

* $\vec{r} = (x, y, z)$ ‚Äî d√ºz x…ôtt √ºz…ôrind…ô ist…ônil…ôn n√∂qt…ônin radius-vektoru,
* $\vec{r_0} = (x_0, y_0, z_0)$ ‚Äî x…ôttin ke√ßdiyi n√∂qt…ônin radius-vektoru,
* $\vec{v} = (a, b, c)$ ‚Äî istiqam…ôt vektoru,
* $t$ ‚Äî parametr.

---

## üìå 3. 2D halda

$$
\vec{r} = (x, y), \quad \vec{r_0} = (x_0, y_0), \quad \vec{v} = (a, b)
$$

$$
(x, y) = (x_0, y_0) + t (a, b)
$$

Bu da komponentl…ôrd…ô:

$$
\begin{cases}
x = x_0 + at \\
y = y_0 + bt
\end{cases}
$$

Y…ôni vektorial t…ônlik ‚Üí parametrik t…ônliyin qƒ±sa yazƒ±lƒ±≈üƒ±dƒ±r.

---

## üìå 4. N√ºmun…ô

X…ôtt $A(1,2)$ n√∂qt…ôsind…ôn ke√ßir v…ô istiqam…ôt vektoru $\vec{v} = (3,4)$-d√ºr.

**Vektorial t…ônlik:**

$$
(x, y) = (1,2) + t(3,4), \quad t \in \mathbb{R}
$$

**Parametrik formasƒ±:**

$$
\begin{cases}
x = 1 + 3t \\
y = 2 + 4t
\end{cases}
$$
