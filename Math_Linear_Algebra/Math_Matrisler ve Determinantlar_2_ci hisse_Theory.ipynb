## Simmetrik matris (1)

>**T…ôrif:** Simmetrik matris el…ô bir kvadrat matrisdir ki, onun …ôsas diaqonalƒ±na g√∂r…ô simmetrik yerl…ô≈ü…ôn elementl…ôri bir-birin…ô b…ôrab…ôrdir.  

**D√ºstur:**  
$$
a_{ij} = a_{ji}, \quad \forall i, j
$$

---

>**T…ôrif:** Simmetrik matris, transpozisi √∂z√ºn…ô b…ôrab…ôr olan kvadrat matrisdir. Y…ôni, $A$ matrisinin simmetrik olmasƒ± √º√ß√ºn bu ≈ü…ôrt √∂d…ônilm…ôlidir:

$$
A^T = A
$$

Bu, o dem…ôkdir ki, matrisin …ôsas diaqonalƒ±na g√∂r…ô …ôks t…ôr…ôfd…ôki elementl…ôr bir-birin…ô b…ôrab…ôrdir, y…ôni:

$$
a_{ij} = a_{ji}, \quad \forall i, j
$$

### **N√ºmun…ô:**
$$
A =
\begin{bmatrix}
1 & 2 & 3 \\
2 & 4 & 5 \\
3 & 5 & 6
\end{bmatrix}
$$

Bu matrisin transpozunu alsaq:
$$
A^T =
\begin{bmatrix}
1 & 2 & 3 \\
2 & 4 & 5 \\
3 & 5 & 6
\end{bmatrix}
$$

G√∂r√ºnd√ºy√º kimi, $A^T = A$, y…ôni bu, simmetrik matrisdir.

### **X√ºsusiyy…ôtl…ôr:**
1. **Simmetrik matrisin diaqonalƒ±nda olan elementl…ôr ist…ônil…ôn …ôd…ôd ola bil…ôr.**
2. **Simmetrik matrisin k…ônar diaqonallardakƒ± elementl…ôri eyni olur** ($a_{ij} = a_{ji}$).
3. **B√ºt√ºn simmetrik matrisl…ôrin √∂z vektorlarƒ± h…ôqiqi …ôd…ôdl…ôrdir** (bu x√ºsusiyy…ôt kvadrat formalar v…ô spektral n…ôz…ôriyy…ôd…ô vacibdir).
4. **Simmetrik matris h…ômi≈ü…ô kvadrat olmalƒ±dƒ±r** (simmetrik olma ≈ü…ôrtini √∂d…ôm…ôk √º√ß√ºn).

Simmetrik matrisl…ôr x…ôtti c…ôbrd…ô, statistik analizd…ô (kovariasiya matrisl…ôri), fizika v…ô m√ºh…ôndislikd…ô √ßox geni≈ü istifad…ô olunur. ∆èg…ôr …ôlav…ô detallara ehtiyacƒ±n varsa, yaza bil…ôrs…ôn! üòä


## √á…ôp Simmetrik Matris (Skew-Symmetric Matrix) (1)

### **√á…ôp Simmetrik Matris (Skew-Symmetric Matrix)**  

>**T…ôrif:** √á…ôp simmetrik matris el…ô bir kvadrat matrisdir ki, onun transpozisi √∂z√ºn√ºn m…ônfisin…ô b…ôrab…ôrdir. Y…ôni, …ôsas diaqonalƒ±na g√∂r…ô simmetrik yerl…ô≈ü…ôn elementl…ôr i≈üar…ôc…ô t…ôrsdir v…ô …ôsas diaqonalƒ±ndakƒ± elementl…ôr 0-dƒ±r.  

**D√ºstur:**  
$$
A^T = -A
$$

Bu o dem…ôkdir ki, matrisin b√ºt√ºn elementl…ôri a≈üaƒüƒ±dakƒ± ≈ü…ôrti √∂d…ôm…ôlidir:  
$$
a_{ij} = -a_{ji}, \quad \forall i, j
$$

### **N√ºmun…ô:**  
$$

\begin{bmatrix}
0 & -2 & 3 \\
2 & 0 & -4 \\
-3 & 4 & 0
\end{bmatrix}
$$

Bu matrisin transpozunu alsaq:  
$$
A^T =
\begin{bmatrix}
0 & 2 & -3 \\

\end{bmatrix}
$$

G√∂r√ºnd√ºy√º kimi, $A^T = -A$, y…ôni bu, √ß…ôp simmetrik matrisdir.  

### **X√ºsusiyy…ôtl…ôr:**  
1. **B√ºt√ºn √ß…ôp simmetrik matrisl…ôrin …ôsas diaqonalƒ±ndakƒ± elementl…ôri sƒ±fƒ±rdƒ±r** ($a_{ii} = 0$).
2. **∆èsas diaqonala g√∂r…ô simmetrik elementl…ôr i≈üar…ôc…ô t…ôrsdir** ($a_{ij} = -a_{ji}$).
3. **T…ôk d…ôr…ôc…ôli √∂l√ß√ºs√º olan √ß…ôp simmetrik matris h…ômi≈ü…ô determinantƒ± 0 olan t…ôkil (singular) matrisdir**.

## Matrisin ƒ∞zi (Trace of a Matrix) (1)



### **Matrisin ƒ∞zi (Trace of a Matrix)**  

>**T…ôrif:** Matrisin izi, onun …ôsas diaqonalƒ±ndakƒ± elementl…ôrin c…ômi kimi t…ôyin olunur.  

**D√ºstur:**  
$$
\text{tr}(A) = \sum_{i=1}^{n} a_{ii}
$$  
Burada:  
- $A$ ‚Äî $n \times n$ √∂l√ß√ºl√º kvadrat matris,  
- $a_{ii}$ ‚Äî matrisin …ôsas diaqonalƒ±ndakƒ± elementl…ôrdir,  
- $\text{tr}(A)$ ‚Äî matrisin izidir.  

### **N√ºmun…ô:**  
$$
A =
\begin{bmatrix}
2 & 4 & 1 \\
0 & 3 & -1 \\
5 & 2 & 6
\end{bmatrix}
$$  

Matrisin izi:  
$$
\text{tr}(A) = 2 + 3 + 6 = 11
$$  

### **X√ºsusiyy…ôtl…ôr:**  
1. **ƒ∞z x…ôtti operatora nisb…ôt…ôn invariantdƒ±r:**  
   $$
   \text{tr}(A + B) = \text{tr}(A) + \text{tr}(B)
   $$  
2. **Skalyar vurulma:**  
   $$
   \text{tr}(cA) = c \text{tr}(A)
   $$  
3. **ƒ∞z matrisin transpozisi √º√ß√ºn d…ôyi≈ümir:**  
   $$
   \text{tr}(A^T) = \text{tr}(A)
   $$  
4. **ƒ∞z matrisin determinantƒ± il…ô qarƒ±≈üdƒ±rƒ±lmamalƒ±dƒ±r, √ß√ºnki o, diaqonal elementl…ôrin c…ômidir, determinant is…ô x√ºsusi hesablanƒ±r.**  
5. **Matrisin izi onun √∂z qiym…ôtl…ôrinin c…ômin…ô b…ôrab…ôrdir:**  
   $$
   \text{tr}(A) = \lambda_1 + \lambda_2 + \dots + \lambda_n
   $$  
   Burada $\lambda_i$-l…ôr matrisin √∂z qiym…ôtl…ôridir.  

Matrisin izi kvadratik formalar, x…ôtti c…ôbr v…ô statistik analizd…ô geni≈ü istifad…ô olunur.

## Determinantƒ±n S…ôtir v…ô S√ºtun Elementl…ôrin…ô G√∂r…ô Ayrƒ±lmasƒ± (Laplas Teoremi) (1)

### **Determinantƒ±n S…ôtir v…ô S√ºtun Elementl…ôrin…ô G√∂r…ô Ayrƒ±lmasƒ± (Laplas Teoremi)** 

>**T…ôrif:** Laplas teoremi, **Determinant** - Matrisin h…ôr hansƒ± bir **s…ôtirinin v…ô ya s√ºtununun**  b√ºt√ºn **elementl…ôrinin** √∂z **C…ôbri tamalayƒ±cƒ±larƒ±** hasill…ôri c…ômin…ô b…ôrab…ôrdir.

---

Determinantƒ±n bir s…ôtir…ô v…ô ya bir s√ºtuna g√∂r…ô ayrƒ±lmasƒ± **Laplas geni≈ül…ônm…ôsi (Laplace expansion)** adlanƒ±r. Bu √ºsul determinantƒ± ki√ßik √∂l√ß√ºl√º determinantlara par√ßalamaƒüa k√∂m…ôk edir.  

#### **D√ºstur:**  
Bir $n \times n$ √∂l√ß√ºl√º $A$ matrisinin determinantƒ±nƒ± $i$-ci s…ôtir v…ô ya $j$-ci s√ºtuna g√∂r…ô hesablamaq √º√ß√ºn:  

$$
\det(A) = \sum_{j=1}^{n} (-1)^{i+j} a_{ij} M_{ij} \quad \text{(s…ôtir √ºzr…ô)}
$$  

$$
\det(A) = \sum_{i=1}^{n} (-1)^{i+j} a_{ij} M_{ij} \quad \text{(s√ºtun √ºzr…ô)}
$$  

Burada:  
- $a_{ij}$ ‚Äî se√ßilmi≈ü s…ôtir v…ô ya s√ºtundakƒ± element,  
- $M_{ij}$ ‚Äî h…ômin element…ô aid **min√∂r**, y…ôni $a_{ij}$-ni saxlayan s…ôtir v…ô s√ºtun √ßƒ±xarƒ±ldƒ±qdan sonra qalan matrisin determinantƒ±,  
- $(-1)^{i+j}$ ‚Äî **k√∂m…ôk√ßi i≈üar…ô** (alternativ i≈üar…ôl…ôrd…ôn ibar…ôt matrisd…ôn g√∂t√ºr√ºl√ºr).  

---

### **N√ºmun…ô: $3 \times 3$ matris √º√ß√ºn determinant hesablanmasƒ±**  

Verilmi≈ü olsun:  

$$
A =
\begin{bmatrix}
1 & 2 & 3 \\
4 & 5 & 6 \\
7 & 8 & 9
\end{bmatrix}
$$  

Determinantƒ± birinci s…ôtir…ô ($i = 1$) g√∂r…ô a√ßsaq:  

$$
\det(A) = 1 \cdot 
\begin{vmatrix}
5 & 6 \\
8 & 9
\end{vmatrix}
- 2 \cdot
\begin{vmatrix}
4 & 6 \\
7 & 9
\end{vmatrix}
+ 3 \cdot
\begin{vmatrix}
4 & 5 \\
7 & 8
\end{vmatrix}
$$  

ƒ∞ndi **ki√ßik determinantlarƒ±** hesablayaq:  

$$
\begin{vmatrix}
5 & 6 \\
8 & 9
\end{vmatrix} = (5 \times 9 - 6 \times 8) = 45 - 48 = -3
$$  

$$
\begin{vmatrix}
4 & 6 \\
7 & 9
\end{vmatrix} = (4 \times 9 - 6 \times 7) = 36 - 42 = -6
$$  

$$
\begin{vmatrix}
4 & 5 \\
7 & 8
\end{vmatrix} = (4 \times 8 - 5 \times 7) = 32 - 35 = -3
$$  

ƒ∞ndi determinantƒ± hesablayaq:  

$$
\det(A) = 1(-3) - 2(-6) + 3(-3)
$$  

$$
\det(A) = -3 + 12 - 9 = 0
$$  

N…ôtic…ô:  
$$
\det(A) = 0
$$  

Bu √ºsul ist…ônil…ôn √∂l√ß√ºl√º kvadrat matris √º√ß√ºn t…ôtbiq edil…ô bil…ôr, sad…ôc…ô matrisin √∂l√ß√ºs√º artdƒ±qca hesablama daha √ßox vaxt aparƒ±r. **Daha b√∂y√ºk matrisl…ôr √º√ß√ºn rekursiv determinant hesablama v…ô ya sƒ±ra reduksiya metodlarƒ± (Gauss eliminasiya) daha effektivdir.**  

∆èlav…ô izah ist…ôs…ôn, yaza bil…ôrs…ôn! üòä

## Laplas Geni≈ül…ônm…ôsi il…ô Determinantƒ±n Hesablanmasƒ± (1)

>**T…ôrif:** Laplas teoremi, **Determinant** - Matrisin h…ôr hansƒ± bir **s…ôtirinin v…ô ya s√ºtununun**  b√ºt√ºn **elementl…ôrinin** √∂z **C…ôbri tamalayƒ±cƒ±larƒ±** il…ô hasill…ôri c…ômin…ô b…ôrab…ôrdir.

---

### **Laplas Geni≈ül…ônm…ôsi il…ô Determinantƒ±n Hesablanmasƒ±**

Laplas geni≈ül…ônm…ôsi determinantƒ±n hesablama √ºsullarƒ±ndan biridir v…ô ist…ônil…ôn √∂l√ß√ºl√º kvadrat matris √º√ß√ºn t…ôtbiq edil…ô bil…ôr. Bu metod, matrisin ist…ônil…ôn bir s…ôtiri v…ô ya s√ºtunu √ºzr…ô geni≈ül…ônm…ô aparƒ±laraq determinantƒ± hesablayƒ±r. 

---

### **Laplas Geni≈ül…ônm…ôsi D√ºsturu:**

Verilmi≈ü $A$-nƒ±n $n \times n$ √∂l√ß√ºl√º kvadrat matrisi √º√ß√ºn determinant bel…ô hesablanƒ±r:

$$
\text{det}(A) = \sum_{j=1}^{n} a_{ij} \cdot C_{ij}
$$

Burada:  
- $a_{ij}$: $A$-nƒ±n $i$-ci s…ôtir v…ô $j$-ci s√ºtunundakƒ± elementidir.  
- $C_{ij}$: $a_{ij}$-nin kofaktorudur, y…ôni:
  $$
  C_{ij} = (-1)^{i+j} \cdot \text{det}(A_{ij})
  $$
- $A_{ij}$: $A$-dan $i$-ci s…ôtir v…ô $j$-ci s√ºtun √ßƒ±xarƒ±laraq alƒ±nan alt matrisdir.

Hesablamanƒ± ist…ônil…ôn bir s…ôtir v…ô ya s√ºtun √ºzr…ô apara bil…ôrsiniz.

---

### **Misal: √ú√ß T…ôrtibli Matris**

Verilmi≈ü olsun:
$$
A = \begin{bmatrix} 
1 & 2 & 3 \\ 
4 & 5 & 6 \\ 
7 & 8 & 9 
\end{bmatrix}
$$

**1. Birinci s…ôtir √ºzr…ô geni≈ül…ônm…ô aparaq ($i = 1$):**

$$
\text{det}(A) = \sum_{j=1}^{3} a_{1j} \cdot C_{1j}
$$

Bu ifad…ôni a√ßƒ±rƒ±q:
$$
\text{det}(A) = a_{11} \cdot C_{11} + a_{12} \cdot C_{12} + a_{13} \cdot C_{13}
$$

**2. H…ôr bir kofaktoru tapƒ±rƒ±q:**

- $C_{11} = (-1)^{1+1} \cdot \text{det} \begin{bmatrix} 5 & 6 \\ 8 & 9 \end{bmatrix}$  
  $$
  C_{11} = \text{det} \begin{bmatrix} 5 & 6 \\ 8 & 9 \end{bmatrix} = (5 \cdot 9) - (6 \cdot 8) = 45 - 48 = -3
  $$

- $C_{12} = (-1)^{1+2} \cdot \text{det} \begin{bmatrix} 4 & 6 \\ 7 & 9 \end{bmatrix}$  
  $$
  C_{12} = -1 \cdot \text{det} \begin{bmatrix} 4 & 6 \\ 7 & 9 \end{bmatrix} = - \big((4 \cdot 9) - (6 \cdot 7)\big) = - (36 - 42) = 6
  $$

- $C_{13} = (-1)^{1+3} \cdot \text{det} \begin{bmatrix} 4 & 5 \\ 7 & 8 \end{bmatrix}$  
  $$
  C_{13} = \text{det} \begin{bmatrix} 4 & 5 \\ 7 & 8 \end{bmatrix} = (4 \cdot 8) - (5 \cdot 7) = 32 - 35 = -3
  $$

**3. Geni≈ül…ônm…ôni t…ôtbiq edirik:**

$$
\text{det}(A) = (1 \cdot -3) + (2 \cdot 6) + (3 \cdot -3)
$$

$$
\text{det}(A) = -3 + 12 - 9 = 0
$$

---

### **N…ôtic…ô:**
Bu matrisin determinanti $\text{det}(A) = 0$-dƒ±r.  
Bu halda matris t…ôrs edil…ô bilm…ôz, √ß√ºnki determinant sƒ±fƒ±rdƒ±r.

## Teorem (1)

> **T…ôrif:** Determinantƒ±n h…ôr hansƒ± bir s…ôtirinin v…ô ya s√ºtununun b√ºt√ºn element…ôrinin ba≈üqa s…ôtir v…ô ya s√ºtunun uyƒüun elemntl…ôrinin C…ôbri tamamlayƒ±cƒ±larƒ± il…ô hasill…ôri c…ômi sƒ±fƒ±ra b…ôrab…ôrdir

---

## **Teoremin izahƒ±**  
Bir **$n \times n$** t…ôrtibli **$A$** kvadrat matrisini g√∂t√ºr…ôk. **$i$-ci s…ôtirin v…ô ya $j$-ci s√ºtunun** elementl…ôrini ba≈üqa s…ôtir v…ô ya s√ºtunun uyƒüun elemntl…ôrinin **c…ôbri tamamlayƒ±cƒ±larƒ± (cofactorlarƒ±) il…ô hasill…ôrinin c…ômi h…ômi≈ü…ô sƒ±fƒ±rdƒ±r**, …ôg…ôr h…ômin s…ôtir v…ô ya s√ºtun **determinant hesablanmayan bir s…ôtir v…ô ya s√ºtun se√ßil…ôrs…ô**.  

Ba≈üqa s√∂zl…ô, …ôg…ôr $A = (a_{ij})$ matrisi varsa, a≈üaƒüƒ±dakƒ± b…ôrab…ôrlik doƒürudur:  

$$
\sum_{k=1}^{n} a_{ik} C_{jk} = 0, \quad i \neq j
$$

v…ô ya  

$$
\sum_{k=1}^{n} a_{ki} C_{kj} = 0, \quad i \neq j
$$

Burada:  
- $C_{jk}$ ‚Äî **$a_{jk}$** elementinin **c…ôbri tamamlayƒ±cƒ±sƒ±dƒ±r**.  
- $a_{ik}$ v…ô $a_{ki}$ ‚Äî m√ºvafiq s…ôtir v…ô s√ºtun elementl…ôridir.  

---

## **S√ºbutun m…ôntiqi**  
1. **Determinantƒ±n Laplas geni≈ül…ônm…ôsi** g√∂st…ôrir ki, bir s…ôtir v…ô ya s√ºtunun elementl…ôri il…ô onlarƒ±n uyƒüun c…ôbri tamamlayƒ±cƒ±larƒ±nƒ±n hasilinin **c…ômi determinantƒ±** verir.  

   $$
   \sum_{k=1}^{n} a_{ik} C_{ik} = \det(A)
   $$

2. **∆èg…ôr ba≈üqa bir s…ôtir v…ô ya s√ºtun se√ßs…ôk** (y…ôni determinant hesabladƒ±ƒüƒ±mƒ±z s…ôtir v…ô ya s√ºtun yox, ba≈üqa birini g√∂t√ºrs…ôk), onda bu ifad…ônin n…ôtic…ôsi **h…ômi≈ü…ô 0 olur**.  

   $$
   \sum_{k=1}^{n} a_{jk} C_{ik} = 0, \quad i \neq j
   $$

3. **Bunun s…ôb…ôbi ondan ibar…ôtdir ki, determinant x…ôtti ifad…ôl…ôrd…ôn ibar…ôt bir √ßoxh…ôdli funksiyadƒ±r v…ô bu funksiyanƒ±n t…ôrkib hiss…ôl…ôri dig…ôr s…ôtir v…ô s√ºtunlar √º√ß√ºn linearly asƒ±lƒ± olur.**  

---

## Eyni T…ôrtibli Kvadrat Matrisl…ôrin Hasilinin Determinantƒ± (1)

**Eyni t…ôrtibli kvadrat matrisl…ôrin hasilinin determinantƒ±** haqqƒ±nda x√ºsusi bir x√ºsusiyy…ôt m√∂vcuddur. Bu x√ºsusiyy…ôt bel…ôdir:

### **Eyni T…ôrtibli Kvadrat Matrisl…ôrin Hasilinin Determinantƒ±:**

∆èg…ôr $A$ v…ô $B$ **eyni t…ôrtibli kvadrat matrisl…ôr**dirs…ô, o zaman bu matrisl…ôrin hasilinin determinantƒ± a≈üaƒüƒ±dakƒ± kimi hesablana bil…ôr:

$$
\det(AB) = \det(A) \cdot \det(B)
$$

Bu, **eyni t…ôrtibli kvadrat matrisl…ôr** √º√ß√ºn √ºmumi bir qaydadƒ±r v…ô **eyni t…ôrtib** dedikd…ô, h…ôm $A$, h…ôm d…ô $B$ matrisl…ôrinin s…ôtir v…ô s√ºtun saylarƒ±nƒ±n b…ôrab…ôr olmasƒ± n…ôz…ôrd…ô tutulur.

### **∆èg…ôr Matrisl…ôr Eynidirs…ô (Y…ôni, $A = B$):**

∆èg…ôr $A$ v…ô $B$ matrisl…ôri eynidirs…ô, y…ôni $A = B$, o zaman bu halda:

$$
\det(A^2) = (\det(A))^2
$$

Bel…ôlikl…ô, iki eyni t…ôrtibli kvadrat matrisin hasilinin determinantƒ±, h…ômin matrisin determinantƒ±nƒ±n kvadratƒ±na b…ôrab…ôr olur.

### **Misal:**

Tutaq ki, $A$ v…ô $B$ matrisl…ôri bel…ôdir:

$$
A =
\begin{bmatrix}
1 & 2 \\
3 & 4
\end{bmatrix}
\quad \text{v…ô} \quad
B =
\begin{bmatrix}
5 & 6 \\
7 & 8
\end{bmatrix}
$$

Bu matrisl…ôrin hasilinin determinantƒ±nƒ± tapmaƒüa √ßalƒ±≈üaq:

1. **Hasilini tapƒ±rƒ±q:**

$$
AB = 
\begin{bmatrix}
1 & 2 \\
3 & 4
\end{bmatrix}
\begin{bmatrix}
5 & 6 \\
7 & 8
\end{bmatrix}
=
\begin{bmatrix}
1 \cdot 5 + 2 \cdot 7 & 1 \cdot 6 + 2 \cdot 8 \\
3 \cdot 5 + 4 \cdot 7 & 3 \cdot 6 + 4 \cdot 8
\end{bmatrix}
=
\begin{bmatrix}
19 & 22 \\
43 & 50
\end{bmatrix}
$$

2. **Hasilin determinantƒ±nƒ± tapƒ±rƒ±q:**

$$
\det(AB) = 19 \cdot 50 - 22 \cdot 43 = 950 - 946 = 4
$$

3. **Ayrƒ±-ayrƒ±lƒ±qda $ \det(A) $ v…ô $ \det(B) $-ni tapƒ±rƒ±q:**

$$
\det(A) = 1 \cdot 4 - 2 \cdot 3 = 4 - 6 = -2
$$

$$
\det(B) = 5 \cdot 8 - 6 \cdot 7 = 40 - 42 = -2
$$

4. **Sonra $ \det(A) \cdot \det(B) $-ni tapƒ±rƒ±q:**

$$
\det(A) \cdot \det(B) = (-2) \cdot (-2) = 4
$$

### **N…ôtic…ô:**

**$\det(AB) = \det(A) \cdot \det(B)$** olduƒüu g√∂r√ºn√ºr, y…ôni iki eyni t…ôrtibli kvadrat matrisin hasilinin determinantƒ±, h…ômin matrisl…ôrin determinantlarƒ±nƒ±n hasilin…ô b…ôrab…ôrdir.

## Vandermonde Determinantƒ±

>**T…ôrif:** **Vandermonde determinantƒ±**, verilmi≈ü $ x_1, x_2, \dots, x_n $ …ôd…ôdl…ôri il…ô qurulmu≈ü bir Vandermonde matrisinin determinantƒ±dƒ±r v…ô bu **determinant**, **h…ômin …ôd…ôdl…ôrin h…ôr birinin f…ôrql…ôrinin hasilin…ô b…ôrab…ôrdir**. Bu determinant, √ßoxluqlar v…ô interpolasiya m…ôs…ôl…ôl…ôrind…ô geni≈ü t…ôtbiq olunur.

**D√ºstur**:
$$
\det(V) = \prod_{1 \leq i < j \leq n} (x_j - x_i)
$$

---

>**T…ôrif:** **Vandermonde determinantƒ±** x√ºsusi bir determinantdƒ±r v…ô bir √ßox t…ôtbiq sah…ôl…ôrind…ô istifad…ô olunur, x√ºsus…ôn √ßoxluqlar v…ô interpolasiya m…ôs…ôl…ôl…ôrind…ô. **Vandermonde matrisinin determinantƒ±, sƒ±ralƒ± …ôd…ôdl…ôrin g√ºcl…ôri il…ô t…ô≈ükil olunan matrisin determinantƒ±nƒ± t…ôsvir edir**.

### **Vandermonde Matrisinin T…ôrifi:**

Bir **Vandermonde matrisinin** √ºmumi formasƒ± bel…ôdir:

$$
V =
\begin{bmatrix}
1 & x_1 & x_1^2 & \dots & x_1^{n-1} \\
1 & x_2 & x_2^2 & \dots & x_2^{n-1} \\
\vdots & \vdots & \vdots & \ddots & \vdots \\
1 & x_n & x_n^2 & \dots & x_n^{n-1}
\end{bmatrix}
$$

Bu matrisin **determinantƒ±** is…ô **g√ºn√ºm√ºzd…ô …ôn √ßox tanƒ±nan formu** il…ô bel…ô t…ôyin olunur:

### **Vandermonde Determinantƒ±:**

Bir **n x n** √∂l√ß√ºl√º Vandermonde matrisinin determinantƒ± a≈üaƒüƒ±dakƒ± ≈ü…ôkild…ô veril…ô bil…ôr:

$$
\det(V) = \prod_{1 \leq i < j \leq n} (x_j - x_i)
$$

Y…ôni, determinant **x‚ÇÅ, x‚ÇÇ, ..., xn** …ôd…ôdl…ôrinin f…ôrql…ôrinin hasilin…ô b…ôrab…ôrdir. 

### **∆èh…ômiyy…ôtli x√ºsusiyy…ôtl…ôr:**

1. **∆èg…ôr** Vandermonde matrisind…ôki **b√ºt√ºn x‚Çñ-l…ôr eyni** olarsa, determinant sƒ±fƒ±r olacaqdƒ±r. √á√ºnki bu halda h…ôr hansƒ± bir s√ºtun dig…ôr bir s√ºtunun bir kopyasƒ±na √ßevril…ôc…ôkdir.
   
2. **Vandermonde determinantƒ±** sƒ±ralƒ± …ôd…ôdl…ôrl…ô qurulan matrisl…ôrin determinantƒ±nƒ± hesablamaƒüa imkan verir, bu da √ßoxsaylƒ± riyazi v…ô m√ºh…ôndislik m…ôs…ôl…ôl…ôrind…ô geni≈ü istifad…ô olunur.

### **N√ºmun…ô:**

Tutaq ki, **n = 3** √º√ß√ºn Vandermonde matrisini v…ô onun determinantƒ±nƒ± hesablayaq:

$$
V =
\begin{bmatrix}
1 & x_1 & x_1^2 \\
1 & x_2 & x_2^2 \\
1 & x_3 & x_3^2
\end{bmatrix}
$$

Determinantƒ±nƒ± hesablayaq:

$$
\det(V) = (x_2 - x_1)(x_3 - x_1)(x_3 - x_2)
$$

Misal √º√ß√ºn, …ôg…ôr **x‚ÇÅ = 1, x‚ÇÇ = 2, x‚ÇÉ = 3** olduqda:

$$
\det(V) = (2 - 1)(3 - 1)(3 - 2) = 1 \cdot 2 \cdot 1 = 2
$$

Bu halda, **determinant 2** olar.

### **√úmumil…ô≈üdirm…ôk:**

**Vandermonde determinantƒ±** sƒ±ralƒ± …ôd…ôdl…ôrd…ôn qurulmu≈ü matrisl…ôrin determinantƒ±nƒ± tapmaq √º√ß√ºn √ßox faydalƒ±dƒ±r v…ô a≈üaƒüƒ±dakƒ± ≈ü…ôkild…ô √ºmumil…ô≈üdiril…ô bil…ôr:

$$
\det(V) = \prod_{1 \leq i < j \leq n} (x_j - x_i)
$$
