## Vektoralın Vektorial hasili (1)

>**Tərif:**  Vektoralın vektorial hasili, iki vektorun hər biri ilə perpendikulyar olan və uzunluğu bu vektorlar arasındakı paralelogramın sahəsinə bərabər olan və ya bu uzunluq $ |\mathbf{A}| |\mathbf{B}| \sin(\alpha) $ -ya bərabər olan bir vektordur və düsturla belə ifadə edilir:

$$
\mathbf{C} = \mathbf{A} \times \mathbf{B} = \mathbf{S}_⏥ = |\mathbf{A}| |\mathbf{B}| \sin(\alpha) \, \hat{\mathbf{n}}
$$

Burada $ \mathbf{A} $ və $ \mathbf{B} $ vektorları, $ \alpha $ - onların arasındakı bucaq, $\hat{\mathbf{n}} $ isə bu vektorlar arasındakı müstəviyə perpendikulyar olan vahid vektordur.

---

### 1) **C vektoru və paralelogramın sahəsi:**
Vektoralın vektorial hasili ilə əlaqədar olaraq, iki vektorun çapraz hasilindən yaranan vektor $ \mathbf{C} $ belə təsvir olunur:
>**Tərif:** C vektoru yaranır ki, onun uzunluğu, A və B vektorlarından yaranan paralelogramın sahəsinə bərabərdir və bu uzunluq $ |\mathbf{A}| |\mathbf{B}| \sin(\alpha) $ ilə verilmişdir.
  Burada $ \alpha $ - A və B vektorları arasındakı bucaqdır.

- **Düstur:** 
$$
\mathbf{A} \times \mathbf{B} = |\mathbf{A}| |\mathbf{B}| \sin(\alpha) \, \hat{\mathbf{n}}
$$
- Burada $ \hat{\mathbf{n}} $ - A və B vektorlarına perpendikulyar olan bir vahid vektordur.




 **Misal:** 
Tutaq ki, $ \mathbf{A} = (2, 3, 4) $ və $ \mathbf{B} = (1, 0, 0) $.
  
İki vektor arasındakı bucaq $ \alpha = 90^\circ $ olduğundan $ \sin(\alpha) = 1 $ olacaq.

Vektoral hasilin uzunluğu:
$$
|\mathbf{A} \times \mathbf{B}| = |\mathbf{A}| |\mathbf{B}| \sin(90^\circ) = \sqrt{2^2 + 3^2 + 4^2} \cdot \sqrt{1^2 + 0^2 + 0^2} \cdot 1 = 5 \times 1 = 5
$$

Bu, paralelogramın sahəsini göstərir.

---

### 2) **C vektorunun perpendikulyarlığı:**
Vektoral hasilin nəticəsi $ \mathbf{A} \times \mathbf{B} $ vektoru, A və B vektorlarına perpendikulyardır. Yəni, bu vektor həm $ \mathbf{A} $-ya, həm də $ \mathbf{B} $-yə dikdir. Bu, vektoral hasilin ən əsas xüsusiyyətlərindən biridir.

- **Perpendikulyarlıq məsələsi:** Vektoral hasilin nəticəsi olan vektor həmçinin müstəviyə perpendikulyar olacaq ki, bu müstəvi $ \mathbf{A} $ və $ \mathbf{B} $ vektorları tərəfindən yaradılır.

- **Misal:** Əgər $ \mathbf{A} = (1, 0, 0) $ və $ \mathbf{B} = (0, 1, 0) $, onda vektoral hasilin nəticəsi:
$$
\mathbf{A} \times \mathbf{B} = (0, 0, 1)
$$
Nəticə $ (0, 0, 1) $-dir, yəni $ \mathbf{A} $ və $ \mathbf{B} $-yə perpendikulyar bir vektor yaranır.

---

### 3) **Sağ və sol orientasiya qaydası (Right-Hand Rule və Left-Hand Rule):**
Bu qaydalar, vektoral hasilin istiqamətini müəyyən etməyə kömək edir. Bu qaydalar, vektorların nəticəsinin yönünü müəyyən edir:

- **Sağ orientasiya (Right-Hand Rule):** 
  Sağ əlinizin barmaqlarını A vektorunun üzərinə bükün və B vektorunu göstərmək üçün əlinizi döndərin. Əlinizlə döndüyünüz zaman, baş barmağınız vektoral hasilin istiqamətini göstərəcək.
  
- **Sol orientasiya (Left-Hand Rule):** 
  Bu, sadəcə əlinizin sol əllə eyni şəkildə hərəkət etməsidir, burada baş barmaq da nəticə vektorunun istiqamətini göstərir, amma sağ əldən əksinə.

- **Misal:** Əgər $ \mathbf{A} = (1, 0, 0) $ və $ \mathbf{B} = (0, 1, 0) $, sağ əlinizə əsaslanaraq, nəticə vektoru $ \mathbf{A} \times \mathbf{B} = (0, 0, 1) $ olacaq və bu, müsbət Z oxunda olacaq.

---

### 4) **Hasilin yazılış qaydası və işarələmə:**
Vektoral hasilin yazılışında standart olaraq "×" simvolu istifadə olunur:
$$
\mathbf{A} \times \mathbf{B}
$$
Bu, iki vektorun çapraz hasilini göstərir. Vektoral hasilin nəticəsi isə yeni bir vektor olur və nəticənin istiqaməti, əvvəlki qeyd etdiyimiz qaydalara əsasən müəyyən edilir.

- **Misal:** Vektorların çapraz hasili üçün yazılış belə olur:
  $$
  \mathbf{A} \times \mathbf{B} = (A_y B_z - A_z B_y, A_z B_x - A_x B_z, A_x B_y - A_y B_x)
  $$

---

###  5) **Düstur (Kartezian Koordinatlarında):**
Kartezian koordinatlarında, A və B vektorları belə verilsin:

$$
\mathbf{A} = (A_x, A_y, A_z)
$$

$$
\mathbf{B} = (B_x, B_y, B_z)
$$

Vektorialın hasili bu şəkildə hesablanır:
$$
\mathbf{A} \times \mathbf{B} = \left( (A_y \cdot B_z - A_z \cdot B_y), (A_z \cdot B_x - A_x \cdot B_z), (A_x \cdot B_y - A_y \cdot B_x) \right)
$$

Bu düsturda, hər bir komponent iki vektorun uyğun komponentlərinin fərqinin hasilidir.

### **Nümunə:**

Tutaq ki, A və B vektorları belədir:
$$
\mathbf{A} = (1, 2, 3), \quad \mathbf{B} = (4, 5, 6)
$$
Vektorialın hasili olacaq:
$$
\mathbf{A} \times \mathbf{B} = \left( (2 \cdot 6 - 3 \cdot 5), (3 \cdot 4 - 1 \cdot 6), (1 \cdot 5 - 2 \cdot 4) \right)
$$
$$
\mathbf{A} \times \mathbf{B} = (12 - 15, 12 - 6, 5 - 8) = (-3, 6, -3)
$$

Bu vektor, A və B vektorlarının təyin etdiyi müstəviyə perpendikulyar olacaq.

## Skalyar hasil və vektorial hasil fərqli əməliyyatlardır və ona görə də onların düsturları da fərqlidir

     \mathbf{A} \times \mathbf{B} = \left( A_y B_z - A_z B_y, A_z B_x - A_x B_z, A_x B_y - A_y B_x \right)  
     $$  
   - **Məsələn:**  
     $$  
     (1,2,3) \times (4,5,6) = (-3, 6, -3)  
     $$  

**Niyə vektorial hasildə komponentlər cəmlənmir?**  
Çünki vektorial hasil iki vektorun koordinat müstəvisində fırlanmasını və yeni bir istiqamətə yönəlmiş vektor yaratmağı ifadə edir. Burada determinantlar vasitəsilə hesablanan komponent fərqləri lazımdır. Skalyar hasil isə yalnız vektorların eyni istiqamətdə nə qədər üst-üstə düşdüyünü ölçür.

## Vektorialın Vektorial hasilinin Əsas xassələri  (1)

### **1. Antisimetriya (Antikommutativlik) xassəsi:**  
İki vektorun vektorial hasili istiqamətcə əks vektor verir:  
$$
\mathbf{A} \times \mathbf{B} = - (\mathbf{B} \times \mathbf{A})
$$  
🔹 **Misal:**  
$$
(1,2,3) \times (4,5,6) = (-3, 6, -3)
$$  
$$
(4,5,6) \times (1,2,3) = (3, -6, 3)
$$  

### **2. Distributivlik (Cəmi paylama) xassəsi:**  
$$
\mathbf{A} \times (\mathbf{B} + \mathbf{C}) = \mathbf{A} \times \mathbf{B} + \mathbf{A} \times \mathbf{C}
$$  
🔹 **Misal:**  
Əgər  
$$
\mathbf{A} = (1,0,0), \quad \mathbf{B} = (0,1,0), \quad \mathbf{C} = (0,0,1)
$$  
olsa,  
$$
\mathbf{A} \times (\mathbf{B} + \mathbf{C}) = (0,1,0) + (0,0,1) = (0,1,1)
$$  
və  
$$
\mathbf{A} \times \mathbf{B} + \mathbf{A} \times \mathbf{C} = (0,0,1) + (0,-1,0) = (0,1,1)
$$  

### **3. Skalyar vuruqla vektorial hasil:**  
Bir skalyar ədəd ($\lambda$) vektorial hasilə təsir etdikdə:  
$$
(\lambda \mathbf{A}) \times \mathbf{B} = \lambda (\mathbf{A} \times \mathbf{B})
$$  
$$
\mathbf{A} \times (\lambda \mathbf{B}) = \lambda (\mathbf{A} \times \mathbf{B})
$$  
🔹 **Misal:**  
Əgər  
$$
\lambda = 3, \quad \mathbf{A} = (1,2,3), \quad \mathbf{B} = (4,5,6)
$$  
$$
(3\mathbf{A}) \times \mathbf{B} = 3 (\mathbf{A} \times \mathbf{B})
$$  
və nəticə eyni qalır.  

### **4. Öz-özünə hasil 0-dır (Null vektor xassəsi):**  
İstənilən vektorun özü ilə vektorial hasili sıfırdır:  
$$
\mathbf{A} \times \mathbf{A} = \mathbf{0}
$$  
🔹 **Misal:**  
$$
(1,2,3) \times (1,2,3) = (0,0,0)
$$

#### Qeyd:

**4-cü xassə (öz-özünə hasil 0-dır)**

Əgər bir vektor özü ilə vektorial hasilə girirsə:  
$$
\mathbf{A} \times \mathbf{A} = |\mathbf{A}||\mathbf{A}|\sin 0^\circ \ \mathbf{n}
$$  
Burada **$\alpha = 0^\circ$** olduğuna görə:  
$$
\sin 0^\circ = 0
$$  
və nəticədə:  
$$
\mathbf{A} \times \mathbf{A} = \mathbf{0}
$$  

Əgər iki vektor bir istiqamətə paraleldirsə, **$\alpha = 0^\circ$ və ya $\alpha = 180^\circ$** olur. Hər iki halda **$\sin \alpha = 0$** olduğuna görə, vektorial hasil sıfır olur.  

Yəni **öz-özünə hasil almaqda** **$\sin 0^\circ = 0$** olduğu üçün sıfır alınır, **$\sin 180^\circ$** isə paralel vektorlar üçün sıfır olur.  


### **5. Paralel vektorların vektorial hasili sıfırdır:**  
Əgər iki vektor paraleldirsə, onların vektorial hasili sıfır olur:  
$$
\mathbf{A} \times \mathbf{B} = \mathbf{0}, \quad \text{əgər} \quad \mathbf{A} \parallel \mathbf{B}
$$  
🔹 **Misal:**  
$$
(2,4,6) \times (1,2,3) = (0,0,0)
$$  
çünki ikinci vektor birincinin 2 qatıdır.  

#### Qeyd:
**5-ci xassədə ($\mathbf{A} \times \mathbf{B} = \mathbf{0}, \quad \text{əgər} \quad \mathbf{A} \parallel \mathbf{B}$)**, bu o deməkdir ki, **$\mathbf{A}$ və $\mathbf{B}$ paralel olduqda** $\alpha = 0^\circ$ və ya $\alpha = 180^\circ$ olur.  

Vektorial hasilin tərifi:  
$$
|\mathbf{A} \times \mathbf{B}| = |\mathbf{A}||\mathbf{B}|\sin \alpha
$$  

- Əgər **$\alpha = 0^\circ$ (vektorlar eyni istiqamətdədir)** → $\sin 0^\circ = 0$, nəticədə $\mathbf{A} \times \mathbf{B} = \mathbf{0}$.  
- Əgər **$\alpha = 180^\circ$ (vektorlar əks istiqamətdədir)** → $\sin 180^\circ = 0$, nəticədə yenə $\mathbf{A} \times \mathbf{B} = \mathbf{0}$.  

**Nümunə:**  
Vektorları götürək:  
$$
\mathbf{A} = (2,4,6), \quad \mathbf{B} = (1,2,3)
$$  
Burada **$\mathbf{B}$, $\mathbf{A}$-nın 0.5 qatıdır** ($\mathbf{B} = \frac{1}{2} \mathbf{A}$), yəni **paraleldirlər**.  

İndi vektorial hasili hesablayırıq:  
$$
\mathbf{A} \times \mathbf{B} =
\begin{vmatrix} 
\mathbf{i} & \mathbf{j} & \mathbf{k} \\ 
2 & 4 & 6 \\ 
1 & 2 & 3 
\end{vmatrix}
$$  
Determinantı hesablayırıq:  
$$
\mathbf{i} (4 \cdot 3 - 6 \cdot 2) - \mathbf{j} (2 \cdot 3 - 6 \cdot 1) + \mathbf{k} (2 \cdot 2 - 4 \cdot 1)
$$  
$$
\mathbf{i} (12 - 12) - \mathbf{j} (6 - 6) + \mathbf{k} (4 - 4) = (0,0,0)
$$  

Nəticə **$\mathbf{0}$** vektorudur.  

🔹 **Nəticə:**  
Əgər vektorlar **paraleldirsə** ($\alpha = 0^\circ$ və ya $\alpha = 180^\circ$), onların **vektorial hasili sıfır vektor olur**, çünki **$\sin \alpha = 0$** olur. 🚀



### **6. Ortogonal vektorların vektorial hasilinin modulu maksimumdur:**  
Əgər iki vektor bir-birinə perpendikulyardırsa ($\alpha = 90^\circ$), onda onların vektorial hasilinin modulu onların uzunluqlarının hasilinə bərabərdir:  
$$
|\mathbf{A} \times \mathbf{B}| = |\mathbf{A}| |\mathbf{B}|
$$  
🔹 **Misal:**  
$$
(1,0,0) \times (0,1,0) = (0,0,1)
$$  
və  
$$
|(1,0,0) \times (0,1,0)| = |(0,0,1)| = 1
$$  

Bunlar vektorial hasilin əsas xassələridir! 🚀

## Vektorların Vektorial hasilinin Koordinatlarla ifadəsi (1)

Vektorların vektorial hasilinin koordinatlarla ifadəsi aşağıdakı düsturla verilir:  

Əgər **A** və **B** vektorları belə verilirsə:  
$$
\mathbf{A} = (A_x, A_y, A_z), \quad \mathbf{B} = (B_x, B_y, B_z)
$$  
onların vektorial hasili aşağıdakı determinant şəklində ifadə olunur:  

$$
\mathbf{A} \times \mathbf{B} =
\begin{vmatrix} 
\mathbf{i} & \mathbf{j} & \mathbf{k} \\ 
A_x & A_y & A_z \\ 
B_x & B_y & B_z 
\end{vmatrix}
$$

Bu determinant açılarsa:  

$$
\mathbf{A} \times \mathbf{B} =
\left( A_y B_z - A_z B_y \right) \mathbf{i} -
\left( A_x B_z - A_z B_x \right) \mathbf{j} +
\left( A_x B_y - A_y B_x \right) \mathbf{k}
$$

Yəni komponentlər üzrə yazsaq:  
$$
\mathbf{A} \times \mathbf{B} =
\left( (A_y B_z - A_z B_y), (A_z B_x - A_x B_z), (A_x B_y - A_y B_x) \right)
$$

---

### **Nümunə**  
Tutaq ki, bizdə belə vektorlar var:  
$$
\mathbf{A} = (1, 2, 3), \quad \mathbf{B} = (4, 5, 6)
$$

Onda vektorial hasil belə olacaq:  

$$
\mathbf{A} \times \mathbf{B} =
\left( (2 \cdot 6 - 3 \cdot 5), (3 \cdot 4 - 1 \cdot 6), (1 \cdot 5 - 2 \cdot 4) \right)
$$

$$
= (12 - 15, 12 - 6, 5 - 8) = (-3, 6, -3)
$$

Bu, **A və B vektorlarına perpendikulyar olan yeni bir vektordur**. 🚀

## İki vektorun yaratdığı Paraleloqramın sahəsi (1)

İki vektorun yaratdığı **paraleloqramın sahəsi** onların **vektorial hasilinin modulu** ilə hesablanır.  

### **Düstur:**  
Əgər **A** və **B** vektorları verilibsə:  
$$
\mathbf{A} = (A_x, A_y, A_z), \quad \mathbf{B} = (B_x, B_y, B_z)
$$  
Onda paraleloqramın sahəsi belə hesablanır:  
$$
S = |\mathbf{A} \times \mathbf{B}|
$$

onların vektorial hasili aşağıdakı determinant şəklində ifadə olunur:  

$$
\mathbf{A} \times \mathbf{B} =
\begin{vmatrix} 
\mathbf{i} & \mathbf{j} & \mathbf{k} \\ 
A_x & A_y & A_z \\ 
B_x & B_y & B_z 
\end{vmatrix}
$$

Bu determinant açılarsa:  

$$
\mathbf{A} \times \mathbf{B} =
\left( A_y B_z - A_z B_y \right) \mathbf{i} -
\left( A_x B_z - A_z B_x \right) \mathbf{j} +
\left( A_x B_y - A_y B_x \right) \mathbf{k}
$$

Yəni komponentlər üzrə yazsaq:  
$$
\mathbf{A} \times \mathbf{B} =
\left( (A_y B_z - A_z B_y), (A_z B_x - A_x B_z), (A_x B_y - A_y B_x) \right)
$$


Vektorial hasilin modulu isə belə hesablanır:  
$$
\mathbf{S}_⏥ = |\mathbf{A} \times \mathbf{B}| =
\sqrt{(A_y B_z - A_z B_y)^2 + (A_z B_x - A_x B_z)^2 + (A_x B_y - A_y B_x)^2}
$$ 

Bu ifadə **paraleloqramın sahəsini** verir. 🚀

---

Vektorial hasilin modulu isə:  
$$
|\mathbf{A} \times \mathbf{B}| = |\mathbf{A}| |\mathbf{B}| \sin \theta
$$  
burada **$ \theta $** — **A** və **B** vektorları arasındakı bucaqdır.  

---

### **Nümunə**  
 
Tutaq ki,  
$$
\mathbf{A} = (1, 2, 3), \quad \mathbf{B} = (4, 5, 6)
$$

Determinant yazırıq:  
$$
\mathbf{A} \times \mathbf{B} =
\begin{vmatrix}
\mathbf{i} & \mathbf{j} & \mathbf{k} \\
1 & 2 & 3 \\
4 & 5 & 6
\end{vmatrix}
$$

Determinantı açırıq:  
$$
\mathbf{A} \times \mathbf{B} =
\mathbf{i} 
\begin{vmatrix} 
2 & 3 \\ 
5 & 6 
\end{vmatrix}
-
\mathbf{j} 
\begin{vmatrix} 
1 & 3 \\ 
4 & 6 
\end{vmatrix}
+
\mathbf{k} 
\begin{vmatrix} 
1 & 2 \\ 
4 & 5 
\end{vmatrix}
$$

Hər bir determinantı hesablayırıq:  

$$
\begin{vmatrix} 
2 & 3 \\ 
5 & 6 
\end{vmatrix}
= (2 \times 6 - 3 \times 5) = (12 - 15) = -3
$$

$$
\begin{vmatrix} 
1 & 3 \\ 
4 & 6 
\end{vmatrix}
= (1 \times 6 - 3 \times 4) = (6 - 12) = -6
$$

$$
\begin{vmatrix} 
1 & 2 \\ 
4 & 5 
\end{vmatrix}
= (1 \times 5 - 2 \times 4) = (5 - 8) = -3
$$

Nəticəni yığırıq:  
$$
\mathbf{A} \times \mathbf{B} = (-3) \mathbf{i} - (-6) \mathbf{j} + (-3) \mathbf{k}
$$
$$
= (-3, 6, -3)
$$

---

### **Paraleloqramın Sahəsi**  
Vektorial hasilin modulu:  
$$
S = \sqrt{(-3)^2 + 6^2 + (-3)^2}
$$
$$
= \sqrt{9 + 36 + 9} = \sqrt{54} = 3\sqrt{6}
$$

Bu **paraleloqramın sahəsidir**. 🚀

## İki vektorun yaratdığı Paraleloqramın sahəsindən Üçbucağın sahəsini tapmaq (1)

İki vektorun yaratdığı **paraleloqramın sahəsindən** üçbucağın sahəsini tapmaq üçün sadəcə onu **yarıya bölmək** kifayətdir.  

### **Düstur:**  
Əgər iki vektor **A** və **B** paraleloqram yaradırsa, həmin vektorların yaratdığı üçbucağın sahəsi belə hesablanır:  

$$
S_{\triangle} = \frac{1}{2} S_⏥ = \frac{1}{2} |\mathbf{A} \times \mathbf{B}|
$$

$$

$$

Burada:  
- **$ a $** və **$ b $** üçbucağın iki qonşu tərəfinin uzunluqlarıdır.

## Vektorların Qarışıq Hasili (Skalyar-Triple Hasil) (1)

### **Tərif:**  
>**Tərif:** Üç vektorun qarışıq hasili (skalyar-triple hasil) iki vektorun vektorial hasilinin üçüncü vektorla skalyar hasili kimi təyin olunur və nəticə skalyar bir qiymət alır.  

### **Düsturlar:**  
1. **Determinant vasitəsilə:**  
$$  
(\mathbf{a} \times \mathbf{b}) \cdot \mathbf{c} =  
\begin{vmatrix}   
a_1 & a_2 & a_3 \\   
b_1 & b_2 & b_3 \\   
c_1 & c_2 & c_3   
\end{vmatrix}  
$$  

2. **Geometrik ifadə:**  
$$  
(\mathbf{a} \times \mathbf{b}) \cdot \mathbf{c} = |\mathbf{a}||\mathbf{b}||\mathbf{c}| \cos\theta \cdot \sin\phi  
$$  
Burada:  
- $ \theta $ — $ \mathbf{b} $ və $ \mathbf{c} $ arasındakı bucaq,  
- $ \phi $ — $ \mathbf{a} $ və $ \mathbf{b} $ arasındakı bucaqdır.  

**Qeyd:** Qarışıq hasilin modulu bu vektorların yaratdığı **Paralelpipedin həcmini** ifadə edir:  

$$  
V = \left| (\mathbf{a} \times \mathbf{b}) \cdot \mathbf{c} \right|  
$$  

---



Budur, mətniniz istədiyiniz formada düzəldilib:

---

### **Vektorların Qarışıq Hasili (Skalyar-Triple Hasil)**  

Vektorların qarışıq hasili, üç vektorun skalyar hasilini və vektorial hasilini birləşdirən bir əməliyyatdır. Bu əməliyyat **"skalyar-triple hasil"** adlanır və üç vektor $ \mathbf{a}, \mathbf{b}, \mathbf{c} $ verildikdə aşağıdakı kimi təyin olunur:

$$
(\mathbf{a} \times \mathbf{b}) \cdot \mathbf{c}
$$

Bu ifadə əvvəlcə $ \mathbf{a} $ və $ \mathbf{b} $ vektorlarının vektorial hasilini (yeni vektor) yaradır, sonra bu yeni vektor ilə $ \mathbf{c} $ vektorunun skalyar hasili hesablanır.  

---

### **1. Determinant İlə Hesablanması**  
Bu üçlü hasil determinant vasitəsilə belə hesablanır:

$$
(\mathbf{a} \times \mathbf{b}) \cdot \mathbf{c} =
\begin{vmatrix} 
a_1 & a_2 & a_3 \\ 
b_1 & b_2 & b_3 \\ 
c_1 & c_2 & c_3 
\end{vmatrix}
$$

Burada:  
- $ a_1, a_2, a_3 $ — $ \mathbf{a} $ vektorunun koordinatlarıdır.  
- $ b_1, b_2, b_3 $ — $ \mathbf{b} $ vektorunun koordinatlarıdır.  
- $ c_1, c_2, c_3 $ — $ \mathbf{c} $ vektorunun koordinatlarıdır.  

Bu determinantın hesablanması üçün aşağıdakı düsturdan istifadə olunur:

$$
a_1(b_2c_3 - b_3c_2) - a_2(b_1c_3 - b_3c_1) + a_3(b_1c_2 - b_2c_1)
$$

#### **Misal:**  
Vektorlarımız olsun:

$$
\mathbf{a} = (2, -1, 3), \quad \mathbf{b} = (1, 4, -2), \quad \mathbf{c} = (3, 0, 5)
$$

Hesablayaq:

$$
\begin{vmatrix} 
2 & -1 & 3 \\ 
1 & 4 & -2 \\ 
3 & 0 & 5 
\end{vmatrix}
$$

Bu determinantın hesablanması:

$$
2(4 \cdot 5 - (-2) \cdot 0) - (-1)(1 \cdot 5 - (-2) \cdot 3) + 3(1 \cdot 0 - 4 \cdot 3)
$$

$$
= 2(20 - 0) - (-1)(5 + 6) + 3(0 - 12)
$$

$$
= 2(20) + 11 - 36
$$

$$
= 40 + 11 - 36 = 15
$$

---

### **2. Geometrik Mənası və Paralelpipedlə Əlaqəsi**  
Bu hasilin ən mühüm tətbiqlərindən biri üç vektorun yaratdığı **paralelpipedin həcminin hesablanmasıdır**.  

- **Paralelpiped** — üç vektor tərəfindən təyin olunan üçölçülü fiqurdur.  
- **Bu skalyar hasilin modulu (mütləq dəyəri) həmin paralelpipedin həcmini verir**:

$$
V = \left| (\mathbf{a} \times \mathbf{b}) \cdot \mathbf{c} \right|
$$

#### **Misal 1:**  
Əvvəlki misaldan istifadə edək:

$$
V = | 15 | = 15
$$

Bu o deməkdir ki, $ \mathbf{a}, \mathbf{b}, \mathbf{c} $ vektorları ilə qurulan paralelpipedin həcmi **15 vahid kubdur**.

Əgər nəticə 0 olsaydı, bu vektorların eyni müstəvidə olduğunu göstərərdi və paralelpiped yaranmazdı.

#### **Misal 2:**  
Vektorlarımız olsun:

$$
\mathbf{a} = (1, 2, 3), \quad \mathbf{b} = (4, 5, 6), \quad \mathbf{c} = (7, 8, 9)
$$

Determinant vasitəsilə hesablayırıq:

$$
\begin{vmatrix} 
1 & 2 & 3 \\ 
4 & 5 & 6 \\ 
7 & 8 & 9 
\end{vmatrix}
$$

Bu determinantın hesablanması:

$$
1(5 \cdot 9 - 6 \cdot 8) - 2(4 \cdot 9 - 6 \cdot 7) + 3(4 \cdot 8 - 5 \cdot 7)
$$

$$
= 1(45 - 48) - 2(36 - 42) + 3(32 - 35)
$$

$$
= 1(-3) - 2(-6) + 3(-3)
$$

$$
= -3 + 12 - 9 = 0
$$

Bu nəticə göstərir ki, vektorlar xətti asılıdır və onlar eyni müstəvidədirlər. Yəni, bu vektorlarla qurulan paralelpipedin həcmi **0**-dır.

## Vektorial Hasil və Qarışıq Hasil (Skalyar-Triple Hasil) Arasındakı Fərq 

Gəlin, sizin suallarınızı aydın şəkildə izah edim:

### 1. **Vektorial Hasil və Qarışıq Hasil (Skalyar-Triple Hasil) Arasındakı Fərq:**

Bəli, həm vektorial hasil, həm də qarışıq hasil üçün **eyni üsul** tətbiq edilir, amma fərqlilik yalnız **hansı vektorların** istifadə olunduğu və **son nəticənin** nə olduğu ilə bağlıdır.

#### **Vektorial Hasil (Cross Product):**

Vektorial hasilin hesablanmasında istifadə edilən ən yaygın üsul **determinant** üsuludur və bunu tez-tez $ i, j, k $ ortogonal (dik) vektorlar ilə göstəririk.

Məsələn, $ \mathbf{a} = (a_1, a_2, a_3) $ və $ \mathbf{b} = (b_1, b_2, b_3) $ vektorları üçün vektorial hasil aşağıdakı kimi hesablanır:
$$
\mathbf{a} \times \mathbf{b} = 
\begin{vmatrix}
\hat{i} & \hat{j} & \hat{k} \\
a_1 & a_2 & a_3 \\
b_1 & b_2 & b_3
\end{vmatrix}
$$
Burada $ \hat{i}, \hat{j}, \hat{k} $ müstəvi üzərindəki ortogonal (dik) vektorlardır. Yəni $ \hat{i}, \hat{j}, \hat{k} $ vektorlar müstəvidəki **x, y, z oxlarını** təmsil edir.

**Niyə $ i, j, k $-dən istifadə edilir?**
- $ i, j, k $ vektorları müstəvidəki üç əsas oxu təmsil edir: 
  - $ \hat{i} $ – x-oxu,
  - $ \hat{j} $ – y-oxu,
  - $ \hat{k} $ – z-oxu.
- Bu vektorlar müstəvidəki **oriyentasiya** (yəni yönəlmə) və **mütləq koordinat sistemi** barədə məlumat verir.

#### **Qarışıq Hasil (Skalyar-Triple Hasil):**

Qarışıq hasil üçün üç vektorun vektorial hasili və sonra bu vektorial hasilin skalyar hasilini hesablayırıq. Əgər $ \mathbf{a} = (a_1, a_2, a_3) $, $ \mathbf{b} = (b_1, b_2, b_3) $ və $ \mathbf{c} = (c_1, c_2, c_3) $ vektorları varsa, o zaman qarışıq hasil belə hesablanır:

$$
(\mathbf{a} \times \mathbf{b}) \cdot \mathbf{c} = 
\begin{vmatrix}
\hat{i} & \hat{j} & \hat{k} \\
a_1 & a_2 & a_3 \\
b_1 & b_2 & b_3
\end{vmatrix}
\cdot
(c_1, c_2, c_3)
$$

Burada:
- **İlk addımda** $ \mathbf{a} \times \mathbf{b} $-nin vektorial hasilini tapırıq.
- **İkinci addımda** isə, bu nəticənin **skalyar hasilini** $ \mathbf{c} $-lə hesablayırıq.

**Niyə qarışıq hasildə $ i, j, k $-dən istifadə edirik?**
- **Qarışıq hasil də vektorial hasilin hesablanmasıdır**, amma sonrasında skalyar hasil tətbiq edirik. Bu səbəbdən, determinantla ilk vektorial hasilin tapılmasında $ i, j, k $-yə ehtiyac duyuruq.

### 2. **Nə Üçün $ i, j, k $-Nin Qarışıq Hasilində Düzgün Hesablanması Vacibdir, Amma Vektorial Hasilində Çıxarılır?**

Bu məsələni belə izah edək:

- **Vektorial hasildə**: $ \hat{i}, \hat{j}, \hat{k} $ vektorları sadəcə **hesablama** məqsədilə istifadə edilir. Əslində, vektorial hasilin nəticəsi **tək bir vektor** olduğu üçün, müstəvidəki istiqamətləri əks etdirən bu $ i, j, k $ komponentlərinə görə nəticə alınır.
  
- **Qarışıq hasilində**: Burada həm vektorial hasilin tapılması, həm də onun skalyar hasilinin hesablanması lazımdır. Beləliklə, qarışıq hasilin hesablanmasında **təməl əməliyyatları dəqiq etmək** və vektorları düzgün yerləşdirmək üçün **koordiantlar (i, j, k)** istifadə olunur. Amma hər iki əməliyyatı yerinə yetirmək bir növ **üçüncü bir vektorun** hesablanmasını tələb edir.

### 3. **Fərq nədir?**

- **Vektorial hasil** vektorlar arasındakı "çapraz" əlaqəni (yəni, onların müstəvisinə dik olan yeni bir vektor yaratmaq) göstərir.
- **Qarışıq hasil** isə bu çapraz əlaqənin **skalyar** ölçüsünü təhlil edir və bu nəticə, ümumiyyətlə **həmin üç vektorun** **komplanar olub-olmaması** barədə məlumat verir. 

Qarışıq hasilin sıfır olması, üç vektorun eyni müstəvidə (komplanar) olduğunu bildirir, amma bu xüsusiyyət yalnız **skalyar hasilin** nəticəsi olaraq əldə olunur. Bu səbəbdən, qarışıq hasilin sıfır olması, bu vektorların bir-birinə **paralel** olduğu və ya **eyni müstəvidə** olduğu anlamına gəlir.

### **Nəticə:**
- **Vektorial hasil** və **qarışıq hasil** çox oxşar metodlarla hesablanır (determinant üsulu), amma qarışıq hasilin sonunda əlavə bir skalyar hasil hesablama əməliyyatı yerinə yetirilir.
- **$ i, j, k $** istifadə olunması həm vektorial hasilin hesablanmasında, həm də qarışıq hasilin düzgün hesablanmasında vacibdir.

## Qarışıq Skalyar Hasilin (Triple Scalar Product) Əsas Xassələri (1)

**Qarışıq Skalyar Hasilin (Triple Scalar Product) Əsas Xassələri**  

Qarışıq skalyar hasil üç vektorun determinantı kimi hesablanır:  

$$
[\mathbf{a}, \mathbf{b}, \mathbf{c}] = \mathbf{a} \cdot (\mathbf{b} \times \mathbf{c})
$$

Bu ifadə **paralelopipedin həcmidir** və aşağıdakı əsas xassələrə malikdir:

### **1. Dövrü dəyişmə xassəsi (Siklik Permutasiya)**
Üç vektoru dövrü şəkildə dəyişsək, hasilin qiyməti dəyişmir:

$$
[\mathbf{a}, \mathbf{b}, \mathbf{c}] = [\mathbf{b}, \mathbf{c}, \mathbf{a}] = [\mathbf{c}, \mathbf{a}, \mathbf{b}]
$$

### **2. İşarənin dəyişməsi xassəsi (Antisimetriya)**
İki vektorun yerini dəyişsək, hasilin işarəsi dəyişir:

$$
[\mathbf{a}, \mathbf{b}, \mathbf{c}] = -[\mathbf{a}, \mathbf{c}, \mathbf{b}]
$$

Ümumiyyətlə, vektorların cüt sayda yerini dəyişmək **işarəni saxlayır**, tək sayda dəyişmək isə **işarəni dəyişir**.

### **3. Xətti xassə (Linearlıq)**
Birinci vektor skalyar ədədə vurularsa:

$$
[\lambda \mathbf{a}, \mathbf{b}, \mathbf{c}] = \lambda [\mathbf{a}, \mathbf{b}, \mathbf{c}]
$$

Eyni qayda ilə digər vektorlar üçün də xətti xassə tətbiq olunur:

$$
[\mathbf{a}, \lambda \mathbf{b} + \mu \mathbf{c}, \mathbf{d}] = \lambda [\mathbf{a}, \mathbf{b}, \mathbf{d}] + \mu [\mathbf{a}, \mathbf{c}, \mathbf{d}]
$$

### **4. Qarışıq skalyar hasil sıfır oması**
Əgər qarışıq skalyar hasil sıfırdırsa:

$$
[\mathbf{a}, \mathbf{b}, \mathbf{c}] = 0
$$

bu, o deməkdir ki, $\mathbf{a}$, $\mathbf{b}$, $\mathbf{c}$ **vektorları** 
- ya:
- **1).Eyni müstəvidə yerləşir (**Komplanardır**).**
- ya:
- **2).Hər hansı ikisi **Kollienardır****
- ya:
- **3).Heç olmas biri  $\vec{0}$ olmalıdır**


### **5. Determinant ifadəsi**
Üç vektorun qarışıq skalyar hasilini determinant şəklində də yaza bilərik:

$$
[\mathbf{a}, \mathbf{b}, \mathbf{c}] =
\begin{vmatrix}
a_1 & a_2 & a_3 \\
b_1 & b_2 & b_3 \\
c_1 & c_2 & c_3
\end{vmatrix}
$$

Bu determinantın sıfır olması vektorların komplanarlığını göstərir.

---

Bu xassələr **vektor analizi**, **həndəsə**, **mexanika** və **fizika** kimi sahələrdə istifadə olunur, xüsusilə **paralelopipedin həcmi** və **müstəvi üzərində yerləşmə şərtləri** üçün vacibdir.

## Komplanar Vektorlarda Qarışıq  Hasilin  0 olması

**Komplanar Vektorlar:**

Əgər **$(\mathbf{a})$, $(\mathbf{b})$, və $(\mathbf{c})$** vektorları komplanardırsa, bu o deməkdir ki, **üç vektor eyni müstəvidə yerləşir**. Vektorların komplanar olması **onların paralel** olmalarını tələb etmir, amma **eyni müstəvidə olmaları** vacibdir.

**Komplanarlıq** şəraitində, $(\mathbf{a} \times \mathbf{b})$ (yəni $(\mathbf{a})$ və $(\mathbf{b})$-nin vektorial hasili) vektorunun istiqaməti bu müstəviyə **perpendikulyar** olacaq. Lakin, **əgər $(\mathbf{c})$ həmin müstəvidədirsə**, o zaman **$(\mathbf{c})$** vektoru **$(\mathbf{a} \times \mathbf{b})$** ilə **perpendikulyar** olacaq, və bu zaman **skalyar hasil sıfır** olacaqdır.

Niyə bu belədir?

1. **$(\mathbf{a} \times \mathbf{b})$ vektorunun perpendikulyar olması**: Vektorial hasilin nəticəsi, hər iki vektora dik olan yeni bir vektordur. Bu halda, $(\mathbf{a} \times \mathbf{b})$ vektorunun istiqaməti $(\mathbf{a})$-nın və $(\mathbf{b})$-nin müstəvisinə dikdir.

2. **Skalyar hasilin sıfır olması**: Qarışıq hasilin skalyar hissəsi belə hesablanır:

$$
(\mathbf{a} \times \mathbf{b}) \cdot \mathbf{c}
$$

Bu, deməkdir ki, əgər $(\mathbf{c})$ müstəvidədirsə və **$(\mathbf{a} \times \mathbf{b})$-yə perpendikulyardırsa**, onların arasındakı bucağın kosinus dəyəri **90** olacaq (çünki onlar perpendikulyardır). Skalyar hasilin sıfır olması **$\cos 90^\circ = 0$** olduğu üçün nəticə sıfır olacaq.

3. **Paralelopiped**: Bu vəziyyətdə paralelopiped **yaranmır**. **Paralelopiped** yalnız 3 vektorun (komplanar olmayan) nəticəsində əmələ gəlir, çünki bu halda vektorlar arasındakı bucaqlar **90 dərəcədən fərqli olur**. Əgər vektorlar komplanardırsa, onların arasındakı əlaqə paralelopipedin həcmini sıfırlayacaq.

---

**Nəticə:**

- **Komplanar vektorlar**: $(\mathbf{a})$, $(\mathbf{b})$, və $(\mathbf{c})$ vektorları komplanar olduqda, qarışıq hasil **sıfır** olur.
- **Paralelopipedin həcmi sıfır olur**: Çünki paralelopipedin həcmi qarışıq hasilə əsaslanır və sıfır olan qarışıq hasil paralelopipedin həcminin sıfır olduğunu göstərir.

Bu, **$(\mathbf{a})$, $(\mathbf{b})$, $(\mathbf{c})$** vektorlarının komplanar olduqda qarışıq hasilin sıfır olduğunu göstərir.

## Kollinear  Vektorlarda Qarışıq Hasilin 0 olması

### Kollinear Vektorlar
Kollinear vektorlar deməkdir ki, bu vektorlar eyni xətt boyunca və ya düz xəttdədirlər. Yəni, bu vektorlar paraleldir. Kollinear vektorlar arasında bucaqlar $0^\circ$ və ya $180^\circ$-dir. Bunun mənası budur ki, bu vektorlar arasında heç bir "hündürlük" yoxdur, onlar düz xətlə uzanırlar.

### Vektor Hasilinin Sıfır Olması
İki vektor arasında vektor hasili $ \mathbf{b} \times \mathbf{c} $, bu iki vektor paralel olduqda sıfır olar. Bunun səbəbi budur ki:
- Vektor hasili iki vektor arasında yaranan paralelopipedin həcmini təmsil edir. Kollinear vektorlar eyni xətt boyunca olduğu üçün belə bir paralelopiped yaratmazlar, çünki bu vektorlar arasında "dərinlik" və ya hündürlük yoxdur.
- Kollinear vektorlar arasında vektor hasili sıfır olur:

$$
\mathbf{b} \times \mathbf{c} = \mathbf{0}.
$$

Bu səbəbdən, iki kollinear vektor arasındakı vektor hasili sıfır olur.

### Qarışıq Hasilin Nəticəsi
Qarışıq hasil aşağıdakı kimi tərif olunduğu üçün:

$$
\mathbf{a} \cdot (\mathbf{b} \times \mathbf{c})
$$

Əgər $\mathbf{b} \times \mathbf{c} = \mathbf{0}$ (çünki $\mathbf{b}$ və $\mathbf{c}$ kollineardır), o zaman:

$$
\mathbf{a} \cdot (\mathbf{0}) = 0.
$$

Beləliklə, kollinear vektorlar arasında qarışıq hasil sıfır olur.

### Sinus Funksiyasının İzahı
İndi, $\sin(\varphi)$ ilə əlaqələndirək:

- Kollinear vektorlar arasında bucaq $ \varphi = 0^\circ $ və ya $ 180^\circ $-dir. Bu zaman sinus funksiyasının dəyəri sıfır olur:
  
  $$
  \sin(0^\circ) = 0, \quad \sin(180^\circ) = 0.
  $$

- Sinus funksiyasının sıfır olması vektorlar arasında heç bir hündürlük və ya diklik olmadığını göstərir. Yəni, vektorlar paralel olduqda, onların arasında "hündürlük" olmadığı üçün bucağın sinus dəyəri sıfır olur.

### Geometrik İzah
Geometrik baxımdan, kollinear vektorlar bir-birinə paralel olduğundan, bu vektorlar arasında paralelopiped (həcmi olan bir cisim) yaranmaz. Bu halda, vektorların təyin etdiyi sahə sıfır olur. Qarışıq hasilin sıfır olması da, bu paralelopipedin həcminin sıfır olması ilə əlaqədardır.

Sonuç olaraq, kollinear vektorlar arasında vektor hasili sıfır olur və bu səbəbdən qarışıq hasil də sıfır olur.

## Vektorların Qarışıq Hasilindən Piramidanın Həcmi (1)

Piramidanın həcmindən **vektorların qarışıq hasilindən (triple scalar product)** istifadə edərək necə istifadə etməyi başa düşmək üçün, piramidanın üçbucaqlı əsasına paralel olan **üç vektor** təyin edirik. Bu üç vektor bir-birinə ortoqonal deyil, amma paralelopipedin həcmindən istifadə edərək piramidanın həcmini hesablaya bilərik.

### 1. **Piramidanın Həcmi və Vektorların Qarışıq Hasilindən İstifadə**

Əgər bir piramidanın əsas sahəsi **$ A $** və hündürlüyü **$ h $** varsa, piramidanın həcmi belə hesablanır:

$$
V = \frac{1}{3} \times A \times h
$$

Lakin piramidanın həcmindən **vektorların qarışıq hasilindən** istifadə edərək, bu həcmə daha ümumi şəkildə yanaşma verə bilərik.

### 2. **Üç Vektorla Vektorlar Üzərində Həcm Hesablama**

Piramidanın həcmindən hesablamaq üçün, bu piramidanın **üçbucaqlı əsasını** təmsil edən iki vektor və hündürlüyünü təmsil edən bir vektor seçirik. Beləliklə, bu üç vektoru **vektorların qarışıq hasilinə** tətbiq edirik.

#### Addımlar:
1. **Vektorlar Seçmək**: 
   - Birinci vektor **$ \mathbf{v_1} $** əsas üzərindəki bir vektoru,
   - İkinci vektor **$ \mathbf{v_2} $** əsas üzərindəki digər vektoru,
   - Üçüncü vektor **$ \mathbf{v_3} $** piramidanın hündürlüyünü göstərən vektoru təmsil edir.

2. **Qarışıq Hasil Hesablamaq**:
   Bu üç vektor ilə qarışıq skalyar hasilini (triple scalar product) istifadə edərək paralelopipedin həcmindən sonra piramidanın həcmində istifadə edilən xüsusi formula əldə edirik:
   
   $$
   V_{\text{paralelopiped}} = | \mathbf{v_1} \cdot (\mathbf{v_2} \times \mathbf{v_3}) |
   $$

   Bu paralelopipedin həcmidir.

3. **Piramidanın Həcmi**:
   Piramidanın həcmindən hesablamamız üçün paralelopipedin həcmindən **$\frac{1}{3}$** qatını götürürük:

   $$
   V_{\text{piramida}} = \frac{1}{6} \times | \mathbf{v_1} \cdot (\mathbf{v_2} \times \mathbf{v_3}) |
   $$

Bu formul vasitəsilə, vektorlar istifadə edərək piramidanın həcmını hesablaya bilərsiniz. Buradakı **$ \mathbf{v_1} $**, **$ \mathbf{v_2} $** və **$ \mathbf{v_3} $** vektorları piramidanın təmsil edən əsas və hündürlük vektorlarını göstərir.
