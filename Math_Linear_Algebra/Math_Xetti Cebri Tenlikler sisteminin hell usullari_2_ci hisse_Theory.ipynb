## Gauss üsulu (Məchulları ardıcıl yox etmə üsulu) (1)

**Gauss üsulu** (məchulları ardıcıl yox etmə üsulu) xətti cəbrdə tənliklər sistemini həll etmək üçün geniş istifadə olunan effektiv bir metoddur. Əsas məqsəd xətti tənliklər sistemini **üst üçbucaq matrisinə** çevirməkdir ki, bu da tənlikləri asanlıqla həll etməyə imkan verir.

---

### **Metodun Addımları:**

#### 1. **Giriş Sistemi:**
Xətti tənliklər sistemi matris şəklində yazılır:
$$
AX = B
$$
Burada:
- $A$ - əmsallar matrisi,
- $X$ - məchullar vektoru,
- $B$ - sağ tərəfdəki dəyərlər vektoru.

Matris şəklində:
$$
\begin{bmatrix}
a_{11} & a_{12} & \dots & a_{1n} \\
a_{21} & a_{22} & \dots & a_{2n} \\
\vdots & \vdots & \ddots & \vdots \\
a_{n1} & a_{n2} & \dots & a_{nn}
\end{bmatrix}
\begin{bmatrix}
x_1 \\
x_2 \\
\vdots \\
x_n
\end{bmatrix}
=
\begin{bmatrix}
b_1 \\
b_2 \\
\vdots \\
b_n
\end{bmatrix}
$$

#### 2. **Matrisin Üst Üçbucaq Formasına Gətirilməsi:**
- **Əsas Sətir Əməliyyatları**:
  - Birinci sətir seçilir və digər sətirlərdə həmin sütundakı əmsalları $0$-a çevirmək üçün istifadə olunur.
  - Daha sonra digər sətirlərdə də eyni əməliyyat təkrarlanır.
  - Məqsəd, aşağıdakı kimi **üst üçbucaq matris** forması əldə etməkdir:
  $$
  \begin{bmatrix}
  a_{11} & a_{12} & \dots & a_{1n} \\
  0 & a_{22}' & \dots & a_{2n}' \\
  \vdots & \vdots & \ddots & \vdots \\
  0 & 0 & \dots & a_{nn}'
  \end{bmatrix}
  $$

#### 3. **Geri Əvəzetmə:**
Üst üçbucaq matrisindən başlayaraq məchullar geriyə doğru həll edilir:
1. Sonuncu tənlikdən ($x_n$) başlayaraq məchul tapılır.
2. Tapılan məchul yuxarı tənliklərdə əvəzetmə edilərək digər məchullar hesablanır.

---

### **Misal:**

Tənliklər sistemi:
$$
2x + y - z = 8 \\
-3x - y + 2z = -11 \\
-2x + y + 2z = -3
$$

#### Addım 1: Əmsallar Matrisi Yazılır
$$
\begin{bmatrix}
2 & 1 & -1 & | & 8 \\
-3 & -1 & 2 & | & -11 \\
-2 & 1 & 2 & | & -3
\end{bmatrix}
$$

#### Addım 2: Sıfırlama Əməliyyatı (Üst Üçbucaq Forması)
- Birinci sütunun altındakı əmsalları $0$-a çevirmək üçün:
  - İkinci sətr: $R_2 = R_2 + \frac{3}{2}R_1$
  - Üçüncü sətr: $R_3 = R_3 + R_1$

Nəticədə:
$$
\begin{bmatrix}
2 & 1 & -1 & | & 8 \\
0 & 0.5 & 0.5 & | & 1 \\
0 & 2 & 1 & | & 5
\end{bmatrix}
$$

- Daha sonra ikinci sütunun altındakı əmsalları sıfırlamaq üçün əməliyyatlar davam etdirilir.

#### Addım 3: Geri Əvəzetmə
Üst üçbucaq matrisindən başlayaraq $x_3$, $x_2$, $x_1$ tapılır.

---

### **Üstünlükləri:**
- Effektiv və sistematikdir.
- Kompüterlə hesablama üçün əlverişlidir.
  
### **Çatışmazlıqları:**
- Çox böyük matrislərdə hesablama yükü arta bilər.
- Əgər determinant $0$-a bərabərdirsə ($A$ təkdirdirsə), həll mövcud olmur.

## Kronecker-Capelli Teoremi (1)

Xətti cəbrdə tənliklər sisteminin həllinin var olub-olmamasını müəyyən edən bir prinsipdir. Bu teorem, tənliklər sisteminin həllinin varlığını matrislərin sıra (rank) anlayışına əsaslanaraq izah edir.

### Teoremin İfadəsi
Xətti tənliklər sistemi:
$$
A \mathbf{x} = \mathbf{b}
$$
burada:
- $A$ — $m \times n$ ölçülü əmsallar matrisi,
- $\mathbf{x}$ — $n \times 1$ ölçülü dəyişənlər vektoru,
- $\mathbf{b}$ — $m \times 1$ ölçülü nəticələr vektoru.

Bu sistemin həllinin varlığı üçün **Kronecker-Capelli teoremi** belə deyir:

1. **Həllin Var Olması Şərti:** 
   Xətti tənliklər sisteminin həllinin olması üçün əmsallar matrisinin ($A$) sıra (rank) dəyəri ilə genişləndirilmiş matrisin ($[A | b]$) sıra dəyəri bərabər olmalıdır:
   $$
   \text{rank}(A) = \text{rank}([A | b]).
   $$
   Əgər bu şərt ödənilmirsə, sistem **uyğunsuzdur** və həlli yoxdur.

2. **Həllin Növü:** 
   - Əgər $\text{rank}(A) = \text{rank}([A | b]) = n$, sistemin **unikal həlli** var.
   - Əgər $\text{rank}(A) = \text{rank}([A | b]) < n$, sistemin **sonsuz sayda həlli** var (çünki sərbəst dəyişənlər mövcuddur).

### Genişləndirilmiş Matris Nədir?
Genişləndirilmiş matris $[A | b]$, əmsallar matrisinə ($A$) nəticələr vektorunun ($\mathbf{b}$) sütun kimi əlavə edilməsi ilə alınır. Məsələn:
$$
A = \begin{bmatrix}
1 & 2 \\
3 & 4
\end{bmatrix}, \quad
\mathbf{b} = \begin{bmatrix}
5 \\
6
\end{bmatrix}, \quad
[A | b] = \begin{bmatrix}
1 & 2 & 5 \\
3 & 4 & 6
\end{bmatrix}.
$$

### Teoremin Tətbiqi
1. Əmsallar matrisinin $\text{rank}(A)$ dəyərini tapın.
2. Genişləndirilmiş matrisin $\text{rank}([A | b])$ dəyərini tapın.
3. Bu iki dəyəri müqayisə edin:
   - $\text{rank}(A) \neq \text{rank}([A | b])$: Sistem həllsizdir.
   - $\text{rank}(A) = \text{rank}([A | b])$: Sistem həllə malikdir.

### Nəticə
**Kronecker-Capelli teoremi**, xətti tənliklər sisteminin həllinin varlığını təsdiq edən və həllin unikallığını müəyyən edən əsas vasitədir. Əgər əlavə izah və ya nümunə istəyirsinizsə, deyə bilərsiniz! 😊

## Bircins Xətti Tənliklər Sistemi 

### **Bircins Xətti Tənliklər Sistemi (Tərif)**  
>**Tərif:** Bircins xətti tənliklər sistemi, azad hədlərin ($b_i$) hamısının sıfır olduğu xətti tənliklər sistemidir. Başqa sözlə, bu cür sistemlərdə bütün tənliklər yalnız bilinmələrin xətti kombinasiyalarından ibarətdir və sağ tərəfdə həmişə \(0\) olur.

**Ümumi forma**:  
$$
\begin{cases}
a_{11}x_1 + a_{12}x_2 + \dots + a_{1n}x_n = 0 \\
a_{21}x_1 + a_{22}x_2 + \dots + a_{2n}x_n = 0 \\
\vdots \\
a_{m1}x_1 + a_{m2}x_2 + \dots + a_{mn}x_n = 0
\end{cases}
$$
Burada:
- $a_{ij}$ — əmsallar ($i$-ci tənliyin $j$-ci bilinməsinin əmsalı),
- $x_j$ — bilinmələr,
- $m$ — tənliklərin sayı,
- $n$ — bilinmələrin sayı.

---

### **Xüsusiyyətləri:**
1. **Həmişə trivially həllə malikdir**:
   $x_1 = x_2 = \dots = x_n = 0$ həlli həmişə bu sistemin həllidir (trivial həll).  

2. **Qeyri-trivial həllər**:
   Əgər tənliklərin əmsallarından alınan matrisin determinantı $$0$$-dırsa (xüsusi halda bilinmələrin sayı tənliklərin sayından böyükdürsə), sistem qeyri-trivial həllərə də malik ola bilər.

3. **Matris forması**:  
   Bircins xətti tənliklər sistemi aşağıdakı matris forması ilə ifadə olunur:
   $$
   A \mathbf{x} = \mathbf{0}
   $$
   Burada:
   - $A$ — $m \times n$ ölçülü əmsallar matrisi,
   - $\mathbf{x}$ — bilinmələr vektoru,
   - $\mathbf{0}$ — sıfırdan ibarət vektor.

---

### **Misal:**
$$
\begin{cases}
2x + 3y - z = 0 \\
x - y + 4z = 0 \\
3x + 5y + 2z = 0
\end{cases}
$$

#### Matris forması:
$$
\begin{bmatrix}
2 & 3 & -1 \\
1 & -1 & 4 \\
3 & 5 & 2
\end{bmatrix}
\begin{bmatrix}
x \\
y \\
z
\end{bmatrix}
=
\begin{bmatrix}
0 \\
0 \\
0
\end{bmatrix}
$$

#### Həlli:
- **Trivial həll**: $x = 0, y = 0, z = 0$.
- **Qeyri-trivial həll**: Matrisin sıraları xətti asılı olduqda qeyri-trivial həll mövcud ola bilər.

---

Bircins xətti tənliklər sistemi fizika, mühəndislik və data elmlərində simmetriya, tarazlıq və optimallaşdırma məsələlərinin həllində geniş istifadə olunur.

## Bircins Xətti Tənliklər Sistemi: Tənliklərin sayı dəyişənlərin sayına bərabərdirsə (m=n) (1)

> **Tərif:** Əgər tənliklərin sayı dəyişənlərin sayına bərabərdirsə ($m = n$) və matrisin sıraları xətti **asılıdırsa**, **qeyri-trivial həll** mümkündür. Bu vəziyyət **xətti asılılıqdan** və **əmsallar matrisinin determinantının sıfır** olmasından qaynaqlanır.

### **İzah:**

1. **Matris sırasının xətti asılılığı:**  
   Əgər $\text{rank}(A) < n$ (matrisin sırası dəyişənlərin sayından kiçikdirsə), bu o deməkdir ki, sistemdəki tənliklər xətti asılıdır. Bu halda, sistem sonsuz sayda qeyri-trivial həllər verə bilər.

2. **Determinantın sıfır olması:**  
   Əgər $\text{det}(A) = 0$, bu o deməkdir ki, matris tərs olunandır və yalnız trivial həll mövcud deyil. Bu halda qeyri-trivial həll üçün parametrləşdirilmiş həllər mövcud ola bilər.

---

### **Misal:**

$$
\begin{aligned}
x + y + z &= 0, \\
2x + 2y + 2z &= 0, \\
3x + 3y + 3z &= 0.
\end{aligned}
$$

#### **Addımlar:**

1. Matris formasına keçək:  
  $$
   A = 
   \begin{bmatrix}
   1 & 1 & 1 \\
   2 & 2 & 2 \\
   3 & 3 & 3
   \end{bmatrix}, \quad
   X = 
   \begin{bmatrix}
   x \\ y \\ z
   \end{bmatrix}, \quad
   AX = 0.
   $$

2. $\text{rank}(A)$-ni yoxlayaq:  
   - $A$-nın determinantını hesablayırıq:  
     $$
     \text{det}(A) = 0.
     $$
     Bu o deməkdir ki, matris xətti asılıdır.

3. Sonsuz sayda həll var:  
   $$
   x + y + z = 0 \quad \text{ (yalnız ilk tənlik xətti müstəqildir)}.
  $$

#### **Nəticə:**  
Qeyri-trivial həll:  
$$
x = -y - z.
$$

Hər hansı $y, z \in \mathbb{R}$-yə görə həll mövcuddur, yəni qeyri-trivial həll mövcuddur.

---

### **Nəticə:**  
Əgər $\text{rank}(A) < n$ (matris xətti asılıdırsa), hətta $m = n$ olsa belə, qeyri-trivial həll mövcud ola bilər. Sistem yalnız o halda yalnız trivial həllə malik olur ki, matris xətti müstəqil olsun ($\text{rank}(A) = n$).

##  Bircins Xətti Tənliklər Sistemi: Tənliklərin sayı dəyişənlərin sayına bərabər deyilsə ($m \neq n$) (1)

### **Bircins Xətti Tənliklər Sistemi və Matrislər:$m \neq n$**

Bir sistemin xətti olması üçün aşağıdakı ümumi formaya sahib olmalıdır:

$$
A \cdot X = B
$$

Burada:
- $A$ - $m \times n$ ölçülü **koeffisiyent matrisi**,
- $X$ - $n \times 1$ ölçülü dəyişənlər vektoru,
- $B$ - $m \times 1$ ölçülü nəticə vektoru.

---

### **1. $m > n$ (Tənliklər çox, dəyişənlər az)**

Bu halda, tənliklər dəyişənlərdən çoxdur. Matrisin ölçüləri $m \times n$ olduğuna görə, bu vəziyyətdə aşağıdakı ehtimallar meydana çıxır:

#### **(a) Həllin olmaması (Matris sıralarının uyğunsuzluğu)**

Əgər $A$ matrisinin sırası (rank) $n$-dən kiçikdirsə, sistemin **həllinin olmaması** mümkündür. Bu halda tənliklər **xətti asılıdır** və **həll yoxdur**.

- **Matris sırası** ($\text{rank}(A)$) matrisin maksimum sayda xətti müstəqil sətir və ya sütun sayını göstərir.
- Əgər matrisin sırası $n$-dən kiçikdirsə, deməli sistem qeyri-məhdud sayda az sayda xətti müstəqil tənlikdən ibarətdir.
- **Misal:**
  $$
  \begin{bmatrix}
  1 & 1 & 1 \\
  2 & 2 & 2 \\
  3 & 3 & 3
  \end{bmatrix}
  \quad \text{(Bu matrisin sırası 1-dir, xətti asılıdır.)}
  $$

Əgər determinant sıfırdır ($\text{det}(A) = 0$) və sıralar uyğun deyilsə, **həll yoxdur**.

#### **(b) Sonsuz sayda həll (Xətti asılı tənliklər)**

Əgər $m > n$ olduqda tənliklər xətti asılıdırsa (yəni, tənliklər bir-birinin xətti kombinasiyasıdır), sistemin **sonsuz sayda həlli** ola bilər.

- **Misal:**
 $$
  A = 
  \begin{bmatrix}
  1 & 1 & 1 \\
  2 & 2 & 2 \\
  3 & 3 & 3
  \end{bmatrix}, \quad
  X = 
  \begin{bmatrix}
  x \\
  y \\
  z
  \end{bmatrix}, \quad
  AX = 0
 $$
  
  Burada $\text{rank}(A) = $, yəni bu tənliklər xətti asılıdır və sistemin **sonsuz sayda həlli** mövcuddur. Həlllər parametrik şəkildə verilə bilər, məsələn:

  $$
  x = -y - z
  $$

---

### **2. $m < n$ (Dəyişənlər çox, tənliklər az)**

Bu halda, dəyişənlər çox, tənliklər azdır. $A$ matrisinin ölçüləri $m \times n$-dir və burada da bəzi ehtimallar mövcuddur:

#### **(a) Sonsuz sayda həll (Xətti müstəqil tənliklər)**

Əgər tənliklər **xətti müstəqildirsə**, onda sistemin **sonsuz sayda həlli** olacaq. Çünki dəyişənlər çoxdur və yalnız bir neçə tənlik onların qiymətini məhdudlaşdırır.

- **Misal:**
  $$
  A =
  \begin{bmatrix}
  1 & 1 \\
  1 & -1
  \end{bmatrix}, \quad
  X = 
  \begin{bmatrix}
  x \\
  y
  \end{bmatrix}, \quad
  B = 
  \begin{bmatrix}
  5 \\
  3
  \end{bmatrix}, \quad
  AX = B
  $$

Bu halda tənliklər xətti müstəqildir və nəticədə **sonsuz sayda həll** mövcuddur. Həlllər parametrik şəkildə verilə bilər.

#### **(b) Həllin olmaması (Xətti asılı tənliklər)**

Əgər tənliklər **xətti asılı**dırsa, onda sistemin **həllinin olmaması** mümkündür. Tənliklər bir-birinin xətti kombinasiyası olarsa və nəticə vektoru uyğun gəlməzsə, həll yoxdur.

- **Misal:**
  $$
  A =
  \begin{bmatrix}
  1 & 1 & 1 \\
  2 & 2 & 2
  \end{bmatrix}, \quad
  B =
  \begin{bmatrix}
  5 \\
  10
  \end{bmatrix}
  $$

Bu matrisin sırası 1-dir ($\text{rank}(A) = 1$) və nəticə vektoru ilə uyğun gəlmir. Bu halda sistemin **həllinin olmaması** mümkündür.

---

### **3. Sistem tərs matrisin olub-olmaması**

Tərs matrisin mövcudluğu xətti tənliklər sisteminin həllinin olub-olmaması ilə birbaşa əlaqəlidir. Tərs matris yalnız **kvadrat matris** (yəni $m = n$) olduqda mövcuddur. Lakin, bu halda belə, matrisin **determinantı** sıfır olmamalıdır.

- **Matrisin tərsi** yalnız $\text{det}(A) \neq 0$ olduqda mövcuddur.
- **Sonsuz sayda həll** və ya **həllin olmaması** matrisin sırasına və determinantına bağlıdır.

---

### **Nəticə:**

- **$m > n$** (Tənliklər çox, dəyişənlər az): Xətti asılı tənliklər varsa, **sonsuz sayda həll** mümkündür. Xətti uyğunsuz tənliklər varsa, **həll yoxdur**.
- **$m < n$** (Dəyişənlər çox, tənliklər az): Xətti müstəqil tənliklər varsa, **sonsuz sayda həll** mövcuddur. Xətti asılı tənliklər varsa və nəticə uyğunsuzdursa, **həll yoxdur**.
