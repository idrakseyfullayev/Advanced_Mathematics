## Gauss üsulu (Məchulları ardıcıl yox etmə üsulu) (1)

**Gauss üsulu** (məchulları ardıcıl yox etmə üsulu) xətti cəbrdə tənliklər sistemini həll etmək üçün geniş istifadə olunan effektiv bir metoddur. Əsas məqsəd xətti tənliklər sistemini **üst üçbucaq matrisinə** çevirməkdir ki, bu da tənlikləri asanlıqla həll etməyə imkan verir.

---

### **Metodun Addımları:**

#### 1. **Giriş Sistemi:**
Xətti tənliklər sistemi matris şəklində yazılır:
$$
AX = B
$$
Burada:
- $A$ - əmsallar matrisi,
- $X$ - məchullar vektoru,
- $B$ - sağ tərəfdəki dəyərlər vektoru.

Matris şəklində:
$$
\begin{bmatrix}
a_{11} & a_{12} & \dots & a_{1n} \\
a_{21} & a_{22} & \dots & a_{2n} \\
\vdots & \vdots & \ddots & \vdots \\
a_{n1} & a_{n2} & \dots & a_{nn}
\end{bmatrix}
\begin{bmatrix}
x_1 \\
x_2 \\
\vdots \\
x_n
\end{bmatrix}
=
\begin{bmatrix}
b_1 \\
b_2 \\
\vdots \\
b_n
\end{bmatrix}
$$

#### 2. **Matrisin Üst Üçbucaq Formasına Gətirilməsi:**
- **Əsas Sətir Əməliyyatları**:
  - Birinci sətir seçilir və digər sətirlərdə həmin sütundakı əmsalları $0$-a çevirmək üçün istifadə olunur.
  - Daha sonra digər sətirlərdə də eyni əməliyyat təkrarlanır.
  - Məqsəd, aşağıdakı kimi **üst üçbucaq matris** forması əldə etməkdir:
  $$
  \begin{bmatrix}
  a_{11} & a_{12} & \dots & a_{1n} \\
  0 & a_{22}' & \dots & a_{2n}' \\
  \vdots & \vdots & \ddots & \vdots \\
  0 & 0 & \dots & a_{nn}'
  \end{bmatrix}
  $$

#### 3. **Geri Əvəzetmə:**
Üst üçbucaq matrisindən başlayaraq məchullar geriyə doğru həll edilir:
1. Sonuncu tənlikdən ($x_n$) başlayaraq məchul tapılır.
2. Tapılan məchul yuxarı tənliklərdə əvəzetmə edilərək digər məchullar hesablanır.

---

### **Misal:**

Tənliklər sistemi:
$$
2x + y - z = 8 \\
-3x - y + 2z = -11 \\
-2x + y + 2z = -3
$$

#### Addım 1: Əmsallar Matrisi Yazılır
$$
\begin{bmatrix}
2 & 1 & -1 & | & 8 \\
-3 & -1 & 2 & | & -11 \\
-2 & 1 & 2 & | & -3
\end{bmatrix}
$$

#### Addım 2: Sıfırlama Əməliyyatı (Üst Üçbucaq Forması)
- Birinci sütunun altındakı əmsalları $0$-a çevirmək üçün:
  - İkinci sətr: $R_2 = R_2 + \frac{3}{2}R_1$
  - Üçüncü sətr: $R_3 = R_3 + R_1$

Nəticədə:
$$
\begin{bmatrix}
2 & 1 & -1 & | & 8 \\
0 & 0.5 & 0.5 & | & 1 \\
0 & 2 & 1 & | & 5
\end{bmatrix}
$$

- Daha sonra ikinci sütunun altındakı əmsalları sıfırlamaq üçün əməliyyatlar davam etdirilir.

#### Addım 3: Geri Əvəzetmə
Üst üçbucaq matrisindən başlayaraq $x_3$, $x_2$, $x_1$ tapılır.

---

### **Üstünlükləri:**
- Effektiv və sistematikdir.
- Kompüterlə hesablama üçün əlverişlidir.
  
### **Çatışmazlıqları:**
- Çox böyük matrislərdə hesablama yükü arta bilər.
- Əgər determinant $0$-a bərabərdirsə ($A$ təkdirdirsə), həll mövcud olmur.
