**Asimptot** (Asymptote) riyaziyyatda bir funksiyann qrafiki il laqli bir anlaydr v myyn bir xtt yaxnlaan, amma bu xtti he vaxt kemyn bir yri n istifad edilir. 

Baqa szl, **asimptot**, funksiyann qrafikinin myyn bir nqty yaxlatqca daha da yaxnlad, lakin he vaxt bu xtti ksimyn bir xtdir.

Asimptotlar sasn  formada olur:

1. **Yatay Asimptot (Horizontal Asymptote):**
   - Bu tip asimptotda funksiya $x \to \infty $ (v ya $ x \to -\infty $) yaxnladqca myyn bir sabit dyr (yni, $ y = c $) yaxnlar.
   - Msln, $ f(x) = \frac{1}{x} $ funksiyasnn $ y = 0 $ olan yatay asimptotu vardr. $ x \to \infty $ olduqda $ f(x) \to 0 $.
   
   **Nmun:**
  $$
   f(x) = \frac{1}{x}
   $$
   $ x \to \infty $ olduqda \( f(x) $ sfra yaxnlar, amma he vaxt sfr olmaz.

2. **aquli Asimptot (Vertical Asymptote):**
   - Bu tip asimptotda funksiya myyn bir $ x = c $ nqtsin yaxnladqca sonsuzlua doru getmkddir. Bu halda, $ x = c $ aquli asimptot xtti olur.
   - Msln, $ f(x) = \frac{1}{x-2} $ funksiyasnn $ x = 2 $ nqtsind aquli asimptotu vardr, nki bu nqty yaxnladqda funksiyann dyri sonsuzlua gedir.
   
   **Nmun:**
  $$
   f(x) = \frac{1}{x-2}
   $$
   Bu funksiya $ x = 2 $ nqtsind aquli asimptot yaradr, nki $ x \to 2 $ olduqda $ f(x) $ sonsuz olur.

3. **yri Asimptot (Oblique Asymptote / Slant Asymptote):**
   - Bu, xsusil bir funksiya yri asimptot il tyin edildikd grlr. yri asimptotlar, funksiyann qrafiki myyn bir x-in sonsuza yaxnlad zaman bir doruya yaxnlar, amma bu doru n aquli, n d yatay olur.
   - yri asimptotlar, n ox polinom funksiyalarnda, xsusil qismn blnm zaman rast glinir.
   
   **Nmun:**
   $$
   f(x) = \frac{x^2 + 1}{x}
   $$
   Bu funksiyann $ x \to \infty $ olduu zaman $ y = x $ bir yri asimptotu vardr.

### Asimptotlarn Taplmas:
Asimptotlar tapmaq n funksiyann limitlrin baxmaq lazmdr:
- **Yatay asimptot n**: Funksiyann $ x \to \infty $ v $ x \to -\infty $ limitlrin baxlr.
- **aquli asimptot n**: Funksiyan myyn bir $ x = c $ nqtsin yaxnladrrq v orada funksiyann qiymtinin sonsuz olub olmadn yoxlayrq.
- **yri asimptot n**: Funksiyann uzun mddtli davrann tdqiq edirik v gr yri bir xtt yaxnlarsa, bu asimptot olur.

**Qsaca**: Asimptotlar, bir funksiya n o qdr vacibdir ki, funksiyann qrafiki he vaxt bu xtlri kems d, onlar funksiyann uzun mddtli davrann gstrir.
