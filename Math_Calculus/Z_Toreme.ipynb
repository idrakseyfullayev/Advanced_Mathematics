## $x = -2$ nöqtəsində $f(x) = x^2$ funksiyasının törəməsi


Sənin dediyin məsələdə çox maraqlı bir məqam var, gəlin dəqiqləşdirək:

Funksiya **$f(x) = x^2$** və onun törəməsi **$f'(x) = 2x$**-dir.

Bu o deməkdir ki, istənilən nöqtədə bu funksiyanın dəyişmə sürəti (yəni maillik) **$2x$** olur.

---

İndi sən deyirsən:

- **$x = -2$** nöqtəsində törəmə **$f'(-2) = 2*(-2) = -4$** olur.
- Bu nə deməkdir?

---

### Törəmənin mənası:
Törəmə funksiyanın qrafikində toxunan xəttin meyilliyini (slope) göstərir.  
**$f'(-2) = -4$** o deməkdir ki, **$x$** koordinatında 1 vahid sağa (artım) getdikdə, **$y$** koordinatı təxmini **4 vahid aşağı** düşür.

Yəni **bəli**, sənin dediyin doğrudur:

> **1 vahid artım üçün 4 vahid azalma baş verir** həmin nöqtədə (təqribən). Bu lokal (yaxın) təqribidir.

---

Əgər istəsən bu qrafiki çəkib vizual da göstərə bilərəm.

## $x = -4$ nöqtəsində $f(x) = x^2$ funksiyasının törəməsi

Çox gözəl sualdır! Gəlin baxaq:

Verilmiş funksiya:  
**$f(x) = x^2$**

Törəmə:  
**$f'(x) = 2x$**

---

İndi **$x = -4$** nöqtəsində baxaq:

**$f'(-4) = 2 * (-4) = -8$**

Bu nə deməkdir?

### Törəmənin mənası:
Bu nəticə deyir ki:

- **$x$** 1 vahid artdıqda (məsələn $-4$-dən $-3$-ə),
- **$f(x)$** təqribən **8 vahid azalır**.

Yəni qrafik həmin nöqtədə **çox kəskin enişlidir** (çox aşağıya doğru gedir).

---

### Qısa desək:
**$x = -4$** nöqtəsində:

- Törəmə **$-8$**-dir,
- Bu da deməkdir ki, **1 vahid artım üçün təxminən 8 vahid azalma** olur (lokal yaxınlaşma ilə).

Əgər istəyirsənsə, bunu real nöqtələrlə müqayisə edək? Məsələn,  
**$f(-4) = 16$**,  
**$f(-3.9) = (-3.9)^2 = 15.21$**,  
Bu zaman **$\Delta x = 0.1$**,  
**$\Delta y \approx -0.79$**,  
yəni dəyişmə sürəti təxminən **$-7.9$**, bu da törəmə **$-8$**-ə yaxındır.

İstəsən bu cür nümunələrlə davam edə bilərik.

## $x = 4$ nöqtəsində $f(x) = x^2$ funksiyasının törəməsi

Gözəl! İndi gəlin **$x = 4$** nöqtəsində **$f(x) = x^2$** funksiyasına baxaq.

---

### 1. Törəmə nədir?

**$f(x) = x^2$**  
**$f'(x) = 2x$**

---

### 2. **$x = 4$** nöqtəsində:

**$f'(4) = 2 * 4 = 8$**

---

### 3. Bu nə deməkdir?

- Bu o deməkdir ki, **$x$** 1 vahid artdıqda (məsələn $4$-dən $5$-ə), **$y$** təqribən **8 vahid artır**.
- Yəni qrafik bu nöqtədə **yuxarıya doğru 8 vahid meyllidir**.
- Törəmə **müsbət** olduğu üçün bu nöqtədə funksiyanın qrafiki **artandır** və **getdikcə daha sürətlə artır**.

---

### Sadə dillə:

- **$x = 4$** nöqtəsində 1 vahidlik artım üçün **təqribən 8 vahid $y$ artımı** baş verir.

Əgər istəsən, $f(4)$ və $f(5)$ hesablaya bilərik:

- $f(4) = 4^2 = 16$  
- $f(5) = 25$  
- $\Delta x = 1$, $\Delta y = 9$ → bu da göstərir ki, **real artım 9**-dur (amma orta meyl **8** idi, çünki törəmə yaxınlaşmadır).

Əlavə sual varsa, davam edək!

## $x = 4$ nöqtəsində 4 vahid artım üçün $f(x) = x^2$ funksiyasının törəməsi


Çox gözəl düşünürsən! İndi gəlin bu sualını addım-addım anlayaq:

---

### Verilmişdir:

Funksiya: **$f(x) = x^2$**  
Törəmə: **$f'(x) = 2x$**

**$x = 4$** nöqtəsində **$f'(4) = 8$**

Bu **törəmə** bizə **lokal dəyişmə sürətini** verir, yəni **çox kiçik dəyişikliklər üçün yaxınlaşmadır (approximation)**.

---

### İndi baxaq:  
**$4$ vahid artım** üçün nə qədər **$y$** dəyişər?

#### Törəmə baxımından:

Törəmə **$8$** olduğuna görə, **$1$ vahid artım üçün ≈ $8$ vahid $y$ artımı** olacaqsa,  
**$4$ vahid artım üçün təqribən:**  
$$
\Delta y \approx 8 \times 4 = 32
$$

Yəni:

> **$x = 4$** nöqtəsində, **$4$ vahid artım üçün təqribən $32$ vahid $y$ artımı** gözlənilir.

---

### Gəlin bunu dəqiqliklə yoxlayaq (real dəyişmə ilə):

- **$f(4) = 16$**  
- **$f(8) = 64$**  
- **$\Delta x = 4$**, **$\Delta y = f(8) - f(4) = 64 - 16 = 48$**

Yəni **real dəyişmə $48$-dir**, amma **törəmə ilə yaxınlaşma $32$** idi.

---

### Nəticə:

- **Törəmə lokal yaxınlaşmadır** (sadəcə yaxın nöqtələr üçün doğrudur),
- **Böyük dəyişikliklərdə ($4$ vahid və daha çox)** törəmə **dəqiq nəticə vermir**, sadəcə təxmini istiqamət göstərir.

İstəsən bunu qrafiklə də göstərə bilərəm. Davam edək?

## Qrafiklə izah

Əla, davam edək!

---

### 1. **Qrafiklə izah (vizual düşüncə)**

Funksiya: **$f(x) = x^2$**  
Bu paraboladır və aşağıdan yuxarıya doğru açılır.  
- **$x = 0$** nöqtəsində minimumdadır  
- **$x > 0$** olduqda getdikcə daha **sürətlə artır**

---

### 2. **Törəmə və toxunan düz xətt**

**$x = 4$** nöqtəsində törəmə **$8$**-dir.  
Bu o deməkdir ki, həmin nöqtədəki **toxunan düz xəttin meyili $8$**-dir.  
Yəni, bu xətt "funksiyanı təqribən izləyir", amma yalnız **$x = 4$**-ə **yaxın** olan nöqtələrdə!

---

### 3. **Törəmə ilə təxmini artım (linear approximation)**

Əgər biz **$f(x) = x^2$** funksiyasını təqribən təhlil etsək:

#### Törəməyə əsaslanan xətti yaxınlaşma düsturu:

$$
f(x + h) \approx f(x) + f'(x) \cdot h
$$

Burada:  
- $x = 4$  
- $f(4) = 16$  
- $f'(4) = 8$  
- $h = 4$ (yəni 4 vahid artım)

$$
f(8) \approx 16 + 8 \cdot 4 = 16 + 32 = 48
$$

Bu **təxmini nəticədir**, amma **real $f(8) = 64$** idi.

---

### 4. **Nəticə və fərq**

| Yanaşma         | Hesablama     | Nəticə |
|----------------|---------------|--------|
| Törəmə ilə     | $$16 + 8 \times 4$$      | $$48$$     |
| Real           | $$8^2 = 64$$       | $$64$$     |
| Fərq           |               | $$+16$$    |

Bu fərq **parabolanın əyriliyindən** yaranır:  
- Törəmə düz xətt kimi davranır,  
- Amma parabolanın qrafiki əyilir, ona görə də real artım **daha böyükdür**.

---

İstəyirsənsə, bu izahı qrafiklə göstərim: həm **funksiya**, həm də **toxunan düz xətt** bir yerdə.

Hazıram çəkməyə?

## $f(x) = x²$ funksiyasını və $x=4$ nöqtəsindəki toxunan xəttini çəkir

```python
import numpy as np
import matplotlib.pyplot as plt

# Funksiya təyini
def f(x):
    return x**2

# Toxunan xətt funksiyası (x=4 nöqtəsində)
def tangent_line(x):
    return 8*(x - 4) + f(4)  # f'(4)=8, f(4)=16

# Veri nöqtələri
x_vals = np.linspace(0, 8, 400)
x_point = 4
x_tangent_point = 8  # Toxunan xəttin x=8-dəki dəyəri

# Qrafik yaratma
plt.figure(figsize=(8, 6))

# Funksiya çəkilişi
plt.plot(x_vals, f(x_vals), label='f(x) = x²', color='blue')

# Toxunan xətt çəkilişi
plt.plot(x_vals, tangent_line(x_vals), '--', label='Toxunan xətt (x=4)', color='orange')

# Əhəmiyyətli nöqtələr
plt.scatter([x_point, x_tangent_point], 
            [f(x_point), f(x_tangent_point)], 
            color='red', zorder=5)
plt.scatter([x_tangent_point], 
            [tangent_line(x_tangent_point)], 
            color='green', zorder=5)

# Şaquli xətlər
plt.axvline(x=x_point, color='gray', linestyle=':', alpha=0.5)
plt.axvline(x=x_tangent_point, color='gray', linestyle=':', alpha=0.5)

# Mətn əlavələri
plt.text(x_point, f(x_point)+5, f'f(4) = {f(x_point)}', ha='center')
plt.text(x_tangent_point, f(x_tangent_point)+5, f'Real: f(8) = {f(x_tangent_point)}', ha='center')
plt.text(x_tangent_point, tangent_line(x_tangent_point)-5, f'Toxunan yaxınlaşma: {tangent_line(x_tangent_point)}', ha='center')

plt.text(4, -10, '4', ha='center')  # x=4 etiketi
plt.text(8, -10, '8', ha='center')  # x=8 etiketi

# Ox və başlıq ayarları
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('f(x) = x² və x=4 nöqtəsindəki toxunan xətt')
plt.legend()
plt.grid(True)
plt.xticks([0, 2, 4, 6, 8])
plt.ylim(-15, 70)

plt.show()
```


Yuxarıdakı qrafikdə:

- **Mavi xətt** – əsas funksiyamızdır: $f(x) = x^2$  
- **Narıncı xətt (kəsik xətt)** – $x = 4$ nöqtəsində çəkilmiş **toxunan düz xətt**, meyili $8$-dir.  
- **Qırmızı nöqtə** – $x = 4$, yəni toxunma nöqtəsi: $f(4) = 16$  
- **Yaşıl nöqtə** – **real qiymət**: $f(x) = x^2$ $\:\Rightarrow\:$ $f(8) = 64$  
- **Bənövşəyi nöqtə** – **törəmə ilə yaxınlaşma**:

$$f(x) = f(x_0) + f'(x_0)(x - x_0)$$
 
$$
f(x) = f(4) + f'(4)(x - 4) = 16 + 8(x - 4)
$$

$$
f(x) = 16 + 8x - 32 = 8x - 16
$$
  
$$
\boxed{f(x) = 8x - 16}
$$
 
$$x = 8$$ üçün:
$$
f(x) = 8 \cdot 8 - 16 = 64 - 16 = \boxed{48}
$$


Göründüyü kimi, **törəmə ilə alınan qiymət (48)**, əyridən aşağıda qalır, çünki əyrilik nəzərə alınmır — bu da **"linear approximation"** metodunun xüsusiyyətidir.

Başqa nöqtələr üçün də etmək istəyirsənsə, deyə bilərsən!

## $x = 4$ nöqtəsində 1 vahid artım üçün $f(x) = x^2$ funksiyasının özündə



Çox gözəl sualdır — indi **funksiyanın özündə**, yəni **f(x) = x²** üçün baxaq:

---

### Sual:

**f(x) = x²** funksiyasında, **x = 4** nöqtəsində **1 vahid artım** üçün **y** nə qədər artır?

---

### Addım-addım:

1. **Başlanğıc nöqtə:**  
   x = 4  
   f(4) = 4² = 16

2. **1 vahid artım:**  
   x = 5  
   f(5) = 5² = 25

3. **Artım miqdarı:**

$$
\Delta y = f(5) - f(4) = 25 - 16 = \boxed{9}
$$

---

### Yəni:

> **x = 4** nöqtəsində **1 vahidlik artım**, **y-də 9 vahid artıma** səbəb olur.

---

### Maraqlı məqam:

- Halbuki **törəmə (f'(4) = 8)** bizə **təxmini 8 vahid artım** deyirdi.
- Amma **real artım 9** oldu — çünki **f(x) = x²** parabolik funksiyadır və əyrilik təsir edir.

---

## Toxunanın tənliyi nəyi hesablayır?



### Toxunanın tənliyi nədir?
Verilmiş bir nöqtədə funksiyaya **toxunan düz xəttin** tənliyidir.

---

### Ümumi formula:
$$
y = f(a) + f'(a)(x - a)
$$

Burada:
- $a$ – toxunduğun nöqtənin $x$-dəyəri,
- $f(a)$ – həmin nöqtədə funksiyanın y-dəyəri,
- $f'(a)$ – həmin nöqtədə funksiyanın törəməsi (meyl, yəni yamac),
- $x$ – ümumi dəyişən.

---

### Bu tənlik **nəyi hesablayır?**
Bu tənlik, **həmin nöqtədə funksiyanı düz xəttlə yaxınlaşdırır.** Yəni:
- Verilmiş nöqtədəki **toxunanın özünü** tapırıq,
- Bu tənliklə, yaxın nöqtələrdə funksiyanın dəyərini **təxmini** hesablaya bilərik.

---

### Məsələn:
Funksiya: $f(x) = x^2$  
Törəmə: $f'(x) = 2x$  
Nöqtə: $x = 2$

$$
f(2) = 4,\quad f'(2) = 4
$$

Toxunanın tənliyi:
$$
y = 4 + 4(x - 2)
\quad \text{(bu, toxunanın tənliyidir)}
$$

---

### Nəticə:
Toxunanın tənliyini yazmaqla:
- Verilmiş nöqtədə funksiyaya toxunan düz xətti tapırıq,
- Bu düz xəttlə funksiyanın yaxın dəyərlərini **təxmini** hesablamaq mümkündür,
- Bu üsul, **təxmini hesablamalarda və diferensial yanaşmalarda** çox istifadə olunur.

İstəsən, real nümunə ilə qrafik üzərində də izah edə bilərəm.

## $f(x) = x^2$ funksiyası üçün $x = 2$ nöqtəsində **toxunanın tənliyi ilə dəyər hesablama**

$$
f'(x) = 2x

$$
f'(2) = 2 \cdot 2 = 4

$$

$$

$$
\boxed{y = 4x - 4}

Bu, $x = 2$ nöqtəsində $f(x) = x^2$ funksiyasına toxunan düz xəttin tənliyidir. İndi bu düz xətt, $x = 2$-yə yaxın dəyərlər üçün funksiyanı **təxmini** təsvir edir.

### **$x = 2$** nöqtəsindən **0.1 artım**, yəni $x = 2.1$ üçün **toxunanın tənliyi ilə təxmini qiymət** nə olacaq?

**4.4 (təxmini)** yerinə **4.41 (əsl)** — çox kiçik fərq.

Bu, göstərir ki:
> **Toxunanın tənliyi kiçik artımlar üçün funksiyanın proqnozunu çox yaxşı təxmin edir.**

# Ümumiləşdirmə

## $f(x) = x^2$ funksiyası, $f'(x) = 2x$ törəməsi, $y \approx f(a) + f'(a) (x - a)$ toxunanın tənliyi nəyi hesablayır



Əla, gəlin sualları **sistemli şəkildə cavablandıraq** və sonda **Python qrafikləri** ilə də göstərək:

---

### ✅ **1) $f(x) = x^2$ funksiyası nəyi hesablayır?**

> Bu funksiyanın **öz-özlüyündə hesabladığı dəyər**, `x` dəyişənin kvadratıdır.

- Məsələn, `x = 4` → $f(4) = 16$

> Yəni bu funksiya **ənənəvi qiymət funksiyası**dır. Hər bir `x` üçün onun **fayda**, **xərc**, **həcm** və s. ola bilər – konkret məsələyə görə dəyişir.

---

### ✅ **2) $f'(x) = 2x$** törəməsi nəyi hesablayır?

> Bu törəmə **funksiyanın verilmiş nöqtədəki dəyişmə sürətini**, yəni **mailliyini** (hansı sürətlə dəyişdiyini) göstərir.

- Məsələn, `x = 4` → $f'(4) = 8$
- Bu o deməkdir ki, `x` dəyişəndə `y` orta hesabla **8 dəfə sürətli** dəyişir.

> ➡️ 1 vahid `x` artanda, təxminən 8 vahid `y` artır.

---

### ✅ **3) Toxunanın tənliyi nəyi hesablayır?**

> Toxunanın tənliyi — bu funksiyanın müəyyən bir nöqtədəki **lokal proqnozudur**.

> Yəni, əyri yerinə **birbaşa düz xəttlə yaxınlaşma**dır.
- Toxunanın tənliyi:  
  $$
  y \approx f(a) + f'(a) (x - a)
  $$
  burada $a = 4$, $f(4) = 16$, $f'(4) = 8$

- Yəni toxunanın tənliyi:
  $$
  y = 16 + 8(x - 4)
  $$
  
  $$
  y = 8x - 16
  $$
  
- Bu **lokal təxmini dəyərlər** verir, məsələn, `x = 8` üçün:
  $$
  y \approx 8x - 16 = 64 -16 = 48
  $$

➡️ Amma əslində $f(8) = 64$, yəni artıq bu təxmin bir qədər yanılır – çünki toxunan düz xəttdir, funksiya isə əyridir.

---

### ✨ Nəticə (x = 4 nöqtəsində 4 vahid artım üçün):

|   | Dəyər | Nəyi göstərir |
|---|-------|----------------|
| $$f(4)$$ | $16$ | Əsas funksiyada $x=4$ üçün y |
| $$f(8)$$ | $64$ | Həqiqi dəyər |
| Törəmə $f'(4) = 8$ | — | Həmin nöqtədə artım sürəti |
| Toxunanla proqnoz | $$ 8 \cdot 8 - 16 = 48$$ | Təxmini qiymət |

Aşağıdakı kod bunları əks etdirir:

1. $f(x) = x^2$ funksiyasının əyrisi  
2. $x = 4$ nöqtəsində toxunanın tənliyi  
3. Törəmə funksiyası $f'(x) = 2x$

```python
import numpy as np
import matplotlib.pyplot as plt

# Funksiya təyini
def f(x):
    return x**2

# Törəmə funksiyası
def derivative(x):
    return 2 * x

# Toxunan xətt funksiyası (x=4 nöqtəsində)
def tangent_line(x):
    return 8*(x - 4) + f(4)  # f'(4)=8, f(4)=16

# Ümumi x dəyərləri
x_vals = np.linspace(0, 8, 400)
x_point = 4
x_tangent_point = 8  # Toxunan xəttin x=8-dəki dəyəri

### 1. f(x) və toxunan xətt qrafiki ###
plt.figure(figsize=(8, 6))

# Funksiya f(x) = x^2
plt.plot(x_vals, f(x_vals), label='f(x) = x²', color='blue')

# Toxunan xətt
plt.plot(x_vals, tangent_line(x_vals), '--', label='Toxunan xətt - f(x) = 8x - 16', color='orange')

# Əhəmiyyətli nöqtələr
plt.scatter([x_point, x_tangent_point], 
            [f(x_point), f(x_tangent_point)], 
            color='red', zorder=5)
plt.scatter([x_tangent_point], 
            [tangent_line(x_tangent_point)], 
            color='green', zorder=5)

# Şaquli xətlər
plt.axvline(x=x_point, color='gray', linestyle=':', alpha=0.5)
plt.axvline(x=x_tangent_point, color='gray', linestyle=':', alpha=0.5)

# Mətnlər
plt.text(x_point, f(x_point)+5, f'f(4) = {f(x_point)}', ha='center')
plt.text(x_tangent_point, f(x_tangent_point)+5, f'Real: f(8) = {f(x_tangent_point)}', ha='center')
plt.text(x_tangent_point, tangent_line(x_tangent_point)-5, f'Toxunan yaxınlaşma: {tangent_line(x_tangent_point)}', ha='center')
plt.text(4, -10, '4', ha='center')
plt.text(8, -10, '8', ha='center')

# Ayarlar
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('f(x) = x² və x=4 nöqtəsindəki toxunan xətt')
plt.legend()
plt.grid(True)
plt.xticks([0, 2, 4, 6, 8])
plt.ylim(-15, 70)

plt.show()


### 2. Törəmə funksiyası f'(x) = 2x qrafiki ###
plt.figure(figsize=(8, 6))

x_vals = np.append(np.array([-2]), x_vals)

plt.plot(x_vals, derivative(x_vals), label="f'(x) = 2x", color='green')

# Əhəmiyyətli nöqtə
plt.scatter([x_point], [derivative(x_point)], color='red', zorder=5)

# Mətn
plt.text(x_point, derivative(x_point)+2, f"f'(4) = {derivative(x_point)}", ha='center')
plt.text(4, -2, '4', ha='center')

# Ayarlar
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.title("f'(x) = 2x (Törəmə funksiyasının qrafiki)")
plt.legend()
plt.grid(True)
plt.xticks([-2, 0, 2, 4, 6, 8])
plt.ylim(-5, 20)

plt.show()
```
