## H…ônd…ôsi ehtimal

> **T…ôrif:** H…ônd…ôsi ehtimal (geometrik ehtimal) ‚Äì **klassik ehtimalƒ±n** daha geni≈ü halƒ±dƒ±r v…ô n…ôtic…ôl…ôrin **sayla deyil, uzunluq, sah…ô v…ô ya h…ôcm kimi √∂l√ß√ºl…ôrl…ô** qiym…ôtl…ôndirildiyi hallarda istifad…ô olunur.

---

## üî∑ **H…ônd…ôsi ehtimal n…ôdir?**

> **T…ôrif:** Bir t…ôsad√ºfi hadis…ônin ehtimalƒ±, onun uyƒüun sah…ôsinin b√ºt√ºn m√ºmk√ºn sah…ôy…ô olan nisb…ôtidir:

$$
P(A) = \frac{\text{Hadis…ôy…ô uyƒüun √∂l√ß√º (uzunluq, sah…ô, h…ôcm)}}{\text{M√ºmk√ºn b√ºt√ºn √∂l√ß√º}}
$$

---

## üìå **∆èsas anlayƒ±≈ülar**

| Anlayƒ±≈ü               | ƒ∞zah                                                         |
| --------------------- | ------------------------------------------------------------ |
| **N…ôtic…ô**            | T…ôsad√ºfi t…ôcr√ºb…ônin m√ºmk√ºn n…ôtic…ôsi.                         |
| **M√ºmk√ºn sah…ô**       | Hadis…ônin ba≈ü ver…ô bil…ôc…ôyi b√ºt√ºn sah…ô (m…ôs: b√ºt√ºn kvadrat). |
| **T…ôl…ôb olunan sah…ô** | Hadis…ônin ba≈ü verdiyi hiss…ô (m…ôs: dair…ônin i√ßi).             |
| **Ehtimal**           | T…ôl…ôb olunan sah…ônin b√ºt√ºn sah…ôy…ô nisb…ôti.                   |

---

## üßÆ **∆èsas d√ºstur:**

### ‚úÖ **H…ônd…ôsi ehtimalƒ±n √ºmumi formulu:**

$$
P(A) = \frac{\text{T…ôl…ôb olunan uzunluq / sah…ô / h…ôcm}}{\text{C…ômi m√ºmk√ºn uzunluq / sah…ô / h…ôcm}}
$$

---

## üìò **Misallar**

### üîπ **1. Uzunluqla baƒülƒ± misal:**

0-dan 10-a q…ôd…ôr bir n√∂qt…ô t…ôsad√ºfi se√ßilir. Bu n√∂qt…ônin 3 il…ô 7 arasƒ±nda olma ehtimalƒ±:

$$
P = \frac{7 - 3}{10 - 0} = \frac{4}{10} = 0.4
$$

---

### üîπ **2. Sah…ô il…ô baƒülƒ± misal:**

1√ó1 √∂l√ß√ºl√º kvadratƒ±n i√ßind…ôn t…ôsad√ºfi n√∂qt…ô se√ßilir. Bu n√∂qt…ônin kvadrat daxilind…ô √ß…ôkilmi≈ü radiusu 0.5 olan dair…ôy…ô d√º≈üm…ô ehtimalƒ±:

$$
P = \frac{\pi \cdot r^2}{1^2} = \frac{\pi \cdot 0.5^2}{1} = \frac{\pi}{4} \approx 0.785
$$

---

### üîπ **3. H…ônd…ôsi ehtimalda klassik yana≈üma m√ºmk√ºn deyil:**

Klassik ehtimalda n…ôtic…ôl…ôrin **sayƒ±nƒ±** sayƒ±rƒ±q:

$$
P = \frac{\text{∆èlveri≈üli n…ôtic…ôl…ôrin sayƒ±}}{\text{M√ºmk√ºn n…ôtic…ôl…ôrin sayƒ±}}
$$

Amma hend…ôsi ehtimalda **√∂l√ß√º** (uzunluq, sah…ô v…ô s.) …ôsas g√∂t√ºr√ºl√ºr.

---

## ≈û…ôrti Ehtimal



### üî∑ **≈û…ôrti ehtimal n…ôdir?**

> **T…ôrif:** **≈û…ôrti ehtimal**, bir hadis…ônin, **ba≈üqa bir hadis…ônin ba≈ü verdiyi m…ôlum olduqdan sonra** ba≈ü verm…ô ehtimalƒ±dƒ±r.


üìò Y…ôni, **"B ba≈ü veribs…ô, A-nƒ±n ba≈ü verm…ô ehtimalƒ± n…ô q…ôd…ôrdir?"** sualƒ±na cavab verir.

---

### üß† **∆èsas anlayƒ±≈ülar**

| Anlayƒ±≈ü                | ƒ∞zah                                         |
| ---------------------- | -------------------------------------------- |
| **Hadis…ô A**           | Maraqlƒ± olan hadis…ô (n…ôtic…ô)                 |
| **Hadis…ô B**           | M…ôlum olan hadis…ô (≈ü…ôrt kimi q…ôbul olunur)   |
| **K…ôsi≈üm…ô $A \cap B$** | A v…ô B hadis…ôl…ôrinin eyni vaxtda ba≈ü verm…ôsi |
| **≈û…ôrti ehtimal**      | B ba≈ü veribs…ô, A-nƒ±n ba≈ü verm…ô ehtimalƒ±      |


---

### üßÆ **∆èsas d√ºstur:**

$$
\boxed{P(A \mid B) = \frac{P(A \cap B)}{P(B)}}
\quad \text{(…ôg…ôr } P(B) > 0 \text{)}
$$

üìå Oxunu≈üu: ‚Äú**P A ≈ü…ôrti B**‚Äù v…ô ya "A-nƒ±n B ba≈ü verdikd…ôn sonra ehtimalƒ±" v…ô ya "B ba≈ü ver…ônd…ôn sonra A-nƒ±n ehtimalƒ±".

---

### üí° **Qeyd:**

* ≈û…ôrti ehtimal √º√ß√ºn **≈ü…ôrt qoyulan hadis…ônin ehtimalƒ± sƒ±fƒ±r olmamalƒ±dƒ±r**:

  $$
  P(B) > 0
  $$

---

### üîÅ **∆èlaq…ôli d√ºstur:**

### üîπ **K…ôsi≈üm…ô d√ºsturu (√ßarpma d√ºsturu):**

$$
P(A \cap B) = P(A \mid B) \cdot P(B)
$$

v…ô ya

$$
P(A \cap B) = P(B \mid A) \cdot P(A)
$$

---


### üìò **Sad…ô izah:**

**Normal ehtimalda**: ‚ÄúA ba≈ü ver…ô bil…ôr?‚Äù
**≈û…ôrti ehtimalda**: ‚ÄúB artƒ±q ba≈ü verib, indi A-nƒ±n ba≈ü verm…ô ≈üansƒ± n…ô q…ôd…ôrdir?‚Äù

---



### üßÆ **Bayes Teoremi (…ôlav…ô):**

∆èg…ôr lazƒ±m olarsa, Bayes teoremi d…ô ≈ü…ôrti ehtimaldan istifad…ô edir:

$$
\boxed{
P(A \mid B) = \frac{P(B \mid A) \cdot P(A)}{P(B)}
}
$$

---




### üß† **Misal 1:**

Bir sinifd…ô:

* 10 oƒülan
* 20 qƒ±z var.

6 n…ôf…ôr  eyn…ôk taxƒ±r, onlardan 4-√º qƒ±zdƒ±r.

üîπ **Sual**: Eyn…ôk taxan ≈üagirdin **qƒ±z** olma ehtimalƒ± n…ôdir?

Bu ≈ü…ôrti ehtimaldƒ±r: $P(\text{qƒ±z} \mid \text{eyn…ôk})$

$$
P(\text{qƒ±z} \mid \text{eyn…ôk}) = \frac{P(\text{qƒ±z v…ô eyn…ôk})}{P(\text{eyn…ôk})} = \frac{4/30}{6/30} = \frac{4}{6} = \boxed{\frac{2}{3}}
$$





∆èla! G…ôlin eyni m…ôs…ôl…ôni **Bayes Teoremi** il…ô h…ôll ed…ôk.
Y…ôni:

> **Eyn…ôk taxan ≈üagirdin qƒ±z olma ehtimalƒ± n…ôdir?**
> Y…ôni: $P(\text{qƒ±z} \mid \text{eyn…ôk})$

---

### üß† **Misal 1:** ‚úÖ Bayes Teoremi il…ô H…ôll:

Bayes teoremi bel…ôdir:

$$
P(A \mid B) = \frac{P(B \mid A) \cdot P(A)}{P(B)}
$$

Bu m…ôs…ôl…ô √º√ß√ºn:

* **A = qƒ±z**
* **B = eyn…ôk**

$$
P(\text{qƒ±z} \mid \text{eyn…ôk}) = \frac{P(\text{eyn…ôk} \mid \text{qƒ±z}) \cdot P(\text{qƒ±z})}{P(\text{eyn…ôk})}
$$



### üî¢ M…ôlumatlar:

* √úmumi ≈üagird sayƒ±: $10 + 20 = 30$
* Qƒ±zlarƒ±n sayƒ±: $20 \Rightarrow P(\text{qƒ±z}) = \frac{20}{30}$
* Eyn…ôk taxan ≈üagirdl…ôrin sayƒ±: 6 ‚áí $P(\text{eyn…ôk}) = \frac{6}{30}$
* Qƒ±zlardan 4-√º eyn…ôk taxƒ±r ‚áí $P(\text{eyn…ôk} \mid \text{qƒ±z}) = \frac{4}{20}$



### üîÅ ∆èv…ôzl…ôy…ôk:

$$
P(\text{qƒ±z} \mid \text{eyn…ôk}) = \frac{\frac{4}{20} \cdot \frac{20}{30}}{\frac{6}{30}}
$$

Sad…ôl…ô≈üdir…ôk:

* Yuxarƒ±: $\frac{4}{20} \cdot \frac{20}{30} = \frac{4}{30}$
* A≈üaƒüƒ±: $\frac{6}{30}$

$$
P(\text{qƒ±z} \mid \text{eyn…ôk}) = \frac{\frac{4}{30}}{\frac{6}{30}} = \frac{4}{6} = \frac{2}{3}
$$



### ‚úÖ N…ôtic…ô:

$$
\boxed{P(\text{qƒ±z} \mid \text{eyn…ôk}) = \frac{2}{3}}
$$

Y…ôni **eyn…ôk taxan bir ≈üagirdin qƒ±z olmasƒ± ehtimalƒ± 2/3-d√ºr**.

## Bayes Teoremi: Misal



### üß† **Bayes Teoremi ‚Äî N√ºmun…ô:**

### üìò **V…ôziyy…ôt:**

Bir ≈ü…ôh…ôrd…ô:

* ∆èhalinin **1%-i** m√º…ôyy…ôn bir x…ôst…ôliy…ô yoluxub.
* Bu x…ôst…ôliyi a≈ükarlayan test m√∂vcuddur:

  * **X…ôst…ô olan** ≈ü…ôxsi **90% ehtimalla d√ºzg√ºn tapƒ±r**. (True Positive)
  * **Saƒülam olan** ≈ü…ôxsi **95% ehtimalla d√ºzg√ºn tanƒ±yƒ±r**. (True Negative)

---

### üîπ **Sual:**

Test **m√ºsb…ôt √ßƒ±xarsa**, h…ômin ≈ü…ôxsin **h…ôqiq…ôt…ôn x…ôst…ô olma ehtimalƒ± n…ô q…ôd…ôrdir?**

Y…ôni biz…ô **$P(\text{X…ôst…ô} \mid \text{Test+})$** lazƒ±mdƒ±r.

---

### ‚ú≥Ô∏è **Veril…ônl…ôri yazƒ±rƒ±q:**

| Hadis…ô                  | Simvol                    | Ehtimal            |
| ----------------------- | ------------------------- | ------------------ |
| ≈û…ôxs x…ôst…ôdir           | $$D$$                       | $$P(D) = 0.01$$      |
| ≈û…ôxs saƒülamdƒ±r          | $$\neg D$$                  | $$P(\neg D) = 0.99$$ |
| Test m√ºsb…ôtdir (Test+)  | $$+$$                       | veril…ôc…ôk a≈üaƒüƒ±da  |
| X…ôst…ô ik…ôn test+ √ßƒ±xƒ±r  | $$P(+ \mid D) = 0.90$$      |                    |
| Saƒülam ik…ôn test+ √ßƒ±xƒ±r | $$P(+ \mid \neg D) = 0.05$$ |                    |

---

### ‚úÖ **Bayes d√ºsturu il…ô hesablayaq:**

$$
\boxed{
P(A \mid B) = \frac{P(B \mid A) \cdot P(A)}{P(B)}
}
$$

$$
P(D \mid +) = \frac{P(+ \mid D) \cdot P(D)}{P(+)}
$$

∆èvv…ôlc…ô **$P(+)$** tapƒ±rƒ±q (testin √ºmumilikd…ô m√ºsb…ôt √ßƒ±xma ehtimalƒ±):

$$
P(+) = P(+ \mid D) \cdot P(D) + P(+ \mid \neg D) \cdot P(\neg D)
$$

$$
= 0.90 \cdot 0.01 + 0.05 \cdot 0.99 = 0.009 + 0.0495 = 0.0585
$$

---

ƒ∞ndi …ôsas ehtimalƒ± hesablayaq:

$$
P(D \mid +) = \frac{0.90 \cdot 0.01}{0.0585} = \frac{0.009}{0.0585} \approx \boxed{0.1538} \approx \boxed{15.4\%}
$$

---

### üìå **N…ôtic…ô:**

Test **m√ºsb…ôt √ßƒ±xsa bel…ô**, ≈ü…ôxsin h…ôqiq…ôt…ôn x…ôst…ô olma ehtimalƒ± sad…ôc…ô **15.4%**-dir!

---

### üîé **Yorum:**

Bu n…ôtic…ô √ßox insan √º√ß√ºn t…ô…ôcc√ºbl√ºd√ºr. Amma s…ôb…ôbi budur ki:

* X…ôst…ôlik **nadir haldƒ±r (1%)**, y…ôni √ßox adam onsuz da saƒülamdƒ±r.
* Testin s…ôhv **saƒülam insana m√ºsb…ôt** verm…ôsi az da olsa (5%) √ßox insana t…ôsir edir (√ß√ºnki saƒülam √ßoxdur!).



## **N…ô √º√ß√ºn cavab yalnƒ±z 15.4% √ßƒ±xdƒ±** v…ô **n…ôtic…ônin niy…ô t…ô…ôcc√ºbl√º olduƒüunu** sad…ô dill…ô izahƒ±



### üîç Sualƒ±mƒ±z n…ô idi?

Test **m√ºsb…ôt** √ßƒ±xarsa, **≈ü…ôxsin h…ôqiq…ôt…ôn x…ôst…ô olma ehtimalƒ± n…ô q…ôd…ôrdir?**

> Cavab: t…ôxmin…ôn **15.4%**

---

### ‚ùì **B…ôs niy…ô bu q…ôd…ôr azdƒ±r? Test m√ºsb…ôtdirs…ô, niy…ô 90% deyil?**

---

### G…ôlin m…ôs…ôl…ôy…ô **saylarla** baxaq ki, tam aydƒ±n olsun:

T…ôs…ôvv√ºr et:

* ≈û…ôh…ôrd…ô **10.000 n…ôf…ôr** var.
* Onlarƒ±n **1%-i** x…ôst…ôdir ‚Üí Y…ôni **100 n…ôf…ôr** x…ôst…ôdir, **9900 n…ôf…ôr** saƒülamdƒ±r.

---

### üî¨ Test nec…ô i≈ül…ôyir?

### 1. **X…ôst…ô olanlara test ver…ôk:**

* 100 n…ôf…ôr x…ôst…ôdir.
* Onlarƒ±n **90%-i testl…ô d√ºzg√ºn tutulur** ‚Üí **90 n…ôf…ôrd…ô test m√ºsb…ôt √ßƒ±xƒ±r** (‚úÖ **doƒüru m√ºsb…ôt**)
* Qalan **10 n…ôf…ôr x…ôst…ô olsa da, test s…ôhv…ôn m…ônfi √ßƒ±xƒ±r** (‚ùå **yalan√ßƒ± m…ônfi**)

### 2. **Saƒülam olanlara test ver…ôk:**

* 9900 n…ôf…ôr saƒülamdƒ±r.
* Testin **5%-i s…ôhv…ôn m√ºsb…ôt √ßƒ±xƒ±r** ‚Üí

  $$
  0.05 \cdot 9900 = 495 \text{ n…ôf…ôrd…ô test s…ôhv…ôn m√ºsb…ôt √ßƒ±xƒ±r} \quad (\text{‚ùå yalan√ßƒ± m√ºsb…ôt})
  $$

---

### ‚ú≥Ô∏è **C…ômi m√ºsb…ôt √ßƒ±xanlar:**

* Doƒürudan x…ôst…ô olanlar: **90 n…ôf…ôr**
* Saƒülam olub testd…ô s…ôhv…ôn m√ºsb…ôt √ßƒ±xanlar: **495 n…ôf…ôr**

> Y…ôni **m√ºsb…ôt test n…ôtic…ôsi olanlar = 90 + 495 = 585 n…ôf…ôr**

---

### ‚úÖ **ƒ∞ndi ehtimal hesabƒ±:**

### üî∑ 1-ci Toplama Teoremi (Mutually Exclusive ‚Äì Qar≈üƒ±lƒ±qlƒ± istisna olan hadis…ôl…ôr)

∆èg…ôr **A** v…ô **B** hadis…ôl…ôri **eyni anda ba≈ü ver…ô bilm…ôzs…ô**, y…ôni **A ‚à© B = √ò** (k…ôsi≈üm…ôsi yoxdur), onda:

> Y…ôni: A v…ô ya B ba≈ü versin ehtimalƒ± = A-nƒ±n ehtimalƒ± + B-nin ehtimalƒ±

$$

$$

$$

### üî∑ 2-ci Toplama Teoremi (√úmumi hal ‚Äì k…ôsi≈üm…ôsi olan hadis…ôl…ôr)

∆èg…ôr **A v…ô B** hadis…ôl…ôrinin **k…ôsi≈üm…ôsi varsa**, y…ôni **eyni anda ba≈ü ver…ô bilirl…ôrs…ô**, onda:

$$

> K…ôsi≈üm…ô √ßƒ±xƒ±lƒ±r ki, t…ôkrar sayƒ±lmasƒ±n.

ƒ∞lk √∂nc…ô anlayƒ±≈ülarƒ± f…ôrql…ôndir…ôk:

| --------------------------- | -------------------------------------------------------------------- |

### üî∑ **Asƒ±lƒ± olmayan hadis…ôl…ôr √º√ß√ºn:**

∆èg…ôr A v…ô B **asƒ±lƒ± olmayan hadis…ôl…ôrdirs…ô**, onda:

### ‚úÖ **∆èg…ôr B ba≈ü vers…ô d…ô, verm…ôs…ô d…ô, A-nƒ±n ehtimalƒ± d…ôyi≈ümirs…ô:**

$$

$$

Bu o dem…ôkdir ki, A v…ô B **m√ºst…ôqil** (independent) hadis…ôl…ôrdir.

---

### üìå **Asƒ±lƒ± olmayan hadis…ôl…ôrin birlikd…ô ba≈ü verm…ôsi:**

∆èg…ôr A v…ô B asƒ±lƒ± deyils…ô:

$$
P(A \cap B) = P(A) \cdot P(B)
$$

---

### üî∑ Amma sualƒ±n …ôsas n√∂qt…ôsi:

### ‚û§ ‚Äú**Asƒ±lƒ± olmayan hadis…ôl…ôrd…ôn birinin ba≈ü verm…ôsi ehtimalƒ±**‚Äù n…ôdir?

∆èg…ôr A v…ô B **asƒ±lƒ± olmayan hadis…ôl…ôrdirs…ô**, onda:

$$
P(A \cup B) = P(A) + P(B) - P(A \cap B)
$$

V…ô √ß√ºnki:

$$
P(A \cap B) = P(A) \cdot P(B)
$$

Onda bu olur:

$$
P(A \cup B) = P(A) + P(B) - P(A) \cdot P(B)
$$

---

### ‚úÖ **Bu formul n…ôyi ifad…ô edir?**

Bu, **iki m√ºst…ôqil hadis…ôd…ôn …ôn az birinin ba≈ü verm…ôsi ehtimalƒ±dƒ±r**.

* Hadis…ô B: ƒ∞kinci z…ôri atanda 5 g…ôlsin ‚Üí  $P(B) = \frac{1}{6}$

Hadis…ôl…ôr m√ºst…ôqildir, √ß√ºnki biri dig…ôrin…ô t…ôsir etmir.

$$
P(A \cup B) = \frac{1}{6} + \frac{1}{6} - \left( \frac{1}{6} \cdot \frac{1}{6} \right) = \frac{2}{6} - \frac{1}{36} = \frac{12 - 1}{36} = \frac{11}{36}
$$

## Ehtimalƒ± Vurma Teoreml…ôri

### EHTƒ∞MALI VURMA TEOREML∆èRƒ∞

>**T…ôrfi:** **Vurma teoremi**, iki v…ô ya daha √ßox hadis…ônin **eyni anda ba≈ü verm…ô ehtimalƒ±nƒ±**, onlarƒ±n bir-birin…ô t…ôsirind…ôn asƒ±lƒ± olaraq **≈ü…ôrti v…ô ya m√ºst…ôqil ehtimallarla** hesablama qaydasƒ±dƒ±r.





### üìò EHTƒ∞MALI VURMA TEOREML∆èRƒ∞

Ehtimalƒ± vurma teoreml…ôri …ôsas…ôn **iki hadis…ônin k…ôsi≈üm…ôsi** (y…ôni **A ‚à© B**) √º√ß√ºn ehtimalƒ± hesablamaq √º√ß√ºnd√ºr.

---

### üî∑ 1. √úmumi Vurma Teoremi (Asƒ±lƒ± hadis…ôl…ôr √º√ß√ºn)

∆èg…ôr **A v…ô B hadis…ôl…ôri asƒ±lƒ±dƒ±rsa**, y…ôni biri dig…ôrin…ô t…ôsir edirs…ô:

$$
P(A \cap B) = P(A) \cdot P(B|A)
$$

> Burada $P(B|A)$ ‚Äì A ba≈ü ver…ônd…ôn sonra B-nin ba≈ü verm…ô ehtimalƒ±dƒ±r (≈ü…ôrti ehtimal).

---

### üîπ Misal:

Bir qutuda 3 aƒü, 2 qara top var. Qaytarmadan 2 top √ß…ôkilir.

* Hadis…ô A: 1-ci top aƒü g…ôlsin.
* Hadis…ô B: 2-ci top aƒü g…ôlsin.

Bu hadis…ôl…ôr **asƒ±lƒ±dƒ±r**, √ß√ºnki 1-ci top √ßƒ±xandan sonra v…ôziyy…ôt d…ôyi≈üir.

$$
P(A) = \frac{3}{5}, \quad P(B|A) = \frac{2}{4}
$$

$$
P(A \cap B) = \frac{3}{5} \cdot \frac{2}{4} = \frac{6}{20} = \frac{3}{10}
$$

---

### üî∑ 2. M√ºst…ôqil Hadis…ôl…ôr √º√ß√ºn Vurma Teoremi

∆èg…ôr **A v…ô B hadis…ôl…ôri bir-birind…ôn asƒ±lƒ± deyils…ô**, y…ôni biri dig…ôrin…ô t…ôsir etmirs…ô:

$$
P(A \cap B) = P(A) \cdot P(B)
$$

> Bu, daha sad…ô haldƒ±r v…ô **asƒ±lƒ± olmayan hadis…ôl…ôr √º√ß√ºn** ke√ß…ôrlidir.

---

### üîπ Misal:

ƒ∞ki ayrƒ± zar atƒ±lƒ±r:

* Hadis…ô A: 1-ci z…ôrd…ô 6 g…ôlsin ‚Üí $P(A) = \frac{1}{6}$
* Hadis…ô B: 2-ci z…ôrd…ô 5 g…ôlsin ‚Üí $P(B) = \frac{1}{6}$

$$
P(A \cap B) = \frac{1}{6} \cdot \frac{1}{6} = \frac{1}{36}
$$

---

### üß† Qaydalar v…ô Anlayƒ±≈ülar

| Anlayƒ±≈ü                     | ƒ∞zah                                                           |             |
| --------------------------- | -------------------------------------------------------------- | ----------- |
| **$$P(A \cap B)$$**           | A v…ô B hadis…ôl…ôrinin **h…ôr ikisi eyni anda** ba≈ü verir         |             |
| **Asƒ±lƒ± hadis…ôl…ôr**         | B-nin ehtimalƒ± A ba≈ü ver…ônd…ôn sonra d…ôyi≈üir: $$( P(B|A) )$$        |
| **Asƒ±lƒ± olmayan hadis…ôl…ôr** | B-nin ehtimalƒ± A-ya baƒülƒ± deyil: $$( P(B|A) = P(B) )$$ |
| **Vurma Teoremi**           | ƒ∞ki hadis…ônin birlikd…ô ba≈ü verm…ô ehtimalƒ± √º√ß√ºn istifad…ô olunur |             |

---

### üßÆ Yekun Formullar

| Hal                     | D√ºstur                          |      |
| ----------------------- | ------------------------------- | ---- |
| Asƒ±lƒ± hadis…ôl…ôr         | $$( P(A \cap B) = P(A) \cdot P(B  | A) )$$ |
| Asƒ±lƒ± olmayan hadis…ôl…ôr | $$P(A \cap B) = P(A) \cdot P(B)$$ |      |

---

### üí° Bonus: √ú√ß v…ô daha √ßox hadis…ô

√ú√ß hadis…ônin birlikd…ô ba≈ü verm…ôsi:

$$
P(A \cap B \cap C) = P(A) \cdot P(B|A) \cdot P(C|A \cap B)
$$

∆èg…ôr hamƒ±sƒ± asƒ±lƒ± deyil (m√ºst…ôqildirs…ô):

$$
P(A \cap B \cap C) = P(A) \cdot P(B) \cdot P(C)
$$
