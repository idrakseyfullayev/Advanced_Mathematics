## Asılı olmayan sınaqların ardıcıllığı



## ✅ Bernoulli nöqteyi-nəzərindən Asılı Olmayan Sınaqların Ardıcıllığı

### 🔹 1. Bernoulli sınağı nədir?

**Bernoulli sınağı** — iki nəticəli (məsələn, **uğur** və **uğursuzluq**) sadə bir ehtimal sınağıdır.

Misallar:

* Sikkə atmaq (Üz və ya yazı),
* Testi keçmək və ya keçməmək,
* Müştərinin məhsulu alması və ya almaması.

> Bernoulli sınağında uğurun ehtimalı:

$$
P(Uğur) = p, \quad P(Uğursuzluq) = 1 - p
$$

---

### 🔹 2. Bernoulli sınaqlarının **ardıcıllığı** nədir?

Bu zaman biz **eyni Bernoulli sınağını** bir neçə dəfə **təkrar edirik**, və **hər sınaq əvvəlkilərdən asılı deyil**.

Bu **asılı olmayan sınaqların ardıcıllığıdır** və buna **Bernoulli prosesi** deyilir.

---

### 🔹 3. Bernoulli sınaqları üçün **ümumi ehtimal**:

Əgər **n** dəfə sınaq aparılırsa və **k** dəfə uğur baş verirsə, bu hadisənin ehtimalı belə olur:

$$
P(k \text{ uğur}) = \binom{n}{k} \cdot p^k \cdot (1 - p)^{n - k}
$$

Bu **binom ehtimal düsturudur** və **Bernoulli sınaqlarının** nəticəsidir.

---

### 📌 Nümunə:

Bir sikkə 5 dəfə atılır. **3 dəfə üz** gəlmə ehtimalı nədir?

* $n = 5$, $k = 3$, $p = 0.5$

$$
P(3 \text{ üz}) = \binom{5}{3} \cdot (0.5)^3 \cdot (0.5)^2 = 10 \cdot (0.5)^5 = 10 \cdot \frac{1}{32} = \frac{10}{32} = 0.3125
$$

---

### 📌 Əsas ideya:

**Asılı olmayan Bernoulli sınaqlarının ardıcıllığı** bizə imkan verir ki:

* Hər sınaq **eyni ehtimal** ilə baş versin,
* Sınaqların nəticələri **bir-birinə təsir etməsin**,
* Ümumi ehtimalı **binom düsturu** ilə hesablayaq.


## Bernoulli təcrübəsi və düsturu



## 📌 1. **Tərif** (Bernoulli təcrübəsi və düsturu)

**Bernoulli düsturu**, **eyni ehtimal** ilə və **bir-birindən asılı olmayan** sınaqların ardıcıllığında, müəyyən sayda **uğurun ehtimalını** tapmaq üçün istifadə olunur.

### 🔹 Qısaca:

Eyni sınaq **n dəfə** təkrarlanır, hər dəfə **uğur ehtimalı p**, və **k uğur** almaq istəyirik → ehtimalını Bernoulli düsturu verir.

---

## 📐 2. **Düstur**

$$
P(k \text{ uğur}) = \binom{n}{k} \cdot p^k \cdot (1 - p)^{n - k}
$$

Burada:

* $n$: ümumi sınaq sayı
* $k$: uğur sayı
* $p$: bir uğurun ehtimalı
* $1 - p$: uğursuzluq ehtimalı
* $\binom{n}{k}$: kombinasiya (n sınaqdan k uğur seçmək)

---

## 🎯 3. Əsas anlayışlar və nöqtələr:

| Mövzu                    | İzah                                                                  |
| ------------------------ | --------------------------------------------------------------------- |
| **Bernoulli sınağı**     | 2 nəticəli təcrübə: uğur (1) və uğursuzluq (0)                        |
| **Sınaqlar asılı deyil** | Hər sınağın nəticəsi digərlərindən təsirlənmir                        |
| **Sabit ehtimal**        | Hər sınaqda uğur ehtimalı **eyni** qalır: $p$                         |
| **Binom bölgüsü**        | Bernoulli sınaqları nəticəsində yaranan ehtimal bölgüsü               |
| **Ən çox ehtimal**       | Ən çox ehtimal sınaqların yarısında uğur olanda olur (əgər $p = 0.5$) |
| **Toplam ehtimal**       | Bütün mümkün k-lər üzrə cəmlənsə, nəticə 1 olacaq:                    |

$$
\sum_{k=0}^{n} \binom{n}{k} p^k (1-p)^{n-k} = 1
$$

---

## ✅ Nümunə

Bir testdə düzgün cavab vermə ehtimalı 0.8-dir. 5 sualdan 4-nə düzgün cavab vermə ehtimalı:

$$
P(4 \text{ uğur}) = \binom{5}{4} \cdot 0.8^4 \cdot 0.2^1 = 5 \cdot 0.4096 \cdot 0.2 = 0.4096
$$


## Bernoulli sınaqları və əlamətlərə görə ehtimal hesablanması



## 🔷 **Əsas anlayış:**

Bir sınaq ardıcıllığında (məsələn, test sualları, sikkə atmaq və s.) uğur ehtimalı $p$, uğursuzluq ehtimalı $1-p$ olmaqla, bu sınaqlar **bir-birindən asılı deyil** və **eyni ehtimal** ilə baş verir. Bu Bernoulli prosesi adlanır.

---

## 🔹 a) $n$ sınaqda $m$-dən **az** uğur baş vermə ehtimalı

$$
P(X < m) = P_n(0) + P_n(1) + \dots + P_n(m - 1)
$$

### 🟢 İzah:

* $X$: uğurların sayı
* Bu formula deyir ki, **0-dan (heç uğur yox)** başlayaraq, **(m-1)-ə qədər** uğur halları üçün ehtimalları toplamalıyıq.

---

## 🔹 b)  $n$ sınaqda $m$-dən **çox** uğur baş vermə ehtimalı  $P(X > m)$

$$
P(X > m) = P_n(m+1) + P_n(m+2) + \dots + P_n(n)
$$

### 🟢 İzah:

* Yəni $m$-dən **çox uğur** halları üçün ehtimal tapılır.
* $X > m$ olduğu halları birləşdiririk.

---

## 🔹 c) $n$ sınaqda  **ən azı** $m$-dəfə uğur baş vermə ehtimalı $P(X \ge m)$

$$
P(X \ge m) = P_n(m) + P_n(m+1) + \dots + P_n(n)
$$

### 🟢 İzah:

* $X \ge m$: ən azı $m$ uğur → yəni $m$, $m+1$, ..., $n$ uğurları ehtiva edir.
* Toplam ehtimal bu halların cəmi olur.

---

## 🔹 d) $n$ sınaqda  **ən çoxu** $m$-dəfə uğur baş vermə ehtimalı $P(X \le m)$

$$
P(X \le m) = P_n(0) + P_n(1) + \dots + P_n(m)
$$

### 🟢 İzah:

* Yəni **ən çox m uğur** olsun.
* Bu da $0$ uğurdan başlayaraq $m$-ə qədər olan ehtimalların cəmidir.

---

## ✅ Bəs bu $P_n(k)$ nədir?

Bu **Bernoulli düsturu** ilə hesablanan **k uğur ehtimalıdır**:

$$
P_n(k) = \binom{n}{k} \cdot p^k \cdot (1 - p)^{n - k}
$$

Burada:

* $\binom{n}{k}$: n-dən k uğur seçmək
* $p^k$: uğurların ehtimalı
* $(1-p)^{n-k}$: uğursuzluqların ehtimalı

---

## 💡 Nəticə

Bu 4 düsturda əsas məqsəd budur:

| Məqsəd             | İstifadə olunan formula |
| ------------------ | ----------------------- |
| **m-dən az uğur**  | $P(X < m)$              |
| **m-dən çox uğur** | $P(X > m)$              |
| **ən az m uğur**   | $P(X \ge m)$            |
| **ən çox m uğur**  | $P(X \le m)$            |

Bu düsturlar statistik təhlil, risk hesabatı, sınaqların qiymətləndirilməsi və keyfiyyətə nəzarət kimi sahələrdə çox istifadə olunur.

## Ən böyük ehtimallı ədəd



### ✅ **Ən böyük ehtimallı ədəd nədir?**



### ✅ TƏRİF:

**Ən böyük ehtimallı ədəd** — Bernoulli sınaqları ardıcıllığında (binom bölgüsündə), uğurların sayı üçün ehtimalın **ən böyük olduğu** $k$ dəyəridir.

Başqa sözlə:

> Binom bölgüsündə $P_n(k) = \binom{n}{k}p^k(1-p)^{n-k}$ funksiyasının maksimum olduğu $k$-ya **ən böyük ehtimallı ədəd** deyilir.



Bernoulli sınaqlarının ardıcıllığında (yəni binom bölgüsündə) uğurların ehtimalları müxtəlif olur. Bizə maraqlı olan:

> **Hansının ehtimalı digərlərindən daha böyükdür?**
> Yəni, **$P_n(k)$** maksimum hansı $k$-da olur?

---

### 🎯 **Tərif:**

**Ən böyük ehtimallı ədəd** — binom bölgüsündə $P_n(k)$-nın **maksimum** olduğu $k$-dır.

---

### 📐 **Düstur:**

Əgər:

* sınaq sayı: $n$
* uğur ehtimalı: $p$

O zaman **ən böyük ehtimallı ədəd** (yəni, $P_n(k)$ maksimum olan $k$) təqribən:

$$
k_{\text{max}} = \lfloor (n+1)p \rfloor
$$

Burada:

* $n$: sınaqların (təcrübələrin) sayı
* $p$: uğur ehtimalı (sabitdir)
* $\lfloor \cdot \rfloor$ — tam hissə funksiyasıdır (yəni aşağıya yuvarlaqlaşdırma).

---

### 📌 Nümunə:


**Misal 1:**
10 dəfə test sualı verilir, düzgün cavab ehtimalı $p = 0.7$.
Ən çox hansı uğur sayı mümkündür?

$$
k_{\text{max}} = \lfloor (10+1) \cdot 0.7 \rfloor = \lfloor 7.7 \rfloor = 7
$$

➡️ Deməli, **ən çox ehtimal 7 düzgün cavabda** olacaq.


**Misal 2:**
10 dəfə sikkə atılır. Üz gəlmə ehtimalı $p = 0.5$

$$
k_{\text{max}} = \lfloor (10 + 1) \cdot 0.5 \rfloor = \lfloor 5.5 \rfloor = 5
$$

📌 Yəni, **ən çox ehtimalla 5 dəfə üz gələcək**.

---

### 🧠 İzah:

Əgər uğur ehtimalı **0.5**-dirsə (məsələn, sikkə atmaq), o zaman:

* **Ən çox ehtimal mərkəzdə olur**
* Yəni $k \approx n/2$

Əgər uğur ehtimalı **0.8** kimi böyükdürsə:

* Ən böyük ehtimallı ədəd daha **yuxarı** olur (çünki daha çox uğur gözlənilir)

Əgər $p = 0.2$ kimi kiçikdirsə:

* Ən böyük ehtimallı ədəd daha **kiçik** olur (çünki uğur çətin baş verir)

---

## Ən böyük ehtimallı ədədin tam say olub-olmaması



### 📌 Verilmiş hissədə yazılıb:

> 1. $(np + p)$ – tam ədəd deyilsə, $m_0$ — **bir dəyər olur**
> 2. $(np + p)$ – tam ədəd olarsa,

- >    ➤ $m_0' = np - q$
- >    ➤ $m_0'' = np + p$
- >    Bu halda $m_0$ — **iki dəyər olur**

---

### ✅ İZAH:

Bu, **ən böyük ehtimallı ədədin** (yəni $k$-nın) **bir və ya iki dəyərli** olması məsələsidir.

---

### 🔹 1) $(n \cdot p + p)$ tam ədəd **deyilsə**:

Bu halda yalnız **bir dəyər** $m_0$ olur. Bu dəyər aşağıdakı kimi hesablanır:

$$
m_0 = \lfloor (n + 1)p \rfloor
$$

Yəni **aşağı yuvarlaqlaşdırılmış** bir dəyər alınır.

> 🔸 Bu vəziyyətdə ehtimalların maksimum olduğu **tək bir k** dəyəri olur.

---

### 🔹 2) $(n \cdot p + p)$ tam ədəd **olarsa**:

Bu zaman **iki ədəd** $m_0'$ və $m_0''$ ən böyük ehtimallı ədəd kimi qəbul olunur.

### Bu dəyərlər:

* $m_0' = np - q$
* $m_0'' = np + p$

və bu 2 dəyər üçün:

$$
P(m_0') = P(m_0'') = \text{ən böyük ehtimal}
$$

> 🔸 Bu halda binom bölgüsünün **iki simmetrik nöqtəsində** eyni ehtimal olur və **ikisi də maksimum olur**.

---

### 🧠 NƏ ÜÇÜN BU VACİBDİR?

* Bu məlumat **teorik təhlillər** üçün vacibdir.
* Məsələn, ehtimalların necə dəyişdiyini analiz etmək, **simmetrik bölgülər**, və **normal təqriblər** zamanı lazım olur.
* Xüsusilə də statistik qiymətləndirmə və sınaq nəticələrinin mərkəzini tapmaq üçün əhəmiyyətlidir.

---

### 🔍 Sadə bir misal:






### 🔢 Verilənlər:

* $n = 5$ (yəni 5 sınaq)
* $p = 0.5$ (uğur ehtimalı)
* $(n+1)p = (5+1) \cdot 0.5 = 6 \cdot 0.5 = 3$

Bu **tam ədəddir**. O zaman ən böyük ehtimallı ədəd **iki dəyərli** olacaq.

---

### ❓ Bəs bu iki dəyəri **necə tapırıq?**

Burada iki maksimum ehtimallı ədəd (ən böyük ehtimallı ədəd) aşağıdakı **qayda ilə** tapılır:

### ✔️ Qayda:

Əgər $(n+1)p$ **tam ədəd** olarsa, o zaman:

$$
m_0' = (n+1)p - 1
$$

$$
m_0'' = (n+1)p
$$

---

## 🔍 Nümunəyə tətbiq:

$$
(n+1)p = 3
$$

Deməli:

* $m_0' = 3 - 1 = 2$
* $m_0'' = 3$

Yəni:

> Ən böyük ehtimallı ədədlər: **2 və 3**
> (Hər ikisi üçün ehtimallar **eyni və maksimum** olacaq)


> Bu o deməkdir ki, həm $P(2)$, həm də $P(3)$ eyni və maksimum ehtimal dəyərləridir.

---

## 📐 Alternativ izah:

Binom bölgüsündə, $P_n(k)$ funksiyası üçün:

$$
P_n(k) = \binom{n}{k} p^k (1-p)^{n-k}
$$

Əgər bu funksiya **iki qonşu k üçün eyni dəyərdədirsə**, onda **ən böyük ehtimal iki nöqtədə olur** — yəni $m_0'$ və $m_0''$.

---

## 📌 Nəticə:

Əgər $(n+1)p$ **tam ədəd deyilsə**, o zaman:

$$
m_0 = \lfloor (n+1)p \rfloor
$$

Əgər **tam ədəddirsə**, onda:

$$
m_0' = (n+1)p - 1, \quad m_0'' = (n+1)p
