## Asƒ±lƒ± olmayan sƒ±naqlarƒ±n ardƒ±cƒ±llƒ±ƒüƒ±



## ‚úÖ Bernoulli n√∂qteyi-n…ôz…ôrind…ôn Asƒ±lƒ± Olmayan Sƒ±naqlarƒ±n Ardƒ±cƒ±llƒ±ƒüƒ±

### üîπ 1. Bernoulli sƒ±naƒüƒ± n…ôdir?

**Bernoulli sƒ±naƒüƒ±** ‚Äî iki n…ôtic…ôli (m…ôs…ôl…ôn, **uƒüur** v…ô **uƒüursuzluq**) sad…ô bir ehtimal sƒ±naƒüƒ±dƒ±r.

Misallar:

* Sikk…ô atmaq (√úz v…ô ya yazƒ±),
* Testi ke√ßm…ôk v…ô ya ke√ßm…ôm…ôk,
* M√º≈üt…ôrinin m…ôhsulu almasƒ± v…ô ya almamasƒ±.

> Bernoulli sƒ±naƒüƒ±nda uƒüurun ehtimalƒ±:

$$
P(Uƒüur) = p, \quad P(Uƒüursuzluq) = 1 - p
$$

---

### üîπ 2. Bernoulli sƒ±naqlarƒ±nƒ±n **ardƒ±cƒ±llƒ±ƒüƒ±** n…ôdir?

Bu zaman biz **eyni Bernoulli sƒ±naƒüƒ±nƒ±** bir ne√ß…ô d…ôf…ô **t…ôkrar edirik**, v…ô **h…ôr sƒ±naq …ôvv…ôlkil…ôrd…ôn asƒ±lƒ± deyil**.

Bu **asƒ±lƒ± olmayan sƒ±naqlarƒ±n ardƒ±cƒ±llƒ±ƒüƒ±dƒ±r** v…ô buna **Bernoulli prosesi** deyilir.

---

### üîπ 3. Bernoulli sƒ±naqlarƒ± √º√ß√ºn **√ºmumi ehtimal**:

∆èg…ôr **n** d…ôf…ô sƒ±naq aparƒ±lƒ±rsa v…ô **k** d…ôf…ô uƒüur ba≈ü verirs…ô, bu hadis…ônin ehtimalƒ± bel…ô olur:

$$
P(k \text{ uƒüur}) = \binom{n}{k} \cdot p^k \cdot (1 - p)^{n - k}
$$

Bu **binom ehtimal d√ºsturudur** v…ô **Bernoulli sƒ±naqlarƒ±nƒ±n** n…ôtic…ôsidir.

---

### üìå N√ºmun…ô:

Bir sikk…ô 5 d…ôf…ô atƒ±lƒ±r. **3 d…ôf…ô √ºz** g…ôlm…ô ehtimalƒ± n…ôdir?

* $n = 5$, $k = 3$, $p = 0.5$

$$
P(3 \text{ √ºz}) = \binom{5}{3} \cdot (0.5)^3 \cdot (0.5)^2 = 10 \cdot (0.5)^5 = 10 \cdot \frac{1}{32} = \frac{10}{32} = 0.3125
$$

---

### üìå ∆èsas ideya:

**Asƒ±lƒ± olmayan Bernoulli sƒ±naqlarƒ±nƒ±n ardƒ±cƒ±llƒ±ƒüƒ±** biz…ô imkan verir ki:

* H…ôr sƒ±naq **eyni ehtimal** il…ô ba≈ü versin,
* Sƒ±naqlarƒ±n n…ôtic…ôl…ôri **bir-birin…ô t…ôsir etm…ôsin**,
* √úmumi ehtimalƒ± **binom d√ºsturu** il…ô hesablayaq.


## Bernoulli t…ôcr√ºb…ôsi v…ô d√ºsturu



## üìå 1. **T…ôrif** (Bernoulli t…ôcr√ºb…ôsi v…ô d√ºsturu)

**Bernoulli d√ºsturu**, **eyni ehtimal** il…ô v…ô **bir-birind…ôn asƒ±lƒ± olmayan** sƒ±naqlarƒ±n ardƒ±cƒ±llƒ±ƒüƒ±nda, m√º…ôyy…ôn sayda **uƒüurun ehtimalƒ±nƒ±** tapmaq √º√ß√ºn istifad…ô olunur.

### üîπ Qƒ±saca:

Eyni sƒ±naq **n d…ôf…ô** t…ôkrarlanƒ±r, h…ôr d…ôf…ô **uƒüur ehtimalƒ± p**, v…ô **k uƒüur** almaq ist…ôyirik ‚Üí ehtimalƒ±nƒ± Bernoulli d√ºsturu verir.

---

## üìê 2. **D√ºstur**

$$
P(k \text{ uƒüur}) = \binom{n}{k} \cdot p^k \cdot (1 - p)^{n - k}
$$

Burada:

* $n$: √ºmumi sƒ±naq sayƒ±
* $k$: uƒüur sayƒ±
* $p$: bir uƒüurun ehtimalƒ±
* $1 - p$: uƒüursuzluq ehtimalƒ±
* $\binom{n}{k}$: kombinasiya (n sƒ±naqdan k uƒüur se√ßm…ôk)

---

## üéØ 3. ∆èsas anlayƒ±≈ülar v…ô n√∂qt…ôl…ôr:

| M√∂vzu                    | ƒ∞zah                                                                  |
| ------------------------ | --------------------------------------------------------------------- |
| **Bernoulli sƒ±naƒüƒ±**     | 2 n…ôtic…ôli t…ôcr√ºb…ô: uƒüur (1) v…ô uƒüursuzluq (0)                        |
| **Sƒ±naqlar asƒ±lƒ± deyil** | H…ôr sƒ±naƒüƒ±n n…ôtic…ôsi dig…ôrl…ôrind…ôn t…ôsirl…ônmir                        |
| **Sabit ehtimal**        | H…ôr sƒ±naqda uƒüur ehtimalƒ± **eyni** qalƒ±r: $p$                         |
| **Binom b√∂lg√ºs√º**        | Bernoulli sƒ±naqlarƒ± n…ôtic…ôsind…ô yaranan ehtimal b√∂lg√ºs√º               |
| **∆èn √ßox ehtimal**       | ∆èn √ßox ehtimal sƒ±naqlarƒ±n yarƒ±sƒ±nda uƒüur olanda olur (…ôg…ôr $p = 0.5$) |
| **Toplam ehtimal**       | B√ºt√ºn m√ºmk√ºn k-l…ôr √ºzr…ô c…ôml…ôns…ô, n…ôtic…ô 1 olacaq:                    |

$$
\sum_{k=0}^{n} \binom{n}{k} p^k (1-p)^{n-k} = 1
$$

---

$$


## Bernoulli sƒ±naqlarƒ± v…ô …ôlam…ôtl…ôr…ô g√∂r…ô ehtimal hesablanmasƒ±



## üî∑ **∆èsas anlayƒ±≈ü:**

Bir sƒ±naq ardƒ±cƒ±llƒ±ƒüƒ±nda (m…ôs…ôl…ôn, test suallarƒ±, sikk…ô atmaq v…ô s.) uƒüur ehtimalƒ± $p$, uƒüursuzluq ehtimalƒ± $1-p$ olmaqla, bu sƒ±naqlar **bir-birind…ôn asƒ±lƒ± deyil** v…ô **eyni ehtimal** il…ô ba≈ü verir. Bu Bernoulli prosesi adlanƒ±r.

---

## üîπ a) $n$ sƒ±naqda $m$-d…ôn **az** uƒüur ba≈ü verm…ô ehtimalƒ±

$$
P(X < m) = P_n(0) + P_n(1) + \dots + P_n(m - 1)
$$

### üü¢ ƒ∞zah:

* $X$: uƒüurlarƒ±n sayƒ±
* Bu formula deyir ki, **0-dan (he√ß uƒüur yox)** ba≈ülayaraq, **(m-1)-…ô q…ôd…ôr** uƒüur hallarƒ± √º√ß√ºn ehtimallarƒ± toplamalƒ±yƒ±q.

---

## üîπ b)  $n$ sƒ±naqda $m$-d…ôn **√ßox** uƒüur ba≈ü verm…ô ehtimalƒ±  $P(X > m)$

$$
P(X > m) = P_n(m+1) + P_n(m+2) + \dots + P_n(n)
$$

### üü¢ ƒ∞zah:

* Y…ôni $m$-d…ôn **√ßox uƒüur** hallarƒ± √º√ß√ºn ehtimal tapƒ±lƒ±r.
* $X > m$ olduƒüu hallarƒ± birl…ô≈üdiririk.

---

## üîπ c) $n$ sƒ±naqda  **…ôn azƒ±** $m$-d…ôf…ô uƒüur ba≈ü verm…ô ehtimalƒ± $P(X \ge m)$

$$
P(X \ge m) = P_n(m) + P_n(m+1) + \dots + P_n(n)
$$

### üü¢ ƒ∞zah:

* $X \ge m$: …ôn azƒ± $m$ uƒüur ‚Üí y…ôni $m$, $m+1$, ..., $n$ uƒüurlarƒ± ehtiva edir.
* Toplam ehtimal bu hallarƒ±n c…ômi olur.

---

## üîπ d) $n$ sƒ±naqda  **…ôn √ßoxu** $m$-d…ôf…ô uƒüur ba≈ü verm…ô ehtimalƒ± $P(X \le m)$

$$
P(X \le m) = P_n(0) + P_n(1) + \dots + P_n(m)
$$

### üü¢ ƒ∞zah:

* Y…ôni **…ôn √ßox m uƒüur** olsun.
* Bu da $0$ uƒüurdan ba≈ülayaraq $m$-…ô q…ôd…ôr olan ehtimallarƒ±n c…ômidir.

---

## ‚úÖ B…ôs bu $P_n(k)$ n…ôdir?

Bu **Bernoulli d√ºsturu** il…ô hesablanan **k uƒüur ehtimalƒ±dƒ±r**:

$$
P_n(k) = \binom{n}{k} \cdot p^k \cdot (1 - p)^{n - k}
$$

Burada:

* $\binom{n}{k}$: n-d…ôn k uƒüur se√ßm…ôk
* $p^k$: uƒüurlarƒ±n ehtimalƒ±
* $(1-p)^{n-k}$: uƒüursuzluqlarƒ±n ehtimalƒ±

---

## ∆èn b√∂y√ºk ehtimallƒ± …ôd…ôd



### ‚úÖ **∆èn b√∂y√ºk ehtimallƒ± …ôd…ôd n…ôdir?**



### ‚úÖ T∆èRƒ∞F:

**∆èn b√∂y√ºk ehtimallƒ± …ôd…ôd** ‚Äî Bernoulli sƒ±naqlarƒ± ardƒ±cƒ±llƒ±ƒüƒ±nda (binom b√∂lg√ºs√ºnd…ô), uƒüurlarƒ±n sayƒ± √º√ß√ºn ehtimalƒ±n **…ôn b√∂y√ºk olduƒüu** $k$ d…ôy…ôridir.

Ba≈üqa s√∂zl…ô:

> Binom b√∂lg√ºs√ºnd…ô $P_n(k) = \binom{n}{k}p^k(1-p)^{n-k}$ funksiyasƒ±nƒ±n maksimum olduƒüu $k$-ya **…ôn b√∂y√ºk ehtimallƒ± …ôd…ôd** deyilir.



Bernoulli sƒ±naqlarƒ±nƒ±n ardƒ±cƒ±llƒ±ƒüƒ±nda (y…ôni binom b√∂lg√ºs√ºnd…ô) uƒüurlarƒ±n ehtimallarƒ± m√ºxt…ôlif olur. Biz…ô maraqlƒ± olan:

> **Hansƒ±nƒ±n ehtimalƒ± dig…ôrl…ôrind…ôn daha b√∂y√ºkd√ºr?**
> Y…ôni, **$P_n(k)$** maksimum hansƒ± $k$-da olur?

---

### üéØ **T…ôrif:**

**∆èn b√∂y√ºk ehtimallƒ± …ôd…ôd** ‚Äî binom b√∂lg√ºs√ºnd…ô $P_n(k)$-nƒ±n **maksimum** olduƒüu $k$-dƒ±r.

---

### üìê **D√ºstur:**

∆èg…ôr:

* sƒ±naq sayƒ±: $n$
* uƒüur ehtimalƒ±: $p$

O zaman **…ôn b√∂y√ºk ehtimallƒ± …ôd…ôd** (y…ôni, $P_n(k)$ maksimum olan $k$) t…ôqrib…ôn:

$$
k_{\text{max}} = \lfloor (n+1)p \rfloor
$$

Burada:

* $n$: sƒ±naqlarƒ±n (t…ôcr√ºb…ôl…ôrin) sayƒ±
* $p$: uƒüur ehtimalƒ± (sabitdir)
* $\lfloor \cdot \rfloor$ ‚Äî tam hiss…ô funksiyasƒ±dƒ±r (y…ôni a≈üaƒüƒ±ya yuvarlaqla≈üdƒ±rma).

---

$$
k_{\text{max}} = \lfloor (10+1) \cdot 0.7 \rfloor = \lfloor 7.7 \rfloor = 7
$$

‚û°Ô∏è Dem…ôli, **…ôn √ßox ehtimal 7 d√ºzg√ºn cavabda** olacaq.

$$

üìå Y…ôni, **…ôn √ßox ehtimalla 5 d…ôf…ô √ºz g…ôl…ôc…ôk**.

## ∆èn b√∂y√ºk ehtimallƒ± …ôd…ôdin tam say olub-olmamasƒ±



### üìå Verilmi≈ü hiss…ôd…ô yazƒ±lƒ±b:

> 1. $(np + p)$ ‚Äì tam …ôd…ôd deyils…ô, $m_0$ ‚Äî **bir d…ôy…ôr olur**
> 2. $(np + p)$ ‚Äì tam …ôd…ôd olarsa,

- >    ‚û§ $m_0' = np - q$
- >    ‚û§ $m_0'' = np + p$
- >    Bu halda $m_0$ ‚Äî **iki d…ôy…ôr olur**

---

### ‚úÖ ƒ∞ZAH:

Bu, **…ôn b√∂y√ºk ehtimallƒ± …ôd…ôdin** (y…ôni $k$-nƒ±n) **bir v…ô ya iki d…ôy…ôrli** olmasƒ± m…ôs…ôl…ôsidir.

---

### üîπ 1) $(n \cdot p + p)$ tam …ôd…ôd **deyils…ô**:

Bu halda yalnƒ±z **bir d…ôy…ôr** $m_0$ olur. Bu d…ôy…ôr a≈üaƒüƒ±dakƒ± kimi hesablanƒ±r:

$$
m_0 = \lfloor (n + 1)p \rfloor
$$

Y…ôni **a≈üaƒüƒ± yuvarlaqla≈üdƒ±rƒ±lmƒ±≈ü** bir d…ôy…ôr alƒ±nƒ±r.

> üî∏ Bu v…ôziyy…ôtd…ô ehtimallarƒ±n maksimum olduƒüu **t…ôk bir k** d…ôy…ôri olur.

---

### üîπ 2) $(n \cdot p + p)$ tam …ôd…ôd **olarsa**:

Bu zaman **iki …ôd…ôd** $m_0'$ v…ô $m_0''$ …ôn b√∂y√ºk ehtimallƒ± …ôd…ôd kimi q…ôbul olunur.

### Bu d…ôy…ôrl…ôr:

* $m_0' = np - q$
* $m_0'' = np + p$

v…ô bu 2 d…ôy…ôr √º√ß√ºn:

$$
P(m_0') = P(m_0'') = \text{…ôn b√∂y√ºk ehtimal}
$$

> üî∏ Bu halda binom b√∂lg√ºs√ºn√ºn **iki simmetrik n√∂qt…ôsind…ô** eyni ehtimal olur v…ô **ikisi d…ô maksimum olur**.

---

### üß† N∆è √ú√á√úN BU VACƒ∞BDƒ∞R?

* Bu m…ôlumat **teorik t…ôhlill…ôr** √º√ß√ºn vacibdir.
* M…ôs…ôl…ôn, ehtimallarƒ±n nec…ô d…ôyi≈üdiyini analiz etm…ôk, **simmetrik b√∂lg√ºl…ôr**, v…ô **normal t…ôqribl…ôr** zamanƒ± lazƒ±m olur.
* X√ºsusil…ô d…ô statistik qiym…ôtl…ôndirm…ô v…ô sƒ±naq n…ôtic…ôl…ôrinin m…ôrk…ôzini tapmaq √º√ß√ºn …ôh…ômiyy…ôtlidir.

---

### üîç Sad…ô bir misal:






### üî¢ Veril…ônl…ôr:

* $n = 5$ (y…ôni 5 sƒ±naq)
* $p = 0.5$ (uƒüur ehtimalƒ±)
* $(n+1)p = (5+1) \cdot 0.5 = 6 \cdot 0.5 = 3$

Bu **tam …ôd…ôddir**. O zaman …ôn b√∂y√ºk ehtimallƒ± …ôd…ôd **iki d…ôy…ôrli** olacaq.

---

### ‚ùì B…ôs bu iki d…ôy…ôri **nec…ô tapƒ±rƒ±q?**

Burada iki maksimum ehtimallƒ± …ôd…ôd (…ôn b√∂y√ºk ehtimallƒ± …ôd…ôd) a≈üaƒüƒ±dakƒ± **qayda il…ô** tapƒ±lƒ±r:

### ‚úîÔ∏è Qayda:

∆èg…ôr $(n+1)p$ **tam …ôd…ôd** olarsa, o zaman:

$$
m_0' = (n+1)p - 1
$$

$$
m_0'' = (n+1)p
$$

---

## üîç N√ºmun…ôy…ô t…ôtbiq:

$$
(n+1)p = 3
$$

Dem…ôli:

* $m_0' = 3 - 1 = 2$
* $m_0'' = 3$

Y…ôni:

> ∆èn b√∂y√ºk ehtimallƒ± …ôd…ôdl…ôr: **2 v…ô 3**
> (H…ôr ikisi √º√ß√ºn ehtimallar **eyni v…ô maksimum** olacaq)


> Bu o dem…ôkdir ki, h…ôm $P(2)$, h…ôm d…ô $P(3)$ eyni v…ô maksimum ehtimal d…ôy…ôrl…ôridir.

---

## üìê Alternativ izah:

Binom b√∂lg√ºs√ºnd…ô, $P_n(k)$ funksiyasƒ± √º√ß√ºn:

$$
P_n(k) = \binom{n}{k} p^k (1-p)^{n-k}
$$

∆èg…ôr bu funksiya **iki qon≈üu k √º√ß√ºn eyni d…ôy…ôrd…ôdirs…ô**, onda **…ôn b√∂y√ºk ehtimal iki n√∂qt…ôd…ô olur** ‚Äî y…ôni $m_0'$ v…ô $m_0''$.

---

## üìå N…ôtic…ô:

∆èg…ôr $(n+1)p$ **tam …ôd…ôd deyils…ô**, o zaman:

$$
m_0 = \lfloor (n+1)p \rfloor
$$

∆èg…ôr **tam …ôd…ôddirs…ô**, onda:

$$
m_0' = (n+1)p - 1, \quad m_0'' = (n+1)p
$$
