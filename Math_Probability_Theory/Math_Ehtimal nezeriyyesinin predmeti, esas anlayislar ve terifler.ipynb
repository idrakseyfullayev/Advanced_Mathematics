## Ehtimal nəzəriyyəsi

> **Ehtimal nəzəriyyəsi** — **təsadüfi hadisələrin riyazi qanunauyğunluqlarını** öyrənən bir riyaziyyat sahəsidir. Bu nəzəriyyə, qeyri-müəyyənliklə bağlı olan prosesləri modelləşdirmək və bu proseslərdə müəyyən nəticələrin baş vermə ehtimalını hesablamaq üçün istifadə olunur.


---
> **Ehtimal nəzəriyyəsi** — **təsadüfi hadisələrin baş vermə ehtimallarını riyazi üsullarla** ölçən və qeyri-müəyyənliyi idarə etməyə imkan verən elmdir.
---

### 📌 Əsas anlayışlar:

#### 1. **Təsadüfi hadisə**  
Əvvəlcədən nəticəsi məlum olmayan, lakin müəyyən nəticələr ala biləcək bir hadisə.  
Məsələn:  
- Zəri atdıqda 6 düşməsi  
- Bir adamın yağışlı havada çətir götürməsi  

#### 2. **Nümunə məkanı (Sample space)**  
Bütün mümkün nəticələrin çoxluğu.  
Məsələn: Zəri atdıqda mümkün nəticələr → {1, 2, 3, 4, 5, 6}

#### 3. **Ehtimal (Probability)**  
Hadisənin baş vermə dərəcəsini ifadə edən ədəddir və 0 ilə 1 arasında dəyişir.  
Formul:  
$$
P(A) = \frac{\text{A hadisəsinə uyğun nəticələrin sayı}}{\text{Ümumi mümkün nəticələrin sayı}}
$$

#### 4. **Əks hadisə**  
Bir hadisənin baş verməməsi halı.  
$$
P(A') = 1 - P(A)
$$

#### 5. **Birgə hadisələr və müstəqil hadisələr**
- **Müstəqil hadisələr:** Biri digərinə təsir etmirsə (məsələn, iki fərqli zərin atılması).
- **Asılı hadisələr:** Birinin nəticəsi digərinə təsir edirsə.

---

### 💡 Nə üçün ehtimal nəzəriyyəsi vacibdir?
- Statistik analizlərdə
- Süni intellekt və maşın öyrənmədə
- Qumar oyunlarında və risk təhlilində
- Maliyyə, sığorta və mühəndislik sahələrində qərar qəbul etmək üçün

---

Əgər istəsən, ehtimal nəzəriyyəsində istifadə olunan **kombinatorika, Bayes teoremi, binomial və normal paylanma** kimi mövzuları da izah edə bilərəm. Davam edək?

## Hadisələrin birləşməsi, kəsişməsi, çıxılması və tamamlayıcısı (A-ya daxil olmayan hadisələr) ehtimal nəzəriyyəsinin əsas anlayışlarındandır.


### 🔹 1. **A və B hadisəsinin birləşməsi (A ∪ B):**  
Bu, **A və ya B** hadisələrinin (və ya hər ikisinin) baş verməsi deməkdir.  

**Riyazi yazılışı:**  
$$ A \cup B $$  
**Məna:** A baş verə bilər, B baş verə bilər, ya da hər ikisi baş verə bilər.

---

### 🔹 2. **A və B hadisəsinin kəsişməsi (A ∩ B):**  
Bu, **həm A, həm də B** hadisələrinin eyni anda baş verməsini bildirir.  

**Riyazi yazılışı:**  
$$ A \cap B $$  
**Məna:** Həm A, həm də B hadisələri baş verir.

---

### 🔹 3. **A-dan B-nin çıxılması (A \ B):**  
Bu, **A baş verir, amma B baş vermir** deməkdir. Yəni A hadisəsində olub, amma B hadisəsinə aid olmayan hallardır.  

**Riyazi yazılışı:**  
$$ A \setminus B $$  
və ya  
$$ A - B $$  
**Məna:** A hadisəsində olub, B-yə aid olmayan hallar.

---

### 🔹 4. **A hadisəsinə daxil olmayan hallar (A'nın tamamlayıcısı):**  
Bu, **A baş verməyən** hallardır.  

**Riyazi yazılışı:**  
$$ A' $$  
və ya  
$$ \overline{A} $$  
və ya  
$$ A^c $$  
**Məna:** A hadisəsi baş vermir.

---

### 💡 Qrafik baxımından (Venn diaqramı):  
- **Birləşmə (A ∪ B):** A dairəsi + B dairəsi (ikisinin içi dolu)  
- **Kəsişmə (A ∩ B):** A ilə B-nin ortaq hissəsi (ortada üst-üstə düşən yer)  
- **A - B:** A-nın içində olub, B ilə ortaq olmayan hissə  
- **A'nın tamamlayıcısı (A')**: A-nın xaricində qalan bütün hissə

---

## Uyuşan, Uyuşmayan, Cüt-cüt uyuşmayan hadisələr, Eyni ehtimallı hadisələr


### 🔹 1. **Uyuşan hadisələr (Kəsişən hadisələr)**  
İki hadisə **eyni anda baş verə bilərsə**, bu hadisələr **uyuşan** sayılır.  
**Riyazi işarəsi:**  
- A və B hadisələri üçün:  
  $$ A \cap B \neq \emptyset $$  
  (Yəni, A ilə B-nin kəsişməsi boş deyil.)

**Ehtimalı:**  
$$
P(A \cap B)
$$

**Nümunə:**  
A: Zar atıldıqda cüt ədəd gəlsin. → {2, 4, 6}  
B: Zar atıldıqda 3-dən böyük ədəd gəlsin. → {4, 5, 6}  
A ∩ B = {4, 6} → Uyuşan hadisələr.

---

### 🔹 2. **Uyuşmayan hadisələr (Disjoint / bir-birini istisna edən hadisələr)**  
İki hadisə **eyni anda baş verə bilməzsə**, onlar **uyușmayan** hadisələrdir.  
**Riyazi işarəsi:**  
- $A \cap B = \emptyset$

**Ehtimalı:**  
$$
P(A \cup B) = P(A) + P(B)
$$

**Nümunə:**  
A: Zar atıldıqda 2 gəlsin. → {2}  
B: Zar atıldıqda 5 gəlsin. → {5}  
A ∩ B = ∅ → Uyuşmayan hadisələr.

---

### 🔹 3. **Cüt-cüt uyuşmayan hadisələr (Pairwise disjoint events)**  
Bir neçə hadisə varsa və **hər bir iki hadisə arasında kəsişmə yoxdursa**, yəni **bütün cütlər arasında uyuşmayanlıq varsa**, bunlar **cüt-cüt uyuşmayan** adlanır.

**Riyazi işarəsi:**  
Hadisələr $A₁, A₂, ..., Aₙ$ üçün:
$$
\forall i \neq j,\quad A_i \cap A_j = \emptyset
$$

**Ehtimalı (toplam ehtimal):**  
$$
P\left(\bigcup_{i=1}^{n} A_i\right) = \sum_{i=1}^{n} P(A_i)
$$

**Nümunə:**  
$A₁$: Zar atıldıqda 1 gəlsin → {1}  
$A₂$: Zar atıldıqda 3 gəlsin → {3}  
$A₃$: Zar atıldıqda 6 gəlsin → {6}  
Bunlar cüt-cüt uyuşmayan hadisələrdir.

---

### 🔹 4. **Eyni ehtimallı hadisələr (Bərabər ehtimallı hadisələr)**  
Əgər bütün nəticələrin ehtimalı eynidirsə, onlar **eyni ehtimallı hadisələr** adlanır.

**Riyazi işarəsi:**  
Əgər hadisə A, bərabər ehtimallı $n$ nəticədən $m$-ni əhatə edirsə:
$$
P(A) = \frac{m}{n}
$$

**Nümunə:**  
Zar atıldıqda hər bir nəticənin (1, 2, 3, 4, 5, 6) ehtimalı eynidir:  
$$
P(1) = P(2) = \cdots = P(6) = \frac{1}{6}
$$

---

### 🔹 Vizual olaraq ümumi cədvəl:

| Hadisə tipi                 | Şərt                                      | İşarəsi                        | Ehtimal düsturu                          |
|----------------------------|--------------------------------------------|--------------------------------|------------------------------------------|
| Uyuşan                     | A və B birlikdə baş verə bilər             | $$ A \cap B \neq \emptyset $$   | $$ P(A \cap B) $$                          |
| Uyuşmayan                  | A və B eyni anda baş verə bilməz           | $$ A \cap B = \emptyset $$      | $$ P(A \cup B) = P(A) + P(B) $$            |
| Cüt-cüt uyuşmayan          | Hər iki hadisə cüt-cüt uyuşmur             | $$ A_i \cap A_j = \emptyset $$  | $$ P(\cup A_i) = \sum P(A_i) $$            |
| Eyni ehtimallı             | Bütün nəticələrin ehtimalı eynidir         | -                              | $$ P(A) = \frac{m}{n} $$                  |

##  Kolmoqorovun 3 Ehtimal Aksiomu

📘 Tutaq ki, bizdə ehtimal məkanı $S$ var və bu məkanın altkümələri hadisələrdir. Aşağıdakı 3 aksiom bütün ehtimal funksiyalarının əsasını təşkil edir:

---

### 🔹 **Aksiom 1: Qeyri-mənfilik (Non-negativity)**

Hər bir hadisənin ehtimalı mənfi ola bilməz.  
**Riyazi ifadə:**
$$
\forall A \subseteq S,\quad P(A) \geq 0
$$

Yəni istənilən hadisənin ehtimalı **sıfır və ya sıfırdan böyük** olmalıdır.

---

### 🔹 **Aksiom 2: Tam ehtimal (Normalization)**

Məkanın (bütün mümkün nəticələrin) ehtimalı **1**-dir.  
**Riyazi ifadə:**
$$
P(S) = 1
$$

Bu aksiom deyir ki, hansısa nəticə **mütləq** baş verəcək. (Yəni, hadisələrin ümumi məkanı istisnasız baş verir.)

---

### 🔹 **Aksiom 3: Sayılabilən əlavəlik (Countable additivity)**

Bir-birini istisna edən (uyușmayan) hadisələrin birləşməsinin ehtimalı onların ehtimallarının cəminə bərabərdir.  
**Riyazi ifadə:**  
Əgər $A_1, A_2, A_3, \ldots$ hadisələri cüt-cüt uyuşmayan hadisələrdirsə:

$$
P\left( \bigcup_{i=1}^{\infty} A_i \right) = \sum_{i=1}^{\infty} P(A_i)
$$

> **Bu, çoxlu hadisələr üçün də işləyir, əgər bu hadisələr bir-birini istisna edirsə.**

---

### 🧠 **Yaddaqalan şəkildə xatırlamaq:**

| Aksiom | Adı                   | Riyazi ifadəsi                                  |
|--------|------------------------|--------------------------------------------------|
| 1      | Qeyri-mənfilik         | $$P(A) \geq 0$$                                    |
| 2      | Tam ehtimal            | $$P(S) = 1$$                                      |
| 3      | Sayılabilən əlavəlik   | $P(\bigcup A_i) = \sum P(A_i)$, əgər $A_i \cap A_j = \emptyset$ |

---

### ✅ **Nəticə:**
Bu aksiomlar əsasında ehtimal funksiyaları qurulur və bütün ehtimal qanunları (məsələn, $P(A^c) = 1 - P(A)$ və s.) buradan əldə edilir.

## Ehtimalın Tezlik və Statistik Tərifi

## 📌 1. **Ehtimalın Tezlik Tərifi (Frequential Definition)**

Bu tərif **təcrübəyə əsaslanır**. Bir hadisənin ehtimalı onun **təcrübədə neçə dəfə baş verdiyi** ilə ölçülür.

### 🔹 Tərif:
> Bir hadisənin ehtimalı — çoxlu sayda təcrübə aparıldıqda, həmin hadisənin **nisbi tezliyinə** yaxınlaşır.

### 📐 Riyazi ifadə:
Əgər bir hadisə $A$ təcrübə zamanı $n$ dəfə icra edilib və bu hadisə $m$ dəfə baş veribsə:

$$
P(A) \approx \frac{m}{n}
\quad \text{(n çox olduqca)}
$$

Burada:
- $m$ – A hadisəsinin baş verdiyi say,
- $n$ – ümumi sınaq sayı,
- $\frac{m}{n}$ – nisbi tezlik (relative frequency).

### 📌 Xüsusiyyət:
- **Empirikdir** – real müşahidə və təcrübələrə əsaslanır.
- Təcrübə sayı artdıqca, **əsl ehtimala yaxınlaşır** (Bu, **Qanunların Böyük Ədədlər Qanunu** ilə əlaqəlidir).

### 🎲 Nümunə:
Zarı 1000 dəfə atırsan. 6 rəqəmi 160 dəfə gəlib.  
$$
P(6) \approx \frac{160}{1000} = 0.16
$$

---

## 📌 2. **Ehtimalın Statistik Tərifi (Statistical / Empirik Ehtimal)**

Bu anlayış tezlik tərifi ilə oxşardır, amma **müşahidə olunan məlumatlara** əsaslanır. Yəni, hadisənin ehtimalı **statistik məlumatlar üzərində** hesablanır.

### 🔹 Tərif:
> Əgər real həyatda hadisələrin baş vermə tezliyi məlumdursa, onların ehtimalı **statistik məlumat** əsasında qiymətləndirilir.

### 🔍 Fərq:
- **Tezlik tərifi** — təcrübə əsasında, çoxlu sınaqlarla yanaşır.
- **Statistik tərif** — mövcud **data və müşahidə nəticələri** ilə işləyir.

### 📊 Nümunə:

## Ehtimalın Sadə Xassələri


**sadə və əsas xassələrini** qısa və aydın şəkildə izah edək. Bu xassələr **Kolmoqorov aksiomlarına** və ehtimalın məntiqinə əsaslanır.

---

## ✅ **Ehtimalın Sadə Xassələri**

### 🔹 1. **Ehtimal 0 ilə 1 arasındadır:**

$$
0 \leq P(A) \leq 1
$$

- Heç vaxt 0-dan kiçik və ya 1-dən böyük ola bilməz.
- $P(A) = 0$ → hadisə heç vaxt baş vermir.
- $P(A) = 1$ → hadisə mütləq baş verir.

---

### 🔹 2. **Mümkün (bütün) hadisənin ehtimalı 1-dir:**

$$
P(S) = 1
$$

Burada $S$ – bütün mümkün halların cəmidir (sample space). Yəni, nəsə baş verməsi **mütləqdir**.

---

### 🔹 3. **İmkansız hadisənin ehtimalı 0-dır:**

$$
P(\emptyset) = 0
$$

Əgər hadisə heç vaxt baş verə bilmirsə, onun ehtimalı 0 olur.

---

### 🔹 4. **Əlavəlik xassəsi (Additivlik):**

Əgər A və B **uyuşmayan hadisələrdirsə** (yəni eyni vaxtda baş verə bilməzlər):

$$
P(A \cup B) = P(A) + P(B)
$$

> 🔸 Ümumi halda (uyuşa bilirlərsə):
$$
P(A \cup B) = P(A) + P(B) - P(A \cap B)
$$

---

### 🔹 5. **Tərs hadisənin ehtimalı:**

Əgər $A$ hadisəsi baş vermirsə, onda onun **tərs hadisəsi** $\bar{A}$ baş verir:

$$
P(\bar{A}) = 1 - P(A)
$$

---

### 🔹 6. **Boşluğa qarşı birlik:**

Əgər $A \subseteq B$ isə (yəni A, B-nin daxilindədirsə), o zaman:

$$
P(A) \leq P(B)
$$

---

### 🔹 7. **Əgər hadisələr bərabərsə:**

$$
A = B \Rightarrow P(A) = P(B)
$$

---

### 🔹 8. **Tam qrup təşkil edən hadisələr:**


## ✅ Tam qrup təşkil edən hadisələr nə deməkdir?

Əgər bir neçə hadisə (məsələn, $A_1, A_2, ..., A_n$) birlikdə **sample space**-i (yəni bütün mümkün halları) **tamamilə əhatə edirsə** və **bir-birinə uyğunmursa** (yəni eyni anda baş verə bilmirlərsə), o zaman onlar **tam qrup** (və ya **tam sistem**) təşkil edir.

---

### 📌 Rəsmi tərif:

Hadisələr $A_1, A_2, ..., A_n$ **tam qrup təşkil edir**, əgər:

1. **Tam əhatəlilik**:  
   $$
   A_1 \cup A_2 \cup \cdots \cup A_n = S
   $$  
   (Yəni hamısı birlikdə bütün mümkün halları əhatə edir.)

2. **Cüt-cüt uyğunmama (mutually exclusive)**:  
   $$
   A_i \cap A_j = \emptyset \quad \text{(əgər } i \ne j \text{)}
   $$  
   (Yəni hadisələr eyni vaxtda baş verə bilməz.)

---

## 🧠 Yadda saxlamaq üçün:
- **Tam qrup = hadisələr birlikdə bütün halları tutur + bir-birinə qarışmır.**

---

### 🔁 Bu zaman ehtimallarının cəmi **mütləq 1 olur**:

$$
P(A_1) + P(A_2) + \cdots + P(A_n) = 1
$$

Çünki birlikdə bütün ehtimal sahəsini təşkil edirlər.

---

## 🎲 Nümunə 1: Zar atılması

Sample space:  
$$
S = \{1, 2, 3, 4, 5, 6\}
$$

Hadisələr:  
- $A_1 = \{1, 2\}$  
- $A_2 = \{3, 4\}$  
- $A_3 = \{5, 6\}$

✅ Bunlar:
- **Cüt-cüt uyğunmayan** hadisələrdir: $A_1 \cap A_2 = \emptyset$, və s.  
- Birlikdə **S**-i tam əhatə edir: $A_1 \cup A_2 \cup A_3 = S$

📌 Nəticə: $A_1, A_2, A_3$ **tam qrup** təşkil edir.

---

## 📘 Niyə vacibdir?

Tam qrup hadisələri ehtimalın **tamamlayıcı** və **şərtli ehtimal** mövzularında istifadə olunur.

Məsələn:

$$
P(B) = P(B \cap A_1) + P(B \cap A_2) + \cdots + P(B \cap A_n)
$$

və ya

$$
P(B) = P(A_1) \cdot P(B|A_1) + \cdots + P(A_n) \cdot P(B|A_n)
$$

📌 Bu, **tam ehtimallar düsturudur** (total probability formula).

---

İstəyirsənsə, bu tərifi bir neçə **praktik nümunə ilə** də yaza bilərik (məsələn, hava vəziyyəti, test nəticəsi və s.) — hansını seçək?




---

## 🎓 Qısa Cədvəl:

| Xassə növü              | Formula                            | İzah                                      |
|-------------------------|-------------------------------------|-------------------------------------------|
| Sərhəd xassəsi          | $$0 \leq P(A) \leq 1$$           | Ehtimal hər zaman 0 ilə 1 arasındadır     |
| Tam hadisə              | $$P(S) = 1$$                     | Nəsə mütləq baş verəcək                   |
| İmkansız hadisə         | $$P(\emptyset) = 0$$            | Heç vaxt baş verməyən hadisə              |
| Əlavəlik xassəsi        | $$P(A \cup B) = P(A) + P(B)$$   | A və B uyğunmursa                         |
| Tərs hadisə             | $$P(\bar{A}) = 1 - P(A)$$       | A baş vermirsə, 1-dən çıxılır             |
| Əlavə düzəliş           | $$P(A \cup B) = P(A) + P(B) - P(A \cap B)$$ | A və B uyğunursa                          |
| Alt-hadisə üstünlüyü    | $$A \subseteq B \Rightarrow P(A) \leq P(B)$$ | Kiçik hadisənin ehtimalı az olar         |

## 🧠 Kombinasiya nədir və niyə istifadə olunur?

**Kombinasiya** (C), seçimin **sırasının vacib olmadığı** hallarda istifadə olunur.

Yəni biz **nə seçildiyinə** baxırıq, **necə seçildiyinə** yox.

Məsələn: 3 nəfərlik bir komitə yaratmaq istəyirik → A, B, C → bu qrup A-B-C və C-B-A fərqli sayılmır, çünki **eyni üzvlərdir**.

🔹 Kombinasiya formulu:  
$$
C_n^k = \frac{n!}{k!(n-k)!}
$$

Burda:
- $n$: ümumi element sayı  
- $k$: seçilənlərin sayı

---

## 🔍 İndi şəkildəki 2 məsələni izah edək:

---

## ✅ **Məsələ 1:**

- Cəmi: $N$ ədəd  
- Standart (istədiyimiz): $n$  
- Götürülür: $m$  
- Sual: Götürülən $m$ ədədin içində **$n$-si standart olsa**, ehtimal nə qədər olar?

📌 Yəni biz $m$ sayda seçim edirik, bu seçimin içində istədiyimiz tipdən **$n$ dənə** olmalıdır.

---

#### 🔶 Ehtimal formulu:

$$
P(A) = \frac{C_n^m}{C_N^m}
$$

**İzah:**
- Aşağıda ($C_N^m$) — ümumilikdə $N$ şeydən $m$-ni seçməyin **bütün mümkün yollarının** sayı.
- Yuxarıda ($C_n^m$) — yalnız **istədiyimiz $n$ dənə element arasından $m$-ni** seçməyin yolları.

📌 Əgər **istədiyin şeylər içindən** seçim edirsənsə, bu çox sadə kombinasiya olur.

---

## ✅ Nümunə 1 – Sadə kombinasiya (yalnız istədiyinlərdən seçmək)

### 💬 Sual:
Bir qutuda **10 dənə standart** lampa var. Bu 10 lampadan **4 dənəsini** seçirik.  
Sual: **Təsadüfi olaraq seçilən bu 4 lampanın** **hamısının standart olması** ehtimalı nədir?

---

### ✅ Həll:
- **Cəmi elementlər**: $N = 10$
- **Standartlar**: $n = 10$
- **Götürülən**: $m = 4$

> Biz 10 standartdan 4 dənəsini seçirik.

$$
P(A) = \frac{C_{10}^{4}}{C_{10}^{4}} = \frac{210}{210} = 1
$$

### 📌 Nəticə:
Ehtimal **1-dir**, çünki **hamısı standartdır**, və yalnız onlardan seçim edirik.


---

## ✅ **Məsələ 2:**

- Cəmi: $N$  
- Standart (istədiyimiz): $n$  
- Götürülür: $m$  
- Standartdan: $k$ ədəd olsun (bu dəfə $m \ne n$)

📌 Burda vəziyyət bir az daha mürəkkəbdir: sən **$m$** ədəd götürürsən, **onların içində** **dəqiq $k$ dənəsi** standartdan olsun.

---

#### 🔶 Ehtimal formulu:

$$
P(A) = \frac{C_n^k \cdot C_{N-n}^{m-k}}{C_N^m}
$$

**İzah:**
- **Yuxarıda**:  
  - $C_n^k$: İstədiyin (standart) $n$ içindən $k$ dənəsini seçirsən,  
  - $C_{N-n}^{m-k}$: Qalanların içindən (standart olmayanlardan) **$m-k$** ədəd seçirsən.

- **Aşağıda**:  
  - $C_N^m$: Ümumilikdə $N$ şeydən $m$-ni seçməyin yolları.

📌 Bu formada kombinasiya, **şərtlə seçimlərdə** çox faydalıdır, çünki **təkcə neçə seçimin olduğunu** yox, **hansı şərtlə seçildiyini** də hesablamaq olur.

---

## 🧠 Nəticə: Niyə Kombinasiya?

- Kombinasiya **sıra önəmsizdirsə** istifadə olunur.  
- Bu məsələlərdə **kimin gəldiyi vacibdir**, amma **sırası vacib deyil**.  
- Ona görə də **tərtibat yox**, **kombinasiya** (C) istifadə olunur.

---

## ✅ Nümunə 2 – Qarışıq seçim (standart + qeyri-standart)

### 💬 Sual:
Bir qutuda **15 lampa** var:
- **9-u standartdır**,  
- **6-sı standart deyil.**  
Bu lampalardan **5 dənəsini** seçirik.

Sual: **Seçilənlərin içində dəqiq 3-ü standart olsun** ehtimalı nədir?

---

### ✅ Həll:

- Cəmi: $N = 15$
- Standart: $n = 9$
- Götürülən: $m = 5$
- Standartlardan seçilən: $k = 3$

İndi formulu tətbiq edirik:

$$
P(A) = \frac{C_9^3 \cdot C_6^2}{C_{15}^5}
$$

İndi hesablayaq:

- $C_9^3 = \frac{9!}{3!6!} = 84$
- $C_6^2 = \frac{6!}{2!4!} = 15$
- $C_{15}^5 = \frac{15!}{5!10!} = 3003$

$$
P(A) = \frac{84 \cdot 15}{3003} = \frac{1260}{3003} \approx 0.4196
$$

---

### 📌 Nəticə:
Bu seçimlə ehtimal təxminən **41.96%** olur ki, 5 lampa içində **dəqiq 3-ü standart olsun**.

Çünki:
- **$C_9^3$** → 9 standart lampadan 3-ünü seçməyin yolları,
- **$C_6^2$** → 6 qeyri-standart lampadan 2-ni seçməyin yolları,
- **$C_{15}^5$** → ümumilikdə 15 lampadan 5-ni seçməyin bütün mümkün yolları.

Formul:  
$$
P(A) = \frac{\text{ƏLAQƏLİ hallar}}{\text{MÜMKÜN hallar}}
$$
