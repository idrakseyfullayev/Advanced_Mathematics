## Binomial Paylanma


### ✅ **1. Binomial Paylanmanın Tərifi**

>**Tərif:** Binomial paylanma, sabit uğur ehtimalı ilə `n` sayda **eyni tipli və müstəqil** sınaqların nəticəsində **müəyyən sayda uğur** əldə etmə ehtimalını modelləşdirir.

>**Tərif:** Binomial paylanma, **müstəqil və eyni ehtimallı təcrübələrin müəyyən sayda uğur ehtimalını ölçən diskret ehtimal paylanmasıdır.**


---

### ✅ Parametrlər**

* `n`: Sınaqların ümumi sayı
* `p`: Hər bir sınaqda uğur ehtimalı
* `k`: Uğur sayının ehtimalı soruşulan dəyər
* `q = 1 - p`: Uğursuzluq ehtimalı




**Binomial paylanma** adətən belə bir situasiyanı təsvir edir:

* Hər bir sınaqda yalnız iki nəticə mümkündür: **Uğur (Success)** və **Uğursuzluq (Failure)**.
* Sınaqlar **bir-birindən asılı deyil (müstəqildir)**.
* Hər bir sınaqda uğur ehtimalı eynidir və bu ehtimal **p** ilə göstərilir.
* Uğursuzluq ehtimalı isə **q = 1 - p** olacaq.
* Sınaqların sayı **n** ilə, uğurların sayı isə **k** ilə göstərilir.

Belə bir halda təsadüfi dəyişən \$X \sim Bin(n, p)\$ olur — yəni \$X\$ dəyişəni binomial paylanmaya tabedir.

---

### ✅ 2. Binomial Paylanmanın Düsturu

**Ehtimal kütlə funksiyası (PMF)** aşağıdakı kimidir:

$$
P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k}
$$

### 🎯 Burada hər simvol nəyi ifadə edir?

- \$\bullet\$ \$n\$ — Sınaqların ümumi sayı
- \$\bullet\$ \$k\$ — Uğur sayısı
- \$\bullet\$ \$p\$ — Hər bir sınaqda uğur ehtimalı
- \$\bullet\$ \$1 - p\$ — Uğursuzluq ehtimalı
- \$\bullet\$ \$\binom{n}{k}\$ — \$n\$ sınaqdan \$k\$ uğuru seçmək üçün kombinasiya: \$\frac{n!}{k!(n-k)!}\$


### ✅ **. Məsələn:**

**Sual:** Bir adam 10 dəfə zar atır. Hər atışda 6 gəlmə ehtimalı 1/6-dır. 10 atışda **dəqiq 3 dəfə** 6 gəlmə ehtimalı nə qədərdir?

**Cavab:**

$$
n = 10,\quad k = 3,\quad p = \frac{1}{6}
$$

$$
P(X=3) = \binom{10}{3} \cdot \left(\frac{1}{6}\right)^3 \cdot \left(\frac{5}{6}\right)^7
$$

$$
P(X=3) ≈ 0.155
$$

---





### ✅ 3. Riyazi Gözləmə (Expected Value)

Binomial paylanmanın gözlənilən dəyəri (ortalama nəticə):

$$
\mathbb{E}[X] = n \cdot p
$$

### 💡 Niyə belədir?

Hər sınaqda uğur ehtimalı **p**-dir. **n** sınaq varsa, ümumi gözlənilən uğur sayı, yəni orta uğur sayı **n dəfə p** qədər olacaq.

Məsələn:

* 10 dəfə zar atırsan və hər dəfə 6 gəlmə ehtimalı \$\frac{1}{6}\$-dır.
* Onda orta hesabla \$10 \cdot \frac{1}{6} = 1.67\$ dəfə 6 gəlməsi gözlənilir.

---

### ✅ 4. Dispersiya (Varyans)

Binomial paylanmanın dispersiyası:

$$
Var(X) = n \cdot p \cdot (1 - p)
$$

### 💡 Niyə bu düsturdur?

Bu düstur dəyişkənliyin (dağılımın) ölçüsüdür. Aşağıdakı səbəblərə görə bu formadadır:

1. **Dəyişkənlik iki faktordan asılıdır:**

   * Daha çox sınaq varsa (n çoxdursa), dəyişkənlik artır.
   * Uğur ehtimalı 0 və 1-dən nə qədər uzaqdırsa, yəni p orta dəyərlərə (0.5) yaxın olduqda dispersiya maksimum olur.

2. **Binomial paylanma əslində Bernoulli paylanmalarının cəmi** kimidir.

Yəni:

$$
X = X_1 + X_2 + \ldots + X_n \quad \text{(burada hər } X_i \text{ Bernoulli dəyişənidir)}
$$

Bernoulli dəyişən üçün:

* \$\mathbb{E}\[X\_i] = p\$
* \$Var(X\_i) = p(1 - p)\$

Toplam  \$n\$ Bernoulli dəyişəni varsa, onların dispersiyası da cəmlənir:

$$
Var(X) = \sum Var(X_i) = n \cdot p \cdot (1 - p)
$$

### 🧠 Əlavə intuitiv izah:

Bernoulli dəyişəni yalnız **0** və **1** ala bilər. Əgər uğur ehtimalı \$p\$-dirsə, dəyişkənlik (sıçrama) yalnız **0** və **1** arasında olur.

Əgər:

* Uğur ehtimalı \$p = 0\$ və ya \$p = 1\$ olsa, dəyişkənlik **yoxdur** (çünki həmişə 0 və ya həmişə 1 gəlir) → \$Var = 0\$
* Ən böyük dəyişkənlik **p = 0.5** olanda olur — bu halda:

  * \$\mathbb{E}\[X] = 0.5\$
  * Amma dəyərlər ya 0 gəlir (fərq: \$0 - 0.5 = -0.5\$), ya da 1 gəlir (fərq: \$1 - 0.5 = 0.5\$) → fərqlərin kvadratı: \$(0.5)^2 = 0.25\$
  * Gözlənilən kvadrat fərq də bu olur: \$Var(X\_i) = p(1 - p)\$

Bu səbəbdən də ümumi dəyişkənlik \$n \cdot p \cdot (1 - p)\$ formasındadır.

---

| ------------- | ------------------------------------------- | ------------------------------- |
| Paylanma tipi | $$X \sim Bin(n, p)$$                        | n sınaqda p ehtimalla uğur sayı |
| PMF           | $$P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}$$ | k uğurun ehtimalı               |

| Dispersiya    | $$Var(X) = n \cdot p \cdot (1 - p)$$        | Dəyişkənlik ölçüsü              |




## Binomial paylanmanın Dispersiyasını (Variance) niyə $n \cdot p \cdot (1-p)$ olduğunu anlayaq



### 1. Binomial paylanmanın əsasları

Binomial paylanma $n$ sayda **müstəqil** Bernoulli təcrübəsinin (hər birində uğur ehtimalı $p$, uğursuzluq ehtimalı $1-p$) cəmidir:

$$
X = \sum_{i=1}^n X_i
$$

Burada hər $X_i$ yalnız 0 və ya 1 dəyərini ala bilər.

---

### 2. Bernoulli dəyişənin gözlənilən dəyəri və dispersiyası

Bernoulli dəyişəninin (yəni, $X_i \in \{0,1\}$) gözlənilən dəyəri:

$$
E(X_i) = 1 \cdot p + 0 \cdot (1-p) = p
$$

Dispersiyası isə:

$$
Var(X_i) = E(X_i^2) - (E(X_i))^2
$$

Amma $X_i$ yalnız 0 və 1 dəyərini aldığından, $X_i^2 = X_i$, çünki $0^2=0$ və $1^2=1$.

Beləliklə:

$$
E(X_i^2) = E(X_i) = p
$$

O zaman dispersiya:

$$
Var(X_i) = p - p^2 = p(1-p)
$$

---

### 3. Binomial dəyişənin dispersiyası

Binomial dəyişən $X$ $n$ müstəqil Bernoulli dəyişənin cəmidir, ona görə:

$$
Var(X) = Var\left(\sum_{i=1}^n X_i \right) = \sum_{i=1}^n Var(X_i) = n \cdot p(1-p)
$$

---

### **Nəticə:**

* $p$ — uğur ehtimalı,
* $1-p$ — uğursuzluq ehtimalı,
* Dispersiya həm uğurun, həm də uğursuzluğun ehtimalını nəzərə alır. Ona görə də $p(1-p)$ olur.

---

### Intuitiv izah

Dispersiya dəyişənin **dəyişkənliyini** ölçür.

* Əgər $p$ çox kiçikdir (məsələn, 0.01), deməli çox nadir uğur olur, $p(1-p) \approx 0.01 \times 0.99 \approx 0.0099$ — kiçik dispersiya, çünki nəticələr çox vaxt 0 olur.

* Əgər $p = 0.5$ olsa, $p(1-p) = 0.25$ — maksimum dispersiya, çünki 0 və 1 nəticələri bərabər ehtimalla olur, yəni dəyişkənlik maksimumdur.

## Poisson Paylanma



### ✅ **1. Poisson Paylanmanın Tərifi**

> **Tərif:** Poisson paylanması, müəyyən zaman intervalında və ya sahədə **müstəqil şəkildə baş verən hadisələrin sayını** modelləşdirən **diskret ehtimal paylanmasıdır**.

> **Sadə dillə:** Əgər bir hadisə **təsadüfi və nadir** hallarda baş verirsə, məsələn 1 saatda gələn zənglərin sayı, gündə düşən meteorların sayı, 1 səhifədə olan yazı səhvləri — **Poisson paylanması** istifadə olunur.

---

### ✅ **2. Parametrlər**

* `λ` (lambda): Gözlənilən hadisə sayı (orta), yəni **məhdud sahədə və ya vaxtda baş verən hadisələrin orta sayı**.

---

### ✅ **3. Poisson Paylanmanın Düsturu**

**Ehtimal kütlə funksiyası (PMF):**

$$
P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}
$$

---

### 🎯 Burada hər simvol nə deməkdir?

* \$\lambda\$ — Gözlənilən hadisə sayı (orta), məsələn: saatda 5 zəng → λ = 5
* \$k\$ — Hadisələrin konkret sayı (axtardığımız dəyər)
* \$e\$ — Təqribən 2.718 olan riyazi sabit
* \$k!\$ — k faktorial

---

### ✅ **4. Məsələn**

**Sual:** Bir call center saatda orta hesabla 3 zəng alır. Gələn saat ərzində **dəqiq 2 zəng** gəlməsi ehtimalı nə qədərdir?

**Cavab:**

$$
\lambda = 3, \quad k = 2
$$

$$
P(X = 2) = \frac{3^2 \cdot e^{-3}}{2!} = \frac{9 \cdot e^{-3}}{2} ≈ 0.224
$$

---

### ✅ **5. Gözləmə və Dispersiya**

Poisson paylanmasının **gözlənilən dəyəri və dispersiyası**:

| Xüsusiyyət | Düstur                       |
| ---------- | ---------------------------- |
| Gözləmə    | $$\mathbb{E}[X] = \lambda$$ |
| Dispersiya | $$Var(X) = \lambda$$         |

💡 Yəni **orta dəyər və dispersiya eynidir**.

---

### ✅ **6. Poisson və Binomial Fərqi**

| Xüsusiyyət       | Binomial                         | Poisson                               |
| ---------------- | -------------------------------- | ------------------------------------- |
| Sınaq sayı `n`   | Məlum və sonludur                | Yoxdur                                |
| Ehtimal `p`      | Sabit və məlum                   | Yoxdur (əvəzində λ var)               |
| Parametrlər      | `n`, `p`                         | `λ`                                   |
| Model məqsədi    | `n` sınaqda `k` uğur ehtimalı    | Müəyyən vaxt/sahədə `k` hadisə        |
| Formula          | \$\binom{n}{k} p^k (1-p)^{n-k}\$ | \$\frac{\lambda^k e^{-\lambda}}{k!}\$ |
| Qeyri-məhdud `k` | Yox, `k ≤ n`                     | Bəli, `k = 0, 1, 2, ...`              |
| Misal            | 10 dəfə zar atmaq                | 1 saatda neçə zəng gəlməsi            |

---

### ✅ 7. Qısa Cədvəl — Yekun

| Xüsusiyyət    | Düstur                                         | Açıqlama                              |
| ------------- | ---------------------------------------------- | ------------------------------------- |
| Paylanma tipi | $$X \sim Pois(\lambda)$$                         | Gözlənilən hadisə sayı λ olan Poisson |
| PMF           | $$P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}$$ | k hadisənin ehtimalı                  |
| Gözləmə       | $$\mathbb{E}[X] = \lambda$$                      | Orta hadisə sayı                      |
| Dispersiya    | $$Var(X) = \lambda$$                             | Hadisə sayının dəyişkənliyi           |

---



## Poisson paylanmasının Riyazi Gözləmə və Dispersiyasının niyə `λ` olduğunu izah edək



### ✅ 1. Poisson dəyişənin tərifi

Poisson dəyişən \$X \sim \text{Pois}(\lambda)\$ — müəyyən zaman və ya sahədə **orta hadisə sayı λ** olan təsadüfi dəyişəndir. Yəni:

$$
P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}, \quad k = 0, 1, 2, \ldots
$$

---

### ✅ 2. Gözlənilən dəyər (Expected Value)

Poisson dəyişən üçün gözlənilən dəyər belədir:

$$
\mathbb{E}[X] = \lambda
$$

#### 🔍 **Sübut:**

Poisson paylanmada ehtimal funksiyası belədir:

$$
P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}, \quad k = 0, 1, 2, \ldots
$$

Gözlənilən dəyər:

$$
\mathbb{E}[X] = \sum_{k=0}^{\infty} k \cdot P(X = k) = \sum_{k=0}^{\infty} k \cdot \frac{\lambda^k e^{-\lambda}}{k!}
$$

İndi bu ifadəni sadələşdirək. Qeyd edək ki, \$k = 0\$ üçün \$k \cdot \dots = 0\$ olduğuna görə, cəmi \$k = 1\$-dən başlada bilərik:

$$
\mathbb{E}[X] = \sum_{k=1}^{\infty} \frac{k \lambda^k e^{-\lambda}}{k!}
$$

İfadəni bir az dəyişərək:

$$
= \sum_{k=1}^{\infty} \frac{\lambda \cdot \lambda^{k-1} e^{-\lambda}}{(k-1)!}
$$

Burada \$\lambda\$ sabitdir və cəmdən kənara çıxarılır:

$$
= \lambda e^{-\lambda} \sum_{k=1}^{\infty} \frac{\lambda^{k-1}}{(k-1)!}
$$

Substitusiya edək: \$m = k - 1 \Rightarrow k = m + 1\$ və \$m = 0, 1, 2, \dots\$:

$$
= \lambda e^{-\lambda} \sum_{m=0}^{\infty} \frac{\lambda^m}{m!}
$$

Bu isə **eksponensial funksiyanın tərifi** ilə bərabərdir:

$$
\sum_{m=0}^{\infty} \frac{\lambda^m}{m!} = e^{\lambda}
$$

Onda:

$$
\mathbb{E}[X] = \lambda e^{-\lambda} \cdot e^{\lambda} = \lambda
$$

✅ **Nəticə:** \$\mathbb{E}\[X] = \lambda$

---


### ✅ 3. Dispersiyanın düsturu

Dispersiya:

$$
Var(X) = \mathbb{E}[X^2] - (\mathbb{E}[X])^2
$$

Deməli, bizə **\$\mathbb{E}\[X^2]\$** lazımdır.

Amma gəlin başqa bir daha **sadə və anlayışlı** yoldan gedək: **Poisson dəyişəni əslində limit kimi yaranır**.



### ✅ Poisson → Binomial limitindən baxış

Poisson paylanması əslində **Binomial paylanmanın limit halıdır**, yəni:

Əgər \$X \sim \text{Bin}(n, p)\$ və \$n \to \infty\$, \$p \to 0\$ elə ki, \$n \cdot p = \lambda\$ sabit qalsın, o zaman:

$$
\text{Bin}(n, p) \longrightarrow \text{Pois}(\lambda)
$$



### ✅ Binomialdan nəticə çıxaraq

Binomial dəyişən üçün:

$$
Var(X) = n \cdot p \cdot (1 - p)
$$

Əgər \$p \to 0\$, o zaman \$(1 - p) \to 1\$ və \$n \cdot p \to \lambda\$

Onda:

$$
Var(X) = n \cdot p \cdot (1 - p) \approx n \cdot p = \lambda
$$

Bu limitdən görə bilirik ki, **Poisson paylanmanın dispersiyası da gözlənilən dəyəri kimi `λ` olur**:

$$
Var(X) = \lambda
$$

---

### ✅ 6. Intuitiv izah

* Poisson paylanması — “nadir, amma təsadüfi” hadisələri modelləşdirir.
* Belə hadisələrdə həm **orta gözləmə sayı**, həm də **dəyişkənlik** təxminən eyni olur.

Məsələn:

* Gündə orta hesabla 2 dəfə mesaj gəlir → həm gözləmə, həm də dəyişkənlik \$\approx 2\$ olur.
* Günə 20 zəng gəlirsə — orta da 20, dəyişkənlik də o qədərdir.

🧠 Niyə?

Çünki bu hadisələr **tamamilə təsadüfi** və **bir-birindən asılı olmayan** şəkildə baş verdiyindən, dəyişkənlik hadisələrin ortalama tezliyi ilə **eyni tempdə artır**.

---

| Xüsusiyyət | Dəyər                        | İzah                                                |
| ---------- | ---------------------------- | --------------------------------------------------- |

| Dispersiya | $$Var(X) = \lambda$$         | Dəyişkənlik hadisələrin ortası ilə eyni ölçüdə olur |

## Həndəsi Paylanma



### ✅ **1. Həndəsi Paylanmanın Tərifi**

> **Tərif:** Həndəsi paylanma — uğurun sabit ehtimalla baş verdiyi müstəqil sınaqlarda **ilk uğurun neçəinci sınaqda baş verəcəyini** modelləşdirir.

> **Başqa sözlə:** Həndəsi paylanma, **ilk uğur** əldə olunana qədər olan **sınaqların sayını** göstərən **diskret ehtimal paylanmasıdır.**

---

### ✅ **2. Parametrlər**

* `p`: Hər bir sınaqda uğur ehtimalı
* `q = 1 - p`: Uğursuzluq ehtimalı
* `k`: İlk uğurun baş verdiyi sınaq nömrəsi (1, 2, 3, ...)

📌 Yəni:

* Hər sınaq **uğur (1)** və ya **uğursuzluq (0)** ilə nəticələnir.
* Sınaqlar **müstəqildir**.
* Sual budur: **İlk uğur neçəinci dəfə baş verəcək?**

---

### ✅ **3. Ehtimal Kütlə Funksiyası (PMF)**

Həndəsi paylanmanın ehtimal funksiyası:

$$
P(X = k) = (1 - p)^{k - 1} \cdot p
$$

📌 Yəni: İlk \$k-1\$ sınaq uğursuz olmalı, və **k-cı** sınaqda **ilk uğur** baş verməlidir.

---

### ✅ **4. Məsələn**

**Sual:** Bir adam təkrar-təkrar zar atır. Hər atışda 6 gəlmə ehtimalı \$\frac{1}{6}\$-dır. İlk dəfə 6 gəlməsinin **4-cü atışda** baş vermə ehtimalı nə qədərdir?

**Cavab:**

* \$p = \frac{1}{6}, \quad k = 4\$
* İlk 3 sınaqda **uğursuzluq** (yəni 6 gəlməməlidir): \$\left(\frac{5}{6}\right)^3\$
* 4-cü atışda **6 gəlməlidir**: \$\left(\frac{1}{6}\right)\$

$$
P(X = 4) = \left(\frac{5}{6}\right)^3 \cdot \left(\frac{1}{6}\right) ≈ 0.096
$$

---

### ✅ **5. Riyazi Gözləmə (Expected Value)**

Həndəsi paylanmanın gözlənilən dəyəri:

$$
\mathbb{E}[X] = \frac{1}{p}
$$

📌 Yəni, ortalama olaraq ilk uğur **1/p** sınaqdan sonra baş verir.

**Misal:**
Hər atışda 6 gəlmə ehtimalı \$\frac{1}{6}\$-dır → orta hesabla **6 atışdan sonra** ilk 6 gələcək.

---

### ✅ **6. Dispersiya (Varyans)**

Dispersiya (dağılımın ölçüsü):

$$
Var(X) = \frac{1 - p}{p^2}
$$

📌 Yəni ehtimal kiçikdirsə, ilk uğurun başvermə müddəti çox dəyişkən ola bilər.

---

### 🧠 **Əlavə İntuitiv İzah**

* Hər sınaq 2 nəticəlidir (uğur / uğursuzluq)
* İstədiyimiz odur ki, **birinci uğur** neçəinci dəfə baş verəcək?
* **Əgər p çox balacadırsa** (uğur nadir baş verirsə), uğur çox gözlədilə bilər → gözləmə və dispersiya böyükdür.
* **Əgər p böyükdürsə**, uğur tez baş verəcək → gözləmə və dispersiya kiçikdir.

---

### ✅ 7. Qısa Cədvəl — Yekun

| Xüsusiyyət    | Düstur                               | Açıqlama                                |
| ------------- | ------------------------------------ | --------------------------------------- |
| Paylanma tipi | $$X \sim Geom(p)$$                    | İlk uğurun baş verdiyi sınaq nömrəsi    |
| PMF           | $$P(X = k) = (1 - p)^{k - 1} \cdot p$$ | İlk uğurun k-cı dəfə baş vermə ehtimalı |
| Gözləmə       | $$\mathbb{E}[X] = \frac{1}{p}$$        | Orta uğur əldə etmə müddəti             |
| Dispersiya    | $$Var(X) = \frac{1 - p}{p^2}$$         | Uğura qədər olan dəyişkənlik            |




## Həndəsi Paylanmada niyə Riyazi Gözləmə  $E(X) = \frac{1}{p}$, Dispersiya isə $Var(X) = \frac{1-p}{p^2}$ olur



### ✅ 1. Gözlənilən dəyər (niyə \$E(X) = \frac{1}{p}\$?)

### 📌 Problem:

> İlk uğur neçəci təcrübədə baş verir?

### 🧠 Fikir:

Əgər uğur ehtimalı \$p\$-dirsə, onda "orta hesabla neçə dəfə cəhd etməliyik ki, bir uğur olsun?" sualının cavabı budur.

---

### 📐 Riyazi sübut:

Əgər \$X \sim \text{Geom}(p)\$-dirsə, onda ehtimal funksiyası:

$$
P(X = k) = (1-p)^{k-1} \cdot p
$$

Gözlənilən dəyər belə hesablanır:

$$
E(X) = \sum_{k=1}^{\infty} k \cdot (1-p)^{k-1} \cdot p
$$

Bu **həndəsi silsilənin** xüsusi cəmi kimi tanınır və nəticə belə çıxır:

$$
E(X) = \frac{1}{p}
$$

---

### 🤯 Intuitiv izah:

Tutaq ki, sənin uğur ehtimalın \$p = 0.2\$-dir. Yəni:

* 1 cəhddə uğur şansı: 20%
* 2 cəhddə uğur: 16%
* 3 cəhddə uğur: 12.8%
* və s.

Belə çıxır ki, **çox nadir uğur** olur.

Onda təbii sual:

> Orta hesabla neçə dəfə cəhd etməliyik ki, uğur baş versin?

**Əgər 1 cəhddə 20% uğur varsa → 100 cəhddə 20 uğur gözlənilir → hər uğura 5 cəhd düşür.**

Yəni:

$$
E(X) = \frac{1}{p}
$$

---

### ✅ 2. Dispersiya (niyə $Var(X) = \frac{1-p}{p^2}$?)

### 📐 Riyazi sübut:

Əvvəl \$E(X)\$ bilirik:

$$
E(X) = \frac{1}{p}
$$

Sonra isə:

$$
E(X^2) = \sum_{k=1}^{\infty} k^2 \cdot (1-p)^{k-1} \cdot p = \frac{1 + p}{p^2}
$$

Onda dispersiya:

$$
Var(X) = E(X^2) - (E(X))^2 = \frac{1 + p}{p^2} - \left(\frac{1}{p}\right)^2 = \frac{1 - p}{p^2}
$$

---

### 🤯 Intuitiv izah:

Dispersiya **dəyişkənliyin ölçüsüdür**. Yəni dəyərlər ortalama ətrafında necə "yayılıb".

Əgər \$p\$ **böyükdür** (məsələn, 0.8), uğur tez-tez baş verir. Deməli nəticə çox dəyişmir: bəzən 1, bəzən 2 olur → **az dispersiya**

Əgər \$p\$ **kiçikdir** (məsələn, 0.05), uğur çox nadirdir. Bəzən 3-cü təcrübədə, bəzən 100-cü təcrübədə uğur ola bilər → **böyük dispersiya**

Bu da göstərir ki:

$$
Var(X) = \frac{1-p}{p^2}
$$

Burada:

* \$p^2\$ aşağıda olduğuna görə \$p\$ azaldıqca dispersiya **çox artır**.
* Yəni \$p\$ → 0 olduqda dispersiya \$→ \infty\$

---

### 📌 Cəmləşdirsək:

| **Baxış**        | **Dəyərlər**                                          |
| ---------------- | ----------------------------------------------------- |
| Gözlənilən dəyər | $$E(X) = \frac{1}{p}$$                                |
| Dispersiya       | $$Var(X) = \frac{1-p}{p^2}$$                          |
| Gözlənti niyə?   | Orta hesabla uğura qədər \$1/p\$ qədər sınaq lazımdır |
| Dispersiya niyə? | Uğur gec gəldikcə nəticələr daha çox "yayılır"        |

## Hiperhəndəsi Paylanma



### ✅ **1. Hiperhəndəsi Paylanmanın Tərifi**

> **Tərif:** Hiperhəndəsi paylanma, sonlu ölçülü bir qrupdan (populyasiyadan) **əvəzsiz** şəkildə seçilən nümunə içərisində müəyyən sayda uğur baş vermə ehtimalını modelləşdirən **diskret ehtimal paylanmasıdır**.

> **Sadə dillə:** Təsadüfi olaraq **əvəzsiz** (yəni geri qoymadan) seçim etdikdə, müəyyən sayda uğur əldə etmə ehtimalını ölçür.

---

### ✅ **2. Parametrlər**

| Simvol | Mənası                                          |
| ------ | ----------------------------------------------- |
| `N`    | Ümumi əşya (populyasiya) sayı                   |
| `K`    | Populyasiyada uğurlu (maraqlı) obyektlərin sayı |
| `n`    | Seçilən nümunənin ölçüsü                        |
| `k`    | Nümunədə `k` uğur əldə etmə ehtimalı axtarılır  |

---

**Şərtlər:**

* Populyasiya **sonludur** və ölçüsü `N`-dir.
* Populyasiyada `K` ədəd **uğurlu** obyekt var, qalan `N-K` isə uğursuzdur.
* `n` ədəd **əvəzsiz** seçim edilir.
* Marağımız `k` ədəd uğur seçmək ehtimalıdır.

Bu zaman təsadüfi dəyişən aşağıdakı kimi yazılır:

$$
X \sim \text{Hypergeometric}(N, K, n)
$$

---

### ✅ **3. Hiperhəndəsi Paylanmanın Düsturu (PMF)**

$$
P(X = k) = \frac{\binom{K}{k} \cdot \binom{N - K}{n - k}}{\binom{N}{n}}
$$

### 🎯 Açıqlama:

* \$\binom{K}{k}\$ — Uğurlu obyektlərdən `k` ədəd seçməyin sayı
* \$\binom{N-K}{n-k}\$ — Uğursuzlardan qalan `n-k` ədəd seçməyin sayı
* \$\binom{N}{n}\$ — Ümumilikdə `n` obyekt seçməyin ümumi sayı

---

### ✅ **4. Məsələn:**

**Sual:** Bir qutuda 7 ağ, 5 qara kürə var (cəmi 12 kürə). Əvəzsiz 4 kürə seçilir. Bu 4 kürənin içində **dəqiq 2 ağ** olma ehtimalı nədir?

**Cavab:**

$$
N = 12,\quad K = 7,\quad n = 4,\quad k = 2
$$

$$
P(X = 2) = \frac{\binom{7}{2} \cdot \binom{5}{2}}{\binom{12}{4}} = \frac{21 \cdot 10}{495} = \frac{210}{495} ≈ 0.424
$$

---

### ✅ **5. Riyazi Gözləmə (Expected Value)**

$$
\mathbb{E}[X] = n \cdot \frac{K}{N}
$$

### 💡 Niyə belədir?

Əgər `n` ədəd seçim edirsənsə və ümumi populyasiyada uğur ehtimalı `K / N`-dirsə, orta hesabla **n dəfə bu ehtimal qədər** uğur gözlənilir. Bu, binomiala çox bənzəyir, lakin əvəzsiz seçim olduğu üçün hiperhəndəsidir.

---

### ✅ **6. Dispersiya (Varyans)**

$$
Var(X) = n \cdot \frac{K}{N} \cdot \left(1 - \frac{K}{N}\right) \cdot \frac{N - n}{N - 1}
$$

### 💡 Niyə bu düsturdur?

Bu düstur, **binomial dispersiyanın** düzəliş edilmiş formasıdır. Sonuncu vurğu — \$\frac{N - n}{N - 1}\$ — **əvəzsiz seçim** səbəbindən dəyişkənliyin azaldığını nəzərə alır. Yəni, əgər **əvəzi olsa idi**, binomial forması olardı:

$$
n \cdot \frac{K}{N} \cdot \left(1 - \frac{K}{N}\right)
$$

Amma əvəzsiz seçim olduğuna görə, dispersiya bu faktorla azaldılır.

---

### ✅ **7. Qısa Cədvəl — Yekun**

| Xüsusiyyət    | Düstur                                                                                      | Açıqlama                                        |
| ------------- | ------------------------------------------------------------------------------------------- | ----------------------------------------------- |
| Paylanma tipi | $$X \sim \text{Hypergeometric}(N, K, n)$$                                                     | N populyasiyada K uğurludan n-i seçmək          |
| PMF           | $$P(X = k) = \frac{\binom{K}{k} \cdot \binom{N-K}{n-k}}{\binom{N}{n}}$$                       | Nümunədə k uğur ehtimalı                        |
| Gözləmə       | $$\mathbb{E}[X] = n \cdot \frac{K}{N}$$                                                       | Orta uğur sayı                                  |
| Dispersiya    | $$Var(X) = n \cdot \frac{K}{N} \cdot \left(1 - \frac{K}{N}\right) \cdot \frac{N - n}{N - 1}$$ | Dəyişkənlik ölçüsü (əvəzsiz seçimi nəzərə alır) |



## Hiperhəndəsi Paylanmanın niyə Riyazi Gözləməsi $\mathbb{E}[X] = n \cdot \frac{K}{N}$   və Dispersiyası isə $Var(X) = n \cdot \frac{K}{N} \cdot \left(1 - \frac{K}{N}\right) \cdot \frac{N - n}{N - 1}$ olur?



### ✅ 1. Hiperhəndəsi Paylanmanın Təyinatı (qısa xatırlatma)

Hiperhəndəsi dəyişən belə təyin olunur:

$$
X \sim \text{Hypergeometric}(N, K, n)
$$

| Simvol | İzah                                                     |
| ------ | -------------------------------------------------------- |
| `N`    | Populyasiyanın ümumi sayı                                |
| `K`    | Populyasiyadakı uğurlu obyektlərin sayı                  |
| `n`    | Əvəzsiz seçilmiş nümunənin ölçüsü                        |
| `X`    | Bu nümunədəki uğurlu obyektlərin sayı (təsadüfi dəyişən) |

---

### ✅ 2. Gözlənilən Dəyər (Expected Value)

### 📌 **Düstur:**

$$
\mathbb{E}[X] = n \cdot \frac{K}{N}
$$

---

### ✅ **Sübut (Riyazi Yanaşma)**

Hiperhəndəsi dəyişəni belə başa düşək: Biz `n` ədəd obyekt seçirik və hər biri uğurlu ola bilər. Bunu **indikator dəyişənlər** ilə ifadə edək:

* Götürək:

  $$
  X = X_1 + X_2 + \cdots + X_n
  $$

$$
X_i =
\begin{cases}
1, & \text{əgər } i\text{-ci seçilən uğurludursa} \\
0, & \text{əks halda}
\end{cases}
$$

Bu ifadə deyir ki:

* $X_i = 1$ → əgər seçilən obyekt **uğurlu** (məsələn, qırmızı kürə) olarsa,
* $X_i = 0$ → əgər **uğursuz** olarsa.

Bu dəyişənlər **indikator dəyişənlərdir** — yəni yalnız 0 və 1 alırlar.


* Hər bir \$X\_i\$ dəyişəni **Bernoulli** tiplidir, amma **əvəzsiz seçim** olduğuna görə bu dəyişənlər bir-birindən **müstəqil deyil**.

Bununla belə, **xətti gözləmə xassəsinə** görə:

$$
\mathbb{E}[X] = \mathbb{E}[X_1 + \cdots + X_n] = \sum_{i=1}^{n} \mathbb{E}[X_i]
$$

İndi əsas məsələ \$\mathbb{E}\[X\_i]\$ tapmaqdır:

* Hər seçilmiş obyektin **uğurlu olma ehtimalı**:

$$
P(X_i = 1) = \frac{K}{N}
$$

(yəni, uğurlu olanların sayı / ümumi say)

Deməli:

$$
\mathbb{E}[X_i] = 1 \cdot \frac{K}{N} + 0 \cdot \left(1 - \frac{K}{N}\right) = \frac{K}{N}
$$

Və sonunda:

$$
\mathbb{E}[X] = n \cdot \frac{K}{N}
$$

---

### ✅ Intuitiv Yanaşma

* Tutaq ki, qutuda `N = 100` kürə var, bunun `K = 30`-u qırmızıdır.
* `n = 10` kürə əvəzsiz seçirsənsə, hər birinin qırmızı gəlmə ehtimalı ortalama `30/100 = 0.3`-dür.
* Orta hesabla `10 × 0.3 = 3` qırmızı kürə seçəcəyik.

---

### ✅ 3. Dispersiya (Varyans)

### 📌 **Düstur:**

$$
Var(X) = n \cdot \frac{K}{N} \cdot \left(1 - \frac{K}{N}\right) \cdot \frac{N - n}{N - 1}
$$

---

### ✅ Riyazi Sübut (Xülasə ilə)

Bərabər quruluşla yenə də \$X = \sum X\_i\$ yazırıq. Dispersiya üçün:

$$
Var(X) = \sum Var(X_i) + 2 \sum_{i < j} Cov(X_i, X_j)
$$

Əgər seçim **əvəzi** olsaydı, dəyişənlər müstəqil olardı və bu, **binomial dispersiya** ilə eyni olardı:

$$
n \cdot \frac{K}{N} \cdot \left(1 - \frac{K}{N}\right)
$$

Amma **əvəzsiz seçim** olduğuna görə dəyişkənlik bir qədər **azalır** və bu azalma \$\frac{N - n}{N - 1}\$ vurğu ilə ifadə olunur.

Yəni:

> Dispersiyanın düzəlişi:
>
> $$
> \text{Binomial varians} \cdot \left(\frac{N - n}{N - 1}\right)
> $$

---

### ✅ Intuitiv Yanaşma

* Əgər əvəzi seçim olsa idi — hər dəfə seçdiyimiz kürənin uğurlu olması ehtimalı **dəyişməz** qalardı.
* Amma **əvəzsiz** seçimdə — bir uğurlu seçdikdən sonra növbəti seçimin uğurlu olma ehtimalı **azalır**.
* Bu səbəbdən, **dispersiya azalır**, yəni nəticələr bir-birindən daha az “sıçrayır”.

---

### ✅ 4. Məsələn:

> **Sual:** Bir səbətdə 20 qələm var: 8-i işləkdir (uğurlu), 12-si xarabdır (uğursuz). Əvəzsiz şəkildə 5 qələm seçilir. Gözlənilən işlək qələm sayı və dispersiya nədir?

**Cavab:**

* \$N = 20,\quad K = 8,\quad n = 5\$

* Gözlənilən dəyər:

$$
\mathbb{E}[X] = 5 \cdot \frac{8}{20} = 2
$$

* Dispersiya:

$$
Var(X) = 5 \cdot \frac{8}{20} \cdot \frac{12}{20} \cdot \frac{15}{19} = 5 \cdot 0.4 \cdot 0.6 \cdot \frac{15}{19} ≈ 0.947
$$

---

### ✅ 5. Yekun Cədvəl

| Xüsusiyyət | Düstur                                                                                      | İzah                                             |
| ---------- | ------------------------------------------------------------------------------------------- | ------------------------------------------------ |
| Gözləmə    | $$\mathbb{E}[X] = n \cdot \frac{K}{N}$$                                                       | Nümunədəki orta uğur sayı                        |
| Dispersiya | $$Var(X) = n \cdot \frac{K}{N} \cdot \left(1 - \frac{K}{N}\right) \cdot \frac{N - n}{N - 1}$$ | Dəyişkənliyin əvəzsiz seçimlə azaldılmış forması |
