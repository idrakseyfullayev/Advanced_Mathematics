## Binomial Paylanma


### ‚úÖ **1. Binomial Paylanmanƒ±n T…ôrifi**

>**T…ôrif:** Binomial paylanma, sabit uƒüur ehtimalƒ± il…ô `n` sayda **eyni tipli v…ô m√ºst…ôqil** sƒ±naqlarƒ±n n…ôtic…ôsind…ô **m√º…ôyy…ôn sayda uƒüur** …ôld…ô etm…ô ehtimalƒ±nƒ± modell…ô≈üdirir.

>**T…ôrif:** Binomial paylanma, **m√ºst…ôqil v…ô eyni ehtimallƒ± t…ôcr√ºb…ôl…ôrin m√º…ôyy…ôn sayda uƒüur ehtimalƒ±nƒ± √∂l√ß…ôn diskret ehtimal paylanmasƒ±dƒ±r.**


---

### ‚úÖ Parametrl…ôr**

* `n`: Sƒ±naqlarƒ±n √ºmumi sayƒ±
* `p`: H…ôr bir sƒ±naqda uƒüur ehtimalƒ±
* `k`: Uƒüur sayƒ±nƒ±n ehtimalƒ± soru≈üulan d…ôy…ôr
* `q = 1 - p`: Uƒüursuzluq ehtimalƒ±




**Binomial paylanma** ad…ôt…ôn bel…ô bir situasiyanƒ± t…ôsvir edir:

* H…ôr bir sƒ±naqda yalnƒ±z iki n…ôtic…ô m√ºmk√ºnd√ºr: **Uƒüur (Success)** v…ô **Uƒüursuzluq (Failure)**.
* Sƒ±naqlar **bir-birind…ôn asƒ±lƒ± deyil (m√ºst…ôqildir)**.
* H…ôr bir sƒ±naqda uƒüur ehtimalƒ± eynidir v…ô bu ehtimal **p** il…ô g√∂st…ôrilir.
* Uƒüursuzluq ehtimalƒ± is…ô **q = 1 - p** olacaq.
* Sƒ±naqlarƒ±n sayƒ± **n** il…ô, uƒüurlarƒ±n sayƒ± is…ô **k** il…ô g√∂st…ôrilir.

Bel…ô bir halda t…ôsad√ºfi d…ôyi≈ü…ôn \$X \sim Bin(n, p)\$ olur ‚Äî y…ôni \$X\$ d…ôyi≈ü…ôni binomial paylanmaya tabedir.

---

### ‚úÖ 2. Binomial Paylanmanƒ±n D√ºsturu

**Ehtimal k√ºtl…ô funksiyasƒ± (PMF)** a≈üaƒüƒ±dakƒ± kimidir:

$$
P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k}
$$

### üéØ Burada h…ôr simvol n…ôyi ifad…ô edir?

- \$\bullet\$ \$n\$ ‚Äî Sƒ±naqlarƒ±n √ºmumi sayƒ±
- \$\bullet\$ \$k\$ ‚Äî Uƒüur sayƒ±sƒ±
- \$\bullet\$ \$p\$ ‚Äî H…ôr bir sƒ±naqda uƒüur ehtimalƒ±
- \$\bullet\$ \$1 - p\$ ‚Äî Uƒüursuzluq ehtimalƒ±
- \$\bullet\$ \$\binom{n}{k}\$ ‚Äî \$n\$ sƒ±naqdan \$k\$ uƒüuru se√ßm…ôk √º√ß√ºn kombinasiya: \$\frac{n!}{k!(n-k)!}\$


### ‚úÖ **. M…ôs…ôl…ôn:**

**Sual:** Bir adam 10 d…ôf…ô zar atƒ±r. H…ôr atƒ±≈üda 6 g…ôlm…ô ehtimalƒ± 1/6-dƒ±r. 10 atƒ±≈üda **d…ôqiq 3 d…ôf…ô** 6 g…ôlm…ô ehtimalƒ± n…ô q…ôd…ôrdir?

**Cavab:**

$$
n = 10,\quad k = 3,\quad p = \frac{1}{6}
$$

$$
P(X=3) = \binom{10}{3} \cdot \left(\frac{1}{6}\right)^3 \cdot \left(\frac{5}{6}\right)^7
$$

$$
P(X=3) ‚âà 0.155
$$

---





### ‚úÖ 3. Riyazi G√∂zl…ôm…ô (Expected Value)

Binomial paylanmanƒ±n g√∂zl…ônil…ôn d…ôy…ôri (ortalama n…ôtic…ô):

$$
\mathbb{E}[X] = n \cdot p
$$

### üí° Niy…ô bel…ôdir?

H…ôr sƒ±naqda uƒüur ehtimalƒ± **p**-dir. **n** sƒ±naq varsa, √ºmumi g√∂zl…ônil…ôn uƒüur sayƒ±, y…ôni orta uƒüur sayƒ± **n d…ôf…ô p** q…ôd…ôr olacaq.

M…ôs…ôl…ôn:

* 10 d…ôf…ô zar atƒ±rsan v…ô h…ôr d…ôf…ô 6 g…ôlm…ô ehtimalƒ± \$\frac{1}{6}\$-dƒ±r.
* Onda orta hesabla \$10 \cdot \frac{1}{6} = 1.67\$ d…ôf…ô 6 g…ôlm…ôsi g√∂zl…ônilir.

---

### ‚úÖ 4. Dispersiya (Varyans)

Binomial paylanmanƒ±n dispersiyasƒ±:

$$
Var(X) = n \cdot p \cdot (1 - p)
$$

### üí° Niy…ô bu d√ºsturdur?

Bu d√ºstur d…ôyi≈ük…ônliyin (daƒüƒ±lƒ±mƒ±n) √∂l√ß√ºs√ºd√ºr. A≈üaƒüƒ±dakƒ± s…ôb…ôbl…ôr…ô g√∂r…ô bu formadadƒ±r:

1. **D…ôyi≈ük…ônlik iki faktordan asƒ±lƒ±dƒ±r:**

   * Daha √ßox sƒ±naq varsa (n √ßoxdursa), d…ôyi≈ük…ônlik artƒ±r.
   * Uƒüur ehtimalƒ± 0 v…ô 1-d…ôn n…ô q…ôd…ôr uzaqdƒ±rsa, y…ôni p orta d…ôy…ôrl…ôr…ô (0.5) yaxƒ±n olduqda dispersiya maksimum olur.

2. **Binomial paylanma …ôslind…ô Bernoulli paylanmalarƒ±nƒ±n c…ômi** kimidir.

Y…ôni:

$$
X = X_1 + X_2 + \ldots + X_n \quad \text{(burada h…ôr } X_i \text{ Bernoulli d…ôyi≈ü…ônidir)}
$$

Bernoulli d…ôyi≈ü…ôn √º√ß√ºn:

* \$\mathbb{E}\[X\_i] = p\$
* \$Var(X\_i) = p(1 - p)\$

Toplam  \$n\$ Bernoulli d…ôyi≈ü…ôni varsa, onlarƒ±n dispersiyasƒ± da c…ôml…ônir:

$$
Var(X) = \sum Var(X_i) = n \cdot p \cdot (1 - p)
$$

### üß† ∆èlav…ô intuitiv izah:

Bernoulli d…ôyi≈ü…ôni yalnƒ±z **0** v…ô **1** ala bil…ôr. ∆èg…ôr uƒüur ehtimalƒ± \$p\$-dirs…ô, d…ôyi≈ük…ônlik (sƒ±√ßrama) yalnƒ±z **0** v…ô **1** arasƒ±nda olur.

∆èg…ôr:

* Uƒüur ehtimalƒ± \$p = 0\$ v…ô ya \$p = 1\$ olsa, d…ôyi≈ük…ônlik **yoxdur** (√ß√ºnki h…ômi≈ü…ô 0 v…ô ya h…ômi≈ü…ô 1 g…ôlir) ‚Üí \$Var = 0\$
* ∆èn b√∂y√ºk d…ôyi≈ük…ônlik **p = 0.5** olanda olur ‚Äî bu halda:

  * \$\mathbb{E}\[X] = 0.5\$
  * Amma d…ôy…ôrl…ôr ya 0 g…ôlir (f…ôrq: \$0 - 0.5 = -0.5\$), ya da 1 g…ôlir (f…ôrq: \$1 - 0.5 = 0.5\$) ‚Üí f…ôrql…ôrin kvadratƒ±: \$(0.5)^2 = 0.25\$
  * G√∂zl…ônil…ôn kvadrat f…ôrq d…ô bu olur: \$Var(X\_i) = p(1 - p)\$

Bu s…ôb…ôbd…ôn d…ô √ºmumi d…ôyi≈ük…ônlik \$n \cdot p \cdot (1 - p)\$ formasƒ±ndadƒ±r.

---

## ‚úÖ 5. Qƒ±sa C…ôdv…ôl ‚Äî Yekun

| X√ºsusiyy…ôt    | D√ºstur                                      | A√ßƒ±qlama                        |
| ------------- | ------------------------------------------- | ------------------------------- |
| Paylanma tipi | $$X \sim Bin(n, p)$$                        | n sƒ±naqda p ehtimalla uƒüur sayƒ± |
| PMF           | $$P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}$$ | k uƒüurun ehtimalƒ±               |
| G√∂zl…ôm…ô       | $$\mathbb{E}[X] = n \cdot p$$              | Orta uƒüur sayƒ±                  |
| Dispersiya    | $$Var(X) = n \cdot p \cdot (1 - p)$$        | D…ôyi≈ük…ônlik √∂l√ß√ºs√º              |




## Binomial paylanmanƒ±n Dispersiyasƒ±nƒ± (Variance) niy…ô $n \cdot p \cdot (1-p)$ olduƒüunu anlayaq



### 1. Binomial paylanmanƒ±n …ôsaslarƒ±

Binomial paylanma $n$ sayda **m√ºst…ôqil** Bernoulli t…ôcr√ºb…ôsinin (h…ôr birind…ô uƒüur ehtimalƒ± $p$, uƒüursuzluq ehtimalƒ± $1-p$) c…ômidir:

$$
X = \sum_{i=1}^n X_i
$$

Burada h…ôr $X_i$ yalnƒ±z 0 v…ô ya 1 d…ôy…ôrini ala bil…ôr.

---

### 2. Bernoulli d…ôyi≈ü…ônin g√∂zl…ônil…ôn d…ôy…ôri v…ô dispersiyasƒ±

Bernoulli d…ôyi≈ü…ôninin (y…ôni, $X_i \in \{0,1\}$) g√∂zl…ônil…ôn d…ôy…ôri:

$$
E(X_i) = 1 \cdot p + 0 \cdot (1-p) = p
$$

Dispersiyasƒ± is…ô:

$$
Var(X_i) = E(X_i^2) - (E(X_i))^2
$$

Amma $X_i$ yalnƒ±z 0 v…ô 1 d…ôy…ôrini aldƒ±ƒüƒ±ndan, $X_i^2 = X_i$, √ß√ºnki $0^2=0$ v…ô $1^2=1$.

Bel…ôlikl…ô:

$$
E(X_i^2) = E(X_i) = p
$$

O zaman dispersiya:

$$
Var(X_i) = p - p^2 = p(1-p)
$$

---

### 3. Binomial d…ôyi≈ü…ônin dispersiyasƒ±

Binomial d…ôyi≈ü…ôn $X$ $n$ m√ºst…ôqil Bernoulli d…ôyi≈ü…ônin c…ômidir, ona g√∂r…ô:

$$
Var(X) = Var\left(\sum_{i=1}^n X_i \right) = \sum_{i=1}^n Var(X_i) = n \cdot p(1-p)
$$

---

### **N…ôtic…ô:**

* $p$ ‚Äî uƒüur ehtimalƒ±,
* $1-p$ ‚Äî uƒüursuzluq ehtimalƒ±,
* Dispersiya h…ôm uƒüurun, h…ôm d…ô uƒüursuzluƒüun ehtimalƒ±nƒ± n…ôz…ôr…ô alƒ±r. Ona g√∂r…ô d…ô $p(1-p)$ olur.

---

### Intuitiv izah

Dispersiya d…ôyi≈ü…ônin **d…ôyi≈ük…ônliyini** √∂l√ß√ºr.

* ∆èg…ôr $p$ √ßox ki√ßikdir (m…ôs…ôl…ôn, 0.01), dem…ôli √ßox nadir uƒüur olur, $p(1-p) \approx 0.01 \times 0.99 \approx 0.0099$ ‚Äî ki√ßik dispersiya, √ß√ºnki n…ôtic…ôl…ôr √ßox vaxt 0 olur.

* ∆èg…ôr $p = 0.5$ olsa, $p(1-p) = 0.25$ ‚Äî maksimum dispersiya, √ß√ºnki 0 v…ô 1 n…ôtic…ôl…ôri b…ôrab…ôr ehtimalla olur, y…ôni d…ôyi≈ük…ônlik maksimumdur.

## Poisson Paylanma



### ‚úÖ **1. Poisson Paylanmanƒ±n T…ôrifi**

> **T…ôrif:** Poisson paylanmasƒ±, m√º…ôyy…ôn zaman intervalƒ±nda v…ô ya sah…ôd…ô **m√ºst…ôqil ≈ü…ôkild…ô ba≈ü ver…ôn hadis…ôl…ôrin sayƒ±nƒ±** modell…ô≈üdir…ôn **diskret ehtimal paylanmasƒ±dƒ±r**.

> **Sad…ô dill…ô:** ∆èg…ôr bir hadis…ô **t…ôsad√ºfi v…ô nadir** hallarda ba≈ü verirs…ô, m…ôs…ôl…ôn 1 saatda g…ôl…ôn z…ôngl…ôrin sayƒ±, g√ºnd…ô d√º≈ü…ôn meteorlarƒ±n sayƒ±, 1 s…ôhif…ôd…ô olan yazƒ± s…ôhvl…ôri ‚Äî **Poisson paylanmasƒ±** istifad…ô olunur.

---

### ‚úÖ **2. Parametrl…ôr**

* `Œª` (lambda): G√∂zl…ônil…ôn hadis…ô sayƒ± (orta), y…ôni **m…ôhdud sah…ôd…ô v…ô ya vaxtda ba≈ü ver…ôn hadis…ôl…ôrin orta sayƒ±**.

---

### ‚úÖ **3. Poisson Paylanmanƒ±n D√ºsturu**

**Ehtimal k√ºtl…ô funksiyasƒ± (PMF):**

$$
P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}
$$

---

### üéØ Burada h…ôr simvol n…ô dem…ôkdir?

* \$\lambda\$ ‚Äî G√∂zl…ônil…ôn hadis…ô sayƒ± (orta), m…ôs…ôl…ôn: saatda 5 z…ông ‚Üí Œª = 5
* \$k\$ ‚Äî Hadis…ôl…ôrin konkret sayƒ± (axtardƒ±ƒüƒ±mƒ±z d…ôy…ôr)
* \$e\$ ‚Äî T…ôqrib…ôn 2.718 olan riyazi sabit
* \$k!\$ ‚Äî k faktorial

---

### ‚úÖ **4. M…ôs…ôl…ôn**

**Sual:** Bir call center saatda orta hesabla 3 z…ông alƒ±r. G…ôl…ôn saat …ôrzind…ô **d…ôqiq 2 z…ông** g…ôlm…ôsi ehtimalƒ± n…ô q…ôd…ôrdir?

**Cavab:**

$$
\lambda = 3, \quad k = 2
$$

$$
P(X = 2) = \frac{3^2 \cdot e^{-3}}{2!} = \frac{9 \cdot e^{-3}}{2} ‚âà 0.224
$$

---

### ‚úÖ **5. G√∂zl…ôm…ô v…ô Dispersiya**

Poisson paylanmasƒ±nƒ±n **g√∂zl…ônil…ôn d…ôy…ôri v…ô dispersiyasƒ±**:

| X√ºsusiyy…ôt | D√ºstur                       |
| ---------- | ---------------------------- |
| G√∂zl…ôm…ô    | $$\mathbb{E}[X] = \lambda$$ |
| Dispersiya | $$Var(X) = \lambda$$         |

üí° Y…ôni **orta d…ôy…ôr v…ô dispersiya eynidir**.

---

### ‚úÖ **6. Poisson v…ô Binomial F…ôrqi**

| X√ºsusiyy…ôt       | Binomial                         | Poisson                               |
| ---------------- | -------------------------------- | ------------------------------------- |
| Sƒ±naq sayƒ± `n`   | M…ôlum v…ô sonludur                | Yoxdur                                |
| Ehtimal `p`      | Sabit v…ô m…ôlum                   | Yoxdur (…ôv…ôzind…ô Œª var)               |
| Parametrl…ôr      | `n`, `p`                         | `Œª`                                   |
| Model m…ôqs…ôdi    | `n` sƒ±naqda `k` uƒüur ehtimalƒ±    | M√º…ôyy…ôn vaxt/sah…ôd…ô `k` hadis…ô        |
| Formula          | \$\binom{n}{k} p^k (1-p)^{n-k}\$ | \$\frac{\lambda^k e^{-\lambda}}{k!}\$ |
| Qeyri-m…ôhdud `k` | Yox, `k ‚â§ n`                     | B…ôli, `k = 0, 1, 2, ...`              |
| Misal            | 10 d…ôf…ô zar atmaq                | 1 saatda ne√ß…ô z…ông g…ôlm…ôsi            |

---

### ‚úÖ 7. Qƒ±sa C…ôdv…ôl ‚Äî Yekun

| X√ºsusiyy…ôt    | D√ºstur                                         | A√ßƒ±qlama                              |
| ------------- | ---------------------------------------------- | ------------------------------------- |
| Paylanma tipi | $$X \sim Pois(\lambda)$$                         | G√∂zl…ônil…ôn hadis…ô sayƒ± Œª olan Poisson |
| PMF           | $$P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}$$ | k hadis…ônin ehtimalƒ±                  |
| G√∂zl…ôm…ô       | $$\mathbb{E}[X] = \lambda$$                      | Orta hadis…ô sayƒ±                      |
| Dispersiya    | $$Var(X) = \lambda$$                             | Hadis…ô sayƒ±nƒ±n d…ôyi≈ük…ônliyi           |

---



## Poisson paylanmasƒ±nƒ±n Riyazi G√∂zl…ôm…ô v…ô Dispersiyasƒ±nƒ±n niy…ô `Œª` olduƒüunu izah ed…ôk



### ‚úÖ 1. Poisson d…ôyi≈ü…ônin t…ôrifi

Poisson d…ôyi≈ü…ôn \$X \sim \text{Pois}(\lambda)\$ ‚Äî m√º…ôyy…ôn zaman v…ô ya sah…ôd…ô **orta hadis…ô sayƒ± Œª** olan t…ôsad√ºfi d…ôyi≈ü…ôndir. Y…ôni:

$$
P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}, \quad k = 0, 1, 2, \ldots
$$

---

### ‚úÖ 2. G√∂zl…ônil…ôn d…ôy…ôr (Expected Value)

Poisson d…ôyi≈ü…ôn √º√ß√ºn g√∂zl…ônil…ôn d…ôy…ôr bel…ôdir:

$$
\mathbb{E}[X] = \lambda
$$

#### üîç **S√ºbut:**

Poisson paylanmada ehtimal funksiyasƒ± bel…ôdir:

$$
P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}, \quad k = 0, 1, 2, \ldots
$$

G√∂zl…ônil…ôn d…ôy…ôr:

$$
\mathbb{E}[X] = \sum_{k=0}^{\infty} k \cdot P(X = k) = \sum_{k=0}^{\infty} k \cdot \frac{\lambda^k e^{-\lambda}}{k!}
$$

ƒ∞ndi bu ifad…ôni sad…ôl…ô≈üdir…ôk. Qeyd ed…ôk ki, \$k = 0\$ √º√ß√ºn \$k \cdot \dots = 0\$ olduƒüuna g√∂r…ô, c…ômi \$k = 1\$-d…ôn ba≈ülada bil…ôrik:

$$
\mathbb{E}[X] = \sum_{k=1}^{\infty} \frac{k \lambda^k e^{-\lambda}}{k!}
$$

ƒ∞fad…ôni bir az d…ôyi≈ü…ôr…ôk:

$$
= \sum_{k=1}^{\infty} \frac{\lambda \cdot \lambda^{k-1} e^{-\lambda}}{(k-1)!}
$$

Burada \$\lambda\$ sabitdir v…ô c…ômd…ôn k…ônara √ßƒ±xarƒ±lƒ±r:

$$
= \lambda e^{-\lambda} \sum_{k=1}^{\infty} \frac{\lambda^{k-1}}{(k-1)!}
$$

Substitusiya ed…ôk: \$m = k - 1 \Rightarrow k = m + 1\$ v…ô \$m = 0, 1, 2, \dots\$:

$$
= \lambda e^{-\lambda} \sum_{m=0}^{\infty} \frac{\lambda^m}{m!}
$$

Bu is…ô **eksponensial funksiyanƒ±n t…ôrifi** il…ô b…ôrab…ôrdir:

$$
\sum_{m=0}^{\infty} \frac{\lambda^m}{m!} = e^{\lambda}
$$

Onda:

$$
\mathbb{E}[X] = \lambda e^{-\lambda} \cdot e^{\lambda} = \lambda
$$

‚úÖ **N…ôtic…ô:** \$\mathbb{E}\[X] = \lambda$

---


### ‚úÖ 3. Dispersiyanƒ±n d√ºsturu

Dispersiya:

$$
Var(X) = \mathbb{E}[X^2] - (\mathbb{E}[X])^2
$$

Dem…ôli, biz…ô **\$\mathbb{E}\[X^2]\$** lazƒ±mdƒ±r.

Amma g…ôlin ba≈üqa bir daha **sad…ô v…ô anlayƒ±≈ülƒ±** yoldan ged…ôk: **Poisson d…ôyi≈ü…ôni …ôslind…ô limit kimi yaranƒ±r**.



### ‚úÖ Poisson ‚Üí Binomial limitind…ôn baxƒ±≈ü

Poisson paylanmasƒ± …ôslind…ô **Binomial paylanmanƒ±n limit halƒ±dƒ±r**, y…ôni:

∆èg…ôr \$X \sim \text{Bin}(n, p)\$ v…ô \$n \to \infty\$, \$p \to 0\$ el…ô ki, \$n \cdot p = \lambda\$ sabit qalsƒ±n, o zaman:

$$
\text{Bin}(n, p) \longrightarrow \text{Pois}(\lambda)
$$



### ‚úÖ Binomialdan n…ôtic…ô √ßƒ±xaraq

Binomial d…ôyi≈ü…ôn √º√ß√ºn:

$$
Var(X) = n \cdot p \cdot (1 - p)
$$

∆èg…ôr \$p \to 0\$, o zaman \$(1 - p) \to 1\$ v…ô \$n \cdot p \to \lambda\$

Onda:

$$
Var(X) = n \cdot p \cdot (1 - p) \approx n \cdot p = \lambda
$$

Bu limitd…ôn g√∂r…ô bilirik ki, **Poisson paylanmanƒ±n dispersiyasƒ± da g√∂zl…ônil…ôn d…ôy…ôri kimi `Œª` olur**:

$$
Var(X) = \lambda
$$

## H…ônd…ôsi Paylanma



### ‚úÖ **1. H…ônd…ôsi Paylanmanƒ±n T…ôrifi**

> **T…ôrif:** H…ônd…ôsi paylanma ‚Äî uƒüurun sabit ehtimalla ba≈ü verdiyi m√ºst…ôqil sƒ±naqlarda **ilk uƒüurun ne√ß…ôinci sƒ±naqda ba≈ü ver…ôc…ôyini** modell…ô≈üdirir.

> **Ba≈üqa s√∂zl…ô:** H…ônd…ôsi paylanma, **ilk uƒüur** …ôld…ô olunana q…ôd…ôr olan **sƒ±naqlarƒ±n sayƒ±nƒ±** g√∂st…ôr…ôn **diskret ehtimal paylanmasƒ±dƒ±r.**

---

### ‚úÖ **2. Parametrl…ôr**

* `p`: H…ôr bir sƒ±naqda uƒüur ehtimalƒ±
* `q = 1 - p`: Uƒüursuzluq ehtimalƒ±
* `k`: ƒ∞lk uƒüurun ba≈ü verdiyi sƒ±naq n√∂mr…ôsi (1, 2, 3, ...)

üìå Y…ôni:

* H…ôr sƒ±naq **uƒüur (1)** v…ô ya **uƒüursuzluq (0)** il…ô n…ôtic…ôl…ônir.
* Sƒ±naqlar **m√ºst…ôqildir**.
* Sual budur: **ƒ∞lk uƒüur ne√ß…ôinci d…ôf…ô ba≈ü ver…ôc…ôk?**

---

### ‚úÖ **3. Ehtimal K√ºtl…ô Funksiyasƒ± (PMF)**

H…ônd…ôsi paylanmanƒ±n ehtimal funksiyasƒ±:

$$
P(X = k) = (1 - p)^{k - 1} \cdot p
$$

üìå Y…ôni: ƒ∞lk \$k-1\$ sƒ±naq uƒüursuz olmalƒ±, v…ô **k-cƒ±** sƒ±naqda **ilk uƒüur** ba≈ü verm…ôlidir.

---

### ‚úÖ **4. M…ôs…ôl…ôn**

**Sual:** Bir adam t…ôkrar-t…ôkrar zar atƒ±r. H…ôr atƒ±≈üda 6 g…ôlm…ô ehtimalƒ± \$\frac{1}{6}\$-dƒ±r. ƒ∞lk d…ôf…ô 6 g…ôlm…ôsinin **4-c√º atƒ±≈üda** ba≈ü verm…ô ehtimalƒ± n…ô q…ôd…ôrdir?

**Cavab:**

* \$p = \frac{1}{6}, \quad k = 4\$
* ƒ∞lk 3 sƒ±naqda **uƒüursuzluq** (y…ôni 6 g…ôlm…ôm…ôlidir): \$\left(\frac{5}{6}\right)^3\$
* 4-c√º atƒ±≈üda **6 g…ôlm…ôlidir**: \$\left(\frac{1}{6}\right)\$

$$
P(X = 4) = \left(\frac{5}{6}\right)^3 \cdot \left(\frac{1}{6}\right) ‚âà 0.096
$$

---

### ‚úÖ **5. Riyazi G√∂zl…ôm…ô (Expected Value)**

H…ônd…ôsi paylanmanƒ±n g√∂zl…ônil…ôn d…ôy…ôri:

$$
\mathbb{E}[X] = \frac{1}{p}
$$

üìå Y…ôni, ortalama olaraq ilk uƒüur **1/p** sƒ±naqdan sonra ba≈ü verir.

**Misal:**
H…ôr atƒ±≈üda 6 g…ôlm…ô ehtimalƒ± \$\frac{1}{6}\$-dƒ±r ‚Üí orta hesabla **6 atƒ±≈üdan sonra** ilk 6 g…ôl…ôc…ôk.

---

### ‚úÖ **6. Dispersiya (Varyans)**

Dispersiya (daƒüƒ±lƒ±mƒ±n √∂l√ß√ºs√º):

$$
Var(X) = \frac{1 - p}{p^2}
$$

üìå Y…ôni ehtimal ki√ßikdirs…ô, ilk uƒüurun ba≈üverm…ô m√ºdd…ôti √ßox d…ôyi≈ük…ôn ola bil…ôr.

### üß† **∆èlav…ô ƒ∞ntuitiv ƒ∞zah**

* ƒ∞st…ôdiyimiz odur ki, **birinci uƒüur** ne√ß…ôinci d…ôf…ô ba≈ü ver…ôc…ôk?
* **∆èg…ôr p √ßox balacadƒ±rsa** (uƒüur nadir ba≈ü verirs…ô), uƒüur √ßox g√∂zl…ôdil…ô bil…ôr ‚Üí g√∂zl…ôm…ô v…ô dispersiya b√∂y√ºkd√ºr.
* **∆èg…ôr p b√∂y√ºkd√ºrs…ô**, uƒüur tez ba≈ü ver…ôc…ôk ‚Üí g√∂zl…ôm…ô v…ô dispersiya ki√ßikdir.

G√∂zl…ônil…ôn d…ôy…ôr bel…ô hesablanƒ±r:

$$
E(X) = \sum_{k=1}^{\infty} k \cdot (1-p)^{k-1} \cdot p

Bu **h…ônd…ôsi silsil…ônin** x√ºsusi c…ômi kimi tanƒ±nƒ±r v…ô n…ôtic…ô bel…ô √ßƒ±xƒ±r:

### ‚úÖ 2. Dispersiya (niy…ô $Var(X) = \frac{1-p}{p^2}$?)

### üìê Riyazi s√ºbut:

∆èvv…ôl \$E(X)\$ bilirik:

$$
E(X) = \frac{1}{p}
$$

Sonra is…ô:

$$

$$

Onda dispersiya:

$$
Var(X) = E(X^2) - (E(X))^2 = \frac{1 + p}{p^2} - \left(\frac{1}{p}\right)^2 = \frac{1 - p}{p^2}
$$

---

### ü§Ø Intuitiv izah:

Dispersiya **d…ôyi≈ük…ônliyin √∂l√ß√ºs√ºd√ºr**. Y…ôni d…ôy…ôrl…ôr ortalama …ôtrafƒ±nda nec…ô "yayƒ±lƒ±b".

∆èg…ôr \$p\$ **ki√ßikdir** (m…ôs…ôl…ôn, 0.05), uƒüur √ßox nadirdir. B…ôz…ôn 3-c√º t…ôcr√ºb…ôd…ô, b…ôz…ôn 100-c√º t…ôcr√ºb…ôd…ô uƒüur ola bil…ôr ‚Üí **b√∂y√ºk dispersiya**

Bu da g√∂st…ôrir ki:

$$
Var(X) = \frac{1-p}{p^2}
$$

* \$p^2\$ a≈üaƒüƒ±da olduƒüuna g√∂r…ô \$p\$ azaldƒ±qca dispersiya **√ßox artƒ±r**.
* Y…ôni \$p\$ ‚Üí 0 olduqda dispersiya \$‚Üí \infty\$

## Hiperh…ônd…ôsi Paylanma



### ‚úÖ **1. Hiperh…ônd…ôsi Paylanmanƒ±n T…ôrifi**

> **T…ôrif:** Hiperh…ônd…ôsi paylanma, sonlu √∂l√ß√ºl√º bir qrupdan (populyasiyadan) **…ôv…ôzsiz** ≈ü…ôkild…ô se√ßil…ôn n√ºmun…ô i√ß…ôrisind…ô m√º…ôyy…ôn sayda uƒüur ba≈ü verm…ô ehtimalƒ±nƒ± modell…ô≈üdir…ôn **diskret ehtimal paylanmasƒ±dƒ±r**.

> **Sad…ô dill…ô:** T…ôsad√ºfi olaraq **…ôv…ôzsiz** (y…ôni geri qoymadan) se√ßim etdikd…ô, m√º…ôyy…ôn sayda uƒüur …ôld…ô etm…ô ehtimalƒ±nƒ± √∂l√ß√ºr.

---

### ‚úÖ **2. Parametrl…ôr**

| Simvol | M…ônasƒ±                                          |
| ------ | ----------------------------------------------- |
| `N`    | √úmumi …ô≈üya (populyasiya) sayƒ±                   |
| `K`    | Populyasiyada uƒüurlu (maraqlƒ±) obyektl…ôrin sayƒ± |
| `n`    | Se√ßil…ôn n√ºmun…ônin √∂l√ß√ºs√º                        |
| `k`    | N√ºmun…ôd…ô `k` uƒüur …ôld…ô etm…ô ehtimalƒ± axtarƒ±lƒ±r  |

---

**≈û…ôrtl…ôr:**

* Populyasiya **sonludur** v…ô √∂l√ß√ºs√º `N`-dir.
* Populyasiyada `K` …ôd…ôd **uƒüurlu** obyekt var, qalan `N-K` is…ô uƒüursuzdur.
* `n` …ôd…ôd **…ôv…ôzsiz** se√ßim edilir.
* Maraƒüƒ±mƒ±z `k` …ôd…ôd uƒüur se√ßm…ôk ehtimalƒ±dƒ±r.

Bu zaman t…ôsad√ºfi d…ôyi≈ü…ôn a≈üaƒüƒ±dakƒ± kimi yazƒ±lƒ±r:

$$
X \sim \text{Hypergeometric}(N, K, n)
$$

---

### ‚úÖ **3. Hiperh…ônd…ôsi Paylanmanƒ±n D√ºsturu (PMF)**

$$
P(X = k) = \frac{\binom{K}{k} \cdot \binom{N - K}{n - k}}{\binom{N}{n}}
$$

### üéØ A√ßƒ±qlama:

* \$\binom{K}{k}\$ ‚Äî Uƒüurlu obyektl…ôrd…ôn `k` …ôd…ôd se√ßm…ôyin sayƒ±
* \$\binom{N-K}{n-k}\$ ‚Äî Uƒüursuzlardan qalan `n-k` …ôd…ôd se√ßm…ôyin sayƒ±
* \$\binom{N}{n}\$ ‚Äî √úmumilikd…ô `n` obyekt se√ßm…ôyin √ºmumi sayƒ±

---

### ‚úÖ **4. M…ôs…ôl…ôn:**

**Sual:** Bir qutuda 7 aƒü, 5 qara k√ºr…ô var (c…ômi 12 k√ºr…ô). ∆èv…ôzsiz 4 k√ºr…ô se√ßilir. Bu 4 k√ºr…ônin i√ßind…ô **d…ôqiq 2 aƒü** olma ehtimalƒ± n…ôdir?

**Cavab:**

$$
N = 12,\quad K = 7,\quad n = 4,\quad k = 2
$$

$$
P(X = 2) = \frac{\binom{7}{2} \cdot \binom{5}{2}}{\binom{12}{4}} = \frac{21 \cdot 10}{495} = \frac{210}{495} ‚âà 0.424
$$

---

### ‚úÖ **5. Riyazi G√∂zl…ôm…ô (Expected Value)**

$$
\mathbb{E}[X] = n \cdot \frac{K}{N}
$$

### üí° Niy…ô bel…ôdir?

∆èg…ôr `n` …ôd…ôd se√ßim edirs…ôns…ô v…ô √ºmumi populyasiyada uƒüur ehtimalƒ± `K / N`-dirs…ô, orta hesabla **n d…ôf…ô bu ehtimal q…ôd…ôr** uƒüur g√∂zl…ônilir. Bu, binomiala √ßox b…ônz…ôyir, lakin …ôv…ôzsiz se√ßim olduƒüu √º√ß√ºn hiperh…ônd…ôsidir.

---

### ‚úÖ **6. Dispersiya (Varyans)**

$$
Var(X) = n \cdot \frac{K}{N} \cdot \left(1 - \frac{K}{N}\right) \cdot \frac{N - n}{N - 1}
$$

### üí° Niy…ô bu d√ºsturdur?

Bu d√ºstur, **binomial dispersiyanƒ±n** d√ºz…ôli≈ü edilmi≈ü formasƒ±dƒ±r. Sonuncu vurƒüu ‚Äî \$\frac{N - n}{N - 1}\$ ‚Äî **…ôv…ôzsiz se√ßim** s…ôb…ôbind…ôn d…ôyi≈ük…ônliyin azaldƒ±ƒüƒ±nƒ± n…ôz…ôr…ô alƒ±r. Y…ôni, …ôg…ôr **…ôv…ôzi olsa idi**, binomial formasƒ± olardƒ±:

$$
n \cdot \frac{K}{N} \cdot \left(1 - \frac{K}{N}\right)
$$

Amma …ôv…ôzsiz se√ßim olduƒüuna g√∂r…ô, dispersiya bu faktorla azaldƒ±lƒ±r.

---

### ‚úÖ **7. Qƒ±sa C…ôdv…ôl ‚Äî Yekun**

| X√ºsusiyy…ôt    | D√ºstur                                                                                      | A√ßƒ±qlama                                        |
| ------------- | ------------------------------------------------------------------------------------------- | ----------------------------------------------- |
| Paylanma tipi | $$X \sim \text{Hypergeometric}(N, K, n)$$                                                     | N populyasiyada K uƒüurludan n-i se√ßm…ôk          |
| PMF           | $$P(X = k) = \frac{\binom{K}{k} \cdot \binom{N-K}{n-k}}{\binom{N}{n}}$$                       | N√ºmun…ôd…ô k uƒüur ehtimalƒ±                        |
| G√∂zl…ôm…ô       | $$\mathbb{E}[X] = n \cdot \frac{K}{N}$$                                                       | Orta uƒüur sayƒ±                                  |
| Dispersiya    | $$Var(X) = n \cdot \frac{K}{N} \cdot \left(1 - \frac{K}{N}\right) \cdot \frac{N - n}{N - 1}$$ | D…ôyi≈ük…ônlik √∂l√ß√ºs√º (…ôv…ôzsiz se√ßimi n…ôz…ôr…ô alƒ±r) |



## Hiperh…ônd…ôsi Paylanmanƒ±n niy…ô Riyazi G√∂zl…ôm…ôsi $\mathbb{E}[X] = n \cdot \frac{K}{N}$   v…ô Dispersiyasƒ± is…ô $Var(X) = n \cdot \frac{K}{N} \cdot \left(1 - \frac{K}{N}\right) \cdot \frac{N - n}{N - 1}$ olur?



### ‚úÖ 1. Hiperh…ônd…ôsi Paylanmanƒ±n T…ôyinatƒ± (qƒ±sa xatƒ±rlatma)

Hiperh…ônd…ôsi d…ôyi≈ü…ôn bel…ô t…ôyin olunur:

$$
X \sim \text{Hypergeometric}(N, K, n)
$$

| Simvol | ƒ∞zah                                                     |
| ------ | -------------------------------------------------------- |
| `N`    | Populyasiyanƒ±n √ºmumi sayƒ±                                |
| `K`    | Populyasiyadakƒ± uƒüurlu obyektl…ôrin sayƒ±                  |
| `n`    | ∆èv…ôzsiz se√ßilmi≈ü n√ºmun…ônin √∂l√ß√ºs√º                        |
| `X`    | Bu n√ºmun…ôd…ôki uƒüurlu obyektl…ôrin sayƒ± (t…ôsad√ºfi d…ôyi≈ü…ôn) |

---

### ‚úÖ 2. G√∂zl…ônil…ôn D…ôy…ôr (Expected Value)

### üìå **D√ºstur:**

$$
\mathbb{E}[X] = n \cdot \frac{K}{N}
$$

---

### ‚úÖ **S√ºbut (Riyazi Yana≈üma)**

Hiperh…ônd…ôsi d…ôyi≈ü…ôni bel…ô ba≈üa d√º≈ü…ôk: Biz `n` …ôd…ôd obyekt se√ßirik v…ô h…ôr biri uƒüurlu ola bil…ôr. Bunu **indikator d…ôyi≈ü…ônl…ôr** il…ô ifad…ô ed…ôk:

* G√∂t√ºr…ôk:

  $$
  X = X_1 + X_2 + \cdots + X_n
  $$

$$
X_i =
\begin{cases}
1, & \text{…ôg…ôr } i\text{-ci se√ßil…ôn uƒüurludursa} \\
0, & \text{…ôks halda}
\end{cases}
$$

Bu ifad…ô deyir ki:

* $X_i = 1$ ‚Üí …ôg…ôr se√ßil…ôn obyekt **uƒüurlu** (m…ôs…ôl…ôn, qƒ±rmƒ±zƒ± k√ºr…ô) olarsa,
* $X_i = 0$ ‚Üí …ôg…ôr **uƒüursuz** olarsa.

Bu d…ôyi≈ü…ônl…ôr **indikator d…ôyi≈ü…ônl…ôrdir** ‚Äî y…ôni yalnƒ±z 0 v…ô 1 alƒ±rlar.


* H…ôr bir \$X\_i\$ d…ôyi≈ü…ôni **Bernoulli** tiplidir, amma **…ôv…ôzsiz se√ßim** olduƒüuna g√∂r…ô bu d…ôyi≈ü…ônl…ôr bir-birind…ôn **m√ºst…ôqil deyil**.

Bununla bel…ô, **x…ôtti g√∂zl…ôm…ô xass…ôsin…ô** g√∂r…ô:

$$
\mathbb{E}[X] = \mathbb{E}[X_1 + \cdots + X_n] = \sum_{i=1}^{n} \mathbb{E}[X_i]
$$

ƒ∞ndi …ôsas m…ôs…ôl…ô \$\mathbb{E}\[X\_i]\$ tapmaqdƒ±r:

* H…ôr se√ßilmi≈ü obyektin **uƒüurlu olma ehtimalƒ±**:

$$
P(X_i = 1) = \frac{K}{N}
$$

(y…ôni, uƒüurlu olanlarƒ±n sayƒ± / √ºmumi say)

Dem…ôli:

$$
\mathbb{E}[X_i] = 1 \cdot \frac{K}{N} + 0 \cdot \left(1 - \frac{K}{N}\right) = \frac{K}{N}
$$

V…ô sonunda:

$$
\mathbb{E}[X] = n \cdot \frac{K}{N}
$$

---

### ‚úÖ Intuitiv Yana≈üma

* Tutaq ki, qutuda `N = 100` k√ºr…ô var, bunun `K = 30`-u qƒ±rmƒ±zƒ±dƒ±r.
* `n = 10` k√ºr…ô …ôv…ôzsiz se√ßirs…ôns…ô, h…ôr birinin qƒ±rmƒ±zƒ± g…ôlm…ô ehtimalƒ± ortalama `30/100 = 0.3`-d√ºr.
* Orta hesabla `10 √ó 0.3 = 3` qƒ±rmƒ±zƒ± k√ºr…ô se√ß…ôc…ôyik.

---

### ‚úÖ 3. Dispersiya (Varyans)

### üìå **D√ºstur:**

$$
Var(X) = n \cdot \frac{K}{N} \cdot \left(1 - \frac{K}{N}\right) \cdot \frac{N - n}{N - 1}
$$

---

### ‚úÖ Riyazi S√ºbut (X√ºlas…ô il…ô)

B…ôrab…ôr qurulu≈üla yen…ô d…ô \$X = \sum X\_i\$ yazƒ±rƒ±q. Dispersiya √º√ß√ºn:

$$
Var(X) = \sum Var(X_i) + 2 \sum_{i < j} Cov(X_i, X_j)
$$

∆èg…ôr se√ßim **…ôv…ôzi** olsaydƒ±, d…ôyi≈ü…ônl…ôr m√ºst…ôqil olardƒ± v…ô bu, **binomial dispersiya** il…ô eyni olardƒ±:

$$
n \cdot \frac{K}{N} \cdot \left(1 - \frac{K}{N}\right)
$$

Amma **…ôv…ôzsiz se√ßim** olduƒüuna g√∂r…ô d…ôyi≈ük…ônlik bir q…ôd…ôr **azalƒ±r** v…ô bu azalma \$\frac{N - n}{N - 1}\$ vurƒüu il…ô ifad…ô olunur.

Y…ôni:

> Dispersiyanƒ±n d√ºz…ôli≈üi:
>
> $$
> \text{Binomial varians} \cdot \left(\frac{N - n}{N - 1}\right)
> $$

---

### ‚úÖ Intuitiv Yana≈üma

* ∆èg…ôr …ôv…ôzi se√ßim olsa idi ‚Äî h…ôr d…ôf…ô se√ßdiyimiz k√ºr…ônin uƒüurlu olmasƒ± ehtimalƒ± **d…ôyi≈üm…ôz** qalardƒ±.
* Amma **…ôv…ôzsiz** se√ßimd…ô ‚Äî bir uƒüurlu se√ßdikd…ôn sonra n√∂vb…ôti se√ßimin uƒüurlu olma ehtimalƒ± **azalƒ±r**.
* Bu s…ôb…ôbd…ôn, **dispersiya azalƒ±r**, y…ôni n…ôtic…ôl…ôr bir-birind…ôn daha az ‚Äúsƒ±√ßrayƒ±r‚Äù.

---

### ‚úÖ 4. M…ôs…ôl…ôn:

**Cavab:**

* G√∂zl…ônil…ôn d…ôy…ôr:

$$

$$

$$
