## ‚úÖ **T…ôsad√ºfi k…ômiyy…ôt (Random Variable)** n…ôdir?

> **T…ôrif:** T…ôsad√ºfi k…ômiyy…ôt ‚Äì bir eksperimentin n…ôtic…ôsin…ô …ôd…ôdi qiym…ôt ver…ôn funksiyadƒ±r.

M…ôs…ôl…ôn: Z…ôri atdƒ±qda g…ôl…ôn r…ôq…ôm bir t…ôsad√ºfi k…ômiyy…ôtdir.

---

## üìå T…ôsad√ºfi k…ômiyy…ôtl…ôrin **n√∂vl…ôri**

### 1. **Diskret t…ôsad√ºfi k…ômiyy…ôt (Discrete Random Variable)**

* **K…ôsikli d…ôy…ôrl…ôr alƒ±r** ‚Äì y…ôni yalnƒ±z m√º…ôyy…ôn …ôd…ôdl…ôr ala bil…ôr (m…ôs…ôl…ôn, 0, 1, 2, ...).
* **Misal**:

  * Z…ôrin atƒ±lmasƒ± ‚Üí n…ôtic…ô: 1, 2, 3, 4, 5, 6
  * Bir ail…ôd…ôki u≈üaq sayƒ±: 0, 1, 2, ...

#### Ehtimal k√ºtl…ô funksiyasƒ± (Probability Mass Function ‚Äì PMF)

Diskret t…ôsad√ºfi k…ômiyy…ôtin ehtimallarƒ± **k√ºtl…ô funksiyasƒ±** il…ô verilir:

$$
P(X = x_i)
$$

Burada:

* $X$ ‚Äì t…ôsad√ºfi k…ômiyy…ôtdir,
* $x_i$ ‚Äì ala bil…ôc…ôyi d…ôy…ôrl…ôr,
* $P(X = x_i)$ ‚Äì bu d…ôy…ôrin g…ôlm…ô ehtimalƒ±.

üß† X√ºsusiyy…ôtl…ôr:

* $0 \le P(X = x_i) \le 1$
* $\sum P(X = x_i) = 1$

---

### 2. **K…ôsilm…ôz (davamlƒ±) t…ôsad√ºfi k…ômiyy…ôt (Continuous Random Variable)**

* **Davamlƒ± aralƒ±qda d…ôy…ôrl…ôr ala bil…ôr** ‚Äì m…ôs…ôl…ôn, \[0, 1\], $(-\infty, +\infty)$
* **Misal**:

  * ƒ∞nsanlarƒ±n boyu: 165.2 cm, 165.25 cm, ...
  * Bir cihazƒ±n i≈ül…ôm…ô vaxtƒ±: 3.57 saat, 3.571 saat...

#### Ehtimal sƒ±xlƒ±ƒüƒ± funksiyasƒ± (Probability Density Function ‚Äì PDF)

Bu zaman ehtimallar sƒ±xlƒ±q funksiyasƒ± il…ô g√∂st…ôrilir:

$$
f(x) \ge 0
$$

V…ô ehtimal bir interval √º√ß√ºn verilir:

$$
P(a \le X \le b) = \int_a^b f(x) dx
$$

üß† X√ºsusiyy…ôtl…ôr:

* $f(x) \ge 0$
* $\int_{-\infty}^{+\infty} f(x) dx = 1$

> üìå Davamlƒ± t…ôsad√ºfi k…ômiyy…ôt √º√ß√ºn bir **n√∂qt…ônin ehtimalƒ± sƒ±fƒ±rdƒ±r**:
>
> $$
> P(X = x) = 0
> $$

---

## üìò **Paylanma qanunu (Distribution Law)**

> **T…ôrif:** **Paylanma qanunu** ‚Äì t…ôsad√ºfi k…ômiyy…ôtin ala bil…ôc…ôyi b√ºt√ºn **m√ºmk√ºn d…ôy…ôrl…ôri** v…ô onlarƒ±n h…ôr birinin **ba≈üverm…ô ehtimalƒ±nƒ±** g√∂st…ôr…ôn qanundur, el…ôc…ô d…ô bu **ehtimallarƒ±n c…ômi 1-…ô b…ôrab…ôr** olmalƒ±dƒ±r.

### üîπ Diskret k…ômiyy…ôt √º√ß√ºn:

Bu, sad…ôc…ô a≈üaƒüƒ±dakƒ± kimidir:

| $$x$$ | $$P(X = x)$$ |
| --- | ---------- |
| 0   | 0.2        |
| 1   | 0.5        |
| 2   | 0.3        |

Burada **paylanma qanunu** bu ehtimallar c…ôdv…ôlidir.

### üîπ Davamlƒ± k…ômiyy…ôt √º√ß√ºn:

Paylanma qanunu sƒ±xlƒ±q funksiyasƒ± il…ô ifad…ô olunur v…ô ehtimallar **integralla** tapƒ±lƒ±r.

---

### üìå Qƒ±saca t…ôrif:

> **Paylanma qanunu** ‚Äì t…ôsad√ºfi k…ômiyy…ôtin m√ºmk√ºn d…ôy…ôrl…ôrin…ô uyƒüun g…ôl…ôn ehtimallarƒ± t…ôyin ed…ôn qanundur.
> Diskret halda ehtimal c…ôdv…ôli,
> davamlƒ± halda sƒ±xlƒ±q funksiyasƒ± il…ô ifad…ô olunur.

> T…ôsad√ºfi k…ômiyy…ôtl…ôrin c…ôminin riyazi g√∂zl…ôm…ôsi, onlarƒ±n g√∂zl…ôm…ôl…ôrinin c…ômidir. Bu xass…ô **m√ºst…ôqillik t…ôl…ôb etmir** ‚Äì y…ôni ist…ônil…ôn $X_i$-l…ôr √º√ß√ºn doƒürudur.

## Diskret t…ôsad√ºfi k…ômiyy…ôtin Dispersiyasƒ± (1)



## üìò **Dispersiyanƒ±n T…ôrifi:**

**Dispersiya (D(X))**, t…ôsad√ºfi k…ômiyy…ôtin **orta qiym…ôtd…ôn (riyazi g√∂zl…ôm…ôd…ôn) n…ô q…ôd…ôr sapma etdiyini** √∂l√ß√ºr. Ba≈üqa s√∂zl…ô, dispersiya **d…ôy…ôrl…ôrin yayƒ±lma √∂l√ß√ºs√ºd√ºr**.

> **T…ôrif:**
> Diskret t…ôsad√ºfi k…ômiyy…ôtin dispersiyasƒ± ‚Äî h…ômin k…ômiyy…ôtin m√ºmk√ºn qiym…ôtl…ôrinin **eyni ehtimallarla riyazi g√∂zl…ôm…ôd…ôn kvadratik sapmalarƒ±nƒ±n orta qiym…ôtidir**.

---

## üìê **Dispersiya D√ºsturu:**

Tutaq ki, $X$ diskret t…ôsad√ºfi k…ômiyy…ôtdir v…ô onun m√ºmk√ºn d…ôy…ôrl…ôri $x_1, x_2, ..., x_n$, m√ºvafiq ehtimallarƒ± is…ô $p_1, p_2, ..., p_n$-dir. O zaman:

$$
D(X) = \sum_{i=1}^{n} (x_i - M(X))^2 \cdot p_i
$$

Burada:

* $M(X)$ ‚Äî t…ôsad√ºfi k…ômiyy…ôtin **riyazi g√∂zl…ôm…ôsi**,
* $x_i$ ‚Äî m√ºmk√ºn d…ôy…ôrl…ôr,
* $p_i$ ‚Äî h…ômin d…ôy…ôrl…ôrin ehtimallarƒ±.

---

## ‚úèÔ∏è **Alternativ v…ô daha rahat d√ºstur:**

$$
D(X) = M(X^2) - [M(X)]^2
$$

Burada:

* $M(X^2) = \sum x_i^2 \cdot p_i$ ‚Äî **kvadratlarƒ±n riyazi g√∂zl…ôm…ôsi**,
* $[M(X)]^2$ ‚Äî **riyazi g√∂zl…ôm…ônin kvadratƒ±**.

Bu d√ºstur daha sad…ô v…ô praktik ≈ü…ôkild…ô dispersiyanƒ± hesablamaƒüa imkan verir.

---

## üí° **∆èsas M…ôqamlar:**

1. **Dispersiya h…ômi≈ü…ô sƒ±fƒ±rdan b√∂y√ºk v…ô ya b…ôrab…ôrdir**:

   * Y…ôni $D(X) \geq 0$.
   * ∆èg…ôr $X$ sabit (d…ôyi≈üm…ôy…ôn) bir k…ômiyy…ôtdirs…ô, onda dispersiyasƒ± **0** olar.

2. **∆èg…ôr k…ômiyy…ôt…ô sabit …ôlav…ô edils…ô (X + c)**:

   $$
   D(X + c) = D(X)
   $$

   Y…ôni sabit …ôlav…ô dispersiyanƒ± d…ôyi≈ümir.

3. **∆èg…ôr k…ômiyy…ôt sabitl…ô vurulsa (cX)**:

   $$
   D(cX) = c^2 \cdot D(X)
   $$

   Y…ôni dispersiya **kvadratik ≈ü…ôkild…ô d…ôyi≈üir**.

4. **Dispersiya vahidi**: ∆èg…ôr $X$ √∂l√ß√º il…ô verilirs…ô (m…ôs…ôl…ôn, manat, sm v…ô s.), onda dispersiyanƒ±n vahidi **kvadrat √∂l√ß√ºd…ô** olur. M…ôs…ôl…ôn, …ôg…ôr $X$ manatla √∂l√ß√ºl√ºrs…ô, $D(X)$ **manat¬≤** olacaq.

---

## ‚úÖ **Sad…ô N√ºmun…ô:**

Tutaq ki, $X$ t…ôsad√ºfi k…ômiyy…ôti 1, 2 v…ô 3 d…ôy…ôrl…ôrini alƒ±r v…ô onlarƒ±n ehtimallarƒ± m√ºvafiq olaraq $0.2, 0.5, 0.3$-d√ºr.

### 1. Riyazi g√∂zl…ôm…ô:

$$
M(X) = 1 \cdot 0.2 + 2 \cdot 0.5 + 3 \cdot 0.3 = 2.1
$$

### 2. Kvadratlarƒ±n g√∂zl…ôm…ôsi:

$$
M(X^2) = 1^2 \cdot 0.2 + 2^2 \cdot 0.5 + 3^2 \cdot 0.3 = 0.2 + 2.0 + 2.7 = 4.9
$$

### 3. Dispersiya:

$$
D(X) = M(X^2) - [M(X)]^2 = 4.9 - (2.1)^2 = 4.9 - 4.41 = 0.49
$$


## Diskret t…ôsad√ºfi k…ômiyy…ôtin Dispersiyasƒ± (2)



## üìò **T…ôrif (bir c√ºml…ô il…ô):**

> **T…ôrif:** Diskret t…ôsad√ºfi k…ômiyy…ôtin dispersiyasƒ± ‚Äî onun qiym…ôtl…ôrinin **riyazi g√∂zl…ôm…ôd…ôn orta kvadrat sapmasƒ±nƒ±** g√∂st…ôr…ôn √∂l√ß√ºd√ºr.

---

## üìê **D√ºstur:**

∆èg…ôr t…ôsad√ºfi k…ômiyy…ôt $X$ a≈üaƒüƒ±dakƒ± ehtimal paylanmasƒ±na malikdirs…ô:

$$
X = x_1, x_2, ..., x_n \quad \text{v…ô} \quad P(X = x_i) = p_i
$$

onda dispersiya bel…ô hesablanƒ±r:

$$
\mathrm{Var}(X) = \mathbb{E}[(X - \mathbb{E}[X])^2] = \sum_{i=1}^{n} (x_i - \mu)^2 \cdot p_i
$$

Burada:

* $\mu = \mathbb{E}[X]$ ‚Äî riyazi g√∂zl…ôm…ôdir.

---

### ‚úÖ **Alternativ d√ºstur (praktik hesablamalar √º√ß√ºn):**

$$
\mathrm{Var}(X) = \mathbb{E}[X^2] - (\mathbb{E}[X])^2
$$

Bu √ºsul √ßox zaman daha **sad…ô v…ô s√ºr…ôtli** hesablama imkanƒ± verir.

---

## üîë **∆èsas m…ôqamlar:**

### 1. **Dispersiya h…ômi≈ü…ô sƒ±fƒ±rdan ki√ßik ola bilm…ôz**:

$$
\mathrm{Var}(X) \geq 0
$$

√á√ºnki kvadratlar m…ônfi ola bilm…ôz.

---

### 2. **Dispersiya 0 olarsa, bu n…ô dem…ôkdir?**

Bu o dem…ôkdir ki, t…ôsad√ºfi k…ômiyy…ôt **d…ôyi≈ümir**, y…ôni h…ômi≈ü…ô eyni bir qiym…ôt alƒ±r (sabitdir).

---

### 3. **Sabitl…ô vurulmu≈ü t…ôsad√ºfi k…ômiyy…ôtin dispersiyasƒ±:**

∆èg…ôr $a$ sabit …ôd…ôddirs…ô, onda:

$$
\mathrm{Var}(aX) = a^2 \cdot \mathrm{Var}(X)
$$

> D…ôyi≈ü…ônin b√∂y√ºd√ºlm…ôsi yayƒ±lmanƒ± kvadrat ≈ü…ôkild…ô artƒ±rƒ±r.

---

### 4. **Sabitin …ôlav…ô olunmasƒ± dispersiyanƒ± d…ôyi≈ümir:**

$$
\mathrm{Var}(X + c) = \mathrm{Var}(X)
$$

> √á√ºnki b√ºt√ºn qiym…ôtl…ôr eyni miqdarda s√ºr√º≈ü√ºr, amma yayƒ±lma d…ôyi≈ümir.

---

### 5. **Dispersiya x…ôtti deyil!**

Y…ôni bu doƒüru deyil:

$$
\mathrm{Var}(X + Y) \neq \mathrm{Var}(X) + \mathrm{Var}(Y)
$$

∆èg…ôr $X$ v…ô $Y$ **m√ºst…ôqildirl…ôrs…ô**, **yalnƒ±z o zaman** bel…ô yazmaq olar:

$$
\mathrm{Var}(X + Y) = \mathrm{Var}(X) + \mathrm{Var}(Y)
$$

---

### 6. **Dispersiyanƒ±n vahidi d…ôyi≈üir:**

∆èg…ôr $X$-in vahidi manatdƒ±rsa, $\mathrm{Var}(X)$ **manat¬≤** olur. Bu, onu birba≈üa ≈ü…ôrh etm…ôyi √ß…ôtinl…ô≈üdirir.

Ona g√∂r…ô √ßox zaman **standart sapma** istifad…ô olunur:

$$
\sigma(X) = \sqrt{\mathrm{Var}(X)}
$$

---

## üé≤ **Sad…ô n√ºmun…ô:**

Tutaq ki, z…ôr atƒ±lƒ±r.
$X$: g…ôl…ôn …ôd…ôd

$$
\mathbb{E}[X] = 3.5, \quad \mathbb{E}[X^2] = \frac{1^2 + 2^2 + \dots + 6^2}{6} = \frac{91}{6}
$$

$$
\mathrm{Var}(X) = \mathbb{E}[X^2] - (\mathbb{E}[X])^2 = \frac{91}{6} - 3.5^2 = \frac{91}{6} - 12.25 \approx 2.92
$$




## Diskret t…ôsad√ºfi k…ômiyy…ôtin Dispersiyasƒ±nƒ±n ∆èsas Xass…ôl…ôri



### üìå 1. **Dispersiya h…ômi≈ü…ô m…ônfi olmayan …ôd…ôddir:**

$$
D(X) \geq 0
$$

* Dispersiya he√ß vaxt m…ônfi ola bilm…ôz.
* ∆èg…ôr $D(X) = 0$ olarsa, bu o dem…ôkdir ki, t…ôsad√ºfi k…ômiyy…ôt **d…ôyi≈ümir**, y…ôni **sabit** bir d…ôy…ôr alƒ±r.

---

### üìå 2. **Sabit …ôd…ôdin dispersiyasƒ± 0-dƒ±r:**

$$
D(c) = 0 \quad \text{(c ‚Äî sabit …ôd…ôd)}
$$

* Y…ôni d…ôyi≈üm…ôy…ôn bir …ôd…ôdin yayƒ±lma (variasiya) √∂l√ß√ºs√º yoxdur.

---

### üìå 3. **T…ôsad√ºfi k…ômiyy…ôt…ô sabit …ôlav…ô edil…ônd…ô dispersiya d…ôyi≈ümir:**

$$
D(X + c) = D(X)
$$

* Sabit …ôlav…ô yalnƒ±z **orta qiym…ôti d…ôyi≈üir**, amma **yayƒ±lmanƒ± d…ôyi≈ümir**.

---

### üìå 4. **T…ôsad√ºfi k…ômiyy…ôt sabitl…ô vurulduqda dispersiya kvadrat ≈ü…ôkild…ô d…ôyi≈üir:**

$$
D(cX) = c^2 \cdot D(X)
$$

* Burada $c$ ‚Äî sabit …ôd…ôddir.
* M…ôs…ôl…ôn, …ôg…ôr $X$ 2 il…ô vurularsa, dispersiyasƒ± 4 d…ôf…ô b√∂y√ºy…ôr.

---

### üìå 5. **ƒ∞ki t…ôsad√ºfi k…ômiyy…ôtin c…ôminin dispersiyasƒ±** (onlar **asƒ±lƒ± deyils…ô**):

$$
D(X + Y) = D(X) + D(Y)
$$

* Bu yalnƒ±z **m√ºst…ôqil t…ôsad√ºfi k…ômiyy…ôtl…ôr** √º√ß√ºn ke√ß…ôrlidir.

---

### üìå 6. **ƒ∞ki t…ôsad√ºfi k…ômiyy…ôtin c…ôminin dispersiyasƒ± (√ºmumi hal):**

$$
D(X + Y) = D(X) + D(Y) + 2 \cdot \text{Cov}(X, Y)
$$

* Burada $\text{Cov}(X, Y)$ ‚Äî kovariansdƒ±r (iki k…ômiyy…ôtin birg…ô d…ôyi≈üm…ô √∂l√ß√ºs√º).
* ∆èg…ôr $X$ v…ô $Y$ **m√ºst…ôqildirs…ô**, onda $\text{Cov}(X, Y) = 0$.

---

### üìå 7. **Dispersiya v…ô riyazi g√∂zl…ôm…ô arasƒ±nda …ôlaq…ô:**

$$
D(X) = M(X^2) - [M(X)]^2
$$

* Bu **alternativ d√ºstur** hesablamalarƒ± asanla≈üdƒ±rƒ±r.

---

## Diskret t…ôsad√ºfi k…ômiyy…ôtin Orta Kvadratik Meyli (Standart Deviation)



### üìò **Standart Deviation n…ôdir?**

> **Standart sapma**, t…ôsad√ºfi k…ômiyy…ôtin **√∂z orta qiym…ôtind…ôn n…ô q…ôd…ôr uzaqla≈üdƒ±ƒüƒ±nƒ±** √∂l√ß…ôn statistik g√∂st…ôricidir.
**Standart sapma** ‚Äî t…ôsad√ºfi k…ômiyy…ôtin √∂z orta qiym…ôtind…ôn **n…ô q…ôd…ôr sapdƒ±ƒüƒ±nƒ±** g√∂st…ôr…ôn v…ô **dispersiyanƒ±n kvadrat k√∂k√º** il…ô hesablanan yayƒ±lma √∂l√ß√ºs√ºd√ºr.


Y…ôni:

* Qiym…ôtl…ôr **orta …ôtrafƒ±nda n…ô q…ôd…ôr yayƒ±lƒ±b?**
* **Yayƒ±lma (daƒüƒ±lƒ±m) n…ô q…ôd…ôr g√ºcl√ºd√ºr?** ‚Äî bunu g√∂st…ôrir.

---

### üìê **Standart Deviation d√ºsturu (diskret k…ômiyy…ôt √º√ß√ºn):**

Tutaq ki, t…ôsad√ºfi k…ômiyy…ôt $X$ m√ºmk√ºn d…ôy…ôrl…ôri $x_1, x_2, ..., x_n$ alƒ±r v…ô onlarƒ±n ehtimallarƒ± $p_1, p_2, ..., p_n$-dir.

### Addƒ±mlar:

1. **Riyazi g√∂zl…ôm…ô**:

$$
M(X) = \sum x_i \cdot p_i
$$

2. **Dispersiya**:

$$
D(X) = \sum (x_i - M(X))^2 \cdot p_i
$$

3. **Standart sapma**:

$$
\sigma(X) = \sqrt{D(X)}
$$

---

### üí° **Sad…ô dild…ô des…ôk:**

* Dispersiya ‚Äî **kvadrat sapmalarƒ±n ortasƒ±dƒ±r**,
* Standart sapma ‚Äî onun **kvadrat k√∂k√ºd√ºr**,
* Vahidi **orijinal k…ômiyy…ôtin vahidi il…ô eynidir** (m…ôs: ya≈ü, manat, sm v…ô s.).

---

### ‚úÖ **Sad…ô n√ºmun…ô il…ô hesablayaq:**

Tutaq ki, t…ôsad√ºfi k…ômiyy…ôt $X$ a≈üaƒüƒ±dakƒ± c…ôdv…ôl √ºzr…ô verilir:

| $$x_i$$ | $$p_i$$ |
| ----- | ----- |
| 1     | 0.2   |
| 2     | 0.5   |
| 4     | 0.3   |

### 1. Riyazi g√∂zl…ôm…ô:

$$
M(X) = 1 \cdot 0.2 + 2 \cdot 0.5 + 4 \cdot 0.3 = 0.2 + 1 + 1.2 = 2.4
$$

### 2. Dispersiya:

$$
D(X) = (1 - 2.4)^2 \cdot 0.2 + (2 - 2.4)^2 \cdot 0.5 + (4 - 2.4)^2 \cdot 0.3
$$

$$
= (1.96) \cdot 0.2 + (0.16) \cdot 0.5 + (2.56) \cdot 0.3 = 0.392 + 0.08 + 0.768 = 1.24
$$

### 3. Standart sapma:

$$
\sigma(X) = \sqrt{1.24} \approx 1.113
$$

---

### üìå ∆èsas xass…ôl…ôr:

1. $\sigma(X) \geq 0$
2. $\sigma(X + c) = \sigma(X)$ (sabit …ôlav…ô yayƒ±lmanƒ± d…ôyi≈ümir)
3. $\sigma(cX) = |c| \cdot \sigma(X)$ (vurma yayƒ±lmanƒ± d…ôyi≈üir)
4. Eyni vahidl…ô √∂l√ß√ºl√ºr (manat, sm, ya≈ü v…ô s.)

---

### ü§î Yadda saxlamaq √º√ß√ºn:

* **Dispersiya**: d…ôyi≈üm…ôl…ôrin kvadrat ortasƒ±
* **Standart sapma**: onun kvadrat k√∂k√º ‚Äî daha **intuitiv √∂l√ß√º**
