## Diskret təsadüfi kəmiyyət nədir?

**Diskret təsadüfi kəmiyyət** — elə bir təsadüfi kəmiyyətdir ki, o yalnız **sayılacaq qədər sonlu və ya sayılabilən sonsuz** ədəd **ayrı-ayrı** qiymətlər qəbul edə bilər.

### 📌 Rəsmi tərif:

> **Diskret təsadüfi kəmiyyət** – mümkün qiymətlərinin sayı **sayılabilən** olan və bu qiymətlərin **hər birinə müəyyən ehtimal** aid edilə bilən təsadüfi kəmiyyətdir.

---

### 🔹 Nümunə:

Zəri atdıqda düşən nömrə $X \in \{1, 2, 3, 4, 5, 6\}$ — **diskret təsadüfi kəmiyyət**dir.
Çünki yalnız **ayrı-ayrı və sonlu 6 qiymət** ala bilər və bu qiymətlərin hər birinin ehtimalı var:

$$
P(X=1)=\frac{1}{6}, \; P(X=2)=\frac{1}{6}, \dots, P(X=6)=\frac{1}{6}
$$

---

### ✅ Əsas xüsusiyyətlər:

* Qiymətləri **sayılabiləndir** (məsələn: 0, 1, 2, 3, …).
* Hər qiymətə **ehtimal** uyğun gəlir.
* Ehtimalların **cəmi 1** olmalıdır:

$$
\sum_{i} P(X = x_i) = 1
$$


## Kəsilməz (davamlı) təsadüfi kəmiyyət nədir?



**Kəsilməz (və ya davamlı) təsadüfi kəmiyyət** – qiymətləri **sonsuz sayda** ola bilən və **müəyyən bir intervalda bütün real ədədləri** qəbul edə bilən təsadüfi kəmiyyətə deyilir.

---

### 📌 Rəsmi tərif:

> Təsadüfi kəmiyyət $X$ istənilən iki real ədəd $a$ və $b$ üçün $a < X < b$ şərtini ödəyən **sonsuz sayda qiymət** ala bilirsə, onda bu təsadüfi kəmiyyət **kəsilməzdir**.

---

### 🔹 Nümunə:

Bir məhsulun çəkisi (məsələn, 1 kq ilə 2 kq arasında istənilən real ədəd ola bilər: 1.1 kq, 1.156 kq, 1.78952 kq və s.) – bu **kəsilməz təsadüfi kəmiyyət**dir.

---

### ✅ Əsas xüsusiyyətlər:

* Qiymətləri **real ədədlər aralığında** dəyişir (məsələn: $X \in [0, 1]$)
* Tək-tək qiymətlərin ehtimalı **0**-dır:

$$
P(X = a) = 0
$$

* Ehtimallar **sıxlıq funksiyası (PDF)** vasitəsilə verilir.
* Ehtimallar **inteqral** ilə hesablanır:

$$
P(a < X < b) = \int_a^b f(x) \, dx
$$

Burada $f(x)$ — sıxlıq funksiyasıdır və aşağıdakı şərtləri ödəməlidir:

* $f(x) \geq 0$
* $\int_{-\infty}^{\infty} f(x) \, dx = 1$

---

### 🔁 Diskret vs Kəsilməz | Cədvəl:

| Xüsusiyyət                | Diskret Təsadüfi Kəmiyyət      | Kəsilməz Təsadüfi Kəmiyyət |
| ------------------------- | ------------------------------ | -------------------------- |
| Qiymətlərin sayı          | Sayılabilən (0, 1, 2, ...)     | Sonsuz çox (real ədədlər)  |
| Ehtimal forması           | Ehtimal kütlə funksiyası (PMF) | Sıxlıq funksiyası (PDF)    |
| Ehtimalların hesablanması | Sadə cəmlə                     | İntegralla                 |
| $P(X = a)$                | $> 0$ ola bilər                | Həmişə $= 0$               |



## PMF və PDF nədir?



### 📌 PMF nədir?

### **PMF** — *Probability Mass Function*

(azərbaycanca: **Ehtimal Kütlə Funksiyası**)

> PMF – **diskret** təsadüfi kəmiyyətin **mümkün qiymətlərinə** uyğun gələn **ehtimalları** göstərən funksiyadır.

### 📋 Tərif:

Əgər $X$ — diskret təsadüfi kəmiyyətdirsə, onda onun PMF-i belə yazılır:

$$
P(X = x_i) = p_i
$$

### ✅ Xassələri:

* $P(X = x_i) \geq 0$
* $\sum_{i} P(X = x_i) = 1$

### 🔹 Nümunə:

Zər atılanda:

$$
P(X = 1) = \frac{1}{6},\quad P(X = 2) = \frac{1}{6},\quad \ldots,\quad P(X = 6) = \frac{1}{6}
$$

---

### 📌 PDF nədir?

### **PDF** — *Probability Density Function*

(azərbaycanca: **Ehtimal Sıxlıq Funksiyası**)

> PDF – **kəsilməz** (davamlı) təsadüfi kəmiyyət üçün **ehtimal sıxlığını** göstərən funksiyadır.

### 📋 Tərif:

Əgər $X$ — davamlı təsadüfi kəmiyyətdirsə, onda ehtimal belə hesablanır:

$$
P(a < X < b) = \int_a^b f(x) \, dx
$$

Burada $f(x)$ — sıxlıq funksiyasıdır (**PDF**).

### ✅ Xassələri:

* $f(x) \geq 0$
* $\int_{-\infty}^{\infty} f(x) \, dx = 1$
* Tək bir nöqtənin ehtimalı:

$$
P(X = a) = 0
$$

### 🔹 Nümunə:

Birləşmiş \[0,1\] intervalında bərabər paylanma:

$$
f(x) =
\begin{cases}
1, & 0 \leq x \leq 1 \\
0, & əks halda
\end{cases}
$$

Bu zaman:

$$
P(0.2 < X < 0.5) = \int_{0.2}^{0.5} 1 \, dx = 0.3
$$



###   **$\int_{0.2}^{0.5} 1 \, dx = 0.3$** bərabərliyinin **nəyə görə doğru olduğunu** sadə izahı:

### 📌 İnkişafı:

Verilən inteqral:

$$
\int_{0.2}^{0.5} 1 \, dx
$$

Bu, **1 funksiyasının** $x$ oxunda **\[0.2, 0.5\]** aralığında **sahəsini** ifadə edir.

---

### ✅ Sadə riyazi izah:

Əgər sən bu funksiyanı sadəcə belə düşünsən:

$$
f(x) = 1 \quad \text{(yəni y = 1)}
$$

Bu zaman bu **sabit funksiyadır**, və inteqralın **qiyməti sadəcə bu düsturla** tapılır:

$$
\int_a^b c \, dx = c \cdot (b - a)
$$

Burada:

* $c = 1$
* $a = 0.2$
* $b = 0.5$

Ona görə:

$$
\int_{0.2}^{0.5} 1 \, dx = 1 \cdot (0.5 - 0.2) = 0.3
$$

---

### 🔍 Bu nə deməkdir ehtimal baxımından?

Bu inteqral, təsadüfi kəmiyyət $X$ — vahid aralıqda **bərabər paylanıbsa** (yəni PDF: $f(x) = 1$, $x \in [0,1]$), onda:

$$
P(0.2 < X < 0.5) = \int_{0.2}^{0.5} 1 \, dx = 0.3
$$

Yəni **0.2 ilə 0.5 arasında düşmə ehtimalı 0.3-dir**.

---

### 🔁 Fərqi sadə dillə:

|                | **PMF** (Diskret)      | **PDF** (Kəsilməz)               |
| -------------- | ---------------------- | -------------------------------- |
| Qiymətlər      | Ayrı-ayrı (1, 2, 3, …) | Sonsuz çox (real ədədlər)        |
| Ehtimal hesabı | $$P(X = x)$$             | $$P(a < X < b) = \int_a^b f(x)dx$$ |
| $$P(X = x)$$     | Müsbət ola bilər       | Həmişə 0                         |




## Kəsilməz təsadüfi kəmiyyətin Riyazi Gözləməsi 


### 📌 Təsadüfi Kəmiyyətlərin Sıxlıq Funksiyası (PDF) 

> **Tərif:** **Sıxlıq funksiyası** — **kəsilməz (davamlı) təsadüfi kəmiyyətin** ehtimallarını **sıxlıq şəklində** ifadə edən **riyazi funksiyadır**. Bu funksiya ehtimalların müəyyən intervalda necə "paylandığını" göstərir.


---

**Kəsilməz təsadüfi kəmiyyətin riyazi gözləməsi** (və ya **gözlənilən qiyməti**, **matematik gözləmə**) — ehtimal nəzəriyyəsində və statistikada **bir təsadüfi hadisənin orta nəticəsini** göstərən mərkəzi meyl ölçüsüdür. Bu anlayış, kəsilməz (yəni sonsuz sayda mümkün qiymət alan) təsadüfi kəmiyyətlər üçün xüsusi şəkildə tərtib edilir. 

Kəsilməz təsadüfi kəmiyyətin **riyazi gözləməsi** (mathematical expectation) və ya **orta qiyməti** (mean), onun bütün mümkün qiymətlərinin çəkili ortası kimi hesablanır, burada çəkilər ehtimal sıxlıq funksiyası (probability density function - PDF) ilə təyin olunur. 

---

### 🔹 **Kəsilməz təsadüfi kəmiyyətin Riyazi Gözləmə Tərifi:**

> **Tərif:** Kəsilməz təsadüfi kəmiyyət $X$ üçün onun **sıxlıq funksiyası** (PDF — Probability Density Function) $f(x)$ ilə verildikdə, bu təsadüfi kəmiyyətin **riyazi gözləməsi** aşağıdakı kimi hesablanır:

$$
\mathbb{E}[X] = \int_{-\infty}^{+\infty} x \cdot f(x) \, dx
$$

Burada:
- $ x $: Təsadüfi kəmiyyətin mümkün qiymətləri,
- \$ f(x) $: Ehtimal sıxlıq funksiyası ( $ f(x) \geq 0 $ və $ \int_{-\infty}^{\infty} f(x) \, dx = 1 $ ).

Bu inteqral **orta qiymət** funksiyasını verir. Yəni, $X$ təsadüfi kəmiyyətinin hansı dəyərləri daha çox alacağını və bu dəyərlərin "çəkisini" nəzərə alaraq orta nəticəni tapırıq.

---

### 🔹 **Əsas Şərtlər:**

1. $f(x) \geq 0$ olmalıdır hər $x$ üçün (sıxlıq funksiyası mənfi ola bilməz)
2. $\int_{-\infty}^{+\infty} f(x)\, dx = 1$ (ümumi ehtimal 1 olmalıdır)
3. Riyazi gözləmə yalnız o halda mövcuddur ki, aşağıdakı inteqral **mütləq yaxınlaşsın**:

   $$
   \int_{-\infty}^{+\infty} |x| \cdot f(x)\, dx < \infty
   $$

---

### 🔹 **İzah (nümunə ilə):**

Tutaq ki, $X$ təsadüfi kəmiyyəti **\[0,1\]** aralığında bərabər ehtimalla (yəni uniform paylanma ilə) qiymət alır. Onun sıxlıq funksiyası belədir:

$$
f(x) = 
\begin{cases}
1, & 0 \leq x \leq 1 \\
0, & \text{əks halda}
\end{cases}
$$

Riyazi gözləməsi:

$$
\mathbb{E}[X] = \int_{0}^{1} x \cdot 1 \, dx = \left[ \frac{x^2}{2} \right]_0^1 = \frac{1}{2}
$$

Bu o deməkdir ki, bu paylanmada **gözlənilən dəyər (orta)** 0.5-dir.

---

### 🔹 **Məna və Tətbiq:**


* Riyazi gözləmə, təsadüfi kəmiyyətin uzunmüddətli orta qiymətini göstərir. Məsələn, çox sayda təcrübə zamanı $ X $-in aldığı qiymətlərin ortalaması $ E(X) $-ə yaxınlaşır.
* Riyazi gözləmə — **çoxlu eksperimentlər aparıldıqda orta nəticə** kimi təyin olunur.
* Misal üçün, təbiətdə dəyişkən bir ölçü (məsələn, boy, gəlir, sürət) üzərində statistik model qurulursa, gözləmə həmin modelin mərkəzi göstəricisidir.
* Bu anlayış **iqtisad, sığorta, mühəndislik, süni intellekt, məlumat elmi və risk təhlili** sahələrində geniş tətbiq olunur.

---

### 🔹 **Əlavə: Əgər $g(X)$ funksiyası varsa?**

Əgər bizə $X$ təsadüfi kəmiyyətinin özü deyil, onun bir funksiyasının gözləməsi lazımdırsa, onda aşağıdakı formuldan istifadə olunur:

$$
\mathbb{E}[g(X)] = \int_{-\infty}^{+\infty} g(x) \cdot f(x) \, dx
$$

Məsələn, dispersiyanı hesablamaq üçün $\mathbb{E}[(X - \mu)^2]$ kimi ifadələr istifadə olunur.

> **Diqqət:** Riyazi gözləmənin mövcud olması üçün $ \int_{-\infty}^{\infty} |x| f(x) \, dx $ inteqralının **sonlu olması lazımdır (yəni, inteqral yığılmalıdır)**.


## Mütləq yığılan İnteqral

"**Mütləq yığılan**" ifadəsi riyaziyyatda (xüsusilə **inteqral** və **cəbri cəmlər**də) **bir funksiyanın və ya sıra/inteqralın qiymətinin sonsuzluqda belə "sabit", sonlu dəyərə yaxınlaşması** deməkdir – amma bu yaxınlaşma **modulda**, yəni **mütləq qiymət** baxımından olmalıdır.

---

### 🔹 Sadə dildə desək:

Bir **inteqral**:

$$
\int_{-\infty}^{+\infty} x \cdot f(x) \, dx
$$

yalnız o zaman **mövcuddur** (yəni "hesablana bilər") ki,

$$
\int_{-\infty}^{+\infty} |x \cdot f(x)| \, dx < \infty
$$

yəni bu inteqralın **modulu** da **sonlu bir ədəd** versin.

---

### 🔹 **Niyə mütləq yığılma vacibdir?**

Çünki ola bilər ki, **sadə inteqral** $\int x f(x) dx$ yığılsın, amma bu **təsadüfi şəkildə müsbət və mənfi ədədlərin kompensasiyası** sayəsində olsun.

Amma biz **əslində funksiyanın real "kütlə"si sonsuz ola bilərmi?** – bunu mütləq yığılma ilə yoxlayırıq.

---

### 🔹 Nümunə:

Tutaq ki,

$$
f(x) = \frac{1}{x^2}, \quad x \in [1, \infty)
$$

$$
\int_{1}^{\infty} \frac{1}{x^2} dx = \left[ -\frac{1}{x} \right]_1^\infty = 1 \quad \text{(yığılır)}
$$

Amma bu funksiyanın

$$
\int_{1}^{\infty} \left| \frac{1}{x^2} \right| dx = \int_{1}^{\infty} \frac{1}{x^2} dx = 1 < \infty
$$

Bu mütləq yığılır — yəni bu inteqral **mütləq yığılan**dır.

---

### 🔹 Əks nümunə:

$$
\int_{1}^{\infty} \frac{\sin x}{x} dx
$$

Bu inteqral **yığılır**, yəni müəyyən bir ədədə yaxınlaşır.

Ancaq

$$
\int_{1}^{\infty} \left| \frac{\sin x}{x} \right| dx = \infty
$$

Burada **mütləq yığılmır**, çünki mütləq qiymət altında inteqral **sonsuzdur**.

---

### ✅ Nəticə:

Riyazi gözləmənin mövcudluğu üçün **mütləq yığılma** vacibdir. Yəni:

$$
\int_{-\infty}^{+\infty} |x \cdot f(x)| dx < \infty
$$

Əgər bu şərt ödənmirsə, **gözlənilən qiymət (matematik gözləmə) yoxdur**, yəni həmin təsadüfi kəmiyyət üçün orta dəyər **hesablana bilməz**.

## Kəsilməz təsadüfi kəmiyyətin Dispersiyası

**Kəsilməz (davamlı) təsadüfi kəmiyyətin dispersiyası** — bu təsadüfi kəmiyyətin orta qiymətdən nə qədər sapdığını ölçən **dispersiya** anlayışının kəsilməz vəziyyətdə riyazi ifadəsidir.


> **Tərif:** Kəsilməz təsadüfi kəmiyyətin dispersiyası — bu kəmiyyətin **orta qiymət ətrafında nə qədər dəyişdiyini** göstərən **statistik ölçüdür**. Riyazi gözləmədən (orta dəyərdən) sapmaları kvadrat şəklində ölçür. **Sadə dillə:** Əgər bir dəyişənin dəyərləri ortadan çox yayılıbsa, onun dispersiyası böyük olacaq.

---

### 🔹 **Tərifi:**

Əgər $X$ – sıxlıq funksiyası $f(x)$ olan **kəsilməz təsadüfi kəmiyyətdirsə**, onun **dispersiyası** aşağıdakı düsturla hesablanır:

$$
\mathrm{Var}(X) = \mathbb{E}[(X - \mathbb{E}[X])^2] = \int_{-\infty}^{\infty} (x - \mathbb{E}[X])^2 f(x) \, dx
$$

Burada:

* $\mathbb{E}[X] = \int_{-\infty}^{\infty} x f(x) \, dx$ — **X-in riyazi gözləməsidir (orta qiyməti)**,
* $f(x)$ — **X-in sıxlıq funksiyasıdır**.

---

### 🔹 **Alternativ (praktik) düstur:**

Dispersiyanı bu düsturla da hesablamaq olur:

$$
\mathrm{Var}(X) = \mathbb{E}[X^2] - (\mathbb{E}[X])^2
$$

Burada:

* $\mathbb{E}[X^2] = \int_{-\infty}^{\infty} x^2 f(x) \, dx$ — **X- in kvadratının gözlənən qiyməti**,
* $\mathbb{E}[X] = \int_{-\infty}^{\infty} x f(x) \$ — **X- in riyazi gözləməsidir (orta qiyməti)**.

---



### ✅ **Əsas Məqamlar:**

1. **Dispersiya mənfi ola bilməz** — çünki kvadratlarla işlənir.
2. **Dispersiya 0 olarsa**, bu o deməkdir ki, təsadüfi kəmiyyət dəyişmir, yəni **sabit bir ədəddir**.
3. **Dispersiya ölçü vahidinin kvadratı ilə ölçülür.**

   * Məs: Əgər $X$ uzunluqsa (m), onda dispersiya m²-dir.
4. **Dispersiya yayılmanın miqyasını göstərir**, amma **təklikdə dəyişkənliyin ölçüsü kimi** istifadə olunmur. Bu məqsədlə standart sapma ($\sigma = \sqrt{\text{Var}(X)}$) daha uyğundur.


---


### 🔸 **Nümunə 1:**

Deyək ki, $X \sim U(0,1)$ — yəni X vahid intervalda bərabər paylanmış kəsilməz təsadüfi kəmiyyətdir. O zaman:

* $f(x) = 1$, $0 \le x \le 1$
* $\mathbb{E}[X] = \int_0^1 x \cdot 1 \, dx = \frac{1}{2}$
* $\mathbb{E}[X^2] = \int_0^1 x^2 \cdot 1 \, dx = \frac{1}{3}$

**Dispersiya:**

$$
\mathrm{Var}(X) = \mathbb{E}[X^2] - (\mathbb{E}[X])^2 = \frac{1}{3} - \left(\frac{1}{2}\right)^2 = \frac{1}{3} - \frac{1}{4} = \frac{1}{12}
$$

---

### ✅ **Nümunə 2:**

Tutaq ki, təsadüfi kəmiyyət $X$ aşağıdakı sıxlıq funksiyası ilə verilib:

$$
f(x) = \begin{cases}
2x, & 0 \le x \le 1 \\
0, & \text{əks halda}
\end{cases}
$$

Bu halda:

* $\mathbb{E}[X] = \int_0^1 x \cdot 2x\,dx = \int_0^1 2x^2\,dx = \frac{2}{3}$
* $\mathbb{E}[X^2] = \int_0^1 x^2 \cdot 2x\,dx = \int_0^1 2x^3\,dx = \frac{1}{2}$
* Dispersiya:

  $$
  \text{Var}(X) = \mathbb{E}[X^2] - (\mathbb{E}[X])^2 = \frac{1}{2} - \left( \frac{2}{3} \right)^2 = \frac{1}{18}
  $$
