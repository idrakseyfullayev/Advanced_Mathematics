## KÉ™silmÉ™z tÉ™sadÃ¼fi kÉ™miyyÉ™tin DispersiyasÄ±

**KÉ™silmÉ™z (davamlÄ±) tÉ™sadÃ¼fi kÉ™miyyÉ™tin dispersiyasÄ±** â€” bu tÉ™sadÃ¼fi kÉ™miyyÉ™tin orta qiymÉ™tdÉ™n nÉ™ qÉ™dÉ™r sapdÄ±ÄŸÄ±nÄ± Ã¶lÃ§É™n **dispersiya** anlayÄ±ÅŸÄ±nÄ±n kÉ™silmÉ™z vÉ™ziyyÉ™tdÉ™ riyazi ifadÉ™sidir.


> **TÉ™rif:** KÉ™silmÉ™z tÉ™sadÃ¼fi kÉ™miyyÉ™tin dispersiyasÄ± â€” bu kÉ™miyyÉ™tin **orta qiymÉ™t É™trafÄ±nda nÉ™ qÉ™dÉ™r dÉ™yiÅŸdiyini** gÃ¶stÉ™rÉ™n **statistik Ã¶lÃ§Ã¼dÃ¼r**. Riyazi gÃ¶zlÉ™mÉ™dÉ™n (orta dÉ™yÉ™rdÉ™n) sapmalarÄ± kvadrat ÅŸÉ™klindÉ™ Ã¶lÃ§Ã¼r. **SadÉ™ dillÉ™:** ÆgÉ™r bir dÉ™yiÅŸÉ™nin dÉ™yÉ™rlÉ™ri ortadan Ã§ox yayÄ±lÄ±bsa, onun dispersiyasÄ± bÃ¶yÃ¼k olacaq.

---

### ğŸ”¹ **TÉ™rifi:**

ÆgÉ™r $X$ â€“ sÄ±xlÄ±q funksiyasÄ± $f(x)$ olan **kÉ™silmÉ™z tÉ™sadÃ¼fi kÉ™miyyÉ™tdirsÉ™**, onun **dispersiyasÄ±** aÅŸaÄŸÄ±dakÄ± dÃ¼sturla hesablanÄ±r:

$$
\mathrm{Var}(X) = \mathbb{E}[(X - \mathbb{E}[X])^2] = \int_{-\infty}^{\infty} (x - \mathbb{E}[X])^2 f(x) \, dx
$$

Burada:

* $\mathbb{E}[X] = \int_{-\infty}^{\infty} x f(x) \, dx$ â€” **X-in riyazi gÃ¶zlÉ™mÉ™sidir (orta qiymÉ™ti)**,
* $f(x)$ â€” **X-in sÄ±xlÄ±q funksiyasÄ±dÄ±r**.

---

### ğŸ”¹ **Alternativ (praktik) dÃ¼stur:**

DispersiyanÄ± bu dÃ¼sturla da hesablamaq olur:

$$
\mathrm{Var}(X) = \mathbb{E}[X^2] - (\mathbb{E}[X])^2
$$

Burada:

* $\mathbb{E}[X^2] = \int_{-\infty}^{\infty} x^2 f(x) \, dx$ â€” **X- in kvadratÄ±nÄ±n gÃ¶zlÉ™nÉ™n qiymÉ™ti**,
* $\mathbb{E}[X] = \int_{-\infty}^{\infty} x f(x) \$ â€” **X- in riyazi gÃ¶zlÉ™mÉ™sidir (orta qiymÉ™ti)**.

---



### âœ… **Æsas MÉ™qamlar:**

1. **Dispersiya mÉ™nfi ola bilmÉ™z** â€” Ã§Ã¼nki kvadratlarla iÅŸlÉ™nir.
2. **Dispersiya 0 olarsa**, bu o demÉ™kdir ki, tÉ™sadÃ¼fi kÉ™miyyÉ™t dÉ™yiÅŸmir, yÉ™ni **sabit bir É™dÉ™ddir**.
3. **Dispersiya Ã¶lÃ§Ã¼ vahidinin kvadratÄ± ilÉ™ Ã¶lÃ§Ã¼lÃ¼r.**

   * MÉ™s: ÆgÉ™r $X$ uzunluqsa (m), onda dispersiya mÂ²-dir.
4. **Dispersiya yayÄ±lmanÄ±n miqyasÄ±nÄ± gÃ¶stÉ™rir**, amma **tÉ™klikdÉ™ dÉ™yiÅŸkÉ™nliyin Ã¶lÃ§Ã¼sÃ¼ kimi** istifadÉ™ olunmur. Bu mÉ™qsÉ™dlÉ™ standart sapma ($\sigma = \sqrt{\text{Var}(X)}$) daha uyÄŸundur.


---


### ğŸ”¸ **NÃ¼munÉ™ 1:**

DeyÉ™k ki, $X \sim U(0,1)$ â€” yÉ™ni X vahid intervalda bÉ™rabÉ™r paylanmÄ±ÅŸ kÉ™silmÉ™z tÉ™sadÃ¼fi kÉ™miyyÉ™tdir. O zaman:

* $f(x) = 1$, $0 \le x \le 1$
* $\mathbb{E}[X] = \int_0^1 x \cdot 1 \, dx = \frac{1}{2}$
* $\mathbb{E}[X^2] = \int_0^1 x^2 \cdot 1 \, dx = \frac{1}{3}$

**Dispersiya:**

$$
\mathrm{Var}(X) = \mathbb{E}[X^2] - (\mathbb{E}[X])^2 = \frac{1}{3} - \left(\frac{1}{2}\right)^2 = \frac{1}{3} - \frac{1}{4} = \frac{1}{12}
$$

---

### âœ… **NÃ¼munÉ™ 2:**

Tutaq ki, tÉ™sadÃ¼fi kÉ™miyyÉ™t $X$ aÅŸaÄŸÄ±dakÄ± sÄ±xlÄ±q funksiyasÄ± ilÉ™ verilib:

$$
f(x) = \begin{cases}
2x, & 0 \le x \le 1 \\
0, & \text{É™ks halda}
\end{cases}
$$

Bu halda:

* $\mathbb{E}[X] = \int_0^1 x \cdot 2x\,dx = \int_0^1 2x^2\,dx = \frac{2}{3}$
* $\mathbb{E}[X^2] = \int_0^1 x^2 \cdot 2x\,dx = \int_0^1 2x^3\,dx = \frac{1}{2}$
* Dispersiya:

  $$
  \text{Var}(X) = \mathbb{E}[X^2] - (\mathbb{E}[X])^2 = \frac{1}{2} - \left( \frac{2}{3} \right)^2 = \frac{1}{18}
  $$
