## Tam Ehtimal Düsturu


>**Tərif:** **Tam Ehtimal Düsturu**, bir hadisənin ümumi ehtimalını, onu doğura biləcək qarşılıqlı xaric olan və birlikdə bütün ehtimal məkanını əhatə edən halların ehtimalları və şərti ehtimalları vasitəsilə hesablama qaydasıdır.





### ✅ **1. Tərif:**

**Tam ehtimal düsturu**, mürəkkəb bir hadisənin ehtimalını, onu təşkil edən **bir-birini istisna edən və birlikdə tam olan hadisələrə** əsaslanaraq hesablamağa imkan verir.

---

### ✅ **2. Düstur:**

Əgər hadisələr $B_1, B_2, ..., B_n$:

* **bölünməz və tam sistem təşkil edirsə** (yəni biri mütləq baş verir),
* və A hadisəsinin ehtimalı bu B-lərdən asılıdırsa,

onda:

$$
P(A) = \sum_{i=1}^{n} P(B_i) \cdot P(A \mid B_i)
$$

---

### ✅ **3. Anlayışlar:**

| Anlayış             | İzah                                                                                                   |
| ------------------- | ------------------------------------------------------------------------------------------------------ |
| **Tam sistem**      | $$B_1, B_2, ..., B_n$$ – bir-birini istisna edən və birlikdə bütün hadisə məkanını əhatə edən hadisələr. |
| **$$P(B_i)$$**        | Hər bir hadisənin baş vermə ehtimalı.                                                                  |
| **$$P(A \mid B_i)$$** | A hadisəsinin, $B_i$ baş verəndən sonra baş vermə ehtimalı (**şərti ehtimal**).                        |
| **$$P(A)$$**          | A hadisəsinin ümumi ehtimalı.                                                                          |

---

### ✅ **4. Sadə Misal:**

Bir xəstənin 3 fərqli klinikada müayinə olunma ehtimalları belədir:

* Klinik A: 40% ehtimalla gedir → $P(B_1) = 0.4$
* Klinik B: 35% ehtimalla gedir → $P(B_2) = 0.35$
* Klinik C: 25% ehtimalla gedir → $P(B_3) = 0.25$

Müvafiq klinikada **düzgün diaqnoz** qoyulma ehtimalları:

* Klinik A: $P(A \mid B_1) = 0.9$
* Klinik B: $P(A \mid B_2) = 0.8$
* Klinik C: $P(A \mid B_3) = 0.6$

**Sual:** Xəstənin düzgün diaqnoz alması ümumi ehtimalı neçədir?

$$
P(A) = 0.4 \cdot 0.9 + 0.35 \cdot 0.8 + 0.25 \cdot 0.6 = 0.36 + 0.28 + 0.15 = \boxed{0.79}
$$

---

## Bayes Düsturu



### ✅ **1. TƏRİF:**

>**TƏRİF:** **Bayes Teoremi**, hadisənin ehtimalını onunla əlaqəli digər hadisənin baş verdiyi məlum olduqda, şərti ehtimallar və ilkin ehtimallara əsasən yenidən hesablama qaydasıdır.

---

### ✅ **2. DÜSTUR:**

Əgər $B_1, B_2, ..., B_n$ qarşılıqlı xaric olan hadisələrdir və birlikdə tam sistem təşkil edirlərsə, A hadisəsi baş veribsə, onda:

$$
P(B_i \mid A) = \frac{P(B_i) \cdot P(A \mid B_i)}{\sum_{j=1}^{n} P(B_j) \cdot P(A \mid B_j)}
$$

Sadə formada (2 hadisə üçün):

$$
P(B \mid A) = \frac{P(B) \cdot P(A \mid B)}{P(A)}
$$


#### 1. $P(B \mid A)$

Bu:
- 👉 “A hadisəsi baş veribsə, **B-nin baş vermə ehtimalı** nə qədərdir?”
- Yəni, **A artıq olub**, indi maraqlanırıq ki, **B-nin baş vermə şansı nə qədərdir?**
- Bu, **şərti ehtimal** adlanır.

---

#### 2. $P(B)$

Bu:
-  👉 B hadisəsinin **ümumilikdə baş vermə ehtimalıdır**.
-  Yəni heç bir başqa hadisəyə baxmadan, sadəcə B-nin nə qədər baş verə biləcəyini göstərir.
-  Buna **"ön ehtimal"** da deyirlər.

---

#### 3. $P(A \mid B)$

Bu:
- 👉 “Əgər B baş veribsə, o zaman **A-nın baş vermə ehtimalı** nə qədərdir?”
- Yəni B-nin olması halında A-nın şansı nə qədərdir.

---

#### 4. $P(A)$

Bu:
- 👉 A hadisəsinin **ümumi ehtimalıdır**.
- Yəni ümumiyyətlə A-nın nə qədər baş verə biləcəyini göstərir.
- Buna **normallaşdırma faktoru** da deyirlər, çünki o nəticəni balanslaşdırmaq üçün lazımdır.

---

#### 💡 Sadə Misal:

Tutaq ki:

* B: İnsan **xəstədir**.
* A: Onun **test nəticəsi müsbət çıxıb**.

Bu düsturla belə bir sualı cavablayırıq:
- 👉 Əgər test müsbətdirsə, insanın **doğrudan xəstə olma ehtimalı nə qədərdir?**

---

### ✅ **3. ANLAYIŞLAR:**

| Anlayış           | İzah                                                                           |
| ----------------- | ------------------------------------------------------------------------------ |
| **$$P(B)$$**        | Hadisə B-nin ilkin (əvvəlcədən bilinən) ehtimalı – **prior** ehtimal.          |
| **$$P(A \mid B)$$** | A hadisəsinin B baş verdikdə ehtimalı – **şərti ehtimal**.                     |
| **$$P(B \mid A)$$** | A hadisəsi baş verdikdən sonra B hadisəsinin ehtimalı – **posterior** ehtimal. |
| **$$P(A)$$**        | A hadisəsinin ümumi ehtimalı – **tam ehtimal düsturu** ilə hesablanır.         |

---

### ✅ **4. Sadə Misal:**

Bir xəstənin **xəstə olma ehtimalı**:

* $P(Xəstə) = 0.01$

Testin düzgün nəticə vermə ehtimalları:

* Test **xəstəyə müsbət** deyirsə: $P(Pozitiv \mid Xəstə) = 0.99$
* Test **sağlam insana da müsbət** deyə bilər: $P(Pozitiv \mid Sağlam) = 0.05$

**Sual:** Test nəticəsi müsbət çıxarsa, həmin şəxsin **həqiqətən xəstə olma ehtimalı** neçədir?

---

### Həll:

$$
P(Xəstə \mid Pozitiv) = \frac{P(Xəstə) \cdot P(Pozitiv \mid Xəstə)}{P(Xəstə) \cdot P(Pozitiv \mid Xəstə) + P(Sağlam) \cdot P(Pozitiv \mid Sağlam)}
$$

$$
= \frac{0.01 \cdot 0.99}{0.01 \cdot 0.99 + 0.99 \cdot 0.05} \approx \frac{0.0099}{0.0099 + 0.0495} = \frac{0.0099}{0.0594} \approx \boxed{0.167}
$$

Yəni test müsbət çıxsa da, həqiqətən xəstə olma ehtimalı sadəcə **16.7%**-dir.



### 🧠 **Bayes Teoremi — Nümunə:**

### 📘 **Vəziyyət:**

Bir şəhərdə:

* Əhalinin **1%-i** müəyyən bir xəstəliyə yoluxub.
* Bu xəstəliyi aşkarlayan test mövcuddur:

  * **Xəstə olan** şəxsi **90% ehtimalla düzgün tapır**. (True Positive)
  * **Sağlam olan** şəxsi **95% ehtimalla düzgün tanıyır**. (True Negative)

---

### 🔹 **Sual:**

Test **müsbət çıxarsa**, həmin şəxsin **həqiqətən xəstə olma ehtimalı nə qədərdir?**

Yəni bizə **$P(\text{Xəstə} \mid \text{Test+})$** lazımdır.

---

### ✳️ **Verilənləri yazırıq:**

| Hadisə                  | Simvol                    | Ehtimal            |
| ----------------------- | ------------------------- | ------------------ |
| Şəxs xəstədir           | $$D$$                       | $$P(D) = 0.01$$      |
| Şəxs sağlamdır          | $$\neg D$$                  | $$P(\neg D) = 0.99$$ |
| Test müsbətdir (Test+)  | $$+$$                       | veriləcək aşağıda  |
| Xəstə ikən test+ çıxır  | $$P(+ \mid D) = 0.90$$      |                    |
| Sağlam ikən test+ çıxır | $$P(+ \mid \neg D) = 0.05$$ |                    |

---

### ✅ **Bayes düsturu ilə hesablayaq:**

$$
\boxed{
P(A \mid B) = \frac{P(B \mid A) \cdot P(A)}{P(B)}
}
$$

$$
P(D \mid +) = \frac{P(+ \mid D) \cdot P(D)}{P(+)}
$$

Əvvəlcə **$P(+)$** tapırıq (testin ümumilikdə müsbət çıxma ehtimalı):

$$
P(+) = P(+ \mid D) \cdot P(D) + P(+ \mid \neg D) \cdot P(\neg D)
$$

$$
= 0.90 \cdot 0.01 + 0.05 \cdot 0.99 = 0.009 + 0.0495 = 0.0585
$$

---

İndi əsas ehtimalı hesablayaq:

$$
P(D \mid +) = \frac{0.90 \cdot 0.01}{0.0585} = \frac{0.009}{0.0585} \approx \boxed{0.1538} \approx \boxed{15.4\%}
$$

---

### 📌 **Nəticə:**

Test **müsbət çıxsa belə**, şəxsin həqiqətən xəstə olma ehtimalı sadəcə **15.4%**-dir!

---

### 🔎 **Yorum:**

Bu nəticə çox insan üçün təəccüblüdür. Amma səbəbi budur ki:

* Xəstəlik **nadir haldır (1%)**, yəni çox adam onsuz da sağlamdır.
* Testin səhv **sağlam insana müsbət** verməsi az da olsa (5%) çox insana təsir edir (çünki sağlam çoxdur!).



### Bu düstur  niyə belədir: 
$$
P(+) = \underbrace{P(+ \mid \text{Xəstə}) \cdot P(\text{Xəstə})}_{\text{True Positive}} + \underbrace{P(+ \mid \text{Sağlam}) \cdot P(\text{Sağlam})}_{\text{False Positive}}
$$




> **Niyə $P(+ \mid \text{Xəstə}) \cdot P(\text{Xəstə})$ kimi yazırıq?**  Yəni, niyə bu iki ehtimalı bir-birinə **vururuq**?



### 🧠 Cavab qısa desək:

Bu, **birləşmiş ehtimal**dır (joint probability):
Bir hadisənin **şərti ehtimalı** ilə **o hadisənin özü**nü **eyni anda** baş vermə ehtimalını hesablamaq üçün **vururuq**:

$$
P(\text{Xəstə və Test müsbətdir}) = P(+ \mid \text{Xəstə}) \cdot P(\text{Xəstə})
$$



### 💡 Gəlin gündəlik həyat nümunəsiylə başa düşək:

Təsəvvür et:

* 1000 nəfərlik bir şəhərdə yaşayırıq.
* Onların **10%-i (yəni 100 nəfər)** bir xəstəliyə yoluxub.
* Bu xəstə insanlar üçün test **90% ehtimalla müsbət çıxır.**

**Sual: Bu 1000 nəfərin neçə nəfəri həm xəstədir, həm də testdə müsbət çıxır?**

👉 Gəlin hesablayaq:

* Xəstə sayı: **1000 × 0.1 = 100 nəfər**
* Onların 90%-i üçün test müsbət çıxacaq → **100 × 0.9 = 90 nəfər**

Yəni **əhalinin 9%-i** həm xəstədir, həm testdə müsbətdir.
Bu **birgə ehtimaldır**:

$$
P(\text{Xəstə və müsbət}) = P(\text{Xəstə}) \cdot P(+ \mid \text{Xəstə})
$$


### 📌 Nəticə olaraq:

**Birgə ehtimal** tapanda:

* Əvvəlcə **birinci hadisənin ehtimalını** götürürük: $P(\text{Xəstə})$
* Sonra **ikinci hadisənin, birinci şərti ilə, ehtimalını** vururuq: $P(+ \mid \text{Xəstə})$

Yəni:

$$
P(\text{Xəstə və +}) = P(\text{Xəstə}) \cdot P(+ \mid \text{Xəstə})
$$


---

## ❓ Sual:

> **Niye $P(+ \mid \text{Sağlam}) \cdot P(\text{Sağlam})$ ilə "səhvən müsbət" ehtimalını tapırıq?**



## 🧠 Cavab:

Bu da **birgə ehtimal**dır — yəni:

> **Şəxs sağlamdır** **və** test onun üçün **səhvən müsbət çıxır**.

Bu cüt hadisə üçün ehtimal belə hesablanır:

$$
P(\text{Sağlam və test müsbət çıxır}) = P(\text{Sağlam}) \cdot P(+ \mid \text{Sağlam})
$$

Yəni:

* **$P(\text{Sağlam})$** — adam ümumiyyətlə sağlamdır
* **$P(+ \mid \text{Sağlam})$** — sağlam adama test səhvən müsbət nəticə verib



## 📊 Sadə nümunə ilə başa salım:

Tutaq ki, 10,000 nəfərlik əhali var.

* 1% xəstədir → 100 nəfər
* 99% sağlamdır → 9900 nəfər

Testin **səhvən müsbət** çıxma ehtimalı:

$$
P(+ \mid \text{Sağlam}) = 5\% = 0.05
$$

Bəs **neçə nəfərdə test səhvən müsbət çıxar?**

$$
9900 \cdot 0.05 = 495 \ nəfərdə
$$

Bu 495 nəfər — **sağlam olduğu halda test müsbət çıxanlar**dır. Onların ehtimalı isə belə yazılır:

$$
P(\text{Sağlam və +}) = P(\text{Sağlam}) \cdot P(+ \mid \text{Sağlam}) = 0.99 \cdot 0.05 = 0.0495
$$



## 🔁 Niyə vururuq?

Çünki bir şəxsin:

* **Sağlam olması** (başqa sözlə: testin səhv işləyə biləcəyi bir halda olması)
* **Testin səhvən müsbət çıxması** – bu iki hadisənin **eyni anda baş vermə ehtimalını** hesablamaq üçün **vururuq**.



## ✅ Nəticə:

$$
P(\text{səhvən müsbət}) = P(\text{sağlam}) \cdot P(+ \mid \text{sağlam})
$$

Bu, **"false positive"** ehtimalını verir və Bayes formulunda **paydada** vacib rol oynayır.



## **Nə üçün cavab yalnız 15.4% çıxdı** və **nəticənin niyə təəccüblü olduğunu** sadə dillə izahı



### 🔍 Sualımız nə idi?

Test **müsbət** çıxarsa, **şəxsin həqiqətən xəstə olma ehtimalı nə qədərdir?**

> Cavab: təxminən **15.4%**

---

### ❓ **Bəs niyə bu qədər azdır? Test müsbətdirsə, niyə 90% deyil?**

---

### Gəlin məsələyə **saylarla** baxaq ki, tam aydın olsun:

Təsəvvür et:

* Şəhərdə **10.000 nəfər** var.
* Onların **1%-i** xəstədir → Yəni **100 nəfər** xəstədir, **9900 nəfər** sağlamdır.

---

### 🔬 Test necə işləyir?

### 1. **Xəstə olanlara test verək:**

* 100 nəfər xəstədir.
* Onların **90%-i testlə düzgün tutulur** → **90 nəfərdə test müsbət çıxır** (✅ **doğru müsbət**)
* Qalan **10 nəfər xəstə olsa da, test səhvən mənfi çıxır** (❌ **yalançı mənfi**)

### 2. **Sağlam olanlara test verək:**

* 9900 nəfər sağlamdır.
* Testin **5%-i səhvən müsbət çıxır** →

  $$
  0.05 \cdot 9900 = 495 \text{ nəfərdə test səhvən müsbət çıxır} \quad (\text{❌ yalançı müsbət})
  $$

---

### ✳️ **Cəmi müsbət çıxanlar:**

* Doğrudan xəstə olanlar: **90 nəfər**
* Sağlam olub testdə səhvən müsbət çıxanlar: **495 nəfər**

> Yəni **müsbət test nəticəsi olanlar = 90 + 495 = 585 nəfər**

---

### ✅ **İndi ehtimal hesabı:**

Bu 585 nəfərin **neçəsi həqiqətən xəstədir?**

$$
P(\text{xəstə} \mid \text{test+}) = \frac{90}{585} \approx 0.154 = 15.4\%
$$

---

### 💡 **YEKUN NƏTİCƏ – Sadə dillə:**

* Test **özlüyündə yaxşıdır** (xəstəni 90% tutur),
* Amma **xəstəlik çox nadirdir** (cəmi 1% insanda var),
* Sağlam adam **çoxdur** (9900 nəfər),
* Bu qədər sağlamdan **cüzi faizlə səhv çıxsa belə**, **səhv nəticələrin sayı çox olur.**

---

### 🧠 Analogia:

Təsəvvür et **içində 1 qara kürə və 99 ağ kürə** olan bir qutun var.
Sən gözübağlı çəkirsən, və “bəlkə qaradır” deyən test 90% dəqiqliklə işləmir.
Əksəriyyət ağ olduğuna görə, **səhv təxminlər** çox olur.
