## Bayes teoremi ilə məsələ



## 🧩 Verilən məsələ:

Qutuda:

* 3 qırmızı (Q), 2 yaşıl (Y) kürə var.
* Gözüyumulu **2 kürə seçilir**.
* **Seçilən 2 kürədən biri qırmızı olduğu məlumdur.**
* Sual: **İkincinin də qırmızı olma ehtimalı nədir?**

---

## 🔶 Məntiq:

Bizə verilən:
**Hadisə B** – seçilən kürələrin **ən az biri qırmızıdır**
**Hadisə A** – seçilən **hər ikisi də qırmızıdır**

Sual budur:

$$
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
$$

---

## 🧮 Hissələri hesablayaq:

### 🔸 1. $P(A)$: Hər ikisi də qırmızı seçilmə ehtimalı

3 qırmızıdan 2-sini seçmək:

$$
P(A) = \frac{C(3, 2)}{C(5, 2)} = \frac{3}{10}
$$

---

### 🔸 2. $P(B|A)$: Əgər ikisi də qırmızıdırsa, "ən azı biri qırmızıdır" ehtimalı

Bu **dəqiq doğrudur**. Çünki **ikisi də qırmızı**dırsa, deməli **ən azı biri** mütləq qırmızıdır.

$$
P(B|A) = 1
$$

---

### 🔸 3. $P(B)$: "Ən azı biri qırmızıdır" ehtimalı

Bunu belə hesablamaq rahatdır:

$$
P(B) = 1 - P(\text{heç biri qırmızı deyil})
$$

Yəni, **ikisi də yaşıl** seçilmə ehtimalını tapıb 1-dən çıxırıq:

$$
P(B) = 1 - \frac{C(2, 2)}{C(5, 2)} = 1 - \frac{1}{10} = \frac{9}{10}
$$

---

## 🔚 İndi Bayes formuluna yerinə qoyaq:

$$
P(A|B) = \frac{1 \cdot \frac{3}{10}}{\frac{9}{10}} = \frac{3}{9} = \boxed{\frac{1}{3}}
$$

---

## ✅ Cavab: $\boxed{\frac{1}{3}}$

Beləliklə, Bayes teoremi vasitəsilə **eyni cavabı** aldıq — bu da göstərir ki, Bayes formalı yanaşma ilə məntiqi sayma metodları nəticəni uyğunlaşdırır.
