## MÃ¼ntÉ™zÉ™m (Uniform) paylanma



### âœ… **1. Uniform (BÉ™rabÉ™rsÄ±xlÄ±qlÄ±) PaylanmanÄ±n TÉ™rifi**

> **TÉ™rif:** Uniform paylanma â€” tÉ™sadÃ¼fi dÉ™yiÅŸÉ™nin mÃ¼É™yyÉ™n **mÃ¼ddÉ™t É™rzindÉ™ eyni ehtimalla** bÃ¼tÃ¼n dÉ™yÉ™rlÉ™ri ala bildiyi ehtimal paylanmasÄ±dÄ±r.

> **TÉ™rif:** Uniform (bÉ™rabÉ™r) paylanma, **bÃ¼tÃ¼n dÉ™yÉ™rlÉ™rin eyni ehtimalla** baÅŸ verdiyi bir paylanmadÄ±r.
> YÉ™ni tÉ™sadÃ¼fi dÉ™yiÅŸÉ™n $X$, $[a, b]$ aralÄ±ÄŸÄ±nda **hÉ™r bir dÉ™yÉ™ri bÉ™rabÉ™r ehtimalla** ala bilÉ™r.

### âœ… ParametrlÉ™r

* `a`: Minimum dÉ™yÉ™r (baÅŸlanÄŸÄ±c nÃ¶qtÉ™si)
* `b`: Maksimum dÉ™yÉ™r (son nÃ¶qtÉ™)

Uniform paylanma adÉ™tÉ™n belÉ™ situasiyanÄ± tÉ™svir edir:

* TÉ™sadÃ¼fi dÉ™yiÅŸÉ™n yalnÄ±z **\[a, b\]** intervalÄ±nda dÉ™yÉ™rlÉ™r alÄ±r.

### âœ… 2. Ehtimal SÄ±xlÄ±q FunksiyasÄ± (PDF)

Kontinual uniform paylanma Ã¼Ã§Ã¼n ehtimal sÄ±xlÄ±q funksiyasÄ±:

$$

> Bu funksiya **\$x\$-É™ qÉ™dÉ™r olan dÉ™yÉ™rlÉ™rin ehtimalÄ±nÄ±** gÃ¶stÉ™rir.

  $$

  $$

âœ… YÉ™ni \$X\$-in \$3\$ vÉ™ \$5\$ arasÄ±nda dÃ¼ÅŸmÉ™ ehtimalÄ± **50%**-dir.
