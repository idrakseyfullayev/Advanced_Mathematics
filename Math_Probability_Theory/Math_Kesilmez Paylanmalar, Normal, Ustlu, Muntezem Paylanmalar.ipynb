## Normal Paylanma



### ✅ **1. Normal Paylanmanın Tərifi**

> **Tərif:** Normal paylanma — **kontinu (kəsilməz) ehtimal paylanmasıdır** və simmetrik, zəngvari (bell-shaped) forma malikdir. Mərkəzi ətrafında daha sıx, kənarlarda isə seyrək paylanmış dəyərləri modelləşdirir.

> **Tərif:** **Normal paylanma**, təsadüfi dəyişənin çoxlu müstəqil və təsadüfi faktorların təsiri ilə formalaşdığı hallarda, **orta ətrafında simmetrik** olaraq düzülən və **zəngvari formada olan ehtimal paylanmasıdır**. Normal paylanma, təsadüfi dəyişənin çoxlu təsirlər altında necə **orta qiymət ətrafında yığışdığını** təsvir edən **kontinu və simmetrik ehtimal paylanmasıdır.**

---

### ✅ Parametrlər

* `μ` — Orta (gözlənilən dəyər, mean)
* `σ` — Standart sapma (dispersiyanın kvadrat kökü)
* `σ²` — Dispersiya (varyans)
* `x` — Ehtimalı hesablanacaq konkret dəyər

---

**Normal paylanma** adətən belə bir halı modelləşdirir:

* Təsadüfi dəyişən kəsilməz (kontinu) qiymətlər ala bilər (məsələn: boy, çəki, IQ).
* Paylanma **simmetrikdir**, yəni ortalamanın sol və sağ tərəfi eyni formadadır.
* Dəyərlərin əksəriyyəti ortalama ətrafında cəmlənir, uç dəyərlər isə az olur.
* Normal paylanma **iki parametr** ilə müəyyən olunur: **orta (μ)** və **standart sapma (σ)**.

Yəni, təsadüfi dəyişən aşağıdakı kimi göstərilir:

$$
X \sim \mathcal{N}(\mu, \sigma^2)
$$

---

### ✅ 2. Normal Paylanmanın Düsturu (PDF)

**Ehtimal sıxlıq funksiyası (PDF)** aşağıdakı kimidir:

$$
f(x) = \frac{1}{\sigma \sqrt{2\pi}} \cdot e^{- \frac{(x - \mu)^2}{2\sigma^2}}
$$

---

### 🎯 Burada hər simvol nəyi bildirir?

* \$\mu\$ — Orta (mean)
* \$\sigma\$ — Standart sapma
* \$x\$ — Hesablanacaq nöqtə
* \$e\$ — Təqribən 2.718 (natural logaritma əsası)
* \$\pi\$ — Təqribən 3.1416
* \$\sigma \sqrt{2\pi}\$ — Normallaşdırma üçün əmsal (cəmin 1 olması üçün)

---

### ✅ Məsələn:

**Sual:** Boy normal paylanmışdır: $\mu = 170 \, \text{sm}, \, \sigma = 10$. 160 sm boylu bir insanın ehtimal sıxlığı nə qədərdir?

**Cavab:**

$$
f(160) = \frac{1}{10 \cdot \sqrt{2\pi}} \cdot e^{- \frac{(160 - 170)^2}{2 \cdot 10^2}} = \frac{1}{25.066} \cdot e^{-0.5}
$$

Bu düsturdan **dəqiq ehtimal yox**, **nisbi sıxlıq** alınır. Müəyyən aralıqda ehtimal tapmaq üçün inteqral və ya **CDF** istifadə olunur (aşağıda izah olunur).


---

### ✅ 3. Riyazi Gözləmə (Expected Value)

$$
\mathbb{E}[X] = \mu
$$

Yəni paylanmanın **mərkəzi**, gözlənilən nəticə **μ** ilə göstərilir.

#### 💡 Necə hesablanır?

Əgər ehtimal sıxlığı funksiyası $f(x)$ verilmişsə, gözləmə belə hesablanır:

$$
\mathbb{E}[X] = \int_{-\infty}^{+\infty} x \cdot f(x) \, dx
$$

Normal paylanmada bu inteqral həmişə **μ** verir.

---

### ✅ 4. Dispersiya (Varyans)

$$
Var(X) = \sigma^2
$$

Bu da dəyişkənliyin ölçüsüdür — dəyərlərin ortalama ətrafında nə qədər yayıldığını göstərir.

#### 💡 Necə hesablanır?

Dispersiya, bu şəkildə hesablanır:

$$
Var(X) = \int_{-\infty}^{+\infty} (x - \mu)^2 \cdot f(x) \, dx
$$

Normal paylanma üçün bu həmişə **σ²**-yə bərabər olur.


---


### ✅ 5. Cumulative Distribution Function (CDF)

Normal paylanmada müəyyən aralıqdakı ehtimalı **CDF** ilə tapırıq:

$$
P(a < X < b) = \Phi\left(\frac{b - \mu}{\sigma}\right) - \Phi\left(\frac{a - \mu}{\sigma}\right)
$$

Burada:

* \$\Phi(z)\$ — **standart normal paylanma** üçün **CDF**-dir
* \$\frac{x - \mu}{\sigma}\$ — **z-score** (dəyərin standart normallaşdırması)

#### 💡 Necə hesablanır?

Standart normal paylanmada kəsilməz ehtimal paylanması üçün **CDF** belə inteqralla ifadə olunur:

$$
P(X \leq x) = \Phi(z) = \int_{-\infty}^{z} \frac{1}{\sqrt{2\pi}} e^{-\frac{t^2}{2}} dt
$$

Bu inteqral **əl ilə hesablana bilmir**, adətən **Z cədvəli**, **kompüter proqramları** və ya **Python/Excel funksiyaları** ilə tapılır.

---

### 🧠 Əlavə izah — Z-score nədir?

**Z-score** — müəyyən dəyərin ortalamadan neçə standart sapma uzaqda olduğunu göstərir:

$$
z = \frac{x - \mu}{\sigma}
$$

Bu dəyərlə **standart normal cədvəlindən** ehtimal tapılır.

---

### ✅ 6. Standart Normal Paylanma

Əgər:

$$
X \sim \mathcal{N}(0, 1)
$$

Yəni $\mu = 0$, $\sigma = 1$, bu halda **standart normal paylanma** adlanır və ehtimallar **Z cədvəli** ilə tapılır.

---


## ✅ 7. Qısa Cədvəl — Yekun

| Xüsusiyyət    | Düstur                                                                 | Açıqlama                   |
| ------------- | ---------------------------------------------------------------------- | -------------------------- |
| Paylanma tipi | $$X \sim \mathcal{N}(\mu, \sigma^2)$$                                  | Normal paylanma            |
| PDF           | $$f(x) = \frac{1}{\sigma\sqrt{2\pi}}e^{-\frac{(x-\mu)^2}{2\sigma^2}}$$ | Ehtimal sıxlığı funksiyası |
| Gözləmə       | $$\mathbb{E}[X] = \int_{-\infty}^{+\infty} x \cdot f(x) \, dx$$         | Orta dəyər                 |
| Dispersiya    | $$Var(X) = \int_{-\infty}^{+\infty} (x - \mu)^2 \cdot f(x) \, dx$$  | Dəyişkənlik ölçüsü          |
| Z-score       | $$z = \frac{x - \mu}{\sigma}$$                                         | Standart normallaşdırma      |
| CDF           | $$P(a < X < b) = \Phi(z_b) - \Phi(z_a)$$                              | Aralıq ehtimalı            |




## Z-score tapıldıqdan sonra, Normal Paylanmanın Sıxlıq funksiyası (PDF) və ya Paylayıcı funksiyası (CDF) ilə ehtimal necə hesablanır.



### ✅ **Qısa və Konkret İzahlı Yazılış:**

* **Z-score düsturu:**

$$
z = \frac{x - \mu}{\sigma}
$$

> Burada:
>
> * $x$ – qiymət
> * $\mu$ – orta
> * $\sigma$ – standart sapma

📌 **Nümunə:**
Orta boy $\mu = 170$ sm, standart sapma $\sigma = 10$ sm, bir nəfərin boyu $x = 185$ sm. Onun z-score-u:

$$
z = \frac{185 - 170}{10} = 1.5
$$

---

### ✅ **Z-score verilibsə ehtimal necə tapılır?**

Z cədvəli və ya aşağıdakı **CDF** funksiyası istifadə olunur:

$$
P(X \leq x) = \Phi(z) = \int_{-\infty}^{z} \frac{1}{\sqrt{2\pi}} e^{-\frac{t^2}{2}} dt
$$

> Bu integral **normal paylanmanın paylayıcı funksiyasıdır** (CDF).


📌 **Nümunə:**
Əgər z-score $z = 1.5$-dirsə, onda:

$$
P(X \leq x) = \Phi(1.5) \approx 0.9332
$$

Yəni bu insanın boyu əhalinin təxminən **93.32%-dən uzundur**.


---

### ✅ **Sıxlıq funksiyası (PDF) isə:**

$$
f(x) = \frac{1}{\sigma \sqrt{2\pi}} \cdot e^{ -\frac{(x - \mu)^2}{2\sigma^2} }
$$

> Bu isə ehtimal **sıxlığı** verir, **birbaşa ehtimal deyil**.

📌 **Nümunə:**
Yenə $\mu = 170, \sigma = 10, x = 160$ üçün:

$$
f(160) = \frac{1}{10 \sqrt{2\pi}} \cdot e^{ -\frac{(160 - 170)^2}{2 \cdot 10^2} } \approx \frac{1}{25.066} \cdot e^{-0.5} \approx 0.0242
$$

Bu **PDF dəyəri** 160 sm-dəki ehtimal sıxlığını göstərir — **birbaşa ehtimal deyil.**

---

### ✅ **Nəticə:**

* Z-score → Normal CDF ilə ehtimal → Z cədvəli və ya `NORM.DIST()` (Excel) ilə tapılır.
* Z cədvəlindəki rəqəm: $P(X \leq x)$ ehtimalıdır.


📌 **Nümunə:**
Z-score $z = 0$ üçün ehtimal:

$$
P(X \leq \mu) = \Phi(0) = 0.5
$$

Yəni **orta dəyərdən aşağı və ya yuxarı düşmək ehtimalı 50%-dir.**

---

### **Qeyd**:
> **PDF** düsturu, bu sadəcə ehtimal sıxlığını verir, **bir nöqtədə ehtimal yox**, ərazi altı ehtimal üçün **CDF (yəni `z table`) istifadə olunur.**

📌 **Nümunə fərqi:**
**160** sm boy üçün PDF $f(160) = 0.0242$. Bu, dəqiq **160** sm boyunun **sıxlıq funksiyasıdır, amma ehtimal deyil**.

* Amma **160 sm ilə 180 sm arasında** boylara malik insanların **nisbəti (ehtimalı)** isə **CDF** ilə hesablanır:

$$
P(160 < X < 180) = \Phi\left(\frac{180 - 170}{10}\right) - \Phi\left(\frac{160 - 170}{10}\right) = \Phi(1) - \Phi(-1) \approx 0.8413 - 0.1587 = 0.6826
$$


✅ Yəni, boyu **160 sm ilə 180 sm arasında** olan insanların payı təxminən **68.26%-dir** — bu, **CDF ilə tapılır**, **PDF ilə yox**.

## Üstlü (Exponential) Paylanma



### ✅ **1. Üstlü Paylanmanın Tərifi**

> **Tərif:** Üstlü paylanma, **gözlənilən hadisələr arasında keçən zaman** və ya **növbəti uğurun nə vaxt baş verəcəyi** ilə maraqlanılan situasiyalarda istifadə olunan **kəsilməz ehtimal paylanmasıdır**.

> **Tərif:** Üstlü paylanma, müəyyən bir hadisənin baş verməsi üçün **orta gözləmə vaxtı sabit olan** proseslərdə **vaxt aralığının ehtimalını** ölçür.

---

### ✅ **Parametrlər**

* `λ` (lambda): Hadisənin vahid vaxtda baş vermə intensivliyi (rate).
* `X`: Hadisənin baş verməsinə qədər keçən zaman.

Üstlü paylanma tipik olaraq aşağıdakı halları modelləşdirir:

* Müəyyən hadisənin **nə qədər vaxtdan sonra baş verəcəyini** soruşuruq (məsələn: cihazın nasaz olma vaxtı, müştərinin gəlmə müddəti).
* Hadisələr **Poisson prosesinə uyğun** baş verir (yəni hadisələr **təsadüfi** və **müstəqil** baş verir).
* Gözləmə **kəsilməz zaman dəyişənidir** (məsələn: 2.3 dəqiqədən sonra baş verəcək?).

Belə halda təsadüfi dəyişən:

$$
X \sim Exp(\lambda)
$$

olur — yəni X dəyişəni **üstlü paylanmaya** malikdir.

---

### ✅ 2. Üstlü Paylanmanın Düsturu

**Ehtimal sıxlıq funksiyası (PDF)** aşağıdakı kimidir:

$$
f(x) = \lambda e^{-\lambda x}, \quad x \geq 0
$$

### 🎯 Burada hər simvol nəyi ifadə edir?

* \$x\$ — Hadisənin baş verməsinə qədər keçən zaman
* \$\lambda\$ — Hadisənin baş vermə sürəti (intensivlik)
* \$f(x)\$ — \$x\$ qədər vaxtda hadisənin **dəqiq baş vermə sıxlığı**

---

### ✅ **. Məsələn:**

**Sual:** Bir saytın sorğu qəbul etmə sistemi orta hesabla **hər 5 saniyədə** bir sorğu qəbul edir. Yəni:

$$
\lambda = \frac{1}{5} = 0.2
$$

**Sual:** Növbəti sorğunun **dəqiq 3 saniyədən sonra gəlmə sıxlığı** nə qədərdir?

**Cavab:**

$$
f(3) = 0.2 \cdot e^{-0.2 \cdot 3} = 0.2 \cdot e^{-0.6} ≈ 0.2 \cdot 0.5488 = 0.1098
$$

---

### ✅ 3. Riyazi Gözləmə (Expected Value)

Üstlü paylanmanın gözlənilən dəyəri (orta vaxt):

$$
\mathbb{E}[X] = \frac{1}{\lambda}
$$

### 💡 Niyə belədir?

Əgər hadisə **\$\lambda\$ tezliyi** ilə baş verirsə, orta gözləmə vaxtı **1 hadisə üçün 1/λ qədər** zaman olacaq.

Məsələn:

* \$\lambda = 0.2\$ → hadisə **hər 5 saniyədə** bir baş verir
* Deməli, orta gözləmə vaxtı \$\mathbb{E}\[X] = \frac{1}{0.2} = 5\$ saniyə olacaq

---

### ✅ 4. Dispersiya (Varyans)

Üstlü paylanmanın dispersiyası (dəyişkənlik):

$$
Var(X) = \frac{1}{\lambda^2}
$$

### 💡 Niyə bu düsturdur?

Dispersiya gözlənilən dəyərin kvadratik fərqini ölçür. Üstlü paylanmada dəyişkənlik də **orta dəyərin kvadratı qədər** olur.

Məsələn:

* \$\lambda = 0.2\$ → \$Var(X) = \frac{1}{0.2^2} = 25\$

Bu o deməkdir ki, **gözləmə vaxtları** ətrafında **böyük dəyişkənlik** ola bilər.

---





### ✅ 5.Üstlü paylanmanın CDF-i nədir?

**Üstlü (exponential) funksiyanın CDF-i (Cumulative Distribution Function)** var və çox **önəmli statistik anlam** daşıyır.

Əgər:

* $X \sim \text{Exp}(\lambda)$ — üstlü paylanmaya malikdir,
* burada $\lambda > 0$ — **orta hadisə sayı** (rate),

onda **CDF (yığılı funksiyası)** belə verilir:

$$
F(x) = P(X \leq x) = 1 - e^{-\lambda x}, \quad x \geq 0
$$

və ya **inteqral şəklində**:

$$
F(x) = \int_0^x \lambda e^{-\lambda t} \, dt
$$

> Bu integral — üstlü paylanmanın **sıxlıq funksiyasının (PDF)** sıfırdan \$x\$-ə qədər inteqralıdır.



### 🔍 **Sadə izah:**

* Bu funksiya bizə **hadisənin müəyyən vaxtdan əvvəl başvermə ehtimalını** verir.
* Məsələn: "Bir liftin gəlməsi 3 dəqiqədən çox çəkməyəcək ehtimalı nədir?" kimi sualları bu CDF ilə hesablamaq olar.



### 📌 **Nümunə:**

Əgər $\lambda = 2$, $x = 1$ üçün:

$$
F(1) = 1 - e^{-2 \cdot 1} = 1 - e^{-2} \approx 1 - 0.135 = 0.865
$$

> **Yəni:** Hadisənin 1 dəqiqədən **tez baş vermə ehtimalı** 86.5%-dir.

---

## Müntəzəm (Uniform) paylanma



### ✅ **1. Uniform (Bərabərsıxlıqlı) Paylanmanın Tərifi**

> **Tərif:** Uniform paylanma — təsadüfi dəyişənin müəyyən **müddət ərzində eyni ehtimalla** bütün dəyərləri ala bildiyi ehtimal paylanmasıdır.

> **Tərif:** Uniform (bərabər) paylanma, **bütün dəyərlərin eyni ehtimalla** baş verdiyi bir paylanmadır.
> Yəni təsadüfi dəyişən $X$, $[a, b]$ aralığında **hər bir dəyəri bərabər ehtimalla** ala bilər.

---

### ✅ Parametrlər

* `a`: Minimum dəyər (başlanğıc nöqtəsi)
* `b`: Maksimum dəyər (son nöqtə)
* `x`: Araşdırılan dəyər: Təsadüfi dəyişənin ala biləcəyi dəyər (`a ≤ x ≤ b`)
* \$X \sim U(a, b)\$ — yəni \$X\$ dəyişəni **uniform paylanmaya** tabedir.

---

Uniform paylanma adətən belə situasiyanı təsvir edir:

* Təsadüfi dəyişən yalnız **\[a, b\]** intervalında dəyərlər alır.
* Bu intervalda **hər bir dəyərin baş vermə ehtimalı eynidir**.
* Paylanma **kontinual (kəsilməz)** və ya **diskret** ola bilər. Burada biz **kontinual uniform paylanmanı** izah edirik.

Bu halda təsadüfi dəyişən belə yazılır:

$$
X \sim \mathcal{U}(a, b)
$$

---

### ✅ 2. Ehtimal Sıxlıq Funksiyası (PDF)

Kontinual uniform paylanma üçün ehtimal sıxlıq funksiyası:

$$
f(x) = 
\begin{cases}
\frac{1}{b - a}, & \text{əgər } a \leq x \leq b \\
0, & \text{əks halda}
\end{cases}
$$

Bu, göstərir ki, **$\[a, b\]\$ aralığındakı bütün dəyərlər üçün ehtimal sıxlığı eynidir.**


### 🎯 Burada nə baş verir?

* \$f(x)\$ dəyişənin ehtimal sıxlığıdır — yəni hansı aralıqda daha çox (və ya az) ehtimalla baş verdiyini göstərir.
* Uniform paylanmada bu **həmişə sabitdir**, yəni dəyişməz: \$\frac{1}{b - a}\$.

---


### ✅ 3. Paylayıcı Funksiya (CDF)

**CDF (Cumulative Distribution Function):**

$$
F(x) = P(X \leq x) =
\begin{cases}
0, & x \leq a \\
\frac{x - a}{b - a}, & a < x \leq b \\
1, & x > b
\end{cases}
$$

> Bu funksiya **\$x\$-ə qədər olan dəyərlərin ehtimalını** göstərir.



---

### ✅ 📌 Nümunə

Tutaq ki, \$X \sim U(2, 6)\$ olsun.

* PDF:
  $f(x) = \frac{1}{6 - 2} = \frac{1}{4}, \quad 2 < x \leq 6$

* Məsələn: \$P(3 < X \leq 5)\$ ehtimalı:

  $$
  F(5) - F(3) = \frac{5 - 2}{4} - \frac{3 - 2}{4} = \frac{3}{4} - \frac{1}{4} = \frac{1}{2}
  $$

✅ Yəni \$X\$-in \$3\$ və \$5\$ arasında düşmə ehtimalı **50%**-dir.


---


### ✅ 4. Riyazi Gözləmə (Expected Value)

Uniform paylanmanın gözlənilən dəyəri (ortalama nəticə):

$$
\mathbb{E}[X] = \frac{a + b}{2}
$$

### 💡 Niyə belədir?

Çünki **bütün dəyərlər eyni ehtimalla** baş verir. Ortalama da **intervalın ortası** olur.

Məsələn:

* \$a = 2, b = 6\$ → Orta: \$(2 + 6)/2 = 4\$


> Orta dəyər, **aralığın ortasıdır.**

---


### ✅ 5. Dispersiya (Varyans)

Uniform paylanmanın dispersiyası:

$$
Var(X) = \frac{(b - a)^2}{12}
$$

### 💡 Niyə bu düsturdur?

Bu dəyişkənliyin ölçüsüdür. Aralıq genişləndikcə dəyişkənlik artır:

* **Aralıq genişliyi = (b - a)**
* **Kvadrata görə artması dəyişkənliyi ölçür**
* 12 isə normallaşdırma faktorudur, riyazi sübutla çıxarılır.

Məsələn:

* \$a = 0, b = 4\$ → Dispersiya: \$(4 - 0)^2 / 12 = 16 / 12 = 1.33\$


> Dəyərlərin **ortalamadan yayılma ölçüsüdür.**

---


## ✅ 6. Qısa Cədvəl — Yekun

| Xüsusiyyət    | Düstur                                     | Açıqlama                                 |
| ------------- | ------------------------------------------ | ---------------------------------------- |
| Paylanma tipi | $$X \sim U(a, b)$$                           | $$[a, b]$$ aralığında bərabər ehtimallar |
| PDF           | $$f(x) = \frac{1}{b - a}, \quad a \le x \le b$$ | Hər dəyərin eyni ehtimalla baş verməsi |
| CDF           | $$F(x) = \frac{x - a}{b - a}$$ (aralıq üçün) | Yığılı ehtimal                           |
| Gözləmə       | $$\mathbb{E}[X] = \frac{a + b}{2}$$          | Orta qiymət                              |
| Dispersiya    | $$Var(X) = \frac{(b - a)^2}{12}$$            | Yayılma ölçüsü                           |


---

### ✅ 7. Məsələn:

**Sual:** Təsadüfi bir adam 5 ilə 15 dəqiqə arası bir vaxtda gəlir. Onun **dəqiq 8 dəqiqəyə qədər gəlmə ehtimalı** nə qədərdir?

**Cavab:**

* \$a = 5,\quad b = 15,\quad x = 8\$

Bu ehtimal:

$$
P(X \le 8) = \frac{8 - 5}{15 - 5} = \frac{3}{10} = 0.3
$$
